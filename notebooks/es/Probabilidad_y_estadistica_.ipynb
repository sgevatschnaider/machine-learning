```python
import random
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle
import seaborn as sns
from datetime import datetime, timedelta
import pandas as pd

# Configurar estilo visual
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")

def generate_birthdays(num_people):
    """Genera una lista de cumpleaños aleatorios para un grupo de num_people personas"""
    return [random.randint(1, 365) for _ in range(num_people)]

def has_duplicates(birthdays):
    """Devuelve True si hay duplicados en la lista de cumpleaños"""
    return len(birthdays) != len(set(birthdays))

def birthday_paradox_simulation(num_people, num_simulations=10000):
    """Calcula la probabilidad de que al menos dos personas compartan cumpleaños"""
    count_duplicates = 0
    duplicate_counts = []

    for _ in range(num_simulations):
        birthdays = generate_birthdays(num_people)
        if has_duplicates(birthdays):
            count_duplicates += 1
            # Contar cuántos duplicados hay
            unique_birthdays = len(set(birthdays))
            duplicates = num_people - unique_birthdays
            duplicate_counts.append(duplicates)

    probability = count_duplicates / num_simulations
    return probability, duplicate_counts

def theoretical_probability(num_people):
    """Calcula la probabilidad teórica de la paradoja del cumpleaños"""
    if num_people > 365:
        return 1.0

    probability = 1.0
    for i in range(num_people):
        probability *= (365 - i) / 365

    return 1 - probability

def create_comprehensive_visualization():
    """Crea visualizaciones comprehensivas de la paradoja del cumpleaños"""

    # Solicitar parámetros al usuario
    print("🎂 PARADOJA DEL CUMPLEAÑOS - ANÁLISIS INTERACTIVO 🎂")
    print("=" * 55)

    try:
        max_people = int(input("¿Cuál es el número máximo de personas a analizar? (recomendado: 70): "))
        if max_people < 2:
            max_people = 70
            print("Número ajustado a 70 personas por defecto.")

        num_simulations = int(input("¿Cuántas simulaciones deseas ejecutar? (recomendado: 10000): "))
        if num_simulations < 100:
            num_simulations = 10000
            print("Número de simulaciones ajustado a 10000 por defecto.")

        specific_group = int(input("¿Para qué número específico de personas quieres ver detalles? (ej: 23): "))
        if specific_group < 2:
            specific_group = 23

    except ValueError:
        print("Valores inválidos. Usando valores por defecto.")
        max_people = 70
        num_simulations = 10000
        specific_group = 23

    print(f"\n🔄 Ejecutando simulaciones para hasta {max_people} personas...")
    print(f"📊 Realizando {num_simulations} simulaciones por cada grupo...")

    # Datos para el gráfico principal
    x = list(range(2, max_people + 1))
    y_simulated = []
    y_theoretical = []

    # Calcular probabilidades
    for n in x:
        prob_sim, _ = birthday_paradox_simulation(n, num_simulations)
        prob_theo = theoretical_probability(n)
        y_simulated.append(prob_sim)
        y_theoretical.append(prob_theo)

        # Mostrar progreso
        if n % 10 == 0:
            print(f"✅ Completado hasta {n} personas")

    # Análisis detallado para el grupo específico
    print(f"\n🔍 Análisis detallado para {specific_group} personas:")
    prob_specific, duplicate_counts = birthday_paradox_simulation(specific_group, num_simulations)
    prob_theo_specific = theoretical_probability(specific_group)

    print(f"📈 Probabilidad simulada: {prob_specific:.1%}")
    print(f"📊 Probabilidad teórica: {prob_theo_specific:.1%}")
    print(f"🎯 Diferencia: {abs(prob_specific - prob_theo_specific):.1%}")

    # Crear figura con múltiples subgráficos
    fig = plt.figure(figsize=(20, 12))
    fig.suptitle('🎂 PARADOJA DEL CUMPLEAÑOS - ANÁLISIS COMPLETO 🎂',
                 fontsize=20, fontweight='bold', y=0.98)

    # Gráfico 1: Comparación simulación vs teoría
    ax1 = plt.subplot(2, 3, 1)
    ax1.plot(x, y_simulated, 'o-', label='Simulación', linewidth=2.5, markersize=4, alpha=0.8)
    ax1.plot(x, y_theoretical, 's-', label='Teórica', linewidth=2.5, markersize=4, alpha=0.8)
    ax1.axhline(y=0.5, color='red', linestyle='--', alpha=0.7, label='50% Probabilidad')
    ax1.axvline(x=23, color='green', linestyle='--', alpha=0.7, label='23 Personas')
    ax1.set_xlabel('Número de Personas')
    ax1.set_ylabel('Probabilidad')
    ax1.set_title('Simulación vs Teoría')
    ax1.legend()
    ax1.grid(True, alpha=0.3)
    ax1.set_ylim(0, 1)

    # Gráfico 2: Zoom en la región crítica
    ax2 = plt.subplot(2, 3, 2)
    critical_range = range(15, 35)
    x_critical = [i for i in critical_range if i <= max_people]
    y_critical_sim = [y_simulated[i-2] for i in x_critical]
    y_critical_theo = [y_theoretical[i-2] for i in x_critical]

    ax2.plot(x_critical, y_critical_sim, 'o-', label='Simulación', linewidth=3, markersize=6)
    ax2.plot(x_critical, y_critical_theo, 's-', label='Teórica', linewidth=3, markersize=6)
    ax2.axhline(y=0.5, color='red', linestyle='--', alpha=0.7)
    ax2.axvline(x=23, color='green', linestyle='--', alpha=0.7)
    ax2.set_xlabel('Número de Personas')
    ax2.set_ylabel('Probabilidad')
    ax2.set_title('Zoom: Región Crítica (15-35 personas)')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    # Gráfico 3: Heatmap de probabilidades
    ax3 = plt.subplot(2, 3, 3)
    # Crear matriz para heatmap
    matrix_size = min(10, max_people//5)
    prob_matrix = np.zeros((matrix_size, matrix_size))
    for i in range(matrix_size):
        for j in range(matrix_size):
            people_count = (i * matrix_size + j) * 2 + 2
            if people_count <= max_people:
                prob_matrix[i, j] = y_theoretical[people_count - 2]

    im = ax3.imshow(prob_matrix, cmap='RdYlBu_r', aspect='auto', vmin=0, vmax=1)
    ax3.set_title('Mapa de Calor de Probabilidades')
    ax3.set_xlabel('Índice de Grupo')
    ax3.set_ylabel('Índice de Grupo')
    plt.colorbar(im, ax=ax3, label='Probabilidad')

    # Gráfico 4: Distribución de duplicados para grupo específico
    ax4 = plt.subplot(2, 3, 4)
    if duplicate_counts:
        ax4.hist(duplicate_counts, bins=20, alpha=0.7, color='skyblue', edgecolor='black')
        ax4.set_xlabel('Número de Duplicados')
        ax4.set_ylabel('Frecuencia')
        ax4.set_title(f'Distribución de Duplicados\n({specific_group} personas)')
        ax4.grid(True, alpha=0.3)

    # Gráfico 5: Comparación con eventos aleatorios
    ax5 = plt.subplot(2, 3, 5)
    random_events = [0.1, 0.25, 0.5, 0.75, 0.9]
    event_labels = ['10%', '25%', '50%', '75%', '90%']
    colors = ['red', 'orange', 'yellow', 'lightgreen', 'green']

    for i, (prob, label, color) in enumerate(zip(random_events, event_labels, colors)):
        # Encontrar el número de personas para esta probabilidad
        people_for_prob = next((j for j, p in enumerate(y_theoretical) if p >= prob), len(y_theoretical))
        ax5.bar(i, people_for_prob + 2, color=color, alpha=0.7, label=f'{label}: {people_for_prob + 2} personas')

    ax5.set_xlabel('Nivel de Probabilidad')
    ax5.set_ylabel('Personas Necesarias')
    ax5.set_title('Personas Necesarias por Probabilidad')
    ax5.set_xticks(range(len(event_labels)))
    ax5.set_xticklabels(event_labels)
    ax5.grid(True, alpha=0.3)

    # Gráfico 6: Estadísticas del grupo específico
    ax6 = plt.subplot(2, 3, 6)
    ax6.axis('off')

    # Crear texto con estadísticas
    stats_text = f"""
    ESTADÍSTICAS PARA {specific_group} PERSONAS:

    🎯 Probabilidad simulada: {prob_specific:.1%}
    📊 Probabilidad teórica: {prob_theo_specific:.1%}
    🔄 Simulaciones realizadas: {num_simulations:,}
    ✅ Casos con duplicados: {int(prob_specific * num_simulations):,}

    DATOS CURIOSOS:
    • Con 23 personas: ~50% probabilidad
    • Con 30 personas: ~70% probabilidad
    • Con 50 personas: ~97% probabilidad
    • Con 70 personas: ~99.9% probabilidad

    ¡La paradoja muestra que la intuición
    humana sobre probabilidades puede
    ser engañosa!
    """

    ax6.text(0.1, 0.9, stats_text, transform=ax6.transAxes, fontsize=11,
             verticalalignment='top', bbox=dict(boxstyle="round,pad=0.5",
             facecolor="lightblue", alpha=0.8))

    plt.tight_layout()
    plt.show()

    # Mostrar tabla de resultados clave
    print("\n" + "="*70)
    print("📋 TABLA DE RESULTADOS CLAVE")
    print("="*70)
    print(f"{'Personas':<10} {'Prob. Teórica':<15} {'Prob. Simulada':<15} {'Diferencia':<10}")
    print("-"*70)

    key_values = [10, 15, 20, 23, 25, 30, 35, 40, 45, 50]
    for people in key_values:
        if people <= max_people:
            idx = people - 2
            if idx < len(y_theoretical):
                theo = y_theoretical[idx]
                sim = y_simulated[idx]
                diff = abs(theo - sim)
                print(f"{people:<10} {theo:<15.1%} {sim:<15.1%} {diff:<10.1%}")

    print("\n🎉 ¡Análisis completado! La paradoja del cumpleaños demuestra que")
    print("   con solo 23 personas hay ~50% de probabilidad de cumpleaños compartidos.")

def main():
    """Función principal"""
    try:
        create_comprehensive_visualization()
    except KeyboardInterrupt:
        print("\n\n❌ Análisis interrumpido por el usuario.")
    except Exception as e:
        print(f"\n❌ Error durante el análisis: {e}")
        print("Verifica que tengas instaladas las librerías necesarias:")
        print("pip install matplotlib seaborn pandas numpy")

if __name__ == "__main__":
    main()
```

```python
import hashlib
import random
import string
import math
import os

# --- PARÁMETROS DEL EXPERIMENTO ---
# Decidimos usar un hash de 32 bits. Esto es crucial.
# Un hash más largo (como SHA-256 completo) haría imposible encontrar una colisión.
# Un hash de 32 bits tiene 2^32 resultados posibles.
# 8 caracteres hexadecimales = 8 * 4 bits = 32 bits.
BITS = 32
NUM_CARACTERES_HASH = BITS // 4

# El "espacio de hashes" es el número total de resultados posibles.
# Es el análogo a los "365 días del año" en la paradoja original.
ESPACIO_HASHES = 2**BITS

def generar_hash_truncado(texto: str) -> str:
    """
    Genera un hash SHA-256 de un texto y lo trunca a un número predefinido
    de caracteres para hacer las colisiones factibles.
    """
    hash_obj = hashlib.sha256(texto.encode())
    return hash_obj.hexdigest()[:NUM_CARACTERES_HASH]

def encontrar_primera_colision() -> dict:
    """
    Busca una colisión de hash y registra cuántos intentos fueron necesarios.
    Muestra un contador en vivo para ilustrar el progreso.

    Returns:
        Un diccionario con la información de la colisión.
    """
    hashes_vistos = {}
    intentos = 0
    while True:
        intentos += 1
        texto_aleatorio = ''.join(random.choices(string.ascii_letters + string.digits, k=12))
        hash_val = generar_hash_truncado(texto_aleatorio)

        # Imprimir progreso para que el usuario vea el proceso
        if intentos % 10000 == 0:
            print(f"  ...intentos realizados: {intentos:,}")

        if hash_val in hashes_vistos:
            # ¡Colisión encontrada!
            return {
                "texto1": texto_aleatorio,
                "texto2": hashes_vistos[hash_val],
                "hash_colision": hash_val,
                "intentos": intentos
            }
        else:
            hashes_vistos[hash_val] = texto_aleatorio

def main():
    """Flujo principal del programa de demostración."""
    print("🎂 **Demostración de la Paradoja del Cumpleaños con Hashes** 🎂")
    print("-" * 60)
    print("La Paradoja del Cumpleaños dice que en un grupo de solo 23 personas,")
    print("la probabilidad de que dos compartan cumpleaños es de más del 50%.")
    print("\nVamos a ver un efecto similar con hashes criptográficos.")
    print("\nNUESTRA ANALOGÍA:")
    print(f"  - Un 'año' tiene {ESPACIO_HASHES:,.0f} 'días' (hashes posibles de {BITS} bits).")
    print("  - Una 'persona' es una cadena de texto aleatoria.")
    print("  - Un 'cumpleaños compartido' es una COLISIÓN de hash.")
    print("-" * 60)

    # Cálculo teórico: ¿Cuántos intentos esperamos necesitar?
    # La aproximación para un 50% de probabilidad es sqrt(pi/2 * N)
    intentos_esperados_teoricos = math.sqrt((math.pi / 2) * ESPACIO_HASHES)

    print(f"Análisis Teórico:")
    print(f"Aunque hay ~4.3 mil millones de hashes posibles, la teoría predice")
    print(f"que probablemente encontraremos una colisión en solo unos \033[1m{intentos_esperados_teoricos:,.0f}\033[0m intentos.")
    input("\nPresiona Enter para iniciar el experimento y buscar la colisión...")

    print("\n **Iniciando la búsqueda...**")
    resultado = encontrar_primera_colision()

    print("\n" + "=" * 60)
    print("🎉 ¡COLISIÓN ENCONTRADA! 🎉")
    print("=" * 60)

    print(f"Se necesitaron \033[1;32m{resultado['intentos']:,}\033[0m intentos para encontrar una colisión.")
    print(f"\nLa predicción teórica era de ~{intentos_esperados_teoricos:,.0f} intentos.")

    print("\nDetalles de la Colisión:")
    print(f"  -> Texto 1: '{resultado['texto1']}'")
    print(f"  -> Texto 2: '{resultado['texto2']}'")
    print(f"  => Ambos producen el mismo hash de {BITS} bits: \033[1;33m{resultado['hash_colision']}\033[0m")

    print("\n" + "-" * 60)
    print("Esta es la Paradoja del Cumpleaños en acción: el número de intentos")
    print("necesarios es mucho, mucho más pequeño de lo que la intuición nos diría.")

if __name__ == "__main__":
    main()
```

```python
import random
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Función para simular lanzamientos y calcular la probabilidad empírica
def simulate_rolls(num_rolls):
    successful_rolls = 0
    empirical_probabilities = []

    for i in range(1, num_rolls + 1):
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        if die1 == 6 and die2 == 6:
            successful_rolls += 1
        # Calculamos y almacenamos la probabilidad empírica en cada paso
        empirical_probability = (successful_rolls / i) * 100
        empirical_probabilities.append(empirical_probability)

    return empirical_probabilities

# Función para actualizar el gráfico con el número de lanzamientos seleccionado
def update_plot(num_rolls):
    empirical_probabilities = simulate_rolls(num_rolls)
    theoretical_probability = [2.78] * num_rolls  # Probabilidad teórica constante

    plt.figure(figsize=(12, 6))
    plt.plot(empirical_probabilities, label='Probabilidad Empírica (Observada)', color='#1f77b4', linewidth=2)
    plt.plot(theoretical_probability, color='red', linestyle='--', label='Probabilidad Teórica (2.78%)')
    plt.xlabel('Número de Lanzamientos', fontsize=12)
    plt.ylabel('Probabilidad (%)', fontsize=12)
    plt.ylim(0, 5)  # Limitar el rango y del eje para claridad
    plt.title(f'Convergencia de la Probabilidad Empírica hacia la Teórica con {num_rolls} Lanzamientos', fontsize=14)
    plt.legend(loc="upper right", fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# Deslizador interactivo para seleccionar el número de lanzamientos
num_rolls_slider = widgets.IntSlider(
    value=100,
    min=10,
    max=10000,
    step=10,
    description='Lanzamientos:',
    style={'description_width': 'initial'}
)

# Botón de reinicio
reset_button = widgets.Button(
    description="Reiniciar",
    button_style='warning',
    tooltip="Reiniciar gráfico"
)

# Función de reinicio para establecer el deslizador en su valor inicial
def reset(event):
    num_rolls_slider.value = 100  # Valor inicial del deslizador

reset_button.on_click(reset)

# Conectar el deslizador y el botón a la función de actualización del gráfico
ui = widgets.HBox([num_rolls_slider, reset_button])
out = widgets.interactive_output(update_plot, {'num_rolls': num_rolls_slider})

# Mostrar la interfaz y el gráfico
display(ui, out)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import ScalarFormatter

def create_data() -> pd.DataFrame:
    """
    Crea y prepara el DataFrame con los datos de muestra,
    calculando la frecuencia relativa.

    Returns:
        pd.DataFrame: DataFrame con las columnas 'n', 'k' y 'Frecuencia Relativa'.
    """
    data = {
        'n': [10, 50, 100, 150, 200, 250, 500, 1000, 2000, 3000, 4000, 5000, 10000],
        'k': [1, 1, 5, 12, 14, 15, 15, 45, 70, 129, 156, 202, 399]
    }
    df = pd.DataFrame(data)
    df['Frecuencia Relativa'] = df['k'] / df['n']
    return df

def plot_frequency_stability(df: pd.DataFrame, stability_value: float = 0.04):
    """
    Genera un gráfico de alta calidad que muestra la estabilidad de la
    frecuencia relativa.

    Args:
        df (pd.DataFrame): DataFrame que contiene 'n' y 'Frecuencia Relativa'.
        stability_value (float): Valor de la línea horizontal de estabilidad.
    """
    # 1. Configuración de estilo y figura (Interfaz Orientada a Objetos)
    sns.set_theme(style="whitegrid")
    fig, ax = plt.subplots(figsize=(12, 7))

    # 2. Línea de estabilidad (graficada primero para que quede al fondo)
    ax.axhline(
        y=stability_value,
        color='crimson',
        linestyle='--',
        linewidth=1.5,
        label=f'Valor de Estabilidad Teórico ({stability_value})',
        zorder=1
    )

    # 3. Gráfico principal de la frecuencia relativa
    ax.plot(
        df['n'],
        df['Frecuencia Relativa'],
        marker='o',
        linestyle='-',
        color='royalblue',
        label='Frecuencia Relativa Observada',
        zorder=2
    )

    # 4. Anotación del último punto para resaltar la convergencia
    last_n = df['n'].iloc[-1]
    last_fr = df['Frecuencia Relativa'].iloc[-1]
    ax.annotate(
        f'Valor final: {last_fr:.4f}',
        xy=(last_n, last_fr),
        xytext=(last_n / 2.5, last_fr + 0.005), # Posición del texto
        arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8),
        fontsize=10,
        fontweight='bold',
        bbox=dict(boxstyle="round,pad=0.3", fc="yellow", ec="black", lw=1, alpha=0.7)
    )

    # 5. Configuración de títulos, etiquetas y escalas
    ax.set_title('Análisis de Estabilidad de la Frecuencia Relativa', fontsize=16, fontweight='bold', pad=20)
    ax.set_xlabel('Tamaño de la Muestra (n)', fontsize=12)
    ax.set_ylabel('Frecuencia Relativa (fr)', fontsize=12)
    ax.set_xscale('log') # Escala logarítmica es ideal para este tipo de datos

    # 6. Mejora de los 'ticks' del eje x para que sean legibles en escala logarítmica
    ax.xaxis.set_major_formatter(ScalarFormatter())
    ax.tick_params(axis='x', which='major', rotation=0) # Evitar rotación si no es necesaria

    # 7. Leyenda y rejilla
    ax.legend(loc='upper right', fontsize=10)
    ax.grid(True, which="both", linestyle='--', linewidth=0.5)

    # 8. Ajustar el layout y mostrar el gráfico
    fig.tight_layout()
    plt.show()

# --- Flujo principal de ejecución ---
if __name__ == '__main__':
    # Crear los datos
    dataframe = create_data()

    # Mostrar la tabla en la consola
    print("Tabla de Frecuencia Absoluta y Relativa:")
    print(dataframe.to_string(index=False))

    # Generar la visualización mejorada
    plot_frequency_stability(dataframe)
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Configuración de la simulación
num_lanzamientos_monedas = 1000
num_lanzamientos_dado = 100

# Simulación Frecuentista: Lanzamiento de una moneda
def frecuentista_monedas(num_lanzamientos):
    resultados = {"cara": 0, "ceca": 0}
    frecuencias = []

    for _ in range(num_lanzamientos):
        resultado = random.choice(["cara", "ceca"])
        resultados[resultado] += 1
        frecuencia_cara = resultados["cara"] / (resultados["cara"] + resultados["ceca"])
        frecuencias.append(frecuencia_cara * 100)

    return frecuencias, resultados["cara"]

# Simulación Bayesiana: Evolución paso a paso
def bayesiana_monedas_evolutiva(num_caras):
    alpha, beta = 1, 1
    posterior_hist = []

    for i, caras in enumerate(num_caras, start=1):
        posterior = (caras + alpha) / (i + alpha + beta)
        posterior_hist.append(posterior * 100)

    return posterior_hist

# Variable Aleatoria: Posibles resultados al lanzar un dado
def resultados_dado():
    valores = [1, 2, 3, 4, 5, 6]
    conteo = {i: 0 for i in valores}
    resultados = []

    for _ in range(num_lanzamientos_dado):
        resultado = random.choice(valores)
        conteo[resultado] += 1
        resultados.append(resultado)

    return resultados, conteo

# Ejecutar simulación Frecuentista
frecuencias, total_caras = frecuentista_monedas(num_lanzamientos_monedas)

# Construir lista acumulada de caras para Bayes paso a paso
caras_acumuladas = []
caracount = 0
for i in range(num_lanzamientos_monedas):
    if frecuencias[i] * (i+1) / 100 > caracount:
        caracount += 1
    caras_acumuladas.append(caracount)

# Ejecutar Bayes paso a paso
bayesianas = bayesiana_monedas_evolutiva(caras_acumuladas)

# Ejecutar simulación de lanzamientos de dado
resultados_dado, conteo_dado = resultados_dado()

# Visualización
plt.figure(figsize=(16, 6))

# Gráfico de probabilidad frecuentista (moneda)
plt.subplot(1, 3, 1)
plt.plot(frecuencias, label="Frecuentista - 'Cara'")
plt.axhline(50, color='red', linestyle='--', label="Probabilidad Teórica (50%)")
plt.xlabel("Número de Lanzamientos")
plt.ylabel("Frecuencia de 'Cara' (%)")
plt.title("Frecuencia Observada de 'Cara'")
plt.legend()
plt.grid()

# Gráfico de probabilidad bayesiana (moneda)
plt.subplot(1, 3, 2)
plt.plot(bayesianas, color='green', label="Bayes - P(Cara)")
plt.axhline(50, color='red', linestyle='--', label="Probabilidad Teórica (50%)")
plt.xlabel("Número de Lanzamientos")
plt.ylabel("Probabilidad Bayesiana (%)")
plt.title("Probabilidad Bayesiana de Moneda Justa")
plt.legend()
plt.grid()

# Gráfico de resultados del dado
plt.subplot(1, 3, 3)
plt.bar(conteo_dado.keys(), conteo_dado.values(), color="skyblue")
plt.xlabel("Resultados posibles (1 a 6)")
plt.ylabel("Frecuencia")
plt.title("Distribución de resultados al lanzar un dado")
plt.grid(axis='y')

plt.tight_layout()
plt.show()

# Mostrar resultado de probabilidad bayesiana final
print(f"Probabilidad Bayesiana final de que la moneda sea justa: {bayesianas[-1]:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import beta
from ipywidgets import interact, FloatSlider

# --- INSTRUCCIONES ---
# Este código está diseñado para ejecutarse en un entorno interactivo
# como Jupyter Notebook o JupyterLab.
# Simplemente ejecuta esta celda y aparecerán dos sliders debajo.
# ¡Mueve los sliders para explorar la distribución Beta!

def interactive_beta_plot(alpha_val: float, beta_val: float):
    """
    Genera un gráfico interactivo de la distribución Beta que se actualiza
    en tiempo real basado en los valores de los sliders para alfa y beta.

    Args:
        alpha_val (float): El parámetro 'alfa' (éxitos + 1).
        beta_val (float): El parámetro 'beta' (fracasos + 1).
    """
    # 1. Configuración de estilo y figura
    sns.set_theme(style="whitegrid")
    fig, ax = plt.subplots(figsize=(12, 7))

    # 2. Dominio de la variable x y cálculo de la PDF
    x = np.linspace(0, 1, 1000)
    pdf = beta.pdf(x, alpha_val, beta_val)

    # 3. Graficar la distribución
    ax.plot(x, pdf, color='royalblue', linewidth=3, label="Densidad de Probabilidad (PDF)")
    ax.fill_between(x, pdf, alpha=0.2, color='royalblue')

    # 4. Cálculos estadísticos y anotaciones
    # Media: El valor esperado
    mean = alpha_val / (alpha_val + beta_val)
    ax.axvline(x=mean, color='green', linestyle='--', linewidth=2, label=f"Media (μ) = {mean:.2f}")

    # Modo: El valor más probable (el pico de la distribución)
    # Se manejan casos especiales para alfa o beta <= 1
    if alpha_val > 1 and beta_val > 1:
        mode = (alpha_val - 1) / (alpha_val + beta_val - 2)
        ax.axvline(x=mode, color='red', linestyle='--', linewidth=2, label=f"Modo = {mode:.2f}")

    # 5. Títulos y etiquetas dinámicas
    ax.set_title(f"Distribución Beta (α={alpha_val:.1f}, β={beta_val:.1f})", fontsize=18, fontweight='bold', pad=20)
    ax.set_xlabel("Valor de la Probabilidad (p)", fontsize=14)
    ax.set_ylabel("Densidad de Probabilidad", fontsize=14)
    ax.set_xlim(0, 1)
    # Fijar el eje Y es clave para una comparación visual honesta
    ax.set_ylim(0, 10)

    # 6. Leyenda y finalización
    ax.legend(fontsize=12)
    plt.show()

# --- Creación de los Widgets Interactivos ---
# 'interact' crea automáticamente los sliders a partir de los argumentos de la función.
# Usamos FloatSlider para un control más preciso.
interact(
    interactive_beta_plot,
    alpha_val=FloatSlider(min=0.1, max=20.0, step=0.1, value=2.0, description='α (Éxitos+1):'),
    beta_val=FloatSlider(min=0.1, max=20.0, step=0.1, value=2.0, description='β (Fracasos+1):')
);
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import beta
from typing import List, Tuple

def plot_beta_distributions(param_sets: List[Tuple[float, float]]):
    """
    Genera un gráfico comparativo de alta calidad de múltiples distribuciones Beta
    para ilustrar el impacto de los parámetros alfa y beta.

    Args:
        param_sets (List[Tuple[float, float]]):
            Una lista de tuplas, donde cada tupla contiene un par (alfa, beta).
    """
    # 1. Configuración de estilo y figura (Interfaz Orientada a Objetos)
    sns.set_theme(style="whitegrid")
    fig, ax = plt.subplots(figsize=(14, 8))

    # 2. Dominio de la variable x (probabilidades de 0 a 1)
    x = np.linspace(0, 1, 1000)

    # 3. Generar una paleta de colores profesional
    colors = sns.color_palette("viridis", len(param_sets))

    # 4. Iterar sobre cada par de parámetros para graficar
    for i, (alpha_val, beta_val) in enumerate(param_sets):
        # Calcular la Función de Densidad de Probabilidad (PDF)
        pdf = beta.pdf(x, alpha_val, beta_val)

        # Graficar la distribución
        ax.plot(x, pdf, label=f"α={alpha_val}, β={beta_val}", color=colors[i], linewidth=2.5)
        ax.fill_between(x, pdf, alpha=0.15, color=colors[i])

        # Calcular y graficar la media (valor esperado)
        mean = alpha_val / (alpha_val + beta_val)
        ax.axvline(
            x=mean,
            color=colors[i],
            linestyle='--',
            linewidth=1.2,
            label=f"Media (μ) = {mean:.2f}"
        )

    # 5. Configuración de títulos, etiquetas y límites
    ax.set_title("Análisis Comparativo de la Distribución Beta", fontsize=18, fontweight='bold', pad=20)
    ax.set_xlabel("Valor de la Probabilidad (p)", fontsize=14)
    ax.set_ylabel("Densidad de Probabilidad", fontsize=14)
    ax.set_xlim(0, 1)
    ax.set_ylim(bottom=0)

    # 6. Crear una leyenda clara y organizada
    # Para evitar leyendas duplicadas, combinamos las etiquetas de línea y media
    handles, labels = ax.get_legend_handles_labels()
    # Agrupamos las etiquetas por distribución
    new_labels = [f"Beta(α={p[0]}, β={p[1]}), μ={p[0]/(p[0]+p[1]):.2f}" for p in param_sets]
    ax.legend([handles[i*2] for i in range(len(param_sets))], new_labels, title="Parámetros y Media", fontsize=11, title_fontsize=13)

    # 7. Ajustar el layout y mostrar el gráfico
    fig.tight_layout()
    plt.show()

# --- Flujo principal de ejecución ---
if __name__ == '__main__':
    # Define los conjuntos de parámetros (alfa, beta) para visualizar
    # Cada par cuenta una historia diferente:
    # (1, 1): Creencia inicial, uniforme. "Cualquier probabilidad es igualmente posible".
    # (5, 2): 4 éxitos, 1 fracaso. "La probabilidad parece estar por encima de 0.5".
    # (2, 5): 1 éxito, 4 fracasos. "La probabilidad parece estar por debajo de 0.5".
    # (10, 10): 9 éxitos, 9 fracasos. "La probabilidad está fuertemente centrada en 0.5".
    # (0.5, 0.5): Creencia de que la probabilidad es extrema (cerca de 0 o 1).
    parameter_sets_to_plot = [
        (1, 1),
        (5, 2),
        (2, 5),
        (10, 10),
        (0.5, 0.5)
    ]

    # Generar la visualización
    plot_beta_distributions(parameter_sets_to_plot)
```

```python
import random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# --- CONSTANTES DEL JUEGO ---
CARAS_DADO = 6
VALOR_ESPERADO_LANZAMIENTO = sum(range(1, CARAS_DADO + 1)) / CARAS_DADO

def obtener_configuracion_juego() -> tuple:
    """
    Interactúa con el usuario para establecer el costo por juego y muestra
    el análisis teórico inicial.

    Returns:
        Una tupla con (costo_por_juego, ganancia_esperada_teorica).
        Retorna (None, None) si la entrada es inválida.
    """
    print("🎲 **Bienvenido al Laboratorio de Apuestas del Dado** 🎲")
    print("-" * 50)
    print(f"El valor promedio que se obtiene al lanzar un dado de {CARAS_DADO} caras es: {VALOR_ESPERADO_LANZAMIENTO:.2f}")
    print("Este es el 'pago justo' que la casa debería darte por lanzamiento.")
    print("-" * 50)

    try:
        costo_str = input("➡️  Introduce el COSTO que pagarás para jugar cada ronda: ")
        costo_por_juego = float(costo_str)
    except ValueError:
        print("❌ Error: Debes introducir un número válido.")
        return None, None

    ganancia_esperada_teorica = VALOR_ESPERADO_LANZAMIENTO - costo_por_juego

    print("\n--- Análisis Teórico ---")
    print(f"Tu ganancia esperada por cada juego es: {VALOR_ESPERADO_LANZAMIENTO:.2f} - {costo_por_juego:.2f} = {ganancia_esperada_teorica:.2f}")

    if ganancia_esperada_teorica < 0:
        print("📉 Estás en un juego DESFAVORABLE (como en un casino). Se espera que pierdas dinero a largo plazo.")
    elif ganancia_esperada_teorica == 0:
        print("⚖️  Estás en un juego JUSTO. Se espera que ni ganes ni pierdas dinero a largo plazo.")
    else:
        print("📈 Estás en un juego FAVORABLE. Se espera que ganes dinero a largo plazo.")

    return costo_por_juego, ganancia_esperada_teorica

def realizar_simulacion(costo_por_juego: float, num_juegos: int) -> dict:
    """
    Realiza la simulación del juego y retorna el historial de ganancias.

    Args:
        costo_por_juego: El costo para participar en cada ronda.
        num_juegos: El número total de juegos a simular.

    Returns:
        Un diccionario con el historial de ganancias acumuladas y promedio.
    """
    historial_ganancia_acumulada = []
    historial_ganancia_promedio = []
    ganancia_acumulada = 0

    for i in range(1, num_juegos + 1):
        resultado_dado = random.randint(1, CARAS_DADO)
        ganancia_ronda = resultado_dado - costo_por_juego
        ganancia_acumulada += ganancia_ronda

        historial_ganancia_acumulada.append(ganancia_acumulada)
        historial_ganancia_promedio.append(ganancia_acumulada / i)

    return {
        "acumulada": historial_ganancia_acumulada,
        "promedio": historial_ganancia_promedio
    }

def crear_visualizacion_didactica(historial: dict, teorica: float, num_juegos: int):
    """
    Crea un gráfico avanzado de doble eje para visualizar los resultados de la simulación.
    """
    sns.set_theme(style="whitegrid")
    fig, ax1 = plt.subplots(figsize=(14, 8))

    juegos_eje_x = np.arange(1, num_juegos + 1)

    # EJE 1: Ganancia Acumulada Total
    color_acumulada = 'royalblue'
    ax1.set_xlabel('Número de Juegos Realizados', fontsize=12)
    ax1.set_ylabel('Ganancia Acumulada Total ($)', color=color_acumulada, fontsize=12)
    ax1.plot(juegos_eje_x, historial["acumulada"], color=color_acumulada, label='Ganancia Acumulada Real')
    ax1.tick_params(axis='y', labelcolor=color_acumulada)
    ax1.axhline(0, color='black', linestyle='--', linewidth=1)
    ax1.set_xscale('log') # La escala logarítmica muestra mejor la convergencia

    # EJE 2: Ganancia Promedio por Juego (la clave didáctica)
    ax2 = ax1.twinx()
    color_promedio = 'green'
    ax2.set_ylabel('Ganancia Promedio por Juego ($)', color=color_promedio, fontsize=12)
    ax2.plot(juegos_eje_x, historial["promedio"], color=color_promedio, label='Ganancia Promedio Real')
    ax2.tick_params(axis='y', labelcolor=color_promedio)

    # Línea Teórica: La demostración de la Ley de los Grandes Números
    ax2.axhline(y=teorica, color='red', linestyle='--', linewidth=2, label=f'Ganancia Esperada Teórica ({teorica:.2f})')

    # Título y Leyenda
    plt.title('Análisis de Simulación: Ganancia Acumulada vs. Ganancia Promedio', fontsize=16, fontweight='bold', pad=20)
    fig.legend(loc="upper right", bbox_to_anchor=(0.9, 0.9))

    fig.tight_layout()
    plt.show()

def main():
    """Flujo principal del programa."""
    costo, teorica = obtener_configuracion_juego()
    if costo is None:
        return

    num_juegos_simulacion = 50000
    print(f"\n Procediendo a simular {num_juegos_simulacion} juegos... ¡Esto puede tardar un momento!\n")

    historial = realizar_simulacion(costo, num_juegos_simulacion)

    # Análisis final
    ganancia_final = historial["acumulada"][-1]
    promedio_final = historial["promedio"][-1]

    print("--- 🏁 Resultados Finales de la Simulación 🏁 ---")
    print(f"Ganancia Total Acumulada: ${ganancia_final:,.2f}")
    print(f"Ganancia Promedio por Juego (Real): ${promedio_final:.4f}")
    print(f"Ganancia Promedio por Juego (Teórica): ${teorica:.4f}")
    print("\nObserva en el gráfico cómo la línea verde (promedio real) converge hacia la línea roja discontinua (promedio teórico).")
    print("¡Esa es la Ley de los Grandes Números en acción!")

    crear_visualizacion_didactica(historial, teorica, num_juegos_simulacion)

if __name__ == "__main__":
    main()
```
