```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generar datos
np.random.seed(42)
X = np.linspace(0, 10, 100).reshape(-1, 1)
y_true = 3 * X + 7  # Relación verdadera
noise = np.random.normal(0, 3, size=y_true.shape)  # Ruido aleatorio
y = y_true + noise  # Datos observados con ruido

# Ajustar un modelo de regresión lineal
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Visualizar
plt.scatter(X, y, label="Datos con ruido", color="blue", alpha=0.6)
plt.plot(X, y_true, label="Relación Verdadera", color="green", linestyle="--", linewidth=2)
plt.plot(X, y_pred, label="Modelo Ajustado", color="red", linewidth=2)
plt.legend()
plt.title("Efecto del Ruido en los Datos")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
from IPython.display import display, HTML

# Código HTML que incluye el texto completo
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Sesgo y Varianza en Machine Learning</title>
    <style>
        body {
            font-family: 'Georgia', serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
        }
        h1, h2, h3, h4 {
            color: #333;
            font-family: 'Arial', sans-serif;
        }
        h1 {
            font-size: 2.5em;
            margin-bottom: 10px;
        }
        h2 {
            font-size: 2em;
            margin-bottom: 10px;
        }
        h3 {
            font-size: 1.5em;
            margin-bottom: 5px;
        }
        h4 {
            font-size: 1.3em;
            margin-bottom: 5px;
        }
        p {
            font-size: 1.2em;
            color: #555;
            margin-bottom: 15px;
        }
        ul {
            font-size: 1.2em;
            color: #555;
            margin-bottom: 15px;
            padding-left: 20px;
        }
        li {
            margin-bottom: 10px;
        }
        hr {
            border: 1px solid #ddd;
            margin: 20px 0;
        }
    </style>
</head>
<body>
    <h1>Sesgo y Varianza en Machine Learning</h1>

    <h2>Introducción al Sesgo y la Varianza</h2>
    <p>
        El sesgo y la varianza son conceptos clave en el aprendizaje automático que describen cómo los modelos procesan los datos
        de entrenamiento y generalizan a nuevos datos.
    </p>
    <ul>
        <li><strong>Sesgo:</strong> Se refiere a los supuestos que un modelo hace sobre los datos para predecir el resultado. Modelos con alto sesgo tienden a simplificar en exceso la realidad, ignorando patrones complejos.</li>
        <li><strong>Varianza:</strong> Mide la sensibilidad de un modelo a los cambios en el conjunto de datos de entrenamiento. Un modelo con alta varianza es demasiado flexible, adaptándose a detalles específicos del entrenamiento, lo que reduce su capacidad de generalización.</li>
    </ul>
    <p>
        Comprender y equilibrar estos conceptos permite construir modelos más efectivos que logren un desempeño óptimo en datos desconocidos.
    </p>
    <hr>

    <h2>Importancia de Comprender el Sesgo y la Varianza</h2>
    <ol>
        <li><strong>Diagnóstico del Modelo</strong>
            <ul>
                <li>Identificar si un modelo está:
                    <ul>
                        <li><strong>Sobreajustando (Overfitting):</strong> Captura ruido y patrones específicos del conjunto de entrenamiento (alta varianza).</li>
                        <li><strong>Subajustando (Underfitting):</strong> No capta suficientes patrones de los datos (alto sesgo).</li>
                    </ul>
                </li>
                <li>Herramientas como curvas de aprendizaje o validación cruzada ayudan a evaluar estos problemas.</li>
            </ul>
        </li>
        <li><strong>Optimización del Modelo</strong>
            <ul>
                <li>Lograr un balance adecuado entre precisión en los datos de entrenamiento y capacidad de generalización.</li>
                <li>Esto implica elegir algoritmos, hiperparámetros y técnicas de regularización adecuados.</li>
            </ul>
        </li>
        <li><strong>Impacto en el Desempeño</strong>
            <ul>
                <li>Un <strong>modelo con alto sesgo:</strong>
                    <ul>
                        <li>Simplifica demasiado los patrones subyacentes en los datos.</li>
                        <li>Desempeño deficiente tanto en entrenamiento como en validación.</li>
                    </ul>
                </li>
                <li>Un <strong>modelo con alta varianza:</strong>
                    <ul>
                        <li>Se ajusta demasiado a los datos de entrenamiento.</li>
                        <li>Excelente desempeño en entrenamiento pero deficiente en validación.</li>
                    </ul>
                </li>
            </ul>
        </li>
    </ol>
    <hr>

    <h2>Fuentes de Errores en Modelos de Machine Learning</h2>
    <ol>
        <li><strong>Error Irreducible</strong>
            <ul>
                <li><strong>Definición:</strong> Proviene de la variabilidad inherente en los datos, como ruido, datos incompletos o mal etiquetados.</li>
                <li><strong>Impacto:</strong> No puede ser eliminado, incluso con el mejor modelo posible.</li>
                <li><strong>Ejemplo:</strong> Variaciones no predecibles en las decisiones humanas.</li>
            </ul>
        </li>
        <li><strong>Error de Sesgo</strong>
            <ul>
                <li><strong>Definición:</strong> Ocurre cuando un modelo hace suposiciones excesivamente simplistas.</li>
                <li><strong>Impacto:</strong> Subajuste, donde el modelo no capta patrones complejos.</li>
                <li><strong>Ejemplo:</strong> Usar una regresión lineal para modelar una relación cuadrática.</li>
            </ul>
        </li>
        <li><strong>Error de Varianza</strong>
            <ul>
                <li><strong>Definición:</strong> Surge cuando un modelo es demasiado complejo y se adapta a detalles específicos del conjunto de entrenamiento.</li>
                <li><strong>Impacto:</strong> Sobreajuste, lo que lleva a un bajo desempeño en datos de prueba.</li>
                <li><strong>Ejemplo:</strong> Árboles de decisión profundos que se ajustan exactamente al entrenamiento.</li>
            </ul>
        </li>
    </ol>
    <hr>

    <h2>Conclusión</h2>
    <p>
        Un balance adecuado entre el sesgo y la varianza es esencial para construir modelos efectivos. Este equilibrio,
        conocido como el <em>bias-variance tradeoff</em>, determina la capacidad del modelo para generalizar correctamente.
        Diseñar modelos con una complejidad adecuada y usar técnicas como validación cruzada son pasos clave para lograr
        este objetivo.
    </p>
</body>
</html>
"""

# Renderizar el HTML en Google Colab
display(HTML(html_content))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
import pandas as pd  # Para mostrar los resultados en una tabla

# Generar datos
np.random.seed(42)
X = np.random.uniform(-3, 3, 100).reshape(-1, 1)
y = 0.5 * X**2 + X + 2 + np.random.normal(0, 1, X.shape)  # Función cuadrática con ruido

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Visualizar los datos
plt.scatter(X, y, label='Datos', color='blue')
plt.title("Datos Sintéticos")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Grados de polinomio
degrees = [1, 3, 15]  # Grado 1 (underfitting), Grado 3 (ajuste óptimo), Grado 15 (overfitting)

# Lista para almacenar errores
results = []

plt.figure(figsize=(15, 5))
for i, degree in enumerate(degrees, 1):
    # Transformar los datos con características polinómicas
    poly = PolynomialFeatures(degree=degree)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)

    # Ajustar el modelo
    model = LinearRegression()
    model.fit(X_train_poly, y_train)

    # Predicciones
    y_train_pred = model.predict(X_train_poly)
    y_test_pred = model.predict(X_test_poly)

    # Calcular errores
    train_error = mean_squared_error(y_train, y_train_pred)
    test_error = mean_squared_error(y_test, y_test_pred)

    # Guardar resultados
    results.append({
        "Grado": degree,
        "Train Error": train_error,
        "Test Error": test_error
    })

    # Visualizar el modelo
    plt.subplot(1, 3, i)
    plt.scatter(X, y, label='Datos', color='blue', alpha=0.6)
    X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
    y_range = model.predict(poly.transform(X_range))
    plt.plot(X_range, y_range, label=f"Modelo (grado {degree})", color='red')
    plt.title(f"Grado {degree}\nTrain Error: {train_error:.2f}, Test Error: {test_error:.2f}")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.legend()

plt.tight_layout()
plt.show()

# Mostrar los resultados en forma de tabla
results_df = pd.DataFrame(results)
print("Resultados de Errores:")
print(results_df)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error
import pandas as pd  # Para mostrar los resultados en una tabla

# Generar datos
np.random.seed(42)
X = np.random.uniform(-3, 3, 100).reshape(-1, 1)
y = 0.5 * X**2 + X + 2 + np.random.normal(0, 1, X.shape)  # Función cuadrática con ruido

# Dividir los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Visualizar los datos
plt.scatter(X, y, label='Datos', color='blue')
plt.title("Datos Sintéticos")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()

# Grados de polinomio
degrees = [1, 3, 15]  # Grado 1 (underfitting), Grado 3 (ajuste óptimo), Grado 15 (overfitting)

# Lista para almacenar errores
results = []

plt.figure(figsize=(15, 5))
for i, degree in enumerate(degrees, 1):
    print(f"\n--- Cálculos para Grado {degree} ---")

    # Transformar los datos con características polinómicas
    poly = PolynomialFeatures(degree=degree)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)

    # Ajustar el modelo
    model = LinearRegression()
    model.fit(X_train_poly, y_train)

    # Predicciones
    y_train_pred = model.predict(X_train_poly)
    y_test_pred = model.predict(X_test_poly)

    # Calcular errores paso a paso
    # Error de Entrenamiento
    train_errors = (y_train - y_train_pred) ** 2  # Diferencias al cuadrado
    train_error = np.mean(train_errors)  # Promedio
    print(f"Diferencias cuadradas de entrenamiento:\n{train_errors}")
    print(f"Train Error (Promedio): {train_error:.4f}")

    # Error de Testeo
    test_errors = (y_test - y_test_pred) ** 2  # Diferencias al cuadrado
    test_error = np.mean(test_errors)  # Promedio
    print(f"Diferencias cuadradas de prueba:\n{test_errors}")
    print(f"Test Error (Promedio): {test_error:.4f}")

    # Guardar resultados
    results.append({
        "Grado": degree,
        "Train Error": train_error,
        "Test Error": test_error
    })

    # Visualizar el modelo
    plt.subplot(1, 3, i)
    plt.scatter(X, y, label='Datos', color='blue', alpha=0.6)
    X_range = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
    y_range = model.predict(poly.transform(X_range))
    plt.plot(X_range, y_range, label=f"Modelo (grado {degree})", color='red')
    plt.title(f"Grado {degree}\nTrain Error: {train_error:.2f}, Test Error: {test_error:.2f}")
    plt.xlabel("X")
    plt.ylabel("y")
    plt.legend()

plt.tight_layout()
plt.show()

# Mostrar los resultados en forma de tabla
results_df = pd.DataFrame(results)
print("\nResultados de Errores:")
print(results_df)
```

1. Configuración Inicial
Generamos datos sintéticos que siguen una relación no lineal, con ruido (error irreducible).

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split, learning_curve, validation_curve

# Generar datos sintéticos
np.random.seed(42)
X = np.linspace(-3, 3, 100).reshape(-1, 1)
y = 0.5 * X**2 + X + 2 + np.random.normal(0, 1, size=X.shape)  # Relación no lineal con ruido

# Dividir datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Visualizar los datos
plt.scatter(X, y, label="Datos", color="blue")
plt.title("Datos Sintéticos con Ruido")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

2. Curva de Validación (Sesgo y Varianza)
Ajustamos modelos polinómicos con diferentes grados (1, 3, 5, ..., 20) para observar cómo cambia el error de entrenamiento y validación según la complejidad del modelo.

```python
from sklearn.model_selection import cross_val_score

# Grados del polinomio a probar
degrees = np.arange(1, 21)
train_errors = []
test_errors = []

for degree in degrees:
    # Crear pipeline con transformación polinómica y regresión lineal
    model = Pipeline([
        ("poly_features", PolynomialFeatures(degree)),
        ("lin_reg", LinearRegression())
    ])

    # Ajustar el modelo
    model.fit(X_train, y_train)

    # Calcular errores
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    train_errors.append(mean_squared_error(y_train, y_train_pred))
    test_errors.append(mean_squared_error(y_test, y_test_pred))

# Graficar los errores
plt.figure(figsize=(10, 6))
plt.plot(degrees, train_errors, label="Error de Entrenamiento", marker='o')
plt.plot(degrees, test_errors, label="Error de Validación", marker='o')
plt.axvline(3, color='gray', linestyle='--', label="Grado Ideal")
plt.title("Curva de Validación (Complejidad del Modelo)")
plt.xlabel("Grado del Polinomio")
plt.ylabel("Error")
plt.legend()
plt.show()
```

```python
import xgboost as xgb
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

# Generar datos sintéticos
X, y = make_regression(n_samples=1000, n_features=20, noise=10, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Modelo con alto sesgo (pocos estimadores, baja complejidad)
model_high_bias = xgb.XGBRegressor(
    n_estimators=10, max_depth=2, learning_rate=0.3, random_state=42
)
model_high_bias.fit(X_train, y_train)
y_pred_bias = model_high_bias.predict(X_test)

# Modelo con alta varianza (muchos estimadores, alta complejidad)
model_high_variance = xgb.XGBRegressor(
    n_estimators=500, max_depth=10, learning_rate=0.1, random_state=42
)
model_high_variance.fit(X_train, y_train)
y_pred_variance = model_high_variance.predict(X_test)

# Modelo balanceado
model_balanced = xgb.XGBRegressor(
    n_estimators=100, max_depth=4, learning_rate=0.1, subsample=0.8, random_state=42
)
model_balanced.fit(X_train, y_train)
y_pred_balanced = model_balanced.predict(X_test)

# Calcular errores
print("High Bias MSE:", mean_squared_error(y_test, y_pred_bias))
print("High Variance MSE:", mean_squared_error(y_test, y_pred_variance))
print("Balanced MSE:", mean_squared_error(y_test, y_pred_balanced))
```
