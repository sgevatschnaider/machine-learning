```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report

# Generar datos sintéticos
X, y = make_classification(n_samples=500, n_features=2, n_classes=2, n_redundant=0, random_state=42)

# Visualizar los datos
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor='k')
plt.title("Datos Sintéticos")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Crear y entrenar un modelo de Boosting (AdaBoost)
# Base estimator: Árbol de decisión de profundidad 1 (stump)
base_model = DecisionTreeClassifier(max_depth=1, random_state=42)
adaboost = AdaBoostClassifier(base_model, n_estimators=10, random_state=42)
adaboost.fit(X, y)

# Predicciones del modelo
y_pred = adaboost.predict(X)

# Evaluar el desempeño del modelo
accuracy = accuracy_score(y, y_pred)
print(f"Precisión de AdaBoost: {accuracy:.2f}")
print("\nReporte de Clasificación:")
print(classification_report(y, y_pred))

# Visualizar las regiones de decisión del modelo AdaBoost
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))
Z = adaboost.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor='k')
plt.title("Regiones de Decisión de AdaBoost")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score

# Generar datos sintéticos
X, y = make_classification(n_samples=500, n_features=2, n_classes=2, n_redundant=0, random_state=42)

# Visualizar los datos
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor='k')
plt.title("Datos Sintéticos")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Crear el modelo AdaBoost
base_model = DecisionTreeClassifier(max_depth=1, random_state=42)  # Weak learner (stump)
adaboost = AdaBoostClassifier(base_model, n_estimators=5, random_state=42)
adaboost.fit(X, y)

# Visualizar los errores corregidos por cada modelo base
plt.figure(figsize=(16, 10))
for i, stump in enumerate(adaboost.estimators_):
    # Predicciones del modelo base (stump)
    y_pred = stump.predict(X)
    errors = (y_pred != y)  # Identificar los errores

    # Visualizar las regiones de decisión del stump
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))
    Z = stump.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.subplot(2, 3, i + 1)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor='k')
    plt.scatter(X[errors, 0], X[errors, 1], color='yellow', edgecolor='black', label="Errores", s=100, alpha=0.6)  # Muestras mal clasificadas
    plt.title(f"Modelo Base {i + 1}")
    plt.xlabel("Feature 1")
    plt.ylabel("Feature 2")
    plt.legend()

plt.tight_layout()
plt.show()

# Evaluar el modelo AdaBoost completo
y_pred_final = adaboost.predict(X)
accuracy = accuracy_score(y, y_pred_final)
print(f"\nPrecisión Final de AdaBoost: {accuracy:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
from sklearn.metrics import accuracy_score

# Generar datos sintéticos
X, y = make_classification(n_samples=500, n_features=2, n_classes=2, n_redundant=0, random_state=42)

# Visualizar los datos
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor='k')
plt.title("Datos Sintéticos")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Crear Bagging
bagging = BaggingClassifier(
    estimator=DecisionTreeClassifier(max_depth=3, random_state=42),  # Cambiar base_estimator a estimator
    n_estimators=10,
    random_state=42
)
bagging.fit(X, y)

# Crear Boosting (AdaBoost)
boosting = AdaBoostClassifier(
    estimator=DecisionTreeClassifier(max_depth=1, random_state=42),  # Cambiar base_estimator a estimator
    n_estimators=10,
    random_state=42
)
boosting.fit(X, y)

# Evaluar ambos modelos
y_pred_bagging = bagging.predict(X)
y_pred_boosting = boosting.predict(X)

accuracy_bagging = accuracy_score(y, y_pred_bagging)
accuracy_boosting = accuracy_score(y, y_pred_boosting)

print(f"Precisión de Bagging: {accuracy_bagging:.2f}")
print(f"Precisión de Boosting: {accuracy_boosting:.2f}")

# Visualizar regiones de decisión
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))

# Predicciones del Bagging
Z_bagging = bagging.predict(np.c_[xx.ravel(), yy.ravel()])
Z_bagging = Z_bagging.reshape(xx.shape)

# Predicciones del Boosting
Z_boosting = boosting.predict(np.c_[xx.ravel(), yy.ravel()])
Z_boosting = Z_boosting.reshape(xx.shape)

# Graficar regiones de decisión
plt.figure(figsize=(16, 6))

# Bagging
plt.subplot(1, 2, 1)
plt.contourf(xx, yy, Z_bagging, alpha=0.3, cmap='coolwarm')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor='k')
plt.title("Regiones de Decisión: Bagging")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# Boosting
plt.subplot(1, 2, 2)
plt.contourf(xx, yy, Z_boosting, alpha=0.3, cmap='coolwarm')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor='k')
plt.title("Regiones de Decisión: Boosting")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.tight_layout()
plt.show()
plt.show()
```
