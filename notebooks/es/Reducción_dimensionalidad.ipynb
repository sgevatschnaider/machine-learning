```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Definir la matriz A
A = np.array([[3, 2],
              [2, 3]])

print("Matriz A:")
print(A)

# Paso 2: Calcular A^T * A y A * A^T
AT_A = A.T @ A
A_AT = A @ A.T

print("\nMatriz A^T * A:")
print(AT_A)

print("\nMatriz A * A^T:")
print(A_AT)

# Paso 3: Calcular los valores y vectores propios
eigvals_AT_A, eigvecs_AT_A = np.linalg.eig(AT_A)
eigvals_A_AT, eigvecs_A_AT = np.linalg.eig(A_AT)

print("\nAutovalores de A^T * A:")
print(eigvals_AT_A)

print("\nAutovectores de A^T * A (Vectores singulares derechos V):")
print(eigvecs_AT_A)

print("\nAutovalores de A * A^T:")
print(eigvals_A_AT)

print("\nAutovectores de A * A^T (Vectores singulares izquierdos U):")
print(eigvecs_A_AT)

# Paso 4: Ordenar los valores singulares (raíz de los autovalores)
singular_values = np.sqrt(eigvals_AT_A)
singular_values_sorted_indices = np.argsort(singular_values)[::-1]
singular_values_sorted = singular_values[singular_values_sorted_indices]

print("\nValores singulares (Σ):")
print(singular_values_sorted)

# Paso 5: Construir las matrices U, Σ y V^T
U = eigvecs_A_AT
V = eigvecs_AT_A
Sigma = np.zeros_like(A, dtype=float)
np.fill_diagonal(Sigma, singular_values_sorted)

print("\nMatriz U (Vectores singulares izquierdos):")
print(U)

print("\nMatriz Σ (Diagonal con valores singulares):")
print(Sigma)

print("\nMatriz V^T (Transpuesta de vectores singulares derechos):")
print(V.T)

# Paso 6: Reconstruir A
A_reconstructed = U @ Sigma @ V.T
print("\nMatriz reconstruida A = UΣV^T:")
print(A_reconstructed)

# Paso 7: Visualización
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Original matrix as heatmap
axes[0].imshow(A, cmap="viridis", aspect='auto')
axes[0].set_title("Matriz Original A")
axes[0].axis('off')

# Reconstructed matrix as heatmap
axes[1].imshow(A_reconstructed, cmap="viridis", aspect='auto')
axes[1].set_title("Matriz Reconstruida A (UΣV^T)")
axes[1].axis('off')

plt.show()
```

```python
from IPython.display import display, HTML

# Código HTML para Google Colab
html_svd_truncado = r"""
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>SVD Truncado</title>
    <script type="text/javascript" async
        src="https://cdnjs.cloudflare.com/ajax/libs/mathjax/3.2.0/es5/tex-mml-chtml.js">
    </script>
    <style>
        body {
            font-family: 'Georgia', serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            color: #333;
        }
        .navbar {
            background-color: #0066cc;
            color: white;
            padding: 10px 15px;
            text-align: center;
            font-size: 1.2em;
        }
        h1 {
            color: #333;
            text-align: center;
            font-size: 2.5em;
            margin-bottom: 20px;
        }
        h2 {
            color: #0056b3;
            margin-top: 15px;
        }
        ul {
            font-size: 1.1em;
            margin-bottom: 20px;
            padding-left: 20px;
        }
        p, pre {
            margin-bottom: 15px;
            font-size: 1.1em;
        }
        .view-link {
            background-color: #ff5722;
            color: white;
            padding: 12px 20px;
            font-size: 1.1em;
            display: inline-block;
            border-radius: 8px;
            text-decoration: none;
            margin: 20px 0;
            transition: background-color 0.3s ease;
            box-shadow: 0px 4px 6px rgba(0, 0, 0, 0.1);
        }
        .view-link:hover {
            background-color: #e64a19;
            box-shadow: 0px 6px 10px rgba(0, 0, 0, 0.15);
        }
    </style>
</head>
<body>
    <div class="navbar">
        Descomposición en Valores Singulares Truncada (SVD Truncado)
    </div>

    <h1>Explicación Teórica Completa</h1>

    <h2>1. ¿Qué es el SVD Truncado?</h2>
    <p>El <strong>SVD truncado</strong> es una técnica que aproxima una matriz reduciendo su dimensionalidad. Se seleccionan los <em>r</em> valores singulares más grandes para construir una representación aproximada de la matriz original:</p>
    <p style="text-align: center;">
    \[
    M \approx M' = U_r \Sigma_r V_r^T
    \]
    </p>
    <ul>
        <li><strong>Reducción:</strong> Menor cantidad de datos a almacenar.</li>
        <li><strong>Retención:</strong> Los valores singulares más grandes conservan la mayor parte de la información.</li>
    </ul>

    <h2>2. Propiedades y Aplicaciones</h2>
    <ul>
        <li><strong>Compresión:</strong> Reduce el tamaño de imágenes, videos y otros datos voluminosos.</li>
        <li><strong>Eliminación de ruido:</strong> Filtra valores singulares pequeños que capturan ruido.</li>
        <li><strong>Recomendadores:</strong> Identifica patrones principales en sistemas como recomendaciones de películas.</li>
    </ul>

    <h2>3. Retención de Información</h2>
    <p>La cantidad de información retenida depende de los valores singulares seleccionados:</p>
    <p style="text-align: center;">
    \[
    \text{Porcentaje de información retenida} = \frac{\sum_{i=1}^r \sigma_i^2}{\sum_{i=1}^k \sigma_i^2}
    \]
    </p>
    <p>Un análisis de varianza acumulada ayuda a determinar el valor óptimo de <em>r</em>.</p>

    <h2>4. Aplicación Intuitiva</h2>
    <p>Imagina una matriz que representa calificaciones de usuarios sobre películas. El SVD truncado identifica relaciones principales como:</p>
    <ul>
        <li>Preferencias por géneros (acción, comedia, etc.).</li>
        <li>Relaciones entre usuarios con intereses similares.</li>
    </ul>

    <h2>5. Desafíos</h2>
    <ul>
        <li><strong>Elección de <em>r</em>:</strong> Encontrar un equilibrio entre retención de información y reducción de datos.</li>
        <li><strong>Costo computacional:</strong> Para matrices grandes, el cálculo completo del SVD puede ser intensivo.</li>
    </ul>

    <p>
        <a href="https://drive.google.com/file/d/1QAXw1It0DySIgKI5YV5JKA67HaV8a-2d/view?usp=drive_link" target="_blank" class="view-link">
            Más información sobre SVD Truncado
        </a>
    </p>
</body>
</html>
"""

# Renderizar el HTML en Google Colab
display(HTML(html_svd_truncado))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Paso 1: Crear una matriz simulada M (usuarios x películas)
np.random.seed(42)  # Para reproducibilidad
M = np.array([
    [5, 3, 0, 1, 4],  # Usuario 1
    [4, 0, 0, 1, 3],  # Usuario 2
    [1, 1, 0, 5, 0],  # Usuario 3
    [1, 0, 0, 4, 0],  # Usuario 4
    [0, 1, 5, 4, 0],  # Usuario 5
    [0, 0, 5, 4, 1],  # Usuario 6
    [0, 0, 0, 0, 5],  # Usuario 7
    [5, 4, 1, 1, 0],  # Usuario 8
    [5, 0, 0, 0, 4],  # Usuario 9
    [0, 1, 5, 4, 4]   # Usuario 10
])

print("Matriz original M (Usuarios x Películas):")
print(M)

# Paso 2: Descomposición SVD
U, Sigma, VT = np.linalg.svd(M, full_matrices=False)

print("\nMatriz U (Vectores singulares izquierdos):")
print(U)
print("\nValores singulares (Σ):")
print(Sigma)
print("\nMatriz V^T (Transpuesta de vectores singulares derechos):")
print(VT)

# Paso 3: Reconstrucción truncada con r valores singulares
def svd_truncada(U, Sigma, VT, r):
    # Solo tomamos los r primeros valores singulares
    U_r = U[:, :r]
    Sigma_r = np.diag(Sigma[:r])
    VT_r = VT[:r, :]
    # Reconstruimos la matriz reducida
    M_truncada = U_r @ Sigma_r @ VT_r
    return M_truncada

# Elegimos r = 2
r = 2
M_truncada = svd_truncada(U, Sigma, VT, r)

print(f"\nMatriz reconstruida M' con r = {r}:")
print(M_truncada)

# Paso 4: Cálculo de la información retenida
informacion_total = np.sum(Sigma**2)
informacion_retenida = np.sum(Sigma[:r]**2)
proporcion_retenida = informacion_retenida / informacion_total

print(f"\nInformación retenida con r = {r}: {proporcion_retenida:.2%}")

# Paso 5: Visualización
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

# Matriz original
axes[0].imshow(M, cmap="viridis", aspect="auto")
axes[0].set_title("Matriz Original M")
axes[0].set_xlabel("Películas")
axes[0].set_ylabel("Usuarios")
axes[0].set_xticks(range(M.shape[1]))
axes[0].set_yticks(range(M.shape[0]))

# Matriz reconstruida
axes[1].imshow(M_truncada, cmap="viridis", aspect="auto")
axes[1].set_title(f"Matriz Reconstruida M' (r = {r})")
axes[1].set_xlabel("Películas")
axes[1].set_ylabel("Usuarios")
axes[1].set_xticks(range(M.shape[1]))
axes[1].set_yticks(range(M.shape[0]))

plt.tight_layout()
plt.show()
```
