```python
from IPython.display import display, HTML

# Código HTML
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>RNNs, LSTM y Transformers</title>
    <style>
        body {
            font-family: 'Georgia', serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            color: #333;
        }
        h1, h2 {
            color: #0066cc;
        }
        h1 {
            text-align: center;
        }
        p {
            text-align: justify;
            margin-bottom: 15px;
        }
        ul {
            margin: 10px 0;
            padding-left: 20px;
        }
        li {
            margin-bottom: 10px;
        }
    </style>
</head>
<body>
    <h1>RNNs, LSTM y Transformers: Evolución del Modelado de Secuencias</h1>

    <h2>1. Redes Neuronales Recurrentes (RNNs): El punto de partida</h2>
    <p>
        Las <strong>RNNs</strong> fueron el enfoque inicial para modelar secuencias, como texto, audio o series temporales.
        ¿Cómo funcionan? Imagina que estás leyendo un libro. Cada palabra que lees está influenciada por las anteriores.
        Las RNNs intentan simular esto al procesar las palabras una por una, recordando información previa.
    </p>
    <h3>Problema: El gradiente que se desvanece</h3>
    <ul>
        <li><strong>En teoría:</strong> Las RNNs deberían poder propagar la información de una palabra (token) al resto de la secuencia, incluso si es muy larga.</li>
        <li><strong>En la práctica:</strong> Al entrenar la red, los gradientes (que miden qué tan importantes son los cambios en los parámetros) se vuelven muy pequeños en secuencias largas. Esto es el famoso <strong>problema del gradiente que se desvanece</strong>.</li>
        <li><strong>Consecuencia:</strong> Las RNNs no logran capturar información precisa de tokens lejanos en secuencias largas.</li>
    </ul>
    <p>
        Un ejemplo temprano de una RNN fue la <strong>Red de Elman (1990)</strong>, que ilustraba esta dificultad.
    </p>

    <h2>2. LSTM (1995): El gran avance</h2>
    <p>
        Para resolver el problema del gradiente que se desvanece, surgieron las <strong>LSTM (Long Short-Term Memory)</strong>. Estas redes introdujeron una arquitectura más compleja que permite manejar dependencias a largo plazo en secuencias.
    </p>
    <h3>¿Qué hace que las LSTM sean especiales?</h3>
    <ul>
        <li><strong>Celdas de memoria (memory cells):</strong> Guardan información importante por más tiempo.</li>
        <li><strong>Puertas (gates):</strong> Deciden qué información agregar, olvidar o conservar en cada paso de la secuencia.</li>
        <li><strong>Multiplicative units (unidades multiplicativas):</strong> Estas unidades permiten ajustar el flujo de información usando operaciones de multiplicación. Esto ayuda a controlar mejor cómo se actualiza la información en las celdas de memoria.</li>
    </ul>

    <h2>3. Limitaciones de las LSTM</h2>
    <p>
        Aunque las LSTM solucionaron el problema del gradiente que se desvanece, tenían un <strong>problema fundamental: el procesamiento secuencial</strong>.
        Al igual que las RNNs, procesaban un token a la vez, desde el primero hasta el último.
    </p>
    <h3>Problema del procesamiento secuencial</h3>
    <ul>
        <li><strong>Ineficiencia:</strong> Procesar secuencias largas es lento porque cada paso depende del anterior.</li>
        <li><strong>No paralelizable:</strong> No pueden procesar todos los tokens de la secuencia al mismo tiempo.</li>
    </ul>

    <h2>4. Transformadores (Transformers): Una revolución (2017)</h2>
    <p>
        Los <strong>Transformers</strong> introdujeron un cambio radical al eliminar la necesidad de procesar secuencias de manera secuencial.
    </p>
    <h3>¿Cómo lo lograron?</h3>
    <ul>
        <li><strong>Procesamiento paralelo:</strong> En lugar de procesar un token a la vez, los Transformers procesan toda la secuencia simultáneamente.</li>
        <li><strong>Atención:</strong> Introdujeron el mecanismo de <strong>atención</strong>, que permite a cada token enfocarse directamente en otros tokens relevantes de la secuencia, independientemente de su posición. Esto supera la limitación de las LSTM para capturar relaciones entre tokens lejanos.</li>
        <li><strong>Cómputo cuadrático:</strong> Los Transformadores calculan relaciones entre todos los pares de tokens en la secuencia, lo que requiere un tiempo de cómputo proporcional al cuadrado del tamaño de la ventana de contexto. Aunque más costoso, esta operación es altamente paralelizable en hardware moderno (como GPUs).</li>
    </ul>

    <h2>5. Inspiraciones previas: Fast Weight Controllers y Dynamic Links</h2>
    <ul>
        <li><strong>Fast Weight Controllers (1992):</strong> Introdujeron la idea de una red neuronal lenta que genera parámetros para una red más rápida. La red rápida puede adaptarse dinámicamente a las entradas actuales, lo que es conceptualmente similar a cómo funcionan los Transformadores.</li>
        <li><strong>Dynamic Links (1981):</strong> Una forma temprana de representar relaciones dinámicas entre nodos de una red neuronal.</li>
    </ul>

    <h2>6. Resumen conceptual</h2>
    <ul>
        <li><strong>Antes de LSTM:</strong> RNNs sufrían por el gradiente que se desvanece, incapaces de modelar secuencias largas eficientemente.</li>
        <li><strong>Con LSTM:</strong> Se resolvió el problema del gradiente, pero persistió la limitación del procesamiento secuencial.</li>
        <li><strong>Con Transformers:</strong> Se introdujo el procesamiento paralelo, eliminando la dependencia de tokens previos y permitiendo modelar relaciones a cualquier distancia dentro de la secuencia.</li>
    </ul>
</body>
</html>
"""

# Renderizar el contenido HTML en Google Colab
display(HTML(html_content))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Configuración de la RNN
def rnn_step(x_t, h_prev, Wxh, Whh, bh):
    """
    Un paso de la RNN:
    - x_t: Entrada actual (vector).
    - h_prev: Estado oculto del paso anterior.
    - Wxh: Pesos para la entrada x_t.
    - Whh: Pesos para el estado oculto anterior.
    - bh: Sesgo para el estado oculto.
    """
    # Producto matricial para la entrada actual
    input_contribution = np.dot(Wxh, x_t)
    # Producto matricial para el estado oculto previo
    hidden_contribution = np.dot(Whh, h_prev)
    # Suma de contribuciones y sesgo
    pre_activation = input_contribution + hidden_contribution + bh
    # Aplicar tanh
    h_t = np.tanh(pre_activation)
    return h_t, input_contribution, hidden_contribution, pre_activation

# Inicialización de parámetros
input_size = 3   # Tamaño del vector de entrada x_t
hidden_size = 2  # Tamaño del estado oculto h_t
time_steps = 5   # Longitud de la secuencia

# Pesos y sesgos aleatorios
np.random.seed(42)
Wxh = np.random.randn(hidden_size, input_size) * 0.5  # Pesos de entrada
Whh = np.random.randn(hidden_size, hidden_size) * 0.5  # Pesos del estado oculto
bh = np.zeros((hidden_size, 1))  # Sesgos

# Secuencia de entradas (cada x_t es un vector aleatorio)
x_sequence = [np.random.randn(input_size, 1) for _ in range(time_steps)]

# Inicialización del estado oculto inicial
h_0 = np.zeros((hidden_size, 1))

# Propagación a través de la secuencia
hidden_states = [h_0]
input_contributions = []
hidden_contributions = []
pre_activations = []

# Detallar cálculos paso a paso
for t in range(time_steps):
    h_t, in_contrib, hid_contrib, pre_act = rnn_step(x_sequence[t], hidden_states[-1], Wxh, Whh, bh)
    hidden_states.append(h_t)
    input_contributions.append(in_contrib)
    hidden_contributions.append(hid_contrib)
    pre_activations.append(pre_act)
    # Mostrar detalles del paso actual
    print(f"Paso {t + 1}:")
    print(f"Entrada (x_{t}):\n{x_sequence[t].flatten()}")
    print(f"Estado oculto previo (h_{t}):\n{hidden_states[-2].flatten()}")
    print(f"Contribución de entrada:\n{in_contrib.flatten()}")
    print(f"Contribución de estado oculto previo:\n{hid_contrib.flatten()}")
    print(f"Pre-activación (suma antes de tanh):\n{pre_act.flatten()}")
    print(f"Estado oculto actual (h_{t + 1}):\n{h_t.flatten()}")
    print("-" * 50)

# Graficar los estados ocultos para entender el flujo
fig, ax = plt.subplots(hidden_size, 1, figsize=(8, 4), sharex=True)
time = np.arange(time_steps + 1)

for i in range(hidden_size):
    ax[i].plot(time, [h[i, 0] for h in hidden_states], marker='o', label=f"h[{i}]")
    ax[i].set_ylabel(f"h[{i}]")
    ax[i].legend()
ax[-1].set_xlabel("Paso de tiempo")
plt.suptitle("Estados ocultos a lo largo del tiempo")
plt.tight_layout()
plt.show()

# Graficar las entradas y contribuciones
plt.figure(figsize=(10, 6))
for t in range(time_steps):
    plt.quiver(t, 0, x_sequence[t][0, 0], x_sequence[t][1, 0], angles='xy', scale_units='xy', scale=1, color="blue", label=f"x_{t}" if t == 0 else "")
    plt.quiver(t, 0.5, input_contributions[t][0, 0], input_contributions[t][1, 0], angles='xy', scale_units='xy', scale=1, color="green", label=f"Input contribución" if t == 0 else "")
    plt.quiver(t, -0.5, hidden_contributions[t][0, 0], hidden_contributions[t][1, 0], angles='xy', scale_units='xy', scale=1, color="orange", label=f"Hidden contribución" if t == 0 else "")
plt.title("Representación de las entradas y contribuciones a lo largo del tiempo")
plt.xlabel("Paso de tiempo")
plt.ylabel("Contribuciones")
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Diccionario de palabras y sus representaciones (embeddings)
word_embeddings = {
    "yo": np.array([[0.2], [0.1], [0.4]]),
    "aprendo": np.array([[0.3], [0.6], [0.1]]),
    "redes": np.array([[0.5], [0.3], [0.7]]),
    "recurrentes": np.array([[0.6], [0.9], [0.4]])
}

# Secuencia de palabras
sentence = ["yo", "aprendo", "redes", "recurrentes"]

# Configuración de la RNN
def rnn_step(x_t, h_prev, Wxh, Whh, bh):
    """
    Un paso de la RNN:
    - x_t: Entrada actual (embedding de una palabra).
    - h_prev: Estado oculto del paso anterior.
    - Wxh: Pesos para la entrada x_t.
    - Whh: Pesos para el estado oculto anterior.
    - bh: Sesgo para el estado oculto.
    """
    # Producto matricial para la entrada actual
    input_contribution = np.dot(Wxh, x_t)
    # Producto matricial para el estado oculto previo
    hidden_contribution = np.dot(Whh, h_prev)
    # Suma de contribuciones y sesgo
    pre_activation = input_contribution + hidden_contribution + bh
    # Aplicar tanh
    h_t = np.tanh(pre_activation)
    return h_t, input_contribution, hidden_contribution, pre_activation

# Inicialización de parámetros
input_size = 3   # Tamaño del vector de entrada x_t (embedding de palabra)
hidden_size = 2  # Tamaño del estado oculto h_t
time_steps = len(sentence)  # Longitud de la secuencia (número de palabras)

# Pesos y sesgos aleatorios
np.random.seed(42)
Wxh = np.random.randn(hidden_size, input_size) * 0.5  # Pesos de entrada
Whh = np.random.randn(hidden_size, hidden_size) * 0.5  # Pesos del estado oculto
bh = np.zeros((hidden_size, 1))  # Sesgos

# Inicialización del estado oculto inicial
h_0 = np.zeros((hidden_size, 1))

# Propagación a través de la secuencia
hidden_states = [h_0]
input_contributions = []
hidden_contributions = []
pre_activations = []

# Detallar cálculos paso a paso
for t, word in enumerate(sentence):
    x_t = word_embeddings[word]  # Embedding de la palabra actual
    h_t, in_contrib, hid_contrib, pre_act = rnn_step(x_t, hidden_states[-1], Wxh, Whh, bh)
    hidden_states.append(h_t)
    input_contributions.append(in_contrib)
    hidden_contributions.append(hid_contrib)
    pre_activations.append(pre_act)
    # Mostrar detalles del paso actual
    print(f"Paso {t + 1}: Procesando la palabra '{word}'")
    print(f"Embedding (x_{t}):\n{x_t.flatten()}")
    print(f"Estado oculto previo (h_{t}):\n{hidden_states[-2].flatten()}")
    print(f"Contribución de entrada:\n{in_contrib.flatten()}")
    print(f"Contribución de estado oculto previo:\n{hid_contrib.flatten()}")
    print(f"Pre-activación (suma antes de tanh):\n{pre_act.flatten()}")
    print(f"Estado oculto actual (h_{t + 1}):\n{h_t.flatten()}")
    print("-" * 50)

# Graficar los estados ocultos para entender el flujo
fig, ax = plt.subplots(hidden_size, 1, figsize=(8, 4), sharex=True)
time = np.arange(time_steps + 1)

for i in range(hidden_size):
    ax[i].plot(time, [h[i, 0] for h in hidden_states], marker='o', label=f"h[{i}]")
    ax[i].set_ylabel(f"h[{i}]")
    ax[i].legend()
ax[-1].set_xlabel("Paso de tiempo")
plt.suptitle("Estados ocultos a lo largo del tiempo")
plt.tight_layout()
plt.show()

# Graficar las contribuciones de entrada y estado oculto previo
plt.figure(figsize=(10, 6))
for t, word in enumerate(sentence):
    x_t = word_embeddings[word]
    plt.quiver(t, 0, x_t[0, 0], x_t[1, 0], angles='xy', scale_units='xy', scale=1, color="blue", label=f"x_{word}" if t == 0 else "")
    plt.quiver(t, 0.5, input_contributions[t][0, 0], input_contributions[t][1, 0], angles='xy', scale_units='xy', scale=1, color="green", label=f"Input contribución" if t == 0 else "")
    plt.quiver(t, -0.5, hidden_contributions[t][0, 0], hidden_contributions[t][1, 0], angles='xy', scale_units='xy', scale=1, color="orange", label=f"Hidden contribución" if t == 0 else "")
plt.title("Representación de las entradas y contribuciones a lo largo del tiempo")
plt.xlabel("Paso de tiempo")
plt.ylabel("Contribuciones")
plt.axhline(0, color='gray', linestyle='--', linewidth=0.5)
plt.legend()
plt.grid()
plt.show()
```

```python
from IPython.display import display, HTML

# Código HTML
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Entendiendo las LSTMs</title>
    <style>
        body {
            font-family: 'Georgia', serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            color: #333;
        }
        h1, h2, h3 {
            color: #0066cc;
        }
        h1 {
            text-align: center;
        }
        p {
            text-align: justify;
            margin-bottom: 15px;
        }
        ul {
            margin: 10px 0;
            padding-left: 20px;
        }
        li {
            margin-bottom: 10px;
        }
        .button {
            text-align: center;
            margin: 20px;
        }
        .button a {
            background-color: #007bff;
            color: white;
            padding: 10px 20px;
            text-decoration: none;
            border-radius: 5px;
            font-weight: bold;
        }
        .button a:hover {
            background-color: #0056b3;
        }
    </style>
</head>
<body>
    <h1>Entendiendo las LSTMs</h1>

    <h2>1. Introducción</h2>
    <p>
        Las redes de memoria a corto y largo plazo (<strong>LSTMs</strong>) son una mejora de las redes neuronales recurrentes (<strong>RNNs</strong>), diseñadas para superar sus limitaciones al manejar dependencias a largo plazo.
    </p>
    <p>
        Introducidas en 1997 por Hochreiter y Schmidhuber, las LSTMs son altamente efectivas en tareas como:
    </p>
    <ul>
        <li>Traducción automática.</li>
        <li>Análisis de series temporales.</li>
        <li>Generación de texto.</li>
        <li>Modelado del lenguaje.</li>
    </ul>

    <h2>2. Estructura de las LSTMs</h2>
    <p>
        A diferencia de las RNNs estándar, donde el módulo repetitivo tiene una estructura simple, las LSTMs tienen cuatro capas interactivas dentro de cada módulo repetitivo:
    </p>
    <ul>
        <li><strong>Estado celular (Ct):</strong> Es como una "cinta transportadora" que transmite información casi sin cambios.</li>
        <li><strong>Estado oculto (ht):</strong> Guarda la memoria a corto plazo y genera las salidas.</li>
        <li><strong>Puertas:</strong> Controlan qué información recordar, olvidar y actualizar.</li>
    </ul>

    <h3>Puertas de las LSTMs</h3>
    <ul>
        <li><strong>Puerta de olvido (ft):</strong> Decide qué información del estado celular eliminar.</li>
        <li><strong>Puerta de entrada (it):</strong> Decide qué información nueva agregar al estado celular.</li>
        <li><strong>Puerta de salida (ot):</strong> Decide qué partes del estado celular se utilizarán para la salida.</li>
    </ul>

    <h2>3. Funcionamiento Paso a Paso</h2>
    <p><strong>Paso 1: Puerta de olvido</strong></p>
    <p>
        Determina qué partes del estado celular anterior (Ct-1) deben olvidarse. Por ejemplo, en un modelo de lenguaje, si el género del sujeto cambia, esta puerta decide eliminar el género anterior.
    </p>

    <p><strong>Paso 2: Puerta de entrada</strong></p>
    <p>
        Decide qué información nueva se almacenará en el estado celular. Genera un candidato de actualización (~Ct) que se agrega al estado celular.
    </p>

    <p><strong>Paso 3: Actualización del estado celular</strong></p>
    <p>
        Combina la información de las puertas de olvido e ingreso para actualizar el estado celular:
        Ct = ft * Ct-1 + it * ~Ct
    </p>

    <p><strong>Paso 4: Puerta de salida</strong></p>
    <p>
        Decide qué partes del estado celular deben influir en la salida. Filtra el estado celular a través de una capa tanh y lo multiplica por la salida de la puerta de salida:
        ht = ot * tanh(Ct)
    </p>

    <h2>4. Variantes de las LSTMs</h2>
    <p>Existen muchas variantes de LSTMs diseñadas para tareas específicas, como:</p>
    <ul>
        <li><strong>Conexiones de peephole:</strong> Las puertas tienen acceso directo al estado celular.</li>
        <li><strong>Unificación de puertas:</strong> Combina las puertas de entrada y olvido, como en las GRUs.</li>
    </ul>

    <h2>5. Aplicaciones y Avances Modernos</h2>
    <p>
        Aunque las LSTMs fueron un avance revolucionario, han sido superadas en muchas tareas por modelos como los Transformers, que usan mecanismos de <strong>atención</strong>. Este enfoque permite al modelo enfocarse en partes específicas de la entrada en lugar de procesar toda la secuencia.
    </p>

    <h2>6. Conclusión</h2>
    <p>
        Las LSTMs marcaron un gran paso en lo que podemos lograr con RNNs. Aunque ahora existen técnicas más avanzadas, como la atención y los Transformers, las LSTMs siguen siendo relevantes para problemas que requieren manejar relaciones temporales complejas.
    </p>

    <div class="button">
        <a href="https://colah.github.io/posts/2015-08-Understanding-LSTMs/" target="_blank">Understanding LSTM Networks</a>
    </div>
</body>
</html>
"""

# Renderizar el contenido HTML en Google Colab
display(HTML(html_content))

```
