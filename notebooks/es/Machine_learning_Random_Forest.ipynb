```python
from IPython.display import display, HTML

# Código HTML que incluye el texto completo
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Ensambles en Machine Learning</title>
    <style>
        body {
            font-family: 'Georgia', serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
        }
        h1, h2, h3 {
            color: #333;
            font-family: 'Arial', sans-serif;
        }
        h1 {
            font-size: 2.5em;
            margin-bottom: 10px;
        }
        h2 {
            font-size: 2em;
            margin-bottom: 10px;
        }
        h3 {
            font-size: 1.5em;
            margin-bottom: 5px;
        }
        p {
            font-size: 1.2em;
            color: #555;
            margin-bottom: 15px;
        }
        ul {
            font-size: 1.2em;
            color: #555;
            margin-bottom: 15px;
            padding-left: 20px;
        }
        li {
            margin-bottom: 10px;
        }
    </style>
</head>
<body>
    <h1>Ensambles en Machine Learning</h1>

    <h2>Introducción a Ensambles</h2>
    <p>
        Los ensambles en Machine Learning son métodos que combinan múltiples modelos (también llamados estimadores)
        con el objetivo de mejorar el rendimiento general del sistema de predicción. En lugar de confiar en un único
        modelo, los métodos de ensamble buscan aprovechar las fortalezas individuales de varios modelos para generar
        un predictor más robusto y preciso.
    </p>
    <p>
        La idea central detrás de los ensambles es que, al unir las predicciones de múltiples modelos, los errores
        individuales tienden a cancelarse, lo que reduce la varianza, el sesgo o ambos, dependiendo de la técnica
        utilizada. Esto se inspira en el concepto de "sabiduría de las masas," donde las decisiones combinadas de
        un grupo pueden ser más precisas que las de un único individuo.
    </p>
    <p>
        Los métodos de ensamble son ampliamente utilizados debido a su eficacia para abordar problemas complejos,
        como clasificación y regresión, y su capacidad para mitigar problemas comunes como el <strong>overfitting</strong>
        y el <strong>underfitting</strong>. Algunas de las técnicas más populares incluyen <strong>Bagging</strong>,
        <strong>Boosting</strong>, <strong>Stacking</strong>, y <strong>Voting</strong>, cada una con enfoques únicos
        para combinar modelos.
    </p>
    <ul>
        <li><strong>Mayor precisión:</strong> Suelen superar el rendimiento de los modelos individuales al combinar sus predicciones.</li>
        <li><strong>Robustez:</strong> Resisten mejor al ruido en los datos.</li>
        <li><strong>Flexibilidad:</strong> Pueden ser aplicados con diferentes tipos de algoritmos base (árboles de decisión,
        regresores lineales, redes neuronales, etc.).</li>
        <li><strong>Generalización:</strong> Mejoran la capacidad del modelo para generalizar a nuevos datos.</li>
    </ul>

    <h2>Necesidad de Diversidad en los Modelos</h2>
    <p>
        Uno de los principios fundamentales para el éxito de los métodos de ensamble es la <strong>diversidad</strong>
        entre los modelos que los componen. Si los modelos base son muy similares entre sí (es decir, si están altamente
        correlacionados), sus errores tenderán a coincidir, lo que disminuye la efectividad del ensamble. En cambio, si
        los modelos son diferentes y tienen enfoques complementarios, los errores individuales se reducirán al combinar
        las predicciones.
    </p>
    <p>La diversidad puede introducirse de varias formas, incluyendo:</p>
    <ul>
        <li><strong>Diferencias en los datos de entrenamiento:</strong>
            <ul>
                <li>Técnicas como el <strong>Bagging</strong> (Bootstrap Aggregating) utilizan muestreo con reemplazo para generar diferentes subconjuntos de datos de entrenamiento.</li>
                <li>En métodos como <strong>Random Forest</strong>, además de muestrear datos, se seleccionan aleatoriamente subconjuntos de características en cada división del árbol.</li>
            </ul>
        </li>
        <li><strong>Diferencias en los algoritmos:</strong>
            <ul>
                <li>Combinar modelos de diferente naturaleza, como árboles de decisión, regresores lineales, y modelos basados en redes neuronales, puede incrementar la diversidad.</li>
                <li>Por ejemplo, un ensamble que combine un Support Vector Machine con un Random Forest podría beneficiarse de los enfoques complementarios de estos modelos.</li>
            </ul>
        </li>
        <li><strong>Diferencias en los hiperparámetros:</strong>
            <ul>
                <li>Entrenar el mismo tipo de modelo (como árboles de decisión) con diferentes configuraciones de hiperparámetros puede producir resultados variados.</li>
                <li>Este enfoque es común en métodos como el <strong>Gradient Boosting</strong>, donde cada iteración ajusta los errores de los modelos previos, creando una secuencia de modelos diversos.</li>
            </ul>
        </li>
        <li><strong>Diferencias en las hipótesis:</strong>
            <ul>
                <li>Cambiar las suposiciones subyacentes del modelo base puede inducir diversidad. Por ejemplo, en <strong>Bagging</strong>, cada modelo puede asumir diferentes hipótesis al aprender patrones a partir de diferentes subconjuntos de datos.</li>
            </ul>
        </li>
        <li><strong>Introducción de ruido controlado:</strong>
            <ul>
                <li>En algunos casos, el ruido aleatorio puede ser introducido deliberadamente en los datos de entrenamiento para fomentar la diversidad. Esto es particularmente útil en métodos como Dropout para redes neuronales, aunque no es exclusivo de ensambles.</li>
            </ul>
        </li>
    </ul>
    <p>
        En resumen, la diversidad entre los modelos base es una piedra angular de los métodos de ensamble. Sin diversidad,
        el ensamble se convierte en una simple réplica de un modelo único, perdiendo su capacidad de mejorar la precisión
        y generalización. Los diseñadores de ensambles deben considerar cuidadosamente cómo introducir y mantener la diversidad
        mientras maximizan la complementariedad de los modelos.
    </p>
</body>
</html>
"""

# Renderizar el HTML en Google Colab
display(HTML(html_content))
```

```python
from IPython.display import display, HTML

# Código HTML
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Bagging y Random Forest</title>
    <style>
        body {
            font-family: 'Georgia', serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
        }
        h1, h2, h3 {
            color: #333;
            font-family: 'Arial', sans-serif;
        }
        h1 {
            font-size: 2.5em;
            margin-bottom: 10px;
        }
        h2 {
            font-size: 2em;
            margin-bottom: 10px;
        }
        h3 {
            font-size: 1.5em;
            margin-bottom: 5px;
        }
        p {
            font-size: 1.2em;
            color: #555;
            margin-bottom: 15px;
        }
        ul {
            font-size: 1.2em;
            color: #555;
            margin-bottom: 15px;
            padding-left: 20px;
        }
        li {
            margin-bottom: 10px;
        }
        strong {
            color: #000;
        }
        em {
            color: #0056b3;
        }
    </style>
</head>
<body>
    <h1>Bagging</h1>

    <h2>Definición y Metodología</h2>
    <p>
        Bagging (abreviatura de <strong>Bootstrap Aggregating</strong>) es una técnica de ensamble que mejora la precisión y estabilidad
        de los modelos de aprendizaje supervisado al reducir la varianza de las predicciones. La idea central es entrenar múltiples
        modelos independientes utilizando diferentes subconjuntos de los datos de entrenamiento y luego combinar sus predicciones.
        Esto se logra mediante el muestreo aleatorio con reemplazo (<em>bootstrap</em>) para generar subconjuntos de datos, donde algunos
        puntos de datos pueden repetirse en una muestra y otros pueden omitirse.
    </p>
    <h3>Pasos principales de Bagging:</h3>
    <ul>
        <li><strong>Muestreo con reemplazo:</strong>
            <ul>
                <li>Se generan varias muestras aleatorias (con reemplazo) a partir del conjunto de datos original.</li>
                <li>Cada muestra bootstrap tiene el mismo tamaño que el conjunto original pero puede contener repeticiones de algunos ejemplos.</li>
            </ul>
        </li>
        <li><strong>Entrenamiento de modelos independientes:</strong>
            <ul>
                <li>Cada muestra bootstrap se utiliza para entrenar un modelo base (por ejemplo, un árbol de decisión).</li>
                <li>Los modelos base pueden ser de alta varianza, como los árboles de decisión profundos, que tienden a ajustarse a los datos de entrenamiento.</li>
            </ul>
        </li>
        <li><strong>Combinación de predicciones:</strong>
            <ul>
                <li>Para tareas de clasificación: las predicciones finales se realizan utilizando votación por mayoría entre los modelos.</li>
                <li>Para tareas de regresión: se toma el promedio de las predicciones.</li>
            </ul>
        </li>
    </ul>
    <h3>Ventajas de Bagging:</h3>
    <ul>
        <li>Reduce el <strong>overfitting</strong> al promediar múltiples modelos.</li>
        <li>Mejora la precisión, especialmente cuando se utilizan modelos base que son inestables o tienen alta varianza.</li>
        <li>Es fácilmente paralelizable porque los modelos se entrenan de forma independiente.</li>
    </ul>
    <h3>Limitaciones:</h3>
    <ul>
        <li>No aborda bien el problema del sesgo alto en los modelos base.</li>
        <li>Puede ser computacionalmente costoso cuando el número de modelos es grande.</li>
    </ul>

    <hr>

    <h1>Random Forest</h1>

    <h2>Funcionamiento y Características</h2>
    <p>
        Random Forest es una extensión del Bagging que introduce aleatoriedad adicional para aumentar la diversidad entre los
        modelos base, que en este caso son árboles de decisión. En Random Forest, además de utilizar diferentes subconjuntos
        de datos para cada árbol, también se selecciona aleatoriamente un subconjunto de características para evaluar en cada
        nodo del árbol. Esto evita que ciertos atributos dominantes influyan excesivamente en las decisiones de los árboles.
    </p>
    <h3>Pasos principales de Random Forest:</h3>
    <ul>
        <li><strong>Muestreo bootstrap:</strong>
            <ul>
                <li>Similar al Bagging, se generan subconjuntos de datos mediante muestreo con reemplazo.</li>
            </ul>
        </li>
        <li><strong>Selección aleatoria de características:</strong>
            <ul>
                <li>En cada división de un árbol de decisión, se selecciona un subconjunto aleatorio de características (k)) de las (m) características totales. Este \(k\) es un hiperparámetro importante.</li>
            </ul>
        </li>
        <li><strong>Construcción de árboles independientes:</strong>
            <ul>
                <li>Se construyen (n) árboles independientes, cada uno utilizando su propio subconjunto de datos y evaluando sólo las características seleccionadas aleatoriamente.</li>
            </ul>
        </li>
        <li><strong>Combinación de predicciones:</strong>
            <ul>
                <li>Clasificación: se utiliza votación por mayoría entre los árboles.</li>
                <li>Regresión: se toma el promedio de las predicciones de todos los árboles.</li>
            </ul>
        </li>
    </ul>
    <h3>Ventajas de Random Forest:</h3>
    <ul>
        <li><strong>Robustez frente al overfitting:</strong> Reducido mediante el promedio de múltiples árboles.</li>
        <li><strong>Reducción de la varianza:</strong> Las decisiones individuales de los árboles tienden a ser menos ruidosas cuando se combinan.</li>
        <li><strong>Importancia de características:</strong> Proporciona métricas sobre la importancia relativa de cada característica en las predicciones.</li>
        <li><strong>Paralelización:</strong> Es fácilmente implementable en paralelo, ya que cada árbol se construye de manera independiente.</li>
    </ul>
    <h3>Desventajas:</h3>
    <ul>
        <li><strong>Complejidad computacional:</strong> Entrenar un gran número de árboles puede ser costoso, especialmente con grandes conjuntos de datos o muchas características.</li>
        <li><strong>Rendimiento con datos pequeños:</strong> No es tan efectivo con conjuntos de datos muy pequeños, ya que la diversidad en las muestras bootstrap será limitada.</li>
    </ul>

    <hr>

    <h2>Problemas con Atributos Fuertes</h2>
    <p>
        Cuando algunos atributos tienen una gran capacidad predictiva (también llamados predictores fuertes), los árboles en el Random Forest tienden a parecerse entre sí. Esto sucede porque los atributos fuertes dominan las decisiones iniciales en la construcción de los árboles, reduciendo la diversidad del ensamble.
    </p>
    <h3>Efectos de los atributos fuertes:</h3>
    <ul>
        <li><strong>Pérdida de diversidad:</strong>
            <ul>
                <li>La selección aleatoria de características en cada nodo mitiga este problema al obligar a los árboles a considerar atributos menos dominantes, pero si los predictores fuertes aparecen consistentemente, los árboles seguirán convergiendo.</li>
            </ul>
        </li>
        <li><strong>Bias en las decisiones:</strong>
            <ul>
                <li>Los atributos fuertes pueden crear divisiones tempranas que impacten negativamente en la generalización del modelo.</li>
            </ul>
        </li>
    </ul>
    <h3>Estrategias para abordar este problema:</h3>
    <ul>
        <li>Ajustar (k): Reducir el número de características seleccionadas (k) en cada división para forzar una mayor exploración de atributos secundarios.</li>
        <li><strong>Normalización de datos:</strong> Asegurarse de que todas las características estén en la misma escala para evitar que los atributos con valores más grandes dominen.</li>
        <li><strong>Métodos híbridos:</strong> Combinar Random Forest con técnicas como Boosting para compensar el sesgo introducido por atributos fuertes.</li>
    </ul>
</body>
</html>
"""

# Renderizar el HTML en Google Colab
display(HTML(html_content))

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score, classification_report

# Generar datos sintéticos
X, y = make_classification(n_samples=500, n_features=2, n_classes=2, n_redundant=0, random_state=42)

# Visualizar los datos
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor='k')
plt.title("Datos Sintéticos")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# Entrenar un único árbol de decisión
model = DecisionTreeClassifier(max_depth=3, random_state=42)
model.fit(X, y)

# Visualizar el árbol entrenado
plt.figure(figsize=(15, 8))
plot_tree(model, filled=True, feature_names=["Feature 1", "Feature 2"], class_names=["Clase 0", "Clase 1"])
plt.title("Árbol de Decisión (Profundidad Máxima: 3)")
plt.show()

# Realizar predicciones
y_pred = model.predict(X)

# Evaluar el desempeño del modelo
accuracy = accuracy_score(y, y_pred)
print(f"Precisión del Árbol de Decisión: {accuracy:.2f}")
print("\nReporte de Clasificación:")
print(classification_report(y, y_pred))

# Visualizar las regiones de decisión del árbol
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', edgecolor='k')
plt.title("Regiones de Decisión de un Árbol de Decisión")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

1. Simulación  de Bagging
Objetivo: Mostrar cómo funciona Bagging con muestreo bootstrap y combinaciones simples.

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score

# Generar datos sintéticos
X, y = make_classification(n_samples=500, n_features=10, random_state=42)

# Crear subconjuntos bootstrap y entrenar modelos independientes
n_estimators = 10
predictions = []
for i in range(n_estimators):
    # Muestreo bootstrap
    idx = np.random.choice(len(X), size=len(X), replace=True)
    X_bootstrap, y_bootstrap = X[idx], y[idx]

    # Entrenar un árbol de decisión
    model = DecisionTreeClassifier(max_depth=3, random_state=i)
    model.fit(X_bootstrap, y_bootstrap)

    # Guardar las predicciones
    predictions.append(model.predict(X))

# Combinar predicciones mediante votación por mayoría
final_predictions = np.array(predictions).T
final_predictions = [np.bincount(row).argmax() for row in final_predictions]

# Evaluar desempeño
print("Accuracy:", accuracy_score(y, final_predictions))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display

# Función para calcular la convergencia de la probabilidad teórica
def bootstrap_convergence_with_values(max_samples):
    # Crear una lista de números de muestras para simular
    num_samples_list = np.linspace(1, max_samples, 100, dtype=int)

    theoretical_probs = []
    approx_prob = 1 / np.e  # Aproximación para N grande
    results = []

    for N in num_samples_list:
        prob_not_selected = (1 - 1 / N) ** N
        theoretical_probs.append(prob_not_selected)
        results.append((N, prob_not_selected))

    # Graficar la convergencia
    plt.figure(figsize=(10, 6))
    plt.plot(num_samples_list, theoretical_probs, label="Probabilidad Teórica $(1 - 1/N)^N$", color="blue")
    plt.axhline(approx_prob, color="red", linestyle="--", label="Aproximación $1/e$")
    plt.title("Convergencia de $(1 - 1/N)^N$ a $1/e$", fontsize=16)
    plt.xlabel("Número de muestras (N)", fontsize=12)
    plt.ylabel("Probabilidad de no ser seleccionado", fontsize=12)
    plt.legend(fontsize=12)
    plt.grid(True)
    plt.show()

    # Retornar los resultados
    return results

# Solicitar el número máximo de muestras al usuario
try:
    max_samples = int(input("Ingresa el número máximo de muestras (ejemplo: 1000): "))
    if max_samples < 1:
        raise ValueError("El número de muestras debe ser mayor a 0.")
except ValueError as e:
    print(f"Entrada inválida: {e}. Usando el valor predeterminado de 1000.")
    max_samples = 1000

# Llamar a la función con el valor ingresado
results = bootstrap_convergence_with_values(max_samples)

# Mostrar los valores calculados en un DataFrame
df_results = pd.DataFrame(results, columns=["Número de Muestras (N)", "Probabilidad Teórica"])
display(df_results)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import pandas as pd

# Generar datos sintéticos
X, y = make_classification(n_samples=500, n_features=10, n_classes=2, n_redundant=0, random_state=42)

# Configuración del Random Forest
n_estimators = 10  # Número de árboles en el bosque
max_features = 5   # Número máximo de características seleccionadas por árbol
max_depth = 3      # Profundidad máxima de cada árbol
random_state = 42  # Semilla para reproducibilidad

predictions = []
models = []  # Almacenar modelos, características y muestras seleccionadas
selected_features_list = []
bootstrap_indices_list = []

# Entrenar múltiples árboles de decisión
print("Características y muestras seleccionadas para cada árbol:")
for i in range(n_estimators):
    # Muestreo bootstrap
    idx = np.random.choice(len(X), size=len(X), replace=True)
    bootstrap_indices_list.append(idx)
    X_bootstrap, y_bootstrap = X[idx], y[idx]

    # Selección aleatoria de características
    feature_indices = np.random.choice(X.shape[1], size=max_features, replace=False)
    selected_features_list.append(feature_indices)
    X_bootstrap_reduced = X_bootstrap[:, feature_indices]

    # Entrenar un árbol de decisión
    model = DecisionTreeClassifier(max_depth=max_depth, random_state=random_state + i)
    model.fit(X_bootstrap_reduced, y_bootstrap)
    models.append((model, feature_indices, idx))  # Guardar el modelo, características y muestras seleccionadas

    # Mostrar las características y muestras seleccionadas
    print(f"Árbol {i+1}: características seleccionadas {feature_indices}")
    print(f"Árbol {i+1}: primeras 5 muestras seleccionadas (índices): {idx[:5]}")

    # Predecir en el conjunto original usando solo las características seleccionadas
    predictions.append(model.predict(X[:, feature_indices]))

# Votación por mayoría para predicción final
predictions = np.array(predictions)
final_predictions = np.apply_along_axis(lambda x: np.bincount(x).argmax(), axis=0, arr=predictions)

# Mostrar resumen de características y muestras seleccionadas
print("\nResumen de características seleccionadas por árbol:")
for i, features in enumerate(selected_features_list, start=1):
    print(f"Árbol {i}: {features}")

print("\nResumen de muestras seleccionadas (primeros 5 índices por árbol):")
for i, idx in enumerate(bootstrap_indices_list, start=1):
    print(f"Árbol {i}: {idx[:5]}")

# Crear tablas resumen
df_features = pd.DataFrame({
    f"Árbol {i+1}": features for i, features in enumerate(selected_features_list)
}).T
df_features.columns = [f"Característica {j+1}" for j in range(max_features)]

df_samples = pd.DataFrame({
    f"Árbol {i+1}": indices[:5] for i, indices in enumerate(bootstrap_indices_list)
}).T
df_samples.columns = [f"Muestra {j+1}" for j in range(5)]

# Mostrar tablas resumen
print("\nTabla Resumen de Características Seleccionadas:")
print(df_features)

print("\nTabla Resumen de Muestras Seleccionadas:")
print(df_samples)

# Mostrar precisión del ensamble
accuracy = accuracy_score(y, final_predictions)
print(f"\nPrecisión del Random Forest: {accuracy:.2f}")

# Visualización de la precisión individual
individual_accuracies = [accuracy_score(y, predictions[i]) for i in range(n_estimators)]
plt.figure(figsize=(10, 6))
plt.bar(range(1, n_estimators + 1), individual_accuracies, label="Modelos Individuales")
plt.axhline(accuracy, color='r', linestyle='--', label="Precisión del Ensamble")
plt.title("Comparación de Precisión: Modelos Individuales vs Ensamble", fontsize=14)
plt.xlabel("Modelo", fontsize=12)
plt.ylabel("Precisión", fontsize=12)
plt.legend(fontsize=12)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import seaborn as sns

# Cargar el conjunto de datos Iris
iris = load_iris()
X = iris.data[:, :2]  # Usar solo las dos primeras características
y = iris.target

# Dividir los datos en conjunto de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Entrenar un Árbol de Decisión
dt = DecisionTreeClassifier(max_depth=3, random_state=42)
dt.fit(X_train, y_train)

# Entrenar un Random Forest con solo las dos primeras características
rf = RandomForestClassifier(n_estimators=10, max_depth=3, random_state=42)
rf.fit(X_train, y_train)

# Crear una cuadrícula para visualizar las regiones de decisión
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 200), np.linspace(y_min, y_max, 200))

# Predicciones del Árbol de Decisión
Z_dt = dt.predict(np.c_[xx.ravel(), yy.ravel()])
Z_dt = Z_dt.reshape(xx.shape)

# Predicciones del Random Forest
Z_rf = rf.predict(np.c_[xx.ravel(), yy.ravel()])
Z_rf = Z_rf.reshape(xx.shape)

# Comparar visualmente
plt.figure(figsize=(16, 6))

# Árbol de Decisión
plt.subplot(1, 2, 1)
plt.contourf(xx, yy, Z_dt, alpha=0.3, cmap=ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF']))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=ListedColormap(['#FF0000', '#00FF00', '#0000FF']), edgecolor='k')
plt.title("Árbol de Decisión")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

# Random Forest
plt.subplot(1, 2, 2)
plt.contourf(xx, yy, Z_rf, alpha=0.3, cmap=ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF']))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=ListedColormap(['#FF0000', '#00FF00', '#0000FF']), edgecolor='k')
plt.title("Random Forest (10 árboles)")
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])

plt.show()

# Evaluar el Random Forest
y_pred_rf = rf.predict(X_test)
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Precisión del Random Forest: {accuracy_rf:.2f}")

# Reporte de Clasificación
print("\nReporte de Clasificación (Random Forest):")
print(classification_report(y_test, y_pred_rf, target_names=iris.target_names))

# Matriz de confusión
conf_matrix_rf = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_rf, annot=True, fmt="d", cmap="coolwarm", xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title("Matriz de Confusión (Random Forest)")
plt.xlabel("Predicción")
plt.ylabel("Verdadero")
plt.show()

# Importancia de las características (solo las dos utilizadas)
feature_importances_rf = pd.Series(rf.feature_importances_, index=iris.feature_names[:2]).sort_values(ascending=False)

# Visualizar la importancia de las características
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances_rf, y=feature_importances_rf.index, palette="viridis")
plt.title("Importancia de las Características (Random Forest)")
plt.xlabel("Importancia")
plt.ylabel("Características")
plt.show()

```

```python
# Importar librerías necesarias
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el conjunto de datos Iris
from sklearn import datasets
iris = datasets.load_iris()

# Crear un DataFrame para facilitar el manejo de los datos
data = pd.DataFrame({
    'sepal length': iris.data[:, 0],
    'sepal width': iris.data[:, 1],
    'petal length': iris.data[:, 2],
    'petal width': iris.data[:, 3],
    'species': iris.target
})

# Mostrar las primeras filas del conjunto de datos
print("Primeras filas del conjunto de datos:")
print(data.head())

# Definir las características (X) y las etiquetas (y)
X = data[['sepal length', 'sepal width', 'petal length', 'petal width']]
y = data['species']

# Dividir el conjunto de datos en entrenamiento y prueba (70% entrenamiento, 30% prueba)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Crear un modelo Random Forest
clf = RandomForestClassifier(n_estimators=100, random_state=42)

# Entrenar el modelo con los datos de entrenamiento
clf.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = clf.predict(X_test)

# Calcular la precisión del modelo
accuracy = accuracy_score(y_test, y_pred)
print(f"\nPrecisión del modelo: {accuracy:.2f}")

# Reporte de clasificación
print("\nReporte de Clasificación:")
print(classification_report(y_test, y_pred, target_names=iris.target_names))

# Matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="coolwarm", xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title("Matriz de Confusión")
plt.xlabel("Predicción")
plt.ylabel("Verdadero")
plt.show()

# Importancia de las características
feature_importances = pd.Series(clf.feature_importances_, index=X.columns).sort_values(ascending=False)
print("\nImportancia de las características:")
print(feature_importances)

# Visualizar la importancia de las características
plt.figure(figsize=(10, 6))
sns.barplot(x=feature_importances, y=feature_importances.index, palette="viridis")
plt.title("Importancia de las Características")
plt.xlabel("Importancia")
plt.ylabel("Características")
plt.show()

# Generar el modelo en las características seleccionadas (sin 'sepal width')
X_selected = data[['petal length', 'petal width', 'sepal length']]  # Quitamos 'sepal width'
X_train_sel, X_test_sel, y_train_sel, y_test_sel = train_test_split(X_selected, y, test_size=0.3, random_state=42, stratify=y)

# Entrenar un nuevo modelo con las características seleccionadas
clf_sel = RandomForestClassifier(n_estimators=100, random_state=42)
clf_sel.fit(X_train_sel, y_train_sel)

# Predicciones y evaluación
y_pred_sel = clf_sel.predict(X_test_sel)
accuracy_sel = accuracy_score(y_test_sel, y_pred_sel)
print(f"\nPrecisión del modelo con características seleccionadas: {accuracy_sel:.2f}")

# Matriz de confusión con características seleccionadas
conf_matrix_sel = confusion_matrix(y_test_sel, y_pred_sel)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_sel, annot=True, fmt="d", cmap="coolwarm", xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title("Matriz de Confusión (Características Seleccionadas)")
plt.xlabel("Predicción")
plt.ylabel("Verdadero")
plt.show()
```
