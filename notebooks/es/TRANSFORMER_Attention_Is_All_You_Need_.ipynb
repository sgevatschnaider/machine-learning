```python
from IPython.display import display, HTML

# Código HTML completo
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Arquitectura del Transformer</title>
    <style>
        body {
            font-family: 'Georgia', serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            background-image: url('https://www.transparenttextures.com/patterns/light-paper-fibers.png');
            background-size: cover;
            background-attachment: fixed;
        }
        h1, h2, h3 {
            color: #333;
        }
        h1 {
            font-size: 2.5em;
            text-align: center;
            margin-bottom: 20px;
        }
        h2 {
            font-size: 1.8em;
            margin-top: 30px;
        }
        h3 {
            font-size: 1.5em;
            margin-top: 20px;
        }
        ul {
            margin-top: 10px;
            padding-left: 20px;
        }
        li {
            margin-bottom: 10px;
        }
        .highlight {
            background-color: #ffe0b2;
            padding: 5px;
            border-radius: 5px;
            font-weight: bold;
        }
        .button {
            display: block;
            width: fit-content;
            margin: 20px auto;
            padding: 10px 20px;
            background-color: #0066cc;
            color: white;
            text-decoration: none;
            font-size: 1.2em;
            font-weight: bold;
            text-align: center;
            border-radius: 5px;
            box-shadow: 0px 4px 6px rgba(0, 0, 0, 0.1);
        }
        .button:hover {
            background-color: #004d99;
        }
    </style>
</head>
<body>
    <h1>Arquitectura del Transformer</h1>

    <h2>1. Entradas y Embedding</h2>
    <p>
        El Transformer comienza con una secuencia de palabras, usualmente representadas como números enteros (tokens) que corresponden a palabras o sub-palabras en un vocabulario.
        Estas palabras se convierten en vectores densos (representaciones numéricas más ricas) a través de una matriz de embeddings.
        <br><span class="highlight">Analogía:</span> Es como mapear palabras a posiciones en un mapa donde las relaciones semánticas están representadas por la proximidad.
    </p>

    <h2>2. Positional Encoding</h2>
    <p>
        A diferencia de otros modelos como RNNs, el Transformer no procesa las palabras en orden secuencial.
        Para entender el orden de las palabras, agrega un codificado posicional que introduce información sobre la posición relativa o absoluta de cada palabra en la secuencia.
        <br><span class="highlight">Analogía:</span> Es como etiquetar cada palabra con una coordenada temporal que dice dónde se encuentra en el texto.
    </p>

    <h2>3. Multi-Head Attention</h2>
    <p>
        Este es el componente clave del Transformer. La atención sirve para que el modelo se enfoque en diferentes partes de la entrada en paralelo.
        <br><span class="highlight">¿Qué es Atención?</span> Imagina que lees una frase compleja. Para comprender una palabra, necesitas mirar otras palabras que le den contexto.
        <br><span class="highlight">Analogía:</span> Es como tener múltiples cámaras enfocadas en la misma escena pero desde ángulos diferentes.
    </p>

    <h2>4. Add & Norm</h2>
    <p>
        Después del paso de atención, se agrega la entrada original al resultado procesado (conexión residual), y el resultado pasa por una capa de normalización.
        <br><span class="highlight">Analogía:</span> Es como construir un edificio sobre una base sólida y luego ajustar el equilibrio para asegurarte de que todo está alineado.
    </p>

    <h2>5. Feed Forward Network</h2>
    <p>
        Cada vector (palabra) pasa por una red neuronal completamente conectada. Esto se realiza independientemente para cada posición en la secuencia.
        <br><span class="highlight">Analogía:</span> Es como aplicar un filtro de realce a cada píxel de una imagen individualmente.
    </p>

    <h2>6. Repetición del Bloque</h2>
    <p>
        El bloque que incluye Multi-Head Attention, Add & Norm, y Feed Forward se repite N veces. Esto permite al modelo profundizar en el entendimiento de las relaciones entre las palabras.
        <br><span class="highlight">Analogía:</span> Es como refinar un dibujo artístico con múltiples capas de detalles.
    </p>

    <h2>7. Decodificador</h2>
    <p>
        La segunda parte del Transformer es el decodificador. Este toma las representaciones internas generadas por el codificador y las convierte en una salida.
        <br><span class="highlight">Masked Multi-Head Attention:</span> Se agrega una atención enmascarada para asegurarse de que el modelo solo vea palabras previas durante la generación de texto.
        <br><span class="highlight">Analogía:</span> Es como escribir una oración en un examen: solo puedes usar información de las palabras que ya has escrito, no de las futuras.
    </p>

    <h2>8. Softmax y Salida</h2>
    <p>
        Finalmente, el Transformer pasa las representaciones finales por una capa completamente conectada (linear) y aplica una función Softmax para convertir las predicciones en probabilidades.
        <br><span class="highlight">Analogía:</span> Es como elegir una palabra en un menú donde cada palabra tiene una probabilidad asignada, y seleccionas la más probable.
    </p>

    <a href="https://arxiv.org/pdf/1706.03762" class="button" target="_blank">Attention Is All You Need</a>
</body>
</html>
"""

# Renderizar el contenido HTML
display(HTML(html_content))
```

```python
# Importamos las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Configuración del vocabulario y matriz de embedding
vocab = ["gato", "perro", "león", "ratón", "auto", "avión", "barco", "bicicleta", "casa", "hotel"]
vocab_size = len(vocab)
embedding_dim = 5  # Dimensiones del embedding (pequeño para simplicidad)

# Creamos una matriz de embedding aleatoria
np.random.seed(42)  # Fijar semilla para reproducibilidad
embedding_matrix = np.random.rand(vocab_size, embedding_dim)

# Visualización: Reducimos las dimensiones de los embeddings a 2D usando PCA
pca = PCA(n_components=2)
reduced_embeddings = pca.fit_transform(embedding_matrix)

# Graficamos los embeddings en 2D
plt.figure(figsize=(10, 6))
for i, word in enumerate(vocab):
    x, y = reduced_embeddings[i]
    plt.scatter(x, y, label=word)
    plt.text(x + 0.02, y + 0.02, word, fontsize=12)

plt.title("Visualización de Embeddings en 2D", fontsize=16)
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.grid(True)
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.tight_layout()
plt.show()

```

```python
from IPython.display import display, HTML
import numpy as np
import matplotlib.pyplot as plt

# HTML para explicar la teoría
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Positional Encoding</title>
    <style>
        body {
            font-family: 'Georgia', serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            background-image: url('https://www.transparenttextures.com/patterns/light-paper-fibers.png');
            background-size: cover;
            background-attachment: fixed;
        }
        h1, h2 {
            color: #333;
        }
        h1 {
            font-size: 2.5em;
            text-align: center;
            margin-bottom: 20px;
        }
        h2 {
            font-size: 1.8em;
            margin-top: 30px;
        }
        p {
            margin-bottom: 15px;
        }
        .highlight {
            background-color: #ffe0b2;
            padding: 5px;
            border-radius: 5px;
            font-weight: bold;
        }
    </style>
</head>
<body>
    <h1>Positional Encoding</h1>

    <h2>Analogía: Marcador de tiempo en una película</h2>
    <p>
        En el Transformer, las palabras no tienen un orden intrínseco, ya que el modelo no es secuencial como una RNN.
        Para darles un "orden", se utiliza un codificado posicional (<span class="highlight">Positional Encoding</span>).
    </p>
    <p>
        Este codificado es como un marcador de tiempo en una película: indica dónde está cada palabra en la secuencia, sin procesar las palabras una por una.
    </p>

    <h2>¿Cómo funciona?</h2>
    <p>
        Usa funciones trigonométricas (<span class="highlight">seno</span> y <span class="highlight">coseno</span>) para generar patrones únicos para cada posición.
        Si tienes una posición <em>pos</em> y una dimensión <em>i</em>, el cálculo es:
    </p>
    <ul>
        <li><span class="highlight">PE(pos, 2i) = sin(pos / 10000^(2i/d))</span></li>
        <li><span class="highlight">PE(pos, 2i+1) = cos(pos / 10000^(2i/d))</span></li>
    </ul>
    <p>
        Donde <em>d</em> es la dimensión del embedding.
    </p>

    <h2>¿Por qué usar seno y coseno?</h2>
    <p>
        Estas funciones son <span class="highlight">periódicas</span> y permiten que el modelo capture relaciones relativas y absolutas entre las palabras, incluso si la longitud de la secuencia cambia.
    </p>

    <h2>Ejemplo práctico:</h2>
    <p>
        En el código siguiente, crearemos un gráfico que muestra cómo los valores de seno y coseno varían con las posiciones en la secuencia para diferentes dimensiones.
    </p>
</body>
</html>
"""

# Mostrar el HTML con la teoría
display(HTML(html_content))

# Función para calcular el Positional Encoding
def positional_encoding(seq_len, d_model):
    """
    Calcula el codificado posicional para una secuencia y una dimensión del modelo.
    :param seq_len: Longitud de la secuencia
    :param d_model: Dimensión del embedding
    :return: Matriz de codificados posicionales de tamaño [seq_len, d_model]
    """
    pos = np.arange(seq_len)[:, np.newaxis]  # Posiciones [seq_len, 1]
    i = np.arange(d_model)[np.newaxis, :]    # Dimensiones [1, d_model]
    angle_rates = 1 / np.power(10000, (2 * (i // 2)) / d_model)  # Tasa de cambio
    angle_rads = pos * angle_rates  # Multiplicamos posición por tasa

    # Aplicamos seno a índices pares y coseno a índices impares
    pos_encoding = np.zeros((seq_len, d_model))
    pos_encoding[:, 0::2] = np.sin(angle_rads[:, 0::2])  # Seno para pares
    pos_encoding[:, 1::2] = np.cos(angle_rads[:, 1::2])  # Coseno para impares

    return pos_encoding

# Parámetros del modelo
seq_len = 50  # Longitud de la secuencia
d_model = 16  # Dimensión del embedding

# Calculamos el Positional Encoding
pos_encoding = positional_encoding(seq_len, d_model)

# Visualización del mapa de calor
plt.figure(figsize=(12, 6))
plt.title("Visualización del Positional Encoding", fontsize=16)
plt.imshow(pos_encoding, cmap='viridis', aspect='auto')
plt.colorbar(label="Valor codificado")
plt.xlabel("Dimensiones del Modelo")
plt.ylabel("Posición en la Secuencia")
plt.tight_layout()
plt.show()

# Gráfica de una sola dimensión
plt.figure(figsize=(10, 6))
for i in range(4):  # Visualizamos las primeras 4 dimensiones
    plt.plot(pos_encoding[:, i], label=f"Dimensión {i}")
plt.title("Variación de Positional Encoding para Diferentes Dimensiones", fontsize=16)
plt.xlabel("Posición en la Secuencia")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from IPython.display import display, HTML

# HTML para la teoría de Multi-Head Attention
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Multi-Head Attention</title>
    <style>
        body {
            font-family: 'Georgia', serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            background-image: url('https://www.transparenttextures.com/patterns/light-paper-fibers.png');
            background-size: cover;
            background-attachment: fixed;
        }
        h1, h2 {
            color: #333;
        }
        h1 {
            font-size: 2.5em;
            text-align: center;
            margin-bottom: 20px;
        }
        h2 {
            font-size: 1.8em;
            margin-top: 30px;
        }
        p, ul {
            margin-bottom: 15px;
        }
        ul {
            margin-left: 20px;
        }
        .highlight {
            background-color: #ffe0b2;
            padding: 5px;
            border-radius: 5px;
            font-weight: bold;
        }
    </style>
</head>
<body>
    <h1>Multi-Head Attention</h1>

    <h2>¿Qué es la atención?</h2>
    <p>
        El mecanismo de atención es como si el modelo estuviera "mirando" diferentes partes de la entrada para decidir cuáles son más relevantes
        al procesar una palabra o un token. En lugar de procesar todas las palabras por igual, la atención permite que el modelo enfoque su
        "atención" en las palabras que son más importantes para el contexto.
    </p>
    <p>
        Por ejemplo, en la oración <span class="highlight">"El gato persigue al ratón"</span>, para entender la palabra <em>"persigue"</em>,
        el modelo necesita prestar atención a <em>"gato"</em> (el sujeto) y a <em>"ratón"</em> (el objeto).
    </p>

    <h2>¿Cómo funciona?</h2>
    <p>
        El mecanismo de atención se basa en tres componentes clave: <span class="highlight">Query (Q)</span>, <span class="highlight">Key (K)</span>,
        y <span class="highlight">Value (V)</span>.
    </p>
    <ul>
        <li>
            <span class="highlight">Query (Q):</span> Representa lo que estamos buscando. Por ejemplo, "¿Qué contexto es importante para esta palabra?"
        </li>
        <li>
            <span class="highlight">Key (K):</span> Representa el contenido disponible en cada palabra (o token) para decidir si es relevante.
        </li>
        <li>
            <span class="highlight">Value (V):</span> Representa la información que necesitamos para construir el contexto final.
        </li>
    </ul>
    <p>
        El modelo calcula la importancia (o peso de atención) para cada palabra en la entrada mediante el producto escalar entre <em>Query</em> y
        <em>Key</em>. Luego, estos pesos se normalizan utilizando <span class="highlight">softmax</span> para que sumen 1.
    </p>

    <h2>¿Qué es el Multi-Head Attention?</h2>
    <p>
        En lugar de calcular la atención una sola vez, el Transformer utiliza múltiples cabezas de atención (<span class="highlight">Multi-Head Attention</span>).
        Cada cabeza calcula la atención de manera independiente, lo que permite que el modelo capture diferentes tipos de relaciones entre las palabras.
    </p>
    <ul>
        <li>
            Por ejemplo, una cabeza puede enfocarse en relaciones gramaticales (e.g., sujeto-verbo), mientras que otra puede capturar relaciones semánticas
            (e.g., "ratón" como objeto del "gato").
        </li>
        <li>
            Los resultados de todas las cabezas se concatenan y se combinan para obtener una representación rica y completa.
        </li>
    </ul>

    <h2>Fórmula de la Atención Escalar</h2>
    <p>
        La atención escalar se calcula como:
    </p>
    <p>
        <span class="highlight">Attention(Q, K, V) = softmax((Q * K^T) / √d_k) * V</span>
    </p>
    <p>
        Donde:
    </p>
    <ul>
        <li>
            <span class="highlight">Q * K^T:</span> Calcula la similitud entre <em>Query</em> y <em>Key</em>.
        </li>
        <li>
            <span class="highlight">√d_k:</span> Escala el resultado para evitar valores extremadamente grandes cuando <em>d_k</em> (la dimensión de las claves) es grande.
        </li>
        <li>
            <span class="highlight">softmax:</span> Normaliza los pesos de atención para que sumen 1.
        </li>
    </ul>

    <h2>Beneficios del Multi-Head Attention</h2>
    <ul>
        <li>
            Permite que el modelo enfoque su atención en diferentes partes de la entrada al mismo tiempo.
        </li>
        <li>
            Captura relaciones complejas entre palabras a diferentes niveles (locales y globales).
        </li>
        <li>
            Mejora la capacidad del modelo para entender oraciones largas y contextos complejos.
        </li>
    </ul>

    <h2>Relación con el Transformer</h2>
    <p>
        En el Transformer, el mecanismo de atención se utiliza tanto en el codificador como en el decodificador:
    </p>
    <ul>
        <li>
            <span class="highlight">Codificador:</span> Las palabras en la entrada se comparan entre sí para generar representaciones enriquecidas.
        </li>
        <li>
            <span class="highlight">Decodificador:</span> Las palabras generadas hasta el momento se comparan entre sí y con la entrada para predecir la siguiente palabra.
        </li>
    </ul>
</body>
</html>
"""

# Mostrar el HTML con la teoría
display(HTML(html_content))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize

# Función para calcular la atención escalar
def scaled_dot_product_attention(Q, K, V):
    """
    Calcula la atención escalar.
    :param Q: Matriz Query (shape: [seq_len, d_k])
    :param K: Matriz Key (shape: [seq_len, d_k])
    :param V: Matriz Value (shape: [seq_len, d_v])
    :return: Resultado de la atención (shape: [seq_len, d_v]) y los pesos de atención
    """
    # Producto escalar entre Q y K^T
    matmul_qk = np.dot(Q, K.T)

    # Escalado por la raíz cuadrada de d_k
    d_k = Q.shape[-1]
    scaled_attention_logits = matmul_qk / np.sqrt(d_k)

    # Aplicamos softmax para obtener los pesos de atención
    attention_weights = np.exp(scaled_attention_logits)
    attention_weights = attention_weights / np.sum(attention_weights, axis=-1, keepdims=True)

    # Multiplicamos los pesos por los valores V
    output = np.dot(attention_weights, V)
    return output, attention_weights

# Función de Multi-Head Attention
def multi_head_attention(Q, K, V, num_heads):
    """
    Implementa el mecanismo de Multi-Head Attention.
    :param Q: Matriz Query (shape: [seq_len, d_model])
    :param K: Matriz Key (shape: [seq_len, d_model])
    :param V: Matriz Value (shape: [seq_len, d_model])
    :param num_heads: Número de cabezas de atención
    :return: Salida concatenada de todas las cabezas (shape: [seq_len, d_model])
    """
    seq_len, d_model = Q.shape
    assert d_model % num_heads == 0, "d_model debe ser divisible entre num_heads"

    # Dividimos las dimensiones en subdimensiones para cada cabeza
    depth = d_model // num_heads
    outputs = []

    for head in range(num_heads):
        # Creamos submatrices para cada cabeza
        Q_sub = Q[:, head * depth: (head + 1) * depth]
        K_sub = K[:, head * depth: (head + 1) * depth]
        V_sub = V[:, head * depth: (head + 1) * depth]

        # Aplicamos atención escalar
        head_output, _ = scaled_dot_product_attention(Q_sub, K_sub, V_sub)
        outputs.append(head_output)

    # Concatenamos las salidas de todas las cabezas
    concatenated = np.concatenate(outputs, axis=-1)
    return concatenated

# Simulación de entradas (Q, K, V)
np.random.seed(42)
seq_len = 6       # Longitud de la secuencia
d_model = 8       # Dimensión del modelo
num_heads = 2     # Número de cabezas de atención

Q = normalize(np.random.rand(seq_len, d_model))  # Matriz Query
K = normalize(np.random.rand(seq_len, d_model))  # Matriz Key
V = normalize(np.random.rand(seq_len, d_model))  # Matriz Value

# Multi-Head Attention
multi_head_output = multi_head_attention(Q, K, V, num_heads)

# Visualización de los pesos de atención de una cabeza
output, attention_weights = scaled_dot_product_attention(Q[:, :d_model//num_heads],
                                                         K[:, :d_model//num_heads],
                                                         V[:, :d_model//num_heads])

# Visualizamos los pesos de atención
plt.figure(figsize=(8, 6))
plt.title("Pesos de Atención de una Cabeza", fontsize=16)
plt.imshow(attention_weights, cmap="viridis", aspect="auto")
plt.colorbar(label="Pesos de Atención")
plt.xlabel("Posiciones (Key)")
plt.ylabel("Posiciones (Query)")
plt.tight_layout()
plt.show()

# Visualización de la salida final
plt.figure(figsize=(10, 6))
for i in range(seq_len):
    plt.plot(multi_head_output[i], label=f"Output posición {i}")
plt.title("Salida de Multi-Head Attention por Posición", fontsize=16)
plt.xlabel("Dimensiones del Modelo")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from IPython.display import display, HTML

# HTML para la teoría de Feed Forward Network
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Feed Forward Network (FFN)</title>
    <style>
        body {
            font-family: 'Georgia', serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            background-image: url('https://www.transparenttextures.com/patterns/light-paper-fibers.png');
            background-size: cover;
            background-attachment: fixed;
        }
        h1, h2 {
            color: #333;
        }
        h1 {
            font-size: 2.5em;
            text-align: center;
            margin-bottom: 20px;
        }
        h2 {
            font-size: 1.8em;
            margin-top: 30px;
        }
        p, ul {
            margin-bottom: 15px;
        }
        ul {
            margin-left: 20px;
        }
        .highlight {
            background-color: #ffe0b2;
            padding: 5px;
            border-radius: 5px;
            font-weight: bold;
        }
    </style>
</head>
<body>
    <h1>Feed Forward Network (FFN)</h1>

    <h2>¿Qué es la Feed Forward Network?</h2>
    <p>
        La <span class="highlight">Feed Forward Network (FFN)</span> es una red neuronal completamente conectada que se aplica de manera independiente a cada posición
        en la secuencia. Es un componente crítico del Transformer que ayuda a transformar las representaciones de cada palabra o token después de aplicar la atención.
    </p>
    <p>
        La FFN trabaja de manera <span class="highlight">independiente</span> para cada palabra en la secuencia. Esto significa que no intercambia información entre posiciones,
        sino que procesa cada posición por separado.
    </p>

    <h2>¿Cómo funciona?</h2>
    <p>
        La FFN consta de dos capas densas (completamente conectadas) con una función de activación no lineal (como <span class="highlight">ReLU</span>) entre ellas.
        Para una entrada <em>x</em>, el cálculo es:
    </p>
    <p>
        <span class="highlight">FFN(x) = max(0, xW₁ + b₁)W₂ + b₂</span>
    </p>
    <p>
        Donde:
    </p>
    <ul>
        <li>
            <span class="highlight">W₁:</span> Matriz de pesos de la primera capa.
        </li>
        <li>
            <span class="highlight">b₁:</span> Vector de sesgos de la primera capa.
        </li>
        <li>
            <span class="highlight">W₂:</span> Matriz de pesos de la segunda capa.
        </li>
        <li>
            <span class="highlight">b₂:</span> Vector de sesgos de la segunda capa.
        </li>
        <li>
            <span class="highlight">max(0, ·):</span> Es la función de activación <em>ReLU</em>, que introduce no linealidad.
        </li>
    </ul>

    <h2>¿Por qué es importante?</h2>
    <p>
        La FFN permite que el Transformer introduzca transformaciones adicionales a las representaciones de las palabras. Esto ayuda a:
    </p>
    <ul>
        <li>
            <span class="highlight">Capturar patrones complejos:</span> Procesa características individuales de cada palabra para mejorar su representación.
        </li>
        <li>
            <span class="highlight">Añadir no linealidad:</span> La función <em>ReLU</em> permite modelar relaciones no lineales en los datos, lo que es fundamental para capturar relaciones complejas en el texto.
        </li>
    </ul>

    <h2>Características de la FFN en el Transformer</h2>
    <ul>
        <li>
            <span class="highlight">Independencia posicional:</span> La FFN procesa cada posición de manera independiente, lo que significa que no afecta a otras palabras en la secuencia.
        </li>
        <li>
            <span class="highlight">Dimensiones mayores:</span> La primera capa normalmente aumenta la dimensión del vector, mientras que la segunda la reduce para devolver el tamaño original.
            <ul>
                <li>Por ejemplo: Si la dimensión de entrada es <em>d_model</em> (e.g., 512), la primera capa puede ampliarla a 2048, y la segunda capa la devuelve a 512.</li>
            </ul>
        </li>
        <li>
            <span class="highlight">Velocidad:</span> Dado que no hay interacción entre posiciones, el cálculo de la FFN es altamente paralelizable.
        </li>
    </ul>

    <h2>Relación con otros componentes</h2>
    <p>
        La FFN se encuentra entre dos capas de <span class="highlight">Multi-Head Attention</span> y está envuelta en conexiones residuales y normalización.
        Esto asegura que las representaciones iniciales (antes de la atención) también contribuyan a la salida final.
    </p>
    <p>
        El flujo sería:
    </p>
    <ul>
        <li>Entrada → Multi-Head Attention → Conexión Residual → FFN → Conexión Residual → Salida</li>
    </ul>
</body>
</html>
"""

# Mostrar el HTML con la teoría
display(HTML(html_content))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Clase para implementar la Feed Forward Network
class FeedForwardNetwork:
    def __init__(self, d_model, d_ff):
        """
        Constructor de la Feed Forward Network.
        :param d_model: Dimensión de entrada/salida (e.g., 512).
        :param d_ff: Dimensión interna de la capa oculta (e.g., 2048).
        """
        self.W1 = np.random.randn(d_model, d_ff) * 0.1  # Pesos de la primera capa
        self.b1 = np.zeros(d_ff)  # Sesgo de la primera capa
        self.W2 = np.random.randn(d_ff, d_model) * 0.1  # Pesos de la segunda capa
        self.b2 = np.zeros(d_model)  # Sesgo de la segunda capa

    def relu(self, x):
        """Función de activación ReLU."""
        return np.maximum(0, x)

    def forward(self, x):
        """
        Propagación hacia adelante en la FFN.
        :param x: Entrada (shape: [batch_size, d_model])
        :return: Salida procesada (shape: [batch_size, d_model])
        """
        # Primera capa: Lineal + ReLU
        hidden = self.relu(np.dot(x, self.W1) + self.b1)
        # Segunda capa: Lineal
        output = np.dot(hidden, self.W2) + self.b2
        return output

# Simulación de datos
np.random.seed(42)
batch_size = 4  # Número de palabras o tokens
d_model = 512   # Dimensión de entrada y salida
d_ff = 2048     # Dimensión oculta

# Creamos datos simulados para una entrada (tokens procesados por Multi-Head Attention)
input_data = np.random.rand(batch_size, d_model)

# Inicializamos la Feed Forward Network
ffn = FeedForwardNetwork(d_model, d_ff)

# Propagación hacia adelante
output_data = ffn.forward(input_data)

# Visualización de la entrada y salida
plt.figure(figsize=(10, 6))
plt.plot(input_data[0], label="Entrada (Token 0)")
plt.plot(output_data[0], label="Salida (Token 0)")
plt.title("Feed Forward Network: Entrada vs. Salida", fontsize=16)
plt.xlabel("Dimensiones del Modelo (d_model)")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Resumen de los datos
print("Entrada (primer token):")
print(input_data[0][:10])  # Mostramos las primeras 10 dimensiones
print("\nSalida (primer token):")
print(output_data[0][:10])  # Mostramos las primeras 10 dimensiones
```

```python
from IPython.display import display, HTML

# HTML para la teoría de Add & Norm
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Add & Norm</title>
    <style>
        body {
            font-family: 'Georgia', serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            background-image: url('https://www.transparenttextures.com/patterns/light-paper-fibers.png');
            background-size: cover;
            background-attachment: fixed;
        }
        h1, h2 {
            color: #333;
        }
        h1 {
            font-size: 2.5em;
            text-align: center;
            margin-bottom: 20px;
        }
        h2 {
            font-size: 1.8em;
            margin-top: 30px;
        }
        p, ul {
            margin-bottom: 15px;
        }
        ul {
            margin-left: 20px;
        }
        .highlight {
            background-color: #ffe0b2;
            padding: 5px;
            border-radius: 5px;
            font-weight: bold;
        }
    </style>
</head>
<body>
    <h1>Add & Norm</h1>

    <h2>¿Qué es Add & Norm?</h2>
    <p>
        El paso de <span class="highlight">Add & Norm</span> es un componente clave del Transformer que mejora la estabilidad y la eficiencia del aprendizaje.
        Este paso combina:
    </p>
    <ul>
        <li>
            <span class="highlight">Conexiones Residuales (Add):</span> Aseguran que la información original de entrada se conserve, incluso después de operaciones complejas como la atención o la red feed forward.
        </li>
        <li>
            <span class="highlight">Normalización (Norm):</span> Ajusta las activaciones de los datos para mejorar la estabilidad numérica y acelerar la convergencia del modelo.
        </li>
    </ul>

    <h2>¿Cómo funciona Add & Norm?</h2>
    <p>
        El flujo del Add & Norm se puede dividir en dos pasos:
    </p>
    <ol>
        <li>
            <span class="highlight">Conexión Residual:</span> La salida de un bloque (como Multi-Head Attention o FFN) se suma directamente a su entrada original.
            Esto permite que el modelo conserve información inicial que podría ser útil para aprender características más profundas.
        </li>
        <li>
            <span class="highlight">Normalización:</span> Después de sumar la conexión residual, se aplica una normalización.
            Normalmente se usa la <span class="highlight">Layer Normalization</span>, que ajusta cada vector de forma independiente para tener una media de 0 y varianza de 1.
        </li>
    </ol>
    <p>
        En términos matemáticos, para una entrada <em>x</em> y una salida del bloque <em>SubLayer(x)</em>, el Add & Norm se calcula como:
    </p>
    <p>
        <span class="highlight">Output = LayerNorm(x + SubLayer(x))</span>
    </p>

    <h2>¿Por qué es importante?</h2>
    <p>
        Add & Norm tiene varias ventajas importantes:
    </p>
    <ul>
        <li>
            <span class="highlight">Conservación de información:</span> Gracias a la conexión residual, la información inicial no se pierde, incluso si las capas profundas generan salidas erróneas durante el entrenamiento.
        </li>
        <li>
            <span class="highlight">Estabilidad numérica:</span> La normalización previene que las activaciones exploten o desaparezcan, mejorando la estabilidad del modelo.
        </li>
        <li>
            <span class="highlight">Entrenamiento más rápido:</span> Layer Normalization reduce la sensibilidad del modelo a las inicializaciones de pesos, lo que acelera la convergencia.
        </li>
    </ul>

    <h2>Relación con otros componentes</h2>
    <p>
        Add & Norm se aplica después de bloques clave como:
    </p>
    <ul>
        <li><span class="highlight">Multi-Head Attention:</span> Para combinar la salida de la atención con la entrada original.</li>
        <li><span class="highlight">Feed Forward Network (FFN):</span> Para ajustar las representaciones procesadas por la FFN antes de enviarlas al siguiente bloque.</li>
    </ul>
    <p>
        Esto asegura que cada paso en el Transformer mantenga una conexión con las representaciones iniciales, mientras mejora y ajusta los datos en cada capa.
    </p>
</body>
</html>
"""

# Mostrar el HTML con la teoría
display(HTML(html_content))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Clase para implementar Layer Normalization
class LayerNormalization:
    def __init__(self, epsilon=1e-6):
        """
        Constructor para la normalización por capa.
        :param epsilon: Pequeño valor para evitar divisiones por cero.
        """
        self.epsilon = epsilon

    def normalize(self, x):
        """
        Aplica la normalización por capa.
        :param x: Entrada (shape: [batch_size, d_model])
        :return: Salida normalizada
        """
        mean = np.mean(x, axis=-1, keepdims=True)  # Media por cada vector
        variance = np.var(x, axis=-1, keepdims=True)  # Varianza por cada vector
        return (x - mean) / np.sqrt(variance + self.epsilon)  # Normalización

# Clase para implementar Add & Norm
class AddAndNorm:
    def __init__(self):
        self.layer_norm = LayerNormalization()

    def forward(self, x, sublayer_output):
        """
        Realiza la conexión residual y normalización.
        :param x: Entrada original (shape: [batch_size, d_model])
        :param sublayer_output: Salida del subcomponente (e.g., Multi-Head Attention o FFN)
        :return: Salida tras Add & Norm
        """
        # Conexión residual
        added = x + sublayer_output
        # Normalización por capa
        normalized = self.layer_norm.normalize(added)
        return normalized

# Simulación de datos
np.random.seed(42)
batch_size = 4  # Número de tokens (batch)
d_model = 512   # Dimensión del modelo

# Entrada simulada
input_data = np.random.rand(batch_size, d_model)  # Entrada original
sublayer_output = np.random.rand(batch_size, d_model) * 0.5  # Salida simulada de un subcomponente

# Inicializamos Add & Norm
add_and_norm = AddAndNorm()

# Aplicamos Add & Norm
output_data = add_and_norm.forward(input_data, sublayer_output)

# Visualización: Comparamos Entrada, Sublayer Output y Salida Final
plt.figure(figsize=(10, 6))
plt.plot(input_data[0], label="Entrada (Token 0)", alpha=0.7)
plt.plot(sublayer_output[0], label="Sublayer Output (Token 0)", alpha=0.7)
plt.plot(output_data[0], label="Salida Normalizada (Token 0)", alpha=0.9)
plt.title("Add & Norm: Entrada, Sublayer Output y Salida Final", fontsize=16)
plt.xlabel("Dimensiones del Modelo (d_model)")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Resumen de datos
print("Entrada (primer token):")
print(input_data[0][:10])  # Mostramos las primeras 10 dimensiones
print("\nSublayer Output (primer token):")
print(sublayer_output[0][:10])  # Mostramos las primeras 10 dimensiones
print("\nSalida Normalizada (primer token):")
print(output_data[0][:10])  # Mostramos las primeras 10 dimensiones
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Implementación del Multi-Head Attention simplificado
def scaled_dot_product_attention(Q, K, V, mask=None):
    """
    Calcula la atención escalar.
    :param Q: Matriz Query (shape: [batch_size, seq_len, d_model])
    :param K: Matriz Key (shape: [batch_size, seq_len, d_model])
    :param V: Matriz Value (shape: [batch_size, seq_len, d_model])
    :param mask: Máscara para atención (opcional)
    :return: Salida de la atención y los pesos
    """
    # Producto escalar entre Q y K^T (ajustamos las dimensiones)
    matmul_qk = np.matmul(Q, K.transpose(0, 2, 1))  # [batch_size, seq_len, seq_len]

    # Escalado por la raíz cuadrada de la dimensión de K
    d_k = Q.shape[-1]
    scaled_attention_logits = matmul_qk / np.sqrt(d_k)

    # Aplicamos máscara si está presente
    if mask is not None:
        scaled_attention_logits += (mask * -1e9)

    # Softmax para obtener los pesos de atención
    attention_weights = np.exp(scaled_attention_logits)
    attention_weights /= np.sum(attention_weights, axis=-1, keepdims=True)

    # Multiplicamos los pesos por V
    output = np.matmul(attention_weights, V)  # [batch_size, seq_len, d_model]
    return output, attention_weights

# Clase para implementar Layer Normalization
class LayerNormalization:
    def __init__(self, epsilon=1e-6):
        self.epsilon = epsilon

    def normalize(self, x):
        mean = np.mean(x, axis=-1, keepdims=True)
        variance = np.var(x, axis=-1, keepdims=True)
        return (x - mean) / np.sqrt(variance + self.epsilon)

# Clase para implementar Feed Forward Network
class FeedForwardNetwork:
    def __init__(self, d_model, d_ff):
        self.W1 = np.random.randn(d_model, d_ff) * 0.1
        self.b1 = np.zeros(d_ff)
        self.W2 = np.random.randn(d_ff, d_model) * 0.1
        self.b2 = np.zeros(d_model)

    def relu(self, x):
        return np.maximum(0, x)

    def forward(self, x):
        hidden = self.relu(np.dot(x, self.W1) + self.b1)
        output = np.dot(hidden, self.W2) + self.b2
        return output

# Clase para implementar un bloque del Codificador
class EncoderBlock:
    def __init__(self, d_model, d_ff):
        self.layer_norm1 = LayerNormalization()
        self.layer_norm2 = LayerNormalization()
        self.ffn = FeedForwardNetwork(d_model, d_ff)

    def forward(self, x):
        attention_output, _ = scaled_dot_product_attention(x, x, x)  # Self-Attention
        x = self.layer_norm1.normalize(x + attention_output)  # Add & Norm

        ffn_output = self.ffn.forward(x)
        x = self.layer_norm2.normalize(x + ffn_output)  # Add & Norm
        return x

# Clase para implementar el Codificador completo
class Encoder:
    def __init__(self, num_layers, d_model, d_ff):
        self.layers = [EncoderBlock(d_model, d_ff) for _ in range(num_layers)]

    def forward(self, x):
        for layer in self.layers:
            x = layer.forward(x)
        return x

# Clase para implementar un bloque del Decodificador
class DecoderBlock:
    def __init__(self, d_model, d_ff):
        self.layer_norm1 = LayerNormalization()
        self.layer_norm2 = LayerNormalization()
        self.layer_norm3 = LayerNormalization()
        self.ffn = FeedForwardNetwork(d_model, d_ff)

    def forward(self, x, encoder_output):
        attention_output, _ = scaled_dot_product_attention(x, x, x)  # Self-Attention
        x = self.layer_norm1.normalize(x + attention_output)  # Add & Norm

        attention_output, _ = scaled_dot_product_attention(x, encoder_output, encoder_output)
        x = self.layer_norm2.normalize(x + attention_output)  # Add & Norm

        ffn_output = self.ffn.forward(x)
        x = self.layer_norm3.normalize(x + ffn_output)  # Add & Norm
        return x

# Clase para implementar el Decodificador completo
class Decoder:
    def __init__(self, num_layers, d_model, d_ff):
        self.layers = [DecoderBlock(d_model, d_ff) for _ in range(num_layers)]

    def forward(self, x, encoder_output):
        for layer in self.layers:
            x = layer.forward(x, encoder_output)
        return x

# Simulación de datos
np.random.seed(42)
batch_size = 4
seq_len = 6
d_model = 512
d_ff = 2048
num_layers = 2

encoder_input = np.random.rand(batch_size, seq_len, d_model)
decoder_input = np.random.rand(batch_size, seq_len, d_model)

# Inicializamos el codificador y decodificador
encoder = Encoder(num_layers, d_model, d_ff)
decoder = Decoder(num_layers, d_model, d_ff)

# Pasamos los datos por el codificador
encoder_output = encoder.forward(encoder_input)

# Pasamos los datos por el decodificador
decoder_output = decoder.forward(decoder_input, encoder_output)

# Visualización de los resultados
plt.figure(figsize=(10, 6))
plt.plot(encoder_output[0, 0, :50], label="Salida del Codificador (Token 0)", alpha=0.7)
plt.plot(decoder_output[0, 0, :50], label="Salida del Decodificador (Token 0)", alpha=0.7)
plt.title("Codificador y Decodificador: Representaciones Finales", fontsize=16)
plt.xlabel("Dimensiones del Modelo (d_model)")
plt.ylabel("Valor")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Función Softmax
def softmax(logits):
    """
    Calcula el Softmax sobre un vector de logits.
    :param logits: Vector de logits (shape: [vocab_size])
    :return: Probabilidades (shape: [vocab_size])
    """
    exp_logits = np.exp(logits - np.max(logits))  # Evitar problemas numéricos
    return exp_logits / np.sum(exp_logits)

# Simulación de datos
np.random.seed(42)
d_model = 512  # Dimensión del modelo
vocab_size = 10000  # Tamaño del vocabulario

# Salida del decodificador simulada (antes de Softmax)
decoder_output = np.random.rand(d_model)  # Vector de tamaño d_model

# Capa lineal para proyectar al tamaño del vocabulario
projection_weights = np.random.randn(d_model, vocab_size) * 0.1  # Pesos de la capa lineal
projection_bias = np.zeros(vocab_size)  # Sesgo de la capa lineal

# Proyección lineal: Convertir d_model -> vocab_size
logits = np.dot(decoder_output, projection_weights) + projection_bias

# Aplicamos Softmax
probabilities = softmax(logits)

# Predicción de la palabra con mayor probabilidad
predicted_word_idx = np.argmax(probabilities)

# Visualización de las probabilidades
plt.figure(figsize=(10, 6))
plt.plot(probabilities[:50], label="Probabilidades (Primeras 50 Palabras)")
plt.title("Distribución de Probabilidades del Softmax", fontsize=16)
plt.xlabel("Índice de Palabra en el Vocabulario")
plt.ylabel("Probabilidad")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Resultados
print(f"Tamaño del vocabulario: {vocab_size}")
print(f"Índice de la palabra predicha: {predicted_word_idx}")
print(f"Probabilidad de la palabra predicha: {probabilities[predicted_word_idx]:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Función Softmax
def softmax(x, mask=None):
    """
    Calcula el Softmax con opción de aplicar una máscara.
    :param x: Matriz de logits (shape: [seq_len, seq_len])
    :param mask: Máscara binaria (shape: [seq_len, seq_len])
    :return: Pesos de atención (softmax)
    """
    if mask is not None:
        x = x + mask  # Aplicar la máscara
    exp_x = np.exp(x - np.max(x))  # Evitar problemas numéricos
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)

# Máscara de Padding
def create_padding_mask(seq):
    """
    Crea una máscara de padding para ignorar tokens "<PAD>".
    :param seq: Secuencia de tokens (shape: [seq_len])
    :return: Máscara (shape: [seq_len, seq_len])
    """
    mask = (seq == 0).astype(float)  # Tokens "<PAD>" se marcan con 1
    return mask[:, np.newaxis] * -1e9  # Expandimos dimensiones y aplicamos -∞

# Máscara Causal
def create_causal_mask(seq_len):
    """
    Crea una máscara causal para prevenir mirar hacia adelante.
    :param seq_len: Longitud de la secuencia
    :return: Máscara (shape: [seq_len, seq_len])
    """
    return np.triu(np.ones((seq_len, seq_len)) * -1e9, k=1)  # Triángulo superior

# Simulación de una matriz de atención
np.random.seed(42)
seq_len = 6
attention_logits = np.random.rand(seq_len, seq_len)  # Simulación de logits de atención

# Ejemplo 1: Máscara de Padding
seq_with_padding = np.array([1, 2, 3, 0, 0, 0])  # "0" representa "<PAD>"
padding_mask = create_padding_mask(seq_with_padding)
attention_with_padding = softmax(attention_logits, padding_mask)

# Ejemplo 2: Máscara Causal
causal_mask = create_causal_mask(seq_len)
attention_with_causal = softmax(attention_logits, causal_mask)

# Visualización: Máscara de Padding
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.heatmap(attention_with_padding, annot=False, cmap="Blues", cbar=True)
plt.title("Pesos de Atención con Máscara de Padding", fontsize=14)
plt.xlabel("Tokens (Keys)")
plt.ylabel("Tokens (Queries)")

# Visualización: Máscara Causal
plt.subplot(1, 2, 2)
sns.heatmap(attention_with_causal, annot=False, cmap="Blues", cbar=True)
plt.title("Pesos de Atención con Máscara Causal", fontsize=14)
plt.xlabel("Tokens (Keys)")
plt.ylabel("Tokens (Queries)")

plt.tight_layout()
plt.show()

# Mostrar las máscaras originales para comparación
print("Máscara de Padding (antes de Softmax):")
print(padding_mask)

print("\nMáscara Causal (antes de Softmax):")
print(causal_mask)
```

El siguiente paso en la arquitectura Transformer, después de la tokenización, es convertir los tokens en vectores densos mediante un proceso llamado embedding, seguido de la codificación posicional. Esto proporciona al modelo representaciones numéricas ricas de los tokens, junto con información sobre su posición en la secuencia.

Paso 1: Embeddings
Los embeddings son representaciones densas (vectores) de los tokens que capturan relaciones semánticas entre las palabras. En el caso de un Transformer, cada token se convierte en un vector de una dimensión fija, como 64, 128, o 512, dependiendo del modelo.

Código para Embeddings:
Usaremos embeddings aleatorios como ejemplo para ilustrar el proceso.

```python
import numpy as np

# Simulación de vocabulario y tokens tokenizados
tokens = ['hola', 'cómo', 'estás', '?', 'mi', 'nombre', 'es', 'carlos']
vocab_size = 10000  # Tamaño del vocabulario
embedding_dim = 64  # Dimensión de los embeddings

# Crear una tabla de embeddings aleatoria
embedding_table = np.random.rand(vocab_size, embedding_dim)

# Simular un mapeo de tokens a índices (en un Transformer real, esto sería parte del tokenizador)
vocab_dict = {word: i for i, word in enumerate(tokens)}
token_indices = [vocab_dict[token] for token in tokens]

# Obtener los embeddings para los tokens
embedded_tokens = embedding_table[token_indices]

print("Embeddings de los tokens:")
print(embedded_tokens)
```

Paso 2: Codificación Posicional
El Transformer no tiene una estructura secuencial como las RNNs, por lo que necesita añadir información sobre la posición de cada token en la secuencia mediante un Positional Encoding. Esto permite al modelo saber qué token corresponde a qué posición en el texto.

Código para Codificación Posicional:
Usaremos funciones sinusoidales como se describe en el paper "Attention Is All You Need".

```python
def positional_encoding(seq_length, embedding_dim):
    pos_enc = np.zeros((seq_length, embedding_dim))
    for pos in range(seq_length):
        for i in range(0, embedding_dim, 2):
            pos_enc[pos, i] = np.sin(pos / 10000**(i / embedding_dim))
            pos_enc[pos, i + 1] = np.cos(pos / 10000**(i / embedding_dim))
    return pos_enc

# Longitud de la secuencia
seq_length = len(tokens)

# Calcular codificación posicional
pos_enc = positional_encoding(seq_length, embedding_dim)

# Sumar codificación posicional a los embeddings
final_embeddings = embedded_tokens + pos_enc

print("Embeddings con codificación posicional:")
print(final_embeddings)
```

Embeddings:

Convierte cada token en un vector numérico denso.
Representa semántica y relaciones entre palabras.
Codificación Posicional:

Añade información sobre la posición relativa de los tokens en la secuencia.
Esto es fundamental para que el modelo capture el orden de las palabras.

Implementación de un Bloque Completo del Encoder
Un bloque del Encoder en el Transformer combina:

```python
import numpy as np

# Simulación de entrada
seq_length = 6
embedding_dim = 64  # Dimensión de los embeddings
num_heads = 8  # Número de cabezas en Multi-Head Attention
hidden_dim = 256  # Dimensión de la capa FFN

# Entrada simulada (embeddings)
embedded_tokens = np.random.rand(seq_length, embedding_dim)  # Entrada del encoder


# Función para dividir en múltiples cabezas
def split_heads(x, num_heads):
    batch_size, seq_length, embedding_dim = x.shape
    head_dim = embedding_dim // num_heads
    x = x.reshape(batch_size, seq_length, num_heads, head_dim)
    return np.transpose(x, (0, 2, 1, 3))  # (batch_size, num_heads, seq_length, head_dim)


# Función de Atención Multi-Cabeza
def multi_head_attention(Q, K, V, num_heads):
    # Dividir Q, K y V en múltiples cabezas
    Q_heads = split_heads(Q, num_heads)
    K_heads = split_heads(K, num_heads)
    V_heads = split_heads(V, num_heads)

    outputs = []
    for i in range(num_heads):
        output, _ = scaled_dot_product_attention(Q_heads[0, i], K_heads[0, i], V_heads[0, i])
        outputs.append(output)

    # Combinar las salidas de todas las cabezas
    multi_head_output = np.concatenate(outputs, axis=-1)
    return multi_head_output


# Escalado Dot-Product Attention
def scaled_dot_product_attention(Q, K, V):
    d_k = K.shape[1]
    scores = np.dot(Q, K.T) / np.sqrt(d_k)
    scores -= np.max(scores, axis=1, keepdims=True)  # Evitar overflow
    weights = np.exp(scores) / np.sum(np.exp(scores), axis=1, keepdims=True)
    output = np.dot(weights, V)
    return output, weights


# Layer Normalization
def layer_normalization(x, epsilon=1e-6):
    mean = np.mean(x, axis=-1, keepdims=True)
    std = np.std(x, axis=-1, keepdims=True)
    gamma = np.ones_like(std)  # Escala
    beta = np.zeros_like(mean)  # Desplazamiento
    norm_x = (x - mean) / (std + epsilon)
    return gamma * norm_x + beta


# Feed-Forward Network (FFN)
def feed_forward(x, hidden_dim=256):
    input_dim = x.shape[-1]
    W1 = np.random.rand(input_dim, hidden_dim)
    b1 = np.random.rand(hidden_dim)
    W2 = np.random.rand(hidden_dim, input_dim)
    b2 = np.random.rand(input_dim)
    hidden = np.maximum(0, np.dot(x, W1) + b1)
    output = np.dot(hidden, W2) + b2
    return output


# Bloque del Encoder
def encoder_block(x, num_heads, hidden_dim):
    # Pesos aleatorios para Q, K, V
    W_Q = np.random.rand(embedding_dim, embedding_dim)
    W_K = np.random.rand(embedding_dim, embedding_dim)
    W_V = np.random.rand(embedding_dim, embedding_dim)

    # Calcular Q, K, V
    Q = np.dot(x, W_Q)
    K = np.dot(x, W_K)
    V = np.dot(x, W_V)

    # Atención Multi-Cabeza
    attention_output = multi_head_attention(Q[np.newaxis, :], K[np.newaxis, :], V[np.newaxis, :], num_heads)

    # Conexión Residual + Normalización (después de Atención)
    residual_attention_output = x + attention_output[0]
    normalized_attention_output = layer_normalization(residual_attention_output)

    # Capa Feed-Forward
    ffn_output = feed_forward(normalized_attention_output, hidden_dim)

    # Conexión Residual + Normalización (después de FFN)
    residual_ffn_output = normalized_attention_output + ffn_output
    final_output = layer_normalization(residual_ffn_output)

    return final_output


# Aplicar un bloque del encoder
encoder_output = encoder_block(embedded_tokens, num_heads, hidden_dim)

print("Salida del bloque del Encoder:")
print(encoder_output)
```

```python
import numpy as np

# Parámetros globales
seq_length = 6
embedding_dim = 64  # Dimensión de los embeddings
num_heads = 8  # Número de cabezas en Multi-Head Attention
hidden_dim = 256  # Dimensión de la capa FFN
num_layers = 6  # Número de bloques en el encoder


# Función de Atención Multi-Cabeza
def scaled_dot_product_attention(Q, K, V):
    d_k = K.shape[1]
    scores = np.dot(Q, K.T) / np.sqrt(d_k)
    scores -= np.max(scores, axis=1, keepdims=True)  # Evitar overflow
    weights = np.exp(scores) / np.sum(np.exp(scores), axis=1, keepdims=True)
    output = np.dot(weights, V)
    return output, weights


def multi_head_attention(Q, K, V, num_heads):
    def split_heads(x, num_heads):
        batch_size, seq_length, embedding_dim = x.shape
        head_dim = embedding_dim // num_heads
        x = x.reshape(batch_size, seq_length, num_heads, head_dim)
        return np.transpose(x, (0, 2, 1, 3))  # (batch_size, num_heads, seq_length, head_dim)

    # Dividir Q, K y V en múltiples cabezas
    Q_heads = split_heads(Q, num_heads)
    K_heads = split_heads(K, num_heads)
    V_heads = split_heads(V, num_heads)

    outputs = []
    for i in range(num_heads):
        output, _ = scaled_dot_product_attention(Q_heads[0, i], K_heads[0, i], V_heads[0, i])
        outputs.append(output)

    # Combinar las salidas de todas las cabezas
    multi_head_output = np.concatenate(outputs, axis=-1)
    return multi_head_output


# Normalización por capa
def layer_normalization(x, epsilon=1e-6):
    mean = np.mean(x, axis=-1, keepdims=True)
    std = np.std(x, axis=-1, keepdims=True)
    gamma = np.ones_like(std)  # Escala
    beta = np.zeros_like(mean)  # Desplazamiento
    norm_x = (x - mean) / (std + epsilon)
    return gamma * norm_x + beta


# Capa Feed-Forward (FFN)
def feed_forward(x, hidden_dim=256):
    input_dim = x.shape[-1]
    W1 = np.random.rand(input_dim, hidden_dim)
    b1 = np.random.rand(hidden_dim)
    W2 = np.random.rand(hidden_dim, input_dim)
    b2 = np.random.rand(input_dim)
    hidden = np.maximum(0, np.dot(x, W1) + b1)
    output = np.dot(hidden, W2) + b2
    return output


# Bloque del Encoder
def encoder_block(x, num_heads, hidden_dim):
    # Pesos aleatorios para Q, K, V
    W_Q = np.random.rand(embedding_dim, embedding_dim)
    W_K = np.random.rand(embedding_dim, embedding_dim)
    W_V = np.random.rand(embedding_dim, embedding_dim)

    # Calcular Q, K, V
    Q = np.dot(x, W_Q)
    K = np.dot(x, W_K)
    V = np.dot(x, W_V)

    # Atención Multi-Cabeza
    attention_output = multi_head_attention(Q[np.newaxis, :], K[np.newaxis, :], V[np.newaxis, :], num_heads)

    # Conexión Residual + Normalización (después de Atención)
    residual_attention_output = x + attention_output[0]
    normalized_attention_output = layer_normalization(residual_attention_output)

    # Capa Feed-Forward
    ffn_output = feed_forward(normalized_attention_output, hidden_dim)

    # Conexión Residual + Normalización (después de FFN)
    residual_ffn_output = normalized_attention_output + ffn_output
    final_output = layer_normalization(residual_ffn_output)

    return final_output


# Encoder Completo
def encoder(x, num_layers, num_heads, hidden_dim):
    for i in range(num_layers):
        x = encoder_block(x, num_heads, hidden_dim)  # Apilar bloques del encoder
    return x


# Entrada simulada (embeddings)
embedded_tokens = np.random.rand(seq_length, embedding_dim)  # Entrada inicial del encoder

# Aplicar el encoder completo
encoder_output = encoder(embedded_tokens, num_layers, num_heads, hidden_dim)

print("Salida final del Encoder completo:")
print(encoder_output)

```

El Decoder en un Transformer sigue una arquitectura similar al Encoder, pero incluye dos diferencias clave:

Masked Multi-Head Attention: Restringe la atención solo a tokens anteriores, para garantizar que el modelo no vea el futuro durante la generación de texto.
Cross-Attention: Combina la representación del encoder con la entrada parcial del decoder para generar la salida.
Arquitectura del Decoder
Un bloque del Decoder incluye:

Masked Multi-Head Attention: Atención sobre la entrada parcial del decoder.
Cross-Attention: Atención sobre la salida del encoder (contexto).
Capa Feed-Forward: Similar al encoder.
Conexión Residual + Normalización: Después de cada paso.


```python
import numpy as np

# Parámetros globales
seq_length = 6
embedding_dim = 64
num_heads = 8
hidden_dim = 256
num_layers = 6


# Atención Escalada Dot-Product
def scaled_dot_product_attention(Q, K, V, mask=None):
    d_k = K.shape[1]
    scores = np.dot(Q, K.T) / np.sqrt(d_k)

    # Aplicar máscara si se proporciona
    if mask is not None:
        scores += mask  # Máscara con valores muy negativos para ignorar ciertas posiciones

    # Softmax para obtener pesos
    scores -= np.max(scores, axis=-1, keepdims=True)  # Evitar overflow
    weights = np.exp(scores) / np.sum(np.exp(scores), axis=-1, keepdims=True)
    output = np.dot(weights, V)
    return output, weights


# Multi-Head Attention
def multi_head_attention(Q, K, V, num_heads, mask=None):
    def split_heads(x, num_heads):
        batch_size, seq_length, embedding_dim = x.shape
        head_dim = embedding_dim // num_heads
        x = x.reshape(batch_size, seq_length, num_heads, head_dim)
        return np.transpose(x, (0, 2, 1, 3))  # (batch_size, num_heads, seq_length, head_dim)

    # Dividir Q, K y V en múltiples cabezas
    Q_heads = split_heads(Q, num_heads)
    K_heads = split_heads(K, num_heads)
    V_heads = split_heads(V, num_heads)

    outputs = []
    for i in range(num_heads):
        output, _ = scaled_dot_product_attention(Q_heads[0, i], K_heads[0, i], V_heads[0, i], mask)
        outputs.append(output)

    # Combinar salidas de todas las cabezas
    multi_head_output = np.concatenate(outputs, axis=-1)
    return multi_head_output


# Normalización por capa
def layer_normalization(x, epsilon=1e-6):
    mean = np.mean(x, axis=-1, keepdims=True)
    std = np.std(x, axis=-1, keepdims=True)
    gamma = np.ones_like(std)
    beta = np.zeros_like(mean)
    norm_x = (x - mean) / (std + epsilon)
    return gamma * norm_x + beta


# Capa Feed-Forward
def feed_forward(x, hidden_dim=256):
    input_dim = x.shape[-1]
    W1 = np.random.rand(input_dim, hidden_dim)
    b1 = np.random.rand(hidden_dim)
    W2 = np.random.rand(hidden_dim, input_dim)
    b2 = np.random.rand(input_dim)
    hidden = np.maximum(0, np.dot(x, W1) + b1)
    output = np.dot(hidden, W2) + b2
    return output


# Bloque del Decoder
def decoder_block(x, encoder_output, num_heads, hidden_dim, mask):
    # Pesos aleatorios para Q, K, V
    W_Q = np.random.rand(embedding_dim, embedding_dim)
    W_K = np.random.rand(embedding_dim, embedding_dim)
    W_V = np.random.rand(embedding_dim, embedding_dim)

    # 1. Masked Multi-Head Attention
    Q = np.dot(x, W_Q)
    K = np.dot(x, W_K)
    V = np.dot(x, W_V)
    masked_attention_output = multi_head_attention(Q[np.newaxis, :], K[np.newaxis, :], V[np.newaxis, :], num_heads, mask)

    # Conexión Residual + Normalización
    residual_masked_attention = x + masked_attention_output[0]
    normalized_masked_attention = layer_normalization(residual_masked_attention)

    # 2. Cross-Attention
    Q_cross = np.dot(normalized_masked_attention, W_Q)
    K_cross = np.dot(encoder_output, W_K)
    V_cross = np.dot(encoder_output, W_V)
    cross_attention_output = multi_head_attention(Q_cross[np.newaxis, :], K_cross[np.newaxis, :], V_cross[np.newaxis, :], num_heads)

    # Conexión Residual + Normalización
    residual_cross_attention = normalized_masked_attention + cross_attention_output[0]
    normalized_cross_attention = layer_normalization(residual_cross_attention)

    # 3. Feed-Forward
    ffn_output = feed_forward(normalized_cross_attention, hidden_dim)

    # Conexión Residual + Normalización
    residual_ffn_output = normalized_cross_attention + ffn_output
    final_output = layer_normalization(residual_ffn_output)

    return final_output


# Decoder Completo
def decoder(x, encoder_output, num_layers, num_heads, hidden_dim):
    # Crear una máscara triangular inferior para la atención enmascarada
    mask = np.triu(np.ones((seq_length, seq_length)) * -np.inf, k=1)

    for i in range(num_layers):
        x = decoder_block(x, encoder_output, num_heads, hidden_dim, mask)
    return x


# Entrada simulada (embeddings)
decoder_input = np.random.rand(seq_length, embedding_dim)  # Entrada inicial del decoder
encoder_output = np.random.rand(seq_length, embedding_dim)  # Salida simulada del encoder

# Aplicar el decoder completo
decoder_output = decoder(decoder_input, encoder_output, num_layers, num_heads, hidden_dim)

print("Salida final del Decoder completo:")
print(decoder_output)
```

Explicación del Flujo
Masked Multi-Head Attention:

Atención enmascarada para garantizar que el modelo solo "vea" tokens previos.
Utiliza una máscara triangular inferior.
Cross-Attention:

Combina la salida del encoder con la entrada parcial del decoder.
Permite al decoder usar el contexto codificado por el encoder.
Feed-Forward Network:

Procesa las representaciones enriquecidas para extraer características avanzadas.
Conexiones Residuales + Normalización:

Después de cada paso (Masked Attention, Cross-Attention, FFN).
Salida del Decoder:

Una representación enriquecida de la secuencia de salida.

```python
import numpy as np

# Parámetros globales
seq_length = 6
embedding_dim = 64
num_heads = 8
hidden_dim = 256
num_layers = 6


# Atención Escalada Dot-Product
def scaled_dot_product_attention(Q, K, V, mask=None):
    d_k = K.shape[1]
    scores = np.dot(Q, K.T) / np.sqrt(d_k)
    if mask is not None:
        scores += mask  # Aplicar máscara si se proporciona
    scores -= np.max(scores, axis=-1, keepdims=True)  # Evitar overflow
    weights = np.exp(scores) / np.sum(np.exp(scores), axis=-1, keepdims=True)
    output = np.dot(weights, V)
    return output, weights


# Multi-Head Attention
def multi_head_attention(Q, K, V, num_heads, mask=None):
    def split_heads(x, num_heads):
        batch_size, seq_length, embedding_dim = x.shape
        head_dim = embedding_dim // num_heads
        x = x.reshape(batch_size, seq_length, num_heads, head_dim)
        return np.transpose(x, (0, 2, 1, 3))  # (batch_size, num_heads, seq_length, head_dim)

    Q_heads = split_heads(Q, num_heads)
    K_heads = split_heads(K, num_heads)
    V_heads = split_heads(V, num_heads)

    outputs = []
    for i in range(num_heads):
        output, _ = scaled_dot_product_attention(Q_heads[0, i], K_heads[0, i], V_heads[0, i], mask)
        outputs.append(output)

    multi_head_output = np.concatenate(outputs, axis=-1)
    return multi_head_output


# Normalización por capa
def layer_normalization(x, epsilon=1e-6):
    mean = np.mean(x, axis=-1, keepdims=True)
    std = np.std(x, axis=-1, keepdims=True)
    gamma = np.ones_like(std)
    beta = np.zeros_like(mean)
    norm_x = (x - mean) / (std + epsilon)
    return gamma * norm_x + beta


# Capa Feed-Forward
def feed_forward(x, hidden_dim=256):
    input_dim = x.shape[-1]
    W1 = np.random.rand(input_dim, hidden_dim)
    b1 = np.random.rand(hidden_dim)
    W2 = np.random.rand(hidden_dim, input_dim)
    b2 = np.random.rand(input_dim)
    hidden = np.maximum(0, np.dot(x, W1) + b1)
    output = np.dot(hidden, W2) + b2
    return output


# Bloque del Encoder
def encoder_block(x, num_heads, hidden_dim):
    W_Q = np.random.rand(embedding_dim, embedding_dim)
    W_K = np.random.rand(embedding_dim, embedding_dim)
    W_V = np.random.rand(embedding_dim, embedding_dim)

    Q = np.dot(x, W_Q)
    K = np.dot(x, W_K)
    V = np.dot(x, W_V)

    attention_output = multi_head_attention(Q[np.newaxis, :], K[np.newaxis, :], V[np.newaxis, :], num_heads)

    residual_attention_output = x + attention_output[0]
    normalized_attention_output = layer_normalization(residual_attention_output)

    ffn_output = feed_forward(normalized_attention_output, hidden_dim)

    residual_ffn_output = normalized_attention_output + ffn_output
    final_output = layer_normalization(residual_ffn_output)

    return final_output


# Encoder Completo
def encoder(x, num_layers, num_heads, hidden_dim):
    for i in range(num_layers):
        x = encoder_block(x, num_heads, hidden_dim)
    return x


# Bloque del Decoder
def decoder_block(x, encoder_output, num_heads, hidden_dim, mask):
    W_Q = np.random.rand(embedding_dim, embedding_dim)
    W_K = np.random.rand(embedding_dim, embedding_dim)
    W_V = np.random.rand(embedding_dim, embedding_dim)

    Q = np.dot(x, W_Q)
    K = np.dot(x, W_K)
    V = np.dot(x, W_V)
    masked_attention_output = multi_head_attention(Q[np.newaxis, :], K[np.newaxis, :], V[np.newaxis, :], num_heads, mask)

    residual_masked_attention = x + masked_attention_output[0]
    normalized_masked_attention = layer_normalization(residual_masked_attention)

    Q_cross = np.dot(normalized_masked_attention, W_Q)
    K_cross = np.dot(encoder_output, W_K)
    V_cross = np.dot(encoder_output, W_V)
    cross_attention_output = multi_head_attention(Q_cross[np.newaxis, :], K_cross[np.newaxis, :], V_cross[np.newaxis, :], num_heads)

    residual_cross_attention = normalized_masked_attention + cross_attention_output[0]
    normalized_cross_attention = layer_normalization(residual_cross_attention)

    ffn_output = feed_forward(normalized_cross_attention, hidden_dim)

    residual_ffn_output = normalized_cross_attention + ffn_output
    final_output = layer_normalization(residual_ffn_output)

    return final_output


# Decoder Completo
def decoder(x, encoder_output, num_layers, num_heads, hidden_dim):
    mask = np.triu(np.ones((seq_length, seq_length)) * -np.inf, k=1)
    for i in range(num_layers):
        x = decoder_block(x, encoder_output, num_heads, hidden_dim, mask)
    return x


# Transformer Completo
def transformer(encoder_input, decoder_input, num_layers, num_heads, hidden_dim):
    # Codificar la entrada con el encoder
    encoder_output = encoder(encoder_input, num_layers, num_heads, hidden_dim)

    # Decodificar la entrada parcial del decoder usando la salida del encoder
    decoder_output = decoder(decoder_input, encoder_output, num_layers, num_heads, hidden_dim)

    return decoder_output


# Entradas simuladas
encoder_input = np.random.rand(seq_length, embedding_dim)  # Entrada del encoder
decoder_input = np.random.rand(seq_length, embedding_dim)  # Entrada inicial del decoder

# Aplicar el Transformer completo
transformer_output = transformer(encoder_input, decoder_input, num_layers, num_heads, hidden_dim)

print("Salida final del Transformer completo:")
print(transformer_output)
```

Explicación del Flujo
Encoder:

Procesa la entrada para generar una representación rica del contexto.
Decoder:

Utiliza la salida del encoder como contexto.
Genera una secuencia de salida basada en la entrada parcial del decoder.
Transformer Completo:

Integra encoder y decoder para un flujo completo de entrada y salida.

Incorporación de Softmax
El último paso en un Transformer es transformar la salida del decoder en probabilidades sobre un vocabulario mediante la función Softmax. Esto es crucial en tareas como traducción automática, donde necesitamos seleccionar el token más probable en cada paso.

Implementación de Softmax
A continuación, agregamos la capa de salida que toma la salida del decoder y la proyecta al espacio del vocabulario, seguida de la aplicación de Softmax.

```python
import numpy as np

# Función Softmax
def softmax(x):
    # Restar el máximo para evitar overflow
    x -= np.max(x, axis=-1, keepdims=True)
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)


# Tamaño del vocabulario
vocab_size = 10000

# Proyección al vocabulario
def project_to_vocab(decoder_output, vocab_size):
    # Pesos y sesgos aleatorios para la proyección (simulados aquí)
    W_vocab = np.random.rand(decoder_output.shape[-1], vocab_size)
    b_vocab = np.random.rand(vocab_size)

    # Transformar la salida del decoder al espacio del vocabulario
    logits = np.dot(decoder_output, W_vocab) + b_vocab
    return logits


# Proyección al vocabulario y Softmax
def transformer_with_softmax(encoder_input, decoder_input, num_layers, num_heads, hidden_dim, vocab_size):
    # Obtener la salida del Transformer completo
    decoder_output = transformer(encoder_input, decoder_input, num_layers, num_heads, hidden_dim)

    # Proyectar la salida del decoder al espacio del vocabulario
    logits = project_to_vocab(decoder_output, vocab_size)

    # Aplicar Softmax para obtener probabilidades
    probabilities = softmax(logits)
    return probabilities


# Entradas simuladas
encoder_input = np.random.rand(seq_length, embedding_dim)  # Entrada del encoder
decoder_input = np.random.rand(seq_length, embedding_dim)  # Entrada inicial del decoder

# Aplicar el Transformer completo con Softmax
output_probabilities = transformer_with_softmax(encoder_input, decoder_input, num_layers, num_heads, hidden_dim, vocab_size)

print("Probabilidades de la salida del Transformer:")
print(output_probabilities)
```

Decodificación: Convertir Probabilidades a Tokens
El paso de decodificación convierte las probabilidades generadas por el Transformer en una secuencia de tokens del vocabulario. Esto se hace seleccionando el token más probable en cada posición de la secuencia o mediante métodos más avanzados como beam search o sampling.

Implementación Básica: Decodificación Greedy
Usaremos una decodificación "greedy" (codiciosa), que selecciona el token con la probabilidad más alta en cada posición.

```python
# Diccionario de vocabulario (simulado)
vocab_dict = {i: f"token_{i}" for i in range(10000)}  # Ejemplo de vocabulario
reverse_vocab_dict = {v: k for k, v in vocab_dict.items()}  # Para traducción inversa

# Decodificación Greedy
def decode_greedy(probabilities, vocab_dict):
    """
    Toma las probabilidades y decodifica la secuencia de salida.
    Args:
        probabilities: Matriz de probabilidades (seq_length x vocab_size)
        vocab_dict: Diccionario que mapea índices a palabras.
    Returns:
        Lista de tokens decodificados.
    """
    # Obtener el índice del token con mayor probabilidad en cada posición
    token_indices = np.argmax(probabilities, axis=-1)

    # Mapear los índices a los tokens del vocabulario
    decoded_tokens = [vocab_dict[idx] for idx in token_indices]
    return decoded_tokens


# Entradas simuladas
encoder_input = np.random.rand(seq_length, embedding_dim)  # Entrada del encoder
decoder_input = np.random.rand(seq_length, embedding_dim)  # Entrada inicial del decoder

# Probabilidades generadas por el Transformer
output_probabilities = transformer_with_softmax(encoder_input, decoder_input, num_layers, num_heads, hidden_dim, vocab_size)

# Decodificar la salida a tokens
decoded_tokens = decode_greedy(output_probabilities, vocab_dict)

print("Tokens decodificados:")
print(decoded_tokens)
```

```python
import numpy as np

# Parámetros
seq_length = 6  # Longitud de la secuencia
embedding_dim = 64  # Dimensión de los embeddings
num_heads = 8  # Número de cabezas en atención multi-cabeza
hidden_dim = 256  # Dimensión de la red Feed-Forward
num_layers = 6  # Número de capas (bloques del encoder y decoder)
vocab_size = 100  # Tamaño del vocabulario
temperature = 1.0  # Temperatura para softmax

# Diccionario de vocabulario simulado
vocab_dict = {i: f"token_{i}" for i in range(vocab_size)}
reverse_vocab_dict = {v: k for k, v in vocab_dict.items()}


# --- Funciones de Transformador ---
def scaled_dot_product_attention(Q, K, V, mask=None):
    d_k = K.shape[1]
    scores = np.dot(Q, K.T) / np.sqrt(d_k)
    if mask is not None:
        scores += mask
    scores -= np.max(scores, axis=-1, keepdims=True)  # Evitar overflow
    weights = np.exp(scores) / np.sum(np.exp(scores), axis=-1, keepdims=True)
    output = np.dot(weights, V)
    return output, weights


def multi_head_attention(Q, K, V, num_heads, mask=None):
    def split_heads(x, num_heads):
        seq_length, embedding_dim = x.shape
        head_dim = embedding_dim // num_heads
        x = x.reshape(seq_length, num_heads, head_dim)
        return np.transpose(x, (1, 0, 2))  # (num_heads, seq_length, head_dim)

    Q_heads = split_heads(Q, num_heads)
    K_heads = split_heads(K, num_heads)
    V_heads = split_heads(V, num_heads)

    outputs = []
    for i in range(num_heads):
        output, _ = scaled_dot_product_attention(Q_heads[i], K_heads[i], V_heads[i], mask)
        outputs.append(output)

    multi_head_output = np.concatenate(outputs, axis=-1)  # (seq_length, embedding_dim)
    return multi_head_output


def layer_normalization(x, epsilon=1e-6):
    mean = np.mean(x, axis=-1, keepdims=True)
    std = np.std(x, axis=-1, keepdims=True)
    gamma = np.ones_like(std)
    beta = np.zeros_like(mean)
    return gamma * (x - mean) / (std + epsilon) + beta


def feed_forward(x, hidden_dim):
    input_dim = x.shape[-1]
    W1 = np.random.rand(input_dim, hidden_dim)
    b1 = np.random.rand(hidden_dim)
    W2 = np.random.rand(hidden_dim, input_dim)
    b2 = np.random.rand(input_dim)
    hidden = np.maximum(0, np.dot(x, W1) + b1)  # ReLU
    output = np.dot(hidden, W2) + b2
    return output


def encoder_block(x, num_heads, hidden_dim):
    W_Q = np.random.rand(embedding_dim, embedding_dim)
    W_K = np.random.rand(embedding_dim, embedding_dim)
    W_V = np.random.rand(embedding_dim, embedding_dim)

    Q = np.dot(x, W_Q)
    K = np.dot(x, W_K)
    V = np.dot(x, W_V)

    attention_output = multi_head_attention(Q, K, V, num_heads)
    residual_attention_output = x + attention_output
    normalized_attention_output = layer_normalization(residual_attention_output)

    ffn_output = feed_forward(normalized_attention_output, hidden_dim)
    residual_ffn_output = normalized_attention_output + ffn_output
    final_output = layer_normalization(residual_ffn_output)

    return final_output


def encoder(x, num_layers, num_heads, hidden_dim):
    for i in range(num_layers):
        x = encoder_block(x, num_heads, hidden_dim)
    return x


def decoder_block(x, encoder_output, num_heads, hidden_dim, mask):
    W_Q = np.random.rand(embedding_dim, embedding_dim)
    W_K = np.random.rand(embedding_dim, embedding_dim)
    W_V = np.random.rand(embedding_dim, embedding_dim)

    Q = np.dot(x, W_Q)
    K = np.dot(x, W_K)
    V = np.dot(x, W_V)

    masked_attention_output = multi_head_attention(Q, K, V, num_heads, mask)
    residual_masked_attention = x + masked_attention_output
    normalized_masked_attention = layer_normalization(residual_masked_attention)

    Q_cross = np.dot(normalized_masked_attention, W_Q)
    K_cross = np.dot(encoder_output, W_K)
    V_cross = np.dot(encoder_output, W_V)
    cross_attention_output = multi_head_attention(Q_cross, K_cross, V_cross, num_heads)

    residual_cross_attention = normalized_masked_attention + cross_attention_output
    normalized_cross_attention = layer_normalization(residual_cross_attention)

    ffn_output = feed_forward(normalized_cross_attention, hidden_dim)
    residual_ffn_output = normalized_cross_attention + ffn_output
    final_output = layer_normalization(residual_ffn_output)

    return final_output


def decoder(x, encoder_output, num_layers, num_heads, hidden_dim):
    mask = np.triu(np.ones((seq_length, seq_length)) * -np.inf, k=1)
    for i in range(num_layers):
        x = decoder_block(x, encoder_output, num_heads, hidden_dim, mask)
    return x


def project_to_vocab(decoder_output, vocab_size):
    W_vocab = np.random.randn(decoder_output.shape[-1], vocab_size) * 0.1
    b_vocab = np.random.randn(vocab_size) * 0.1
    logits = np.dot(decoder_output, W_vocab) + b_vocab
    return logits


def softmax(x, temperature=1.0):
    x = x / temperature
    x -= np.max(x, axis=-1, keepdims=True)
    exp_x = np.exp(x)
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)


def transformer_with_softmax_and_decoding(encoder_input, decoder_input, num_layers, num_heads, hidden_dim, vocab_size):
    encoder_output = encoder(encoder_input, num_layers, num_heads, hidden_dim)
    decoder_output = decoder(decoder_input, encoder_output, num_layers, num_heads, hidden_dim)

    logits = project_to_vocab(decoder_output, vocab_size)
    probabilities = softmax(logits)

    # Decodificación greedy
    token_indices = np.argmax(probabilities, axis=-1)
    decoded_tokens = [vocab_dict[idx] for idx in token_indices]
    return decoded_tokens


# Simular entradas
encoder_input = np.random.rand(seq_length, embedding_dim)
decoder_input = np.random.rand(seq_length, embedding_dim)

# Aplicar el modelo completo
decoded_tokens = transformer_with_softmax_and_decoding(encoder_input, decoder_input, num_layers, num_heads, hidden_dim, vocab_size)

print("Tokens decodificados:")
print(decoded_tokens)
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import ipywidgets as widgets
from IPython.display import display, Markdown

# Cargar modelos preentrenados para diferentes idiomas
models = {
    "English": "Helsinki-NLP/opus-mt-es-en",
    "French": "Helsinki-NLP/opus-mt-es-fr",
    "German": "Helsinki-NLP/opus-mt-es-de",
    "Italian": "Helsinki-NLP/opus-mt-es-it"
}

# Widgets interactivos
input_box = widgets.Text(
    placeholder='Escribe una frase en español...',
    description='Frase:',
    layout=widgets.Layout(width='100%')
)

language_dropdown = widgets.Dropdown(
    options=models.keys(),
    description='Idioma:',
    value="English"
)

translate_button = widgets.Button(
    description="Traducir",
    button_style='success'
)

output_text = widgets.Output()

# Función para traducir
def translate_text(_):
    with output_text:
        output_text.clear_output()
        input_text = input_box.value
        target_language = language_dropdown.value

        # Cargar modelo y tokenizador basado en la selección del idioma
        model_name = models[target_language]
        tokenizer = AutoTokenizer.from_pretrained(model_name)
        model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

        # Tokenizar y generar traducción
        input_ids = tokenizer.encode(input_text, return_tensors="pt")
        outputs = model.generate(input_ids, max_length=50, num_beams=5, early_stopping=True)
        translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

        # Mostrar resultado
        display(Markdown(f"**Texto traducido a {target_language}:** {translated_text}"))

# Conectar el botón con la función
translate_button.on_click(translate_text)

# Mostrar los widgets
display(input_box, language_dropdown, translate_button, output_text)
```
