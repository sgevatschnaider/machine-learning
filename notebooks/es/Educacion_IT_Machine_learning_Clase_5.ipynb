```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generar datos simples (2 características, 2 clases)
X = np.array([
    [0.5, 1.5],  # Clase 0
    [1.0, 1.0],  # Clase 0
    [1.5, 0.5],  # Clase 1
    [2.0, 1.0],  # Clase 1
    [2.5, 1.5],  # Clase 1
])
y = np.array([0, 0, 1, 1, 1])  # Etiquetas

# 2. Inicializar pesos y bias
weights = np.zeros(X.shape[1])  # Vector de pesos, inicializado en 0
bias = 0.0  # Bias inicial
learning_rate = 0.1  # Tasa de aprendizaje
epochs = 1000  # Número de iteraciones
m = len(y)  # Número de ejemplos

# 3. Definir la función sigmoide
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# 4. Implementar la función de costo (log-loss)
def compute_cost(X, y, weights, bias):
    z = np.dot(X, weights) + bias  # Calcular z
    predictions = sigmoid(z)  # Calcular predicciones
    cost = -(1 / m) * np.sum(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))
    return cost

# 5. Implementar el gradiente descendente con prints detallados
costs = []  # Almacenar el costo en cada iteración
print("\nEntrenamiento del modelo:")
print(f"{'Época':<6} | {'Costo':<10} | {'Pesos':<20} | {'Bias':<8}")
print("-" * 50)

for epoch in range(epochs):
    # Paso 1: Calcular z y predicciones
    z = np.dot(X, weights) + bias
    predictions = sigmoid(z)

    # Paso 2: Calcular gradientes
    dw = (1 / m) * np.dot(X.T, (predictions - y))  # Gradiente de los pesos
    db = (1 / m) * np.sum(predictions - y)        # Gradiente del bias

    # Paso 3: Actualizar pesos y bias
    weights -= learning_rate * dw
    bias -= learning_rate * db

    # Paso 4: Calcular y almacenar costo
    cost = compute_cost(X, y, weights, bias)
    costs.append(cost)

    # Mostrar detalles de cada iteración
    if epoch % 100 == 0 or epoch == epochs - 1:
        print(f"{epoch:<6} | {cost:<10.4f} | {weights} | {bias:<8.4f}")

# Visualizar la reducción del costo
plt.figure(figsize=(8, 6))
plt.plot(range(epochs), costs, color='purple')
plt.title('Reducción del costo durante el entrenamiento')
plt.xlabel('Épocas')
plt.ylabel('Costo')
plt.grid(True)
plt.show()

# 6. Resultado final
print("\nPesos finales:", weights)
print("Bias final:", bias)

# Visualizar la línea de decisión final
def plot_decision_boundary(X, y, weights, bias):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='blue', label='Clase 0')
    plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='red', label='Clase 1')

    # Crear una línea de decisión
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x_values = np.linspace(x_min, x_max, 100)
    y_values = -(weights[0] * x_values + bias) / weights[1]
    plt.plot(x_values, y_values, color='green', label='Línea de decisión')

    plt.title('Línea de decisión final')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_decision_boundary(X, y, weights, bias)

# 7. Predicciones finales
predictions = sigmoid(np.dot(X, weights) + bias)
predicted_classes = (predictions >= 0.5).astype(int)
print("\nPredicciones finales:", predicted_classes)
print("Clases reales:", y)
```

```python
from IPython.display import display, HTML

# Contenido HTML
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Entropía y Entropía Cruzada en Machine Learning</title>
    <link href="https://fonts.googleapis.com/css2?family=Roboto:wght@400;700&display=swap" rel="stylesheet">
    <style>
        :root {
            --primary-color: #4CAF50;
            --secondary-color: #388E3C;
            --text-color: #333;
            --background-color: #f5f5f5;
            --accent-color: #00695C;
            --border-color: #ccc;
        }
        body {
            font-family: 'Roboto', Arial, sans-serif;
            line-height: 1.8;
            color: var(--text-color);
            margin: 0;
            padding: 20px;
            background-color: var(--background-color);
        }
        h1, h2, h3 {
            color: var(--accent-color);
            text-align: left;
        }
        h1 {
            font-size: 2.5rem;
            margin-bottom: 20px;
        }
        p, ul, pre {
            margin: 15px 0;
            line-height: 1.6;
        }
        ul {
            padding-left: 20px;
        }
        li {
            margin-bottom: 10px;
        }
        pre {
            background-color: #eee;
            padding: 10px;
            border-radius: 5px;
            overflow-x: auto;
        }
        code {
            background-color: #f4f4f4;
            padding: 2px 5px;
            border-radius: 3px;
            font-family: monospace;
        }
    </style>
</head>
<body>
    <header>
        <h1>Entropía y Entropía Cruzada en Machine Learning</h1>
    </header>
    <main>
        <h2>1. Entropía (Entropy)</h2>
        <p>La <strong>entropía</strong> mide la incertidumbre o desorden en una distribución de probabilidad. Fue introducida por Claude Shannon y se usa para cuantificar la cantidad de información promedio contenida en un conjunto de datos. En términos simples:</p>
        <ul>
            <li>Alta entropía significa alta incertidumbre (los eventos son muy impredecibles).</li>
            <li>Baja entropía significa baja incertidumbre (los eventos son más predecibles).</li>
        </ul>
        <p>La fórmula de la entropía es:</p>
        <pre><code>H(p) = − ∑ p_i log(p_i)</code></pre>
        <p>Donde:</p>
        <ul>
            <li><strong>p_i</strong>: Probabilidad del evento <em>i</em>.</li>
        </ul>
        <h3>Ejemplo práctico:</h3>
        <ul>
            <li><strong>Moneda justa:</strong> <code>p(cara) = 0.5, p(sello) = 0.5</code>.
                <br>La entropía es alta porque es impredecible si saldrá cara o sello.</li>
            <pre><code>H = − (0.5 log(0.5) + 0.5 log(0.5)) = 1</code></pre>
            <li><strong>Moneda cargada:</strong> <code>p(cara) = 0.9, p(sello) = 0.1</code>.
                <br>La entropía es menor porque es más predecible que saldrá cara.</li>
            <pre><code>H = − (0.9 log(0.9) + 0.1 log(0.1)) ≈ 0.47</code></pre>
        </ul>
        <p>En machine learning, la entropía se usa para medir la pureza de un nodo en árboles de decisión. Si un nodo contiene datos homogéneos, la entropía es baja; si es heterogéneo, la entropía es alta.</p>

        <h2>2. Entropía Cruzada (Cross-Entropy)</h2>
        <p>La <strong>entropía cruzada</strong> mide la diferencia entre dos distribuciones de probabilidad: una distribución "real" (<em>p</em>) y una distribución "predicha" (<em>q</em>). Se utiliza para evaluar qué tan bien un modelo predice los resultados reales.</p>
        <p>La fórmula es:</p>
        <pre><code>H(p, q) = − ∑ p_i log(q_i)</code></pre>
        <p>Donde:</p>
        <ul>
            <li><strong>p_i:</strong> Distribución real (etiquetas verdaderas, normalmente 1 o 0 en clasificación).</li>
            <li><strong>q_i:</strong> Distribución predicha por el modelo (probabilidades predichas).</li>
        </ul>
        <h3>Ejemplo práctico:</h3>
        <ul>
            <li>Clase verdadera (<code>y = 1</code>) y probabilidades predichas (<code>q = 0.9</code> para <code>y = 1</code> y <code>q = 0.1</code> para <code>y = 0</code>):</li>
            <pre><code>H(p, q) = − (1 ⋅ log(0.9) + 0 ⋅ log(0.1)) = − log(0.9) ≈ 0.105</code></pre>
            <li>Si el modelo predice mal (<code>q = 0.1</code> para <code>y = 1</code>):</li>
            <pre><code>H(p, q) = − (1 ⋅ log(0.1)) = − log(0.1) ≈ 2.302</code></pre>
        </ul>
        <p>La entropía cruzada penaliza las predicciones incorrectas, especialmente cuando el modelo predice una probabilidad baja para el resultado correcto.</p>

        <h2>Aplicaciones en Machine Learning</h2>
        <ul>
            <li><strong>Clasificación binaria:</strong> En problemas como detección de spam, la entropía cruzada es la función de pérdida más común (log-loss).</li>
            <li><strong>Clasificación multiclase:</strong> En problemas como reconocimiento de imágenes, las probabilidades predichas son un vector (<em>q</em>) generado por una función softmax.</li>
            <ul>
                <li><strong>Ejemplo:</strong></li>
                <pre><code>Etiqueta verdadera: p = [1, 0, 0] (la imagen es un gato).</code></pre>
                <pre><code>Predicciones: q = [0.7, 0.2, 0.1] (modelo predice gato con 70% de probabilidad).</code></pre>
            </ul>
            <li><strong>Redes neuronales:</strong> La entropía cruzada se utiliza como función de pérdida para ajustar los pesos y biases durante el entrenamiento.</li>
        </ul>

        <h2>Resumen</h2>
        <ul>
            <li><strong>Entropía:</strong> Mide la incertidumbre en una distribución.</li>
            <li><strong>Entropía cruzada:</strong> Mide la discrepancia entre las distribuciones real y predicha.</li>
            <li><strong>Aplicación principal:</strong> Evaluación del rendimiento de modelos de clasificación en machine learning.</li>
        </ul>
    </main>
</body>
</html>
"""

# Mostrar el HTML en Google Colab
display(HTML(html_content))
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Generar datos simples (2 características, 2 clases)
X = np.array([
    [0.5, 1.5],  # Clase 0
    [1.0, 1.0],  # Clase 0
    [1.5, 0.5],  # Clase 1
    [2.0, 1.0],  # Clase 1
    [2.5, 1.5],  # Clase 1
])
y = np.array([0, 0, 1, 1, 1])  # Etiquetas

# Visualizar los datos originales
plt.figure(figsize=(8, 6))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='blue', label='Clase 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='red', label='Clase 1')
plt.title('Datos originales')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# 2. Inicializar pesos y bias
weights = np.zeros(X.shape[1])  # Vector de pesos, inicializado en 0
bias = 0.0  # Bias inicial
learning_rate = 0.1  # Tasa de aprendizaje
epochs = 1000  # Número de iteraciones
m = len(y)  # Número de ejemplos

# 3. Definir la función sigmoide
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# 4. Implementar la función de costo (log-loss)
def compute_cost(X, y, weights, bias):
    z = np.dot(X, weights) + bias  # Calcular z
    predictions = sigmoid(z)  # Calcular predicciones
    cost = -(1 / m) * np.sum(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))
    return cost

# Función para graficar la línea de decisión durante el entrenamiento
def plot_decision_boundary(X, y, weights, bias, epoch=None):
    plt.figure(figsize=(8, 6))
    plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], color='blue', label='Clase 0')
    plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], color='red', label='Clase 1')

    # Crear una línea de decisión
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    x_values = np.linspace(x_min, x_max, 100)
    y_values = -(weights[0] * x_values + bias) / weights[1]
    plt.plot(x_values, y_values, color='green', label='Línea de decisión')

    plt.title(f'Línea de decisión (Época {epoch})' if epoch else 'Línea de decisión final')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.grid(True)
    plt.show()

# 5. Implementar el gradiente descendente con prints detallados
costs = []  # Almacenar el costo en cada iteración
print("\nEntrenamiento del modelo:")
print(f"{'Época':<6} | {'Costo':<10} | {'Pesos':<25} | {'Bias':<8}")
print("-" * 60)

for epoch in range(epochs):
    # Paso 1: Calcular z y predicciones
    z = np.dot(X, weights) + bias
    predictions = sigmoid(z)

    # Paso 2: Calcular gradientes
    dw = (1 / m) * np.dot(X.T, (predictions - y))  # Gradiente de los pesos
    db = (1 / m) * np.sum(predictions - y)        # Gradiente del bias

    # Paso 3: Actualizar pesos y bias
    weights -= learning_rate * dw
    bias -= learning_rate * db

    # Paso 4: Calcular y almacenar costo
    cost = compute_cost(X, y, weights, bias)
    costs.append(cost)

    # Mostrar detalles de cada iteración
    if epoch % 100 == 0 or epoch == epochs - 1:
        print(f"{epoch:<6} | {cost:<10.4f} | {weights} | {bias:<8.4f}")

    # Visualizar línea de decisión cada 250 iteraciones
    if epoch % 250 == 0:
        plot_decision_boundary(X, y, weights, bias, epoch)

# Visualizar la reducción del costo
plt.figure(figsize=(8, 6))
plt.plot(range(epochs), costs, color='purple')
plt.title('Reducción del costo durante el entrenamiento')
plt.xlabel('Épocas')
plt.ylabel('Costo')
plt.grid(True)
plt.show()

# 6. Resultado final
print("\nPesos finales:", weights)
print("Bias final:", bias)

# Visualizar la línea de decisión final
plot_decision_boundary(X, y, weights, bias)

# 7. Predicciones finales
predictions = sigmoid(np.dot(X, weights) + bias)
predicted_classes = (predictions >= 0.5).astype(int)
print("\nPredicciones finales:", predicted_classes)
print("Clases reales:", y)

# Visualizar las predicciones en el espacio de características
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=predicted_classes, cmap='coolwarm', label='Clases predichas')
plt.title('Predicciones finales del modelo')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ---------------------------
# CLASE: REGRESIÓN LOGÍSTICA CON SCALE MODIFICADO
# ---------------------------

# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
)

# 1. Generación de datos simulados (clasificación binaria) con desviación estándar mayor
np.random.seed(42)  # Para reproducibilidad

# Generamos datos para dos clases con mayor desviación estándar
X_class1 = np.random.normal(loc=[2, 2], scale=3.0, size=(50, 2))  # Clase 1
X_class2 = np.random.normal(loc=[5, 5], scale=3.0, size=(50, 2))  # Clase 2
X = np.vstack((X_class1, X_class2))  # Combinar las dos clases
y = np.array([0] * 50 + [1] * 50)  # Etiquetas: 0 para clase 1, 1 para clase 2

# Visualización de los datos con mayor dispersión
plt.figure(figsize=(8, 6))
plt.scatter(X[:50, 0], X[:50, 1], color='blue', label='Clase 0')
plt.scatter(X[50:, 0], X[50:, 1], color='red', label='Clase 1')
plt.title('Datos Generados con Mayor Desviación Estándar (scale=2.0)')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# 2. División del conjunto de datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Visualización de la división
plt.figure(figsize=(8, 6))
plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], color='blue', label='Clase 0 (Train)')
plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], color='red', label='Clase 1 (Train)')
plt.scatter(X_test[y_test == 0][:, 0], X_test[y_test == 0][:, 1], color='lightblue', label='Clase 0 (Test)', marker='x')
plt.scatter(X_test[y_test == 1][:, 0], X_test[y_test == 1][:, 1], color='pink', label='Clase 1 (Test)', marker='x')
plt.title('División del conjunto de datos')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# 3. Entrenamiento del modelo de Regresión Logística
modelo = LogisticRegression()
modelo.fit(X_train, y_train)

# Mostrar los coeficientes del modelo
print("\nCoeficientes del modelo:")
print("Pesos:", modelo.coef_)
print("Bias:", modelo.intercept_)

# Visualización de la frontera de decisión
def plot_decision_boundary(X, y, model):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.coolwarm)
    plt.title('Frontera de Decisión con scale=2.0')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

plot_decision_boundary(X, y, modelo)

# 4. Predicciones y evaluación
y_pred = modelo.predict(X_test)
y_prob = modelo.predict_proba(X_test)[:, 1]  # Probabilidades para la clase 1

# Calcular métricas
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("\nMétricas de evaluación con scale=2.0:")
print(f"Exactitud (Accuracy): {accuracy:.2f}")
print(f"Precisión (Precision): {precision:.2f}")
print(f"Recall (Sensibilidad): {recall:.2f}")
print(f"F1-Score: {f1:.2f}")

# Matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Clase 0', 'Clase 1'], yticklabels=['Clase 0', 'Clase 1'])
plt.title('Matriz de Confusión')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.show()

# 5. Curva ROC y AUC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Azar')
plt.title('Curva ROC')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ---------------------------
# CLASE: REGRESIÓN LOGÍSTICA CON MENOR CONJUNTO DE ENTRENAMIENTO
# ---------------------------

# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
)

# 1. Generación de datos simulados (clasificación binaria)
np.random.seed(42)  # Para reproducibilidad

# Generamos datos para dos clases
X_class1 = np.random.normal(loc=[2, 2], scale=1.0, size=(50, 2))  # Clase 1
X_class2 = np.random.normal(loc=[5, 5], scale=1.0, size=(50, 2))  # Clase 2
X = np.vstack((X_class1, X_class2))  # Combinar las dos clases
y = np.array([0] * 50 + [1] * 50)  # Etiquetas: 0 para clase 1, 1 para clase 2

# Visualización de los datos originales
plt.figure(figsize=(8, 6))
plt.scatter(X[:50, 0], X[:50, 1], color='blue', label='Clase 0')
plt.scatter(X[50:, 0], X[50:, 1], color='red', label='Clase 1')
plt.title('Datos Generados para Clasificación Binaria')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# 2. División del conjunto de datos
# Aquí reducimos el tamaño del conjunto de entrenamiento
X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Reducimos aún más el conjunto de entrenamiento
X_train, _, y_train, _ = train_test_split(X_train_full, y_train_full, test_size=0.7, random_state=42)

# Visualización del conjunto reducido
plt.figure(figsize=(8, 6))
plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], color='blue', label='Clase 0 (Train)')
plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], color='red', label='Clase 1 (Train)')
plt.scatter(X_test[y_test == 0][:, 0], X_test[y_test == 0][:, 1], color='lightblue', label='Clase 0 (Test)', marker='x')
plt.scatter(X_test[y_test == 1][:, 0], X_test[y_test == 1][:, 1], color='pink', label='Clase 1 (Test)', marker='x')
plt.title('División del Conjunto con Entrenamiento Reducido')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# 3. Entrenamiento del modelo de Regresión Logística
modelo = LogisticRegression()
modelo.fit(X_train, y_train)

# Mostrar los coeficientes del modelo
print("\nCoeficientes del modelo:")
print("Pesos:", modelo.coef_)
print("Bias:", modelo.intercept_)

# Visualización de la frontera de decisión
def plot_decision_boundary(X, y, model):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.coolwarm)
    plt.title('Frontera de Decisión con Conjunto de Entrenamiento Reducido')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

plot_decision_boundary(X, y, modelo)

# 4. Predicciones y evaluación
y_pred = modelo.predict(X_test)
y_prob = modelo.predict_proba(X_test)[:, 1]  # Probabilidades para la clase 1

# Calcular métricas
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("\nMétricas de evaluación con Conjunto de Entrenamiento Reducido:")
print(f"Exactitud (Accuracy): {accuracy:.2f}")
print(f"Precisión (Precision): {precision:.2f}")
print(f"Recall (Sensibilidad): {recall:.2f}")
print(f"F1-Score: {f1:.2f}")

# Matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Clase 0', 'Clase 1'], yticklabels=['Clase 0', 'Clase 1'])
plt.title('Matriz de Confusión')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.show()

# 5. Curva ROC y AUC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Azar')
plt.title('Curva ROC')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ---------------------------
# CLASE: REGRESIÓN LOGÍSTICA CON CAMBIO EN TASA DE APRENDIZAJE Y ÉPOCAS
# ---------------------------

# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
)

# 1. Generación de datos simulados (clasificación binaria)
np.random.seed(42)  # Para reproducibilidad

# Generamos datos para dos clases
X_class1 = np.random.normal(loc=[2, 2], scale=1.0, size=(50, 2))  # Clase 1
X_class2 = np.random.normal(loc=[5, 5], scale=1.0, size=(50, 2))  # Clase 2
X = np.vstack((X_class1, X_class2))  # Combinar las dos clases
y = np.array([0] * 50 + [1] * 50)  # Etiquetas: 0 para clase 1, 1 para clase 2

# Visualización de los datos originales
plt.figure(figsize=(8, 6))
plt.scatter(X[:50, 0], X[:50, 1], color='blue', label='Clase 0')
plt.scatter(X[50:, 0], X[50:, 1], color='red', label='Clase 1')
plt.title('Datos Generados para Clasificación Binaria')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# 2. División del conjunto de datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Visualización de la división
plt.figure(figsize=(8, 6))
plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], color='blue', label='Clase 0 (Train)')
plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], color='red', label='Clase 1 (Train)')
plt.scatter(X_test[y_test == 0][:, 0], X_test[y_test == 0][:, 1], color='lightblue', label='Clase 0 (Test)', marker='x')
plt.scatter(X_test[y_test == 1][:, 0], X_test[y_test == 1][:, 1], color='pink', label='Clase 1 (Test)', marker='x')
plt.title('División del conjunto de datos')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# 3. Entrenamiento manual con gradiente descendente
weights = np.zeros(X.shape[1])  # Inicializar pesos
bias = 0.0  # Inicializar bias
learning_rate = 0.5  # Tasa de aprendizaje más alta
epochs = 100  # Reducir el número de épocas
m = len(y_train)  # Número de ejemplos de entrenamiento

# Función sigmoide
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Función de costo
def compute_cost(X, y, weights, bias):
    z = np.dot(X, weights) + bias
    predictions = sigmoid(z)
    cost = -(1 / m) * np.sum(y * np.log(predictions) + (1 - y) * np.log(1 - predictions))
    return cost

# Gradiente descendente
costs = []  # Lista para almacenar los costos

for epoch in range(epochs):
    z = np.dot(X_train, weights) + bias
    predictions = sigmoid(z)

    # Calcular gradientes
    dw = (1 / m) * np.dot(X_train.T, (predictions - y_train))
    db = (1 / m) * np.sum(predictions - y_train)

    # Actualizar parámetros
    weights -= learning_rate * dw
    bias -= learning_rate * db

    # Calcular costo
    cost = compute_cost(X_train, y_train, weights, bias)
    costs.append(cost)

    # Mostrar cada 10 épocas
    if epoch % 10 == 0 or epoch == epochs - 1:
        print(f"Época {epoch}, Costo: {cost:.4f}, Pesos: {weights}, Bias: {bias:.4f}")

# Graficar la reducción del costo
plt.figure(figsize=(8, 6))
plt.plot(range(epochs), costs, color='purple')
plt.title('Reducción del Costo Durante el Entrenamiento (Épocas Reducidas, Alta Tasa de Aprendizaje)')
plt.xlabel('Épocas')
plt.ylabel('Costo')
plt.grid(True)
plt.show()

# 4. Entrenamiento con scikit-learn para comparación
modelo = LogisticRegression(max_iter=100)
modelo.fit(X_train, y_train)

# Mostrar los coeficientes del modelo
print("\nCoeficientes aprendidos por scikit-learn:")
print("Pesos:", modelo.coef_)
print("Bias:", modelo.intercept_)

# Visualización de la frontera de decisión
def plot_decision_boundary(X, y, model):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.coolwarm)
    plt.title('Frontera de Decisión (Alta Tasa de Aprendizaje, Épocas Reducidas)')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

plot_decision_boundary(X, y, modelo)

# 5. Predicciones y evaluación
y_pred = modelo.predict(X_test)
y_prob = modelo.predict_proba(X_test)[:, 1]  # Probabilidades para la clase 1

# Calcular métricas
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("\nMétricas de evaluación con Alta Tasa de Aprendizaje y Épocas Reducidas:")
print(f"Exactitud (Accuracy): {accuracy:.2f}")
print(f"Precisión (Precision): {precision:.2f}")
print(f"Recall (Sensibilidad): {recall:.2f}")
print(f"F1-Score: {f1:.2f}")

# Matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Clase 0', 'Clase 1'], yticklabels=['Clase 0', 'Clase 1'])
plt.title('Matriz de Confusión')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.show()

# 6. Curva ROC y AUC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Azar')
plt.title('Curva ROC')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ---------------------------
# CLASE: REGRESIÓN LOGÍSTICA CON DESBALANCE EN LAS CLASES
# ---------------------------

# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
)

# 1. Generación de datos simulados (clasificación binaria) con desbalance
np.random.seed(42)  # Para reproducibilidad

# Generamos datos para dos clases con diferentes tamaños
X_class1 = np.random.normal(loc=[2, 2], scale=1.0, size=(5, 2))  # Clase 0 (minoritaria)
X_class2 = np.random.normal(loc=[5, 5], scale=1.0, size=(95, 2))  # Clase 1 (mayoritaria)
X = np.vstack((X_class1, X_class2))  # Combinar las dos clases
y = np.array([0] * 30 + [1] * 70)  # Etiquetas: 0 para clase 1, 1 para clase 2

# Visualización de los datos desbalanceados
plt.figure(figsize=(8, 6))
plt.scatter(X[:30, 0], X[:30, 1], color='blue', label='Clase 0 (Minoritaria)')
plt.scatter(X[30:, 0], X[30:, 1], color='red', label='Clase 1 (Mayoritaria)')
plt.title('Datos Generados con Desbalance de Clases')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# 2. División del conjunto de datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Visualización de la división
plt.figure(figsize=(8, 6))
plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], color='blue', label='Clase 0 (Train, Minoritaria)')
plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], color='red', label='Clase 1 (Train, Mayoritaria)')
plt.scatter(X_test[y_test == 0][:, 0], X_test[y_test == 0][:, 1], color='lightblue', label='Clase 0 (Test)', marker='x')
plt.scatter(X_test[y_test == 1][:, 0], X_test[y_test == 1][:, 1], color='pink', label='Clase 1 (Test)', marker='x')
plt.title('División del conjunto de datos con Desbalance')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid(True)
plt.show()

# 3. Entrenamiento del modelo de Regresión Logística
modelo = LogisticRegression()
modelo.fit(X_train, y_train)

# Mostrar los coeficientes del modelo
print("\nCoeficientes del modelo:")
print("Pesos:", modelo.coef_)
print("Bias:", modelo.intercept_)

# Visualización de la frontera de decisión
def plot_decision_boundary(X, y, model):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.coolwarm)
    plt.title('Frontera de Decisión con Desbalance de Clases')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.show()

plot_decision_boundary(X, y, modelo)

# 4. Predicciones y evaluación
y_pred = modelo.predict(X_test)
y_prob = modelo.predict_proba(X_test)[:, 1]  # Probabilidades para la clase 1

# Calcular métricas
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("\nMétricas de evaluación con Desbalance de Clases:")
print(f"Exactitud (Accuracy): {accuracy:.2f}")
print(f"Precisión (Precision): {precision:.2f}")
print(f"Recall (Sensibilidad): {recall:.2f}")
print(f"F1-Score: {f1:.2f}")

# Matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Clase 0', 'Clase 1'], yticklabels=['Clase 0', 'Clase 1'])
plt.title('Matriz de Confusión')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.show()

# 5. Curva ROC y AUC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Azar')
plt.title('Curva ROC')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ---------------------------
# CLASE: REGRESIÓN LOGÍSTICA CON CARACTERÍSTICAS IRRELEVANTES
# ---------------------------

# Importar las bibliotecas necesarias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, auc
)

# 1. Generación de datos simulados (clasificación binaria)
np.random.seed(42)  # Para reproducibilidad

# Generamos datos para dos clases
X_class1 = np.random.normal(loc=[2, 2], scale=1.0, size=(50, 2))  # Clase 1
X_class2 = np.random.normal(loc=[5, 5], scale=1.0, size=(50, 2))  # Clase 2
X = np.vstack((X_class1, X_class2))  # Combinar las dos clases
y = np.array([0] * 50 + [1] * 50)  # Etiquetas: 0 para clase 1, 1 para clase 2

# Generar características irrelevantes
irrelevant_features = np.random.normal(loc=0, scale=1.0, size=(X.shape[0], 3))  # 3 características irrelevantes
X = np.hstack((X, irrelevant_features))  # Combinar con las características originales

# Visualización de las dos primeras características principales
plt.figure(figsize=(8, 6))
plt.scatter(X[:50, 0], X[:50, 1], color='blue', label='Clase 0')
plt.scatter(X[50:, 0], X[50:, 1], color='red', label='Clase 1')
plt.title('Datos Generados con Características Irrelevantes')
plt.xlabel('Feature 1 (Principal)')
plt.ylabel('Feature 2 (Principal)')
plt.legend()
plt.grid(True)
plt.show()

# 2. División del conjunto de datos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Visualización del conjunto reducido
plt.figure(figsize=(8, 6))
plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], color='blue', label='Clase 0 (Train)')
plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], color='red', label='Clase 1 (Train)')
plt.scatter(X_test[y_test == 0][:, 0], X_test[y_test == 0][:, 1], color='lightblue', label='Clase 0 (Test)', marker='x')
plt.scatter(X_test[y_test == 1][:, 0], X_test[y_test == 1][:, 1], color='pink', label='Clase 1 (Test)', marker='x')
plt.title('División del conjunto de datos')
plt.xlabel('Feature 1 (Principal)')
plt.ylabel('Feature 2 (Principal)')
plt.legend()
plt.grid(True)
plt.show()

# 3. Entrenamiento del modelo de Regresión Logística
modelo = LogisticRegression()
modelo.fit(X_train, y_train)

# Mostrar los coeficientes del modelo
print("\nCoeficientes del modelo:")
print("Pesos:", modelo.coef_)
print("Bias:", modelo.intercept_)

# Visualización de la frontera de decisión (en las dos primeras características)
def plot_decision_boundary(X, y, model):
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.1), np.arange(y_min, y_max, 0.1))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel(), np.zeros((xx.ravel().shape[0], 3))])  # Ignorar irrelevantes en 2D
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.coolwarm)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k', cmap=plt.cm.coolwarm)
    plt.title('Frontera de Decisión con Características Irrelevantes')
    plt.xlabel('Feature 1 (Principal)')
    plt.ylabel('Feature 2 (Principal)')
    plt.show()

plot_decision_boundary(X, y, modelo)

# 4. Predicciones y evaluación
y_pred = modelo.predict(X_test)
y_prob = modelo.predict_proba(X_test)[:, 1]  # Probabilidades para la clase 1

# Calcular métricas
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("\nMétricas de evaluación con Características Irrelevantes:")
print(f"Exactitud (Accuracy): {accuracy:.2f}")
print(f"Precisión (Precision): {precision:.2f}")
print(f"Recall (Sensibilidad): {recall:.2f}")
print(f"F1-Score: {f1:.2f}")

# Matriz de confusión
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Clase 0', 'Clase 1'], yticklabels=['Clase 0', 'Clase 1'])
plt.title('Matriz de Confusión')
plt.xlabel('Predicción')
plt.ylabel('Real')
plt.show()

# 5. Curva ROC y AUC
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--', label='Azar')
plt.title('Curva ROC')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from IPython.display import display, HTML

# Contenido HTML
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Regresión con K-Vecinos (KNN) - Teoría</title>
    <link href="https://fonts.googleapis.com/css2?family=Roboto:wght@400;700&display=swap" rel="stylesheet">
    <style>
        :root {
            --primary-color: #4CAF50;
            --secondary-color: #388E3C;
            --text-color: #333;
            --background-color: #f5f5f5;
            --accent-color: #00695C;
            --border-color: #ccc;
        }
        body {
            font-family: 'Roboto', Arial, sans-serif;
            line-height: 1.8;
            color: var(--text-color);
            margin: 0;
            padding: 20px;
            background-color: var(--background-color);
        }
        h1, h2, h3 {
            color: var(--accent-color);
            text-align: left;
        }
        h1 {
            font-size: 2.5rem;
            margin-bottom: 20px;
        }
        p, ul, pre {
            margin: 15px 0;
            line-height: 1.6;
        }
        ul {
            padding-left: 20px;
        }
        li {
            margin-bottom: 10px;
        }
        pre {
            background-color: #eee;
            padding: 10px;
            border-radius: 5px;
            overflow-x: auto;
        }
        code {
            background-color: #f4f4f4;
            padding: 2px 5px;
            border-radius: 3px;
            font-family: monospace;
        }
    </style>
</head>
<body>
    <header>
        <h1>Regresión con K-Vecinos (KNN) - Teoría en Detalle</h1>
    </header>
    <main>
        <h2>1. Concepto Básico</h2>
        <p>En <strong>regresión con KNN</strong>, la idea principal es predecir el valor de la variable objetivo (<em>y</em>) para un punto de prueba basándose en los valores de <em>y</em> de los <em>K</em>-vecinos más cercanos encontrados en el conjunto de datos de entrenamiento.</p>
        <ul>
            <li><strong>Vecinos más cercanos:</strong> Se seleccionan en función de una métrica de distancia (usualmente distancia Euclidiana).</li>
            <li><strong>Predicción:</strong> Se toma una agregación de los valores de <em>y</em> de los vecinos seleccionados (promedio, mediana, o ponderación inversa a la distancia).</li>
        </ul>

        <h2>2. Proceso de Regresión con KNN</h2>
        <ol>
            <li><strong>Definición de parámetros:</strong>
                <ul>
                    <li>Se establece <em>K</em>, el número de vecinos.</li>
                    <li>Se elige una métrica de distancia (Euclidiana, Manhattan, etc.).</li>
                </ul>
            </li>
            <li><strong>Calcular distancias:</strong> Se calcula la distancia entre el punto de prueba y todos los puntos en el conjunto de entrenamiento.</li>
            <li><strong>Seleccionar los <em>K</em>-vecinos más cercanos:</strong>
                <ul>
                    <li>Se ordenan los puntos de entrenamiento según su distancia al punto de prueba.</li>
                    <li>Se seleccionan los <em>K</em>-puntos más cercanos.</li>
                </ul>
            </li>
            <li><strong>Predicción:</strong>
                <p>Se utiliza el valor promedio (u otra función) de los valores <em>y</em> de los vecinos seleccionados.</p>
                <pre><code>ŷ = (1/K) ∑ y_i</code></pre>
                <p>Alternativamente, se puede usar una ponderación basada en la distancia:</p>
                <pre><code>ŷ = (∑ w_i y_i) / (∑ w_i)</code></pre>
                <p>donde <code>w_i = 1/d(x, x_i)</code> y <code>d(x, x_i)</code> es la distancia entre el punto de prueba y el vecino <em>i</em>.</p>
            </li>
        </ol>

        <h2>3. Métricas de Distancia</h2>
        <p>Las métricas de distancia determinan cómo se seleccionan los vecinos más cercanos. Las más comunes son:</p>
        <ul>
            <li><strong>Distancia Euclidiana:</strong>
                <pre><code>d(x, x') = √(∑ (x_i - x'_i)^2)</code></pre>
            </li>
            <li><strong>Distancia Manhattan:</strong>
                <pre><code>d(x, x') = ∑ |x_i - x'_i|</code></pre>
            </li>
            <li><strong>Distancia Minkowski:</strong>
                <pre><code>d(x, x') = (∑ |x_i - x'_i|^p)^(1/p)</code></pre>
            </li>
        </ul>

        <h2>4. Elección del Número de Vecinos (<em>K</em>)</h2>
        <ul>
            <li><em>K</em> <strong>muy pequeño:</strong>
                <ul>
                    <li>El modelo será muy sensible al ruido (sobreajuste).</li>
                    <li>Puede producir resultados erráticos.</li>
                </ul>
            </li>
            <li><em>K</em> <strong>muy grande:</strong>
                <ul>
                    <li>El modelo pierde capacidad de capturar patrones locales (subajuste).</li>
                    <li>Se vuelve una regresión promedio global.</li>
                </ul>
            </li>
        </ul>
        <p>La elección ideal de <em>K</em> se determina mediante <strong>validación cruzada</strong>.</p>

        <h2>5. Escalado de Datos</h2>
        <p>Dado que KNN se basa en distancias, las características deben estar en la misma escala para evitar que algunas dominen sobre otras:</p>
        <ul>
            <li><strong>Estandarización:</strong> Transformar los datos para que tengan media cero y desviación estándar unitaria.</li>
            <li><strong>Normalización:</strong> Escalar las características a un rango fijo, como [0, 1].</li>
        </ul>

        <h2>6. Ventajas de KNN para Regresión</h2>
        <ul>
            <li><strong>Simplicidad:</strong> Es fácil de entender e implementar.</li>
            <li><strong>No paramétrico:</strong> No hace suposiciones sobre la relación entre las características y la variable objetivo.</li>
            <li><strong>Buena adaptabilidad:</strong> Funciona bien en problemas donde la relación entre las variables es compleja y no lineal.</li>
        </ul>

        <h2>7. Desventajas de KNN para Regresión</h2>
        <ul>
            <li><strong>Costo computacional:</strong>
                <p>KNN requiere calcular distancias con todos los puntos del conjunto de datos para cada predicción, lo que puede ser lento con grandes conjuntos de datos.</p>
            </li>
            <li><strong>Sensibilidad a la escala de las características:</strong>
                <p>Si las características no están escaladas, aquellas con mayores magnitudes dominarán el cálculo de la distancia.</p>
            </li>
            <li><strong>Elección de <em>K</em> y métrica de distancia:</strong>
                <p>El rendimiento depende fuertemente de estos hiperparámetros.</p>
            </li>
            <li><strong>Sensible a ruido:</strong>
                <p>Los valores atípicos pueden afectar las predicciones al incluir vecinos irrelevantes.</p>
            </li>
        </ul>

        <h2>8. Aplicaciones Comunes de KNN en Regresión</h2>
        <ul>
            <li>Predicción de precios (por ejemplo, precios de casas, autos, etc.).</li>
            <li>Estimaciones continuas en problemas de series temporales.</li>
            <li>Predicción de valores numéricos en sistemas de recomendación.</li>
        </ul>
    </main>
</body>
</html>
"""

# Mostrar el HTML en Google Colab
display(HTML(html_content))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Datos simulados basados en la tabla
data = {
    "Age": [25, 35, 45, 20, 35, 52, 23, 40, 60, 48, 33],
    "Loan": [40000, 60000, 80000, 20000, 120000, 18000, 95000, 62000, 100000, 220000, 150000],
    "HPI": [135, 256, 231, 267, 139, 150, 127, 216, 139, 250, 264]  # House Price Index
}

# Convertir los datos a un DataFrame
df = pd.DataFrame(data)

# Caso desconocido (Age=48, Loan=142,000)
unknown_case = {"Age": 48, "Loan": 142000}

# Función para calcular la distancia Euclidiana
def calcular_distancia_euclidiana(row, unknown_case):
    return np.sqrt((row['Age'] - unknown_case['Age'])**2 + (row['Loan'] - unknown_case['Loan'])**2)

# Calcular distancias para cada fila
df['Distance'] = df.apply(lambda row: calcular_distancia_euclidiana(row, unknown_case), axis=1)

# Ordenar por distancia
df_ordenado = df.sort_values(by='Distance')

# Mostrar tabla con distancias ordenadas
print("Tabla de vecinos ordenados por distancia:")
print(df_ordenado[['Age', 'Loan', 'HPI', 'Distance']])

# Caso 1: K=1 (vecino más cercano)
k1 = df_ordenado.iloc[0]
hpi_k1 = k1['HPI']
print(f"\nPara K=1, el vecino más cercano tiene HPI={hpi_k1}")

# Caso 2: K=3 (promedio de los 3 vecinos más cercanos)
k3 = df_ordenado.iloc[:3]
hpi_k3 = k3['HPI'].mean()
print(f"Para K=3, el promedio de los HPI de los 3 vecinos más cercanos es {hpi_k3:.2f}")

# Visualización de los datos y el caso desconocido
plt.scatter(df['Age'], df['Loan'], c='blue', label='Datos conocidos')
plt.scatter(unknown_case['Age'], unknown_case['Loan'], c='red', label='Caso desconocido', marker='x', s=100)
plt.title('Visualización de Datos y Caso Desconocido')
plt.xlabel('Age')
plt.ylabel('Loan')
plt.legend()
plt.show()

# Visualización con vecinos más cercanos
plt.scatter(df_ordenado['Age'], df_ordenado['Loan'], c='blue', label='Datos conocidos')
plt.scatter(k3['Age'], k3['Loan'], c='green', label='Vecinos más cercanos (K=3)')
plt.scatter(unknown_case['Age'], unknown_case['Loan'], c='red', label='Caso desconocido', marker='x', s=100)
plt.title('Visualización de Vecinos Más Cercanos')
plt.xlabel('Age')
plt.ylabel('Loan')
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import pairwise_distances

# Crear datos de ejemplo
X = np.array([[1, 2], [2, 3], [3, 4], [5, 6], [7, 8], [9, 10]])  # Puntos en 2D
y = [0, 1, 1, 0, 1, 0]  # Etiquetas de los puntos

# Definir el modelo KNN usando distancia de Minkowski
knn = KNeighborsClassifier(n_neighbors=3, p=3)  # p=3 es la distancia de Minkowski con p=3
knn.fit(X, y)

# Realizar una predicción
predicción = knn.predict([[4, 5]])  # Predice la etiqueta para el punto (4,5)
print(predicción)

# Calcular las distancias de Minkowski manualmente
distancias = pairwise_distances(X, [[4, 5]], metric='minkowski', p=3)
print(distancias)
```

```python
# Importar las bibliotecas necesarias
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score

# Cargar el dataset Iris
data = load_iris()
X = data.data  # Características (longitud/anchura de sépalo y pétalo)
y = data.target  # Clases (setosa, versicolor, virginica)

# Dividir el dataset en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Estandarizar los datos (KNN es sensible a la escala de los datos)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Crear el modelo KNN con k=3
knn = KNeighborsClassifier(n_neighbors=3)

# Entrenar el modelo
knn.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = knn.predict(X_test)

# Evaluar el modelo
print("Reporte de clasificación:")
print(classification_report(y_test, y_pred, target_names=data.target_names))
print(f"Exactitud: {accuracy_score(y_test, y_pred):.2f}")
```

```python
from IPython.display import display, HTML

# Contenido HTML
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Uso de la Distancia de Hamming en KNN con Variables Categóricas</title>
    <link href="https://fonts.googleapis.com/css2?family=Roboto:wght@400;700&display=swap" rel="stylesheet">
    <style>
        :root {
            --primary-color: #4CAF50;
            --secondary-color: #388E3C;
            --text-color: #333;
            --background-color: #f5f5f5;
            --accent-color: #00695C;
            --border-color: #ccc;
        }
        body {
            font-family: 'Roboto', Arial, sans-serif;
            line-height: 1.8;
            color: var(--text-color);
            margin: 0;
            padding: 20px;
            background-color: var(--background-color);
        }
        h1, h2, h3 {
            color: var(--accent-color);
            text-align: left;
        }
        h1 {
            font-size: 2.5rem;
            margin-bottom: 20px;
        }
        p, ul, pre {
            margin: 15px 0;
            line-height: 1.6;
        }
        ul {
            padding-left: 20px;
        }
        li {
            margin-bottom: 10px;
        }
        pre {
            background-color: #eee;
            padding: 10px;
            border-radius: 5px;
            overflow-x: auto;
        }
        code {
            background-color: #f4f4f4;
            padding: 2px 5px;
            border-radius: 3px;
            font-family: monospace;
        }
        table {
            width: 100%;
            border-collapse: collapse;
            margin: 15px 0;
        }
        th, td {
            border: 1px solid var(--border-color);
            padding: 8px 12px;
            text-align: left;
        }
        th {
            background-color: var(--primary-color);
            color: white;
        }
    </style>
</head>
<body>
    <header>
        <h1>Uso de la Distancia de Hamming en KNN con Variables Categóricas</h1>
    </header>
    <main>
        <p>En el caso de <strong>K-Nearest Neighbors (KNN)</strong>, cuando trabajamos con variables categóricas (como colores, géneros, o etiquetas que no son numéricas), las medidas de distancia tradicionales como la Euclidiana o Manhattan no son aplicables, ya que estas funcionan únicamente con valores continuos o numéricos.</p>

        <h2>¿Qué es la distancia de Hamming?</h2>
        <p>La <strong>distancia de Hamming</strong> es una medida de similitud para variables categóricas. Calcula el número de posiciones en las que dos cadenas (o valores categóricos) son diferentes. Se usa típicamente cuando los datos tienen valores discretos, como cadenas de texto o categorías.</p>

        <h3>Fórmula:</h3>
        <pre><code>Distancia de Hamming = Número de símbolos diferentes entre dos instancias</code></pre>

        <h3>Ejemplo simple:</h3>
        <p>Supongamos que tenemos dos cadenas categóricas:</p>
        <ul>
            <li><strong>Cadena 1:</strong> rojo, azul, verde</li>
            <li><strong>Cadena 2:</strong> rojo, verde, verde</li>
        </ul>
        <p>La distancia de Hamming es el número de posiciones en las que los valores son diferentes:</p>
        <ul>
            <li><strong>Posición 1:</strong> rojo vs. rojo → Igual (0)</li>
            <li><strong>Posición 2:</strong> azul vs. verde → Diferente (1)</li>
            <li><strong>Posición 3:</strong> verde vs. verde → Igual (0)</li>
        </ul>
        <p><strong>Distancia total:</strong> 1</p>

        <h2>Uso de Hamming en KNN</h2>
        <p>Cuando aplicamos KNN a datos categóricos, usamos la distancia de Hamming en lugar de las métricas tradicionales para determinar los vecinos más cercanos.</p>

        <h3>Ejemplo:</h3>
        <p>Si tienes un conjunto de datos como el siguiente:</p>

        <table>
            <thead>
                <tr>
                    <th>Animal</th>
                    <th>Color</th>
                    <th>Pata/Tipo</th>
                    <th>Clase</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>Gato</td>
                    <td>Negro</td>
                    <td>Cuatro</td>
                    <td>Mamífero</td>
                </tr>
                <tr>
                    <td>Perro</td>
                    <td>Marrón</td>
                    <td>Cuatro</td>
                    <td>Mamífero</td>
                </tr>
                <tr>
                    <td>Pájaro</td>
                    <td>Amarillo</td>
                    <td>Dos</td>
                    <td>Ave</td>
                </tr>
                <tr>
                    <td>Serpiente</td>
                    <td>Verde</td>
                    <td>Sin patas</td>
                    <td>Reptil</td>
                </tr>
            </tbody>
        </table>

        <p>Para clasificar un nuevo animal con las características:</p>
        <ul>
            <li><strong>Color:</strong> Marrón</li>
            <li><strong>Pata/Tipo:</strong> Cuatro</li>
        </ul>
        <p>El algoritmo KNN compara al nuevo animal con todos los existentes y calcula la distancia de Hamming basándose en los valores categóricos. Por ejemplo:</p>

        <table>
            <thead>
                <tr>
                    <th>Animal</th>
                    <th>Distancia de Hamming</th>
                </tr>
            </thead>
            <tbody>
                <tr>
                    <td>Gato</td>
                    <td>1 (1 diferencia: Color)</td>
                </tr>
                <tr>
                    <td>Perro</td>
                    <td>0 (Exactamente igual)</td>
                </tr>
                <tr>
                    <td>Pájaro</td>
                    <td>2 (2 diferencias: Color, Pata/Tipo)</td>
                </tr>
                <tr>
                    <td>Serpiente</td>
                    <td>2 (2 diferencias: Pata/Tipo, Color)</td>
                </tr>
            </tbody>
        </table>

        <p>El animal con menor distancia es el <strong>Perro</strong>.</p>
    </main>
</body>
</html>
"""

# Mostrar el HTML en Google Colab
display(HTML(html_content))
```

```python
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OneHotEncoder

# Datos de ejemplo
X = np.array([
    ['Negro', 'Cuatro'],
    ['Marrón', 'Cuatro'],
    ['Amarillo', 'Dos'],
    ['Verde', 'Sin patas']
])
y = np.array(['Mamífero', 'Mamífero', 'Ave', 'Reptil'])

# Convertir variables categóricas a codificación binaria (One-Hot Encoding)
encoder = OneHotEncoder()
X_encoded = encoder.fit_transform(X).toarray()

# Nuevo animal para clasificar
X_new = np.array([['Marrón', 'Cuatro']])
X_new_encoded = encoder.transform(X_new).toarray()

# Crear modelo KNN con la distancia de Hamming
knn = KNeighborsClassifier(n_neighbors=1, metric='hamming')
knn.fit(X_encoded, y)

# Clasificar el nuevo animal
prediction = knn.predict(X_new_encoded)
print(f"Clase predicha: {prediction[0]}")
```
