```python
# Importación de librerías necesarias
from IPython.display import display, HTML
import html # Para escapar caracteres especiales en los títulos

# --- 1. EXTRACCIÓN Y ESTRUCTURACIÓN DE DATOS ---

# --- Bloque de Conclusión ---
conclusion_binomial_html = """
<div class="content-block">
    <h2>Conclusiones</h2>
    <p>La Distribución Binomial es una herramienta esencial en el análisis de variables dicotómicas y en la inferencia sobre proporciones. Verificar los supuestos que la sustentan es clave para su correcta aplicación, y existen aproximaciones y modelos alternativos cuando dichos supuestos se incumplen. Comprender sus propiedades —incluidas herramientas como la función generadora de momentos, los coeficientes de asimetría y curtosis, y la corrección por continuidad— proporciona una base sólida para estudios estadísticos más complejos.</p>
</div>
"""

# --- Contenido Principal para el Acordeón (TEMA: DISTRIBUCIÓN BINOMIAL) ---
binomial_definitions_data = [
    {
        "titulo": "1. Definición y Notación",
        "contenido": """
            <h4>Definición y Notación</h4>
            <p>La Distribución Binomial, denotada como $X \\sim \\text{Bin}(n, p)$, modela el número de éxitos $k$ obtenidos en $n$ ensayos independientes, cada uno con la misma probabilidad de éxito $p$. [3, 10]</p>
            <p><b>Soporte:</b> $k = 0, 1, 2, \\dots, n$.</p>

            <h4>Función de Masa de Probabilidad (PMF)</h4>
            <p>La PMF asigna una probabilidad a cada valor discreto $k$. La suma de las probabilidades sobre todos los posibles valores de $k$ es 1.</p>
            <div class="formula-block">$$ P(X=k) = \\binom{n}{k} p^k (1-p)^{n-k} $$</div>

            <h4>Esperanza y Varianza</h4>
            <ul>
                <li><b>Esperanza (Media) $E[X]$:</b> Representa el número medio de éxitos esperados.
                    <div class="formula-block">$$ E[X] = np $$</div>
                </li>
                <li><b>Varianza $\\text{Var}(X)$:</b> Mide la dispersión de los resultados alrededor de la media.
                    <div class="formula-block">$$ \\text{Var}(X) = np(1-p) $$</div>
                </li>
            </ul>
             <blockquote>Cuando $p=0.5$, la distribución es simétrica; de lo contrario, presenta asimetría.</blockquote>
        """
    },
    {
        "titulo": "2. Supuestos Fundamentales",
        "contenido": """
            <h4>Supuestos Clave</h4>
            <p>Para que un experimento pueda ser modelado con una distribución binomial, se deben cumplir los siguientes supuestos:</p>
            <ol>
                <li><strong>Independencia:</strong> El resultado de un ensayo no afecta a los demás.</li>
                <li><strong>Probabilidad Constante:</strong> La probabilidad de éxito $p$ es la misma en cada ensayo.</li>
                <li><strong>Tamaño Muestral Fijo:</strong> El número de ensayos $n$ se decide antes de realizar el experimento.</li>
            </ol>
            <blockquote>Si estos supuestos se violan (por ejemplo, en un muestreo sin reemplazo de una población finita), la <strong>Distribución Hipergeométrica</strong> puede ser más adecuada.</blockquote>
        """
    },
    {
        "titulo": "3. Propiedades y Conceptos Asociados",
        "contenido": """
            <h4>Propiedades Adicionales</h4>
            <div class="table-container">
                <table>
                    <thead><tr><th>Propiedad</th><th>Expresión</th><th>Definición y Observación</th></tr></thead>
                    <tbody>
                        <tr>
                            <td><strong>Función Generadora de Momentos (MGF)</strong></td>
                            <td>$ M_X(t) = (1 - p + pe^t)^n $</td>
                            <td>Se define como $M_X(t) = E[e^{tX}]$. [2, 5] Derivarla $r$ veces y evaluarla en $t=0$ permite obtener el $r$-ésimo momento (media, varianza, etc.) de forma sistemática.</td>
                        </tr>
                        <tr>
                            <td><strong>Coeficiente de Asimetría ($\\gamma_1$)</strong></td>
                            <td>$ \\gamma_1 = \\frac{1 - 2p}{\\sqrt{np(1-p)}} $</td>
                            <td>Mide la simetría de la distribución. Un valor de cero implica simetría (lo que ocurre cuando $p=0.5$). Valores positivos indican una cola derecha más larga y negativos, una cola izquierda más larga.</td>
                        </tr>
                        <tr>
                            <td><strong>Exceso de Curtosis ($\\gamma_2$)</strong></td>
                            <td>$ \\gamma_2 = \\frac{1 - 6p(1-p)}{np(1-p)} $</td>
                            <td>Compara el "apuntamiento" y el peso de las colas de la distribución con respecto a una distribución normal. Valores positivos indican colas más pesadas (leptocúrtica).</td>
                        </tr>
                    </tbody>
                </table>
            </div>
        """
    },
    {
        "titulo": "4. Aproximaciones a Otras Distribuciones",
        "contenido": """
            <h4>Aproximación de Poisson</h4>
            <p>Si $n$ es grande y $p$ es pequeño (regla común: $p < 0.05$ y $n$ grande), la distribución binomial puede ser aproximada por una distribución de Poisson con parámetro $\\lambda = np$.</p>
            <div class="formula-block">$$ \\text{Bin}(n, p) \\approx \\text{Pois}(\\lambda = np) $$</div>
            <p><b>Motivación:</b> Esta aproximación es útil para modelar eventos raros en una gran cantidad de intentos.</p>

            <h4>Aproximación Normal (Teorema de De Moivre-Laplace)</h4>
            <p>Si $np \\ge 5$ y $n(1-p) \\ge 5$, la distribución binomial puede aproximarse satisfactoriamente por una distribución normal. [11, 15]</p>
            <div class="formula-block">$$ X \\sim \\mathcal{N}(\\mu = np, \\sigma^2 = np(1-p)) $$</div>
            <p><b>Corrección por Continuidad:</b> Al usar una distribución continua (Normal) para aproximar una discreta (Binomial), se ajustan los límites sumando o restando 0.5 para mejorar la precisión. [12, 16, 22] Por ejemplo, $P(X=k)$ se aproxima como $P(k-0.5 \\le Y \\le k+0.5)$, donde $Y$ es la variable normal.</p>
        """
    },
    {
        "titulo": "5. Inferencia Estadística",
        "contenido": """
            <h4>Aplicaciones en Inferencia</h4>
            <div class="table-container">
                <table>
                    <thead><tr><th>Objetivo</th><th>Método</th><th>Breve Explicación</th></tr></thead>
                    <tbody>
                        <tr>
                            <td><strong>Contraste de proporciones</strong></td>
                            <td>Prueba binomial exacta</td>
                            <td>Calcula la probabilidad de observar un resultado tan o más extremo que el obtenido, asumiendo que la hipótesis nula ($H_0$) es cierta.</td>
                        </tr>
                        <tr>
                            <td><strong>Intervalo de Confianza para $p$</strong></td>
                            <td>Métodos de Wilson, Clopper-Pearson, Wald</td>
                            <td>Proporcionan un rango de valores plausibles para la verdadera proporción poblacional $p$. El método de Clopper-Pearson es exacto pero conservador, mientras que el de Wald es una aproximación que puede fallar con muestras pequeñas.</td>
                        </tr>
                        <tr>
                            <td><strong>Bondad de Ajuste</strong></td>
                            <td>Prueba $\\chi^2$ de Pearson o Bootstrap</td>
                            <td>El bootstrapping es una técnica de remuestreo que permite estimar la distribución de un estadístico de interés, útil cuando las derivaciones teóricas son complejas.</td>
                        </tr>
                    </tbody>
                </table>
            </div>
        """
    },
    {
        "titulo": "6. Ejemplo Ilustrativo con Python",
        "contenido": """
            <h4>Escenario</h4>
            <p>Se lanza una moneda 10 veces ($n=10$) y se obtienen 7 caras ($k=7$). Queremos evaluar la hipótesis nula de que la moneda es justa ($H_0: p=0.5$).</p>
            <h4>Código en Python</h4>
            <p>Utilizamos la librería <code>scipy.stats</code> para realizar los cálculos. [7, 13, 17]</p>
            <div class="content-block" style="background-color: #2d2d2d; color: #f0f0f0; font-family: 'JetBrains Mono', monospace; padding: 1rem; border-radius: 8px;">
                <pre><code>from scipy.stats import binom

n, k, p0 = 10, 7, 0.5

# Probabilidad de obtener exactamente 7 caras
prob_k_7 = binom.pmf(k, n, p0)

# p-valor para una prueba de dos colas
# Es la probabilidad de obtener un resultado >= 7 o <= 3
p_valor = 2 * binom.sf(k - 1, n, p0)

print(f"P(X = 7)  = {prob_k_7:.4f}")
print(f"p-valor   = {p_valor:.4f}")</code></pre>
            </div>
            <h4>Resultados</h4>
            <p>La probabilidad de obtener exactamente 7 caras es $P(X=7) \\approx 0.1172$.</p>
            <p>El p-valor es aproximadamente $0.3438$. Dado que este valor es mayor que los niveles de significancia comunes (como 0.05), no tenemos evidencia suficiente para rechazar la hipótesis de que la moneda es justa.</p>
        """
    },
    {
        "titulo": "7. Ámbitos de Aplicación",
        "contenido": """
            <h4>Ejemplos Prácticos</h4>
            <div class="table-container">
                 <table>
                    <thead><tr><th>Área</th><th>Pregunta Típica (Éxito = 1)</th><th>Consideraciones</th></tr></thead>
                    <tbody>
                        <tr><td><strong>Control de Calidad</strong></td><td>Número de productos defectuosos en un lote.</td><td>Se puede usar la aproximación de Poisson si los defectos son eventos raros.</td></tr>
                        <tr><td><strong>Medicina</strong></td><td>Pacientes que responden positivamente a un tratamiento.</td><td>El tamaño muestral es crucial para la potencia del estudio estadístico.</td></tr>
                        <tr><td><strong>Marketing Digital</strong></td><td>Proporción de clics en una campaña de email.</td><td>Útil para estimar el retorno de la inversión (ROI) y planificar el tamaño de la muestra.</td></tr>
                        <tr><td><strong>Finanzas</strong></td><td>Número de impagos en una cartera de créditos.</td><td>La suposición de independencia puede ser violada si los impagos están correlacionados por factores económicos.</td></tr>
                    </tbody>
                </table>
            </div>
        """
    },
    {
        "titulo": "8. Distribuciones Relacionadas",
        "contenido": """
            <h4>Familia de Distribuciones</h4>
            <ul>
                <li><strong>Distribución de Bernoulli:</strong> Es el caso base de la binomial, con un solo ensayo ($n=1$).</li>
                <li><strong>Distribución Multinomial:</strong> Generaliza la binomial a ensayos con más de dos resultados posibles.</li>
                <li><strong>Distribución Binomial Negativa:</strong> Modela el número de ensayos necesarios para obtener un número fijo de éxitos.</li>
                <li><strong>Distribución Beta-Binomial:</strong> Es una mezcla donde la probabilidad de éxito $p$ no es fija, sino que sigue una distribución Beta. Es útil para modelar la sobredispersión (cuando la varianza observada es mayor que la esperada por el modelo binomial).</li>
            </ul>
        """
    }
]


# --- LÓGICA DE GENERACIÓN (SIN CAMBIOS) ---
def generar_tarjetas_definiciones(datos):
    html_generado = []
    for seccion in datos:
        titulo_escapado = html.escape(seccion['titulo'])
        contenido_html = seccion['contenido']

        plantilla_seccion = f"""
        <div class="topic-card">
            <div class="topic-header">
                <span class="topic-title">{titulo_escapado}</span>
                <i class="fas fa-chevron-down expand-icon"></i>
            </div>
            <div class="topic-content">
                {contenido_html}
            </div>
        </div>
        """
        html_generado.append(plantilla_seccion)
    return "\n".join(html_generado)

# --- PRE-CÁLCULO DEL CONTENIDO DINÁMICO ---
contenido_dinamico_html = generar_tarjetas_definiciones(binomial_definitions_data)

# --- PLANTILLA PROFESIONAL (SIN CAMBIOS) ---
plantilla_profesional = """
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>{main_title}</title>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700;800&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">
  <script>
    window.MathJax = {{
      tex: {{
        inlineMath: [['$', '$'], ['\\\\(', '\\\\)']],
        displayMath: [['$$', '$$'], ['\\\\[', '\\\\]']],
        processEscapes: true,
        processEnvironments: true
      }},
      svg: {{
        fontCache: 'global'
      }}
    }};
  </script>
  <script src="https://cdn.jsdelivr.net/npm/mathjax@3/es5/tex-mml-chtml.js" id="MathJax-script" async></script>
  <style>
    :root {{
      --bg-primary: linear-gradient(135deg, #16a085 0%, #2980b9 100%);
      --bg-secondary: rgba(255, 255, 255, 0.85);
      --bg-tertiary: rgba(248, 250, 252, 0.8);
      --text-primary: #2c3e50;
      --text-secondary: #34495e;
      --text-light: #ffffff;
      --accent-primary: #16a085;
      --accent-secondary: #2980b9;
      --accent-gradient: linear-gradient(135deg, var(--accent-primary) 0%, var(--accent-secondary) 100%);
      --border-color: rgba(226, 232, 240, 0.8);
      --shadow-card: 0 15px 35px rgba(0, 0, 0, 0.08);
      --border-radius: 20px;
      --transition: all 0.4s cubic-bezier(0.25, 0.8, 0.25, 1);
    }}
    [data-theme="dark"] {{
      --bg-primary: linear-gradient(135deg, #0f2027 0%, #203a43 50%, #2c5364 100%);
      --bg-secondary: rgba(26, 32, 44, 0.85);
      --bg-tertiary: rgba(45, 55, 72, 0.7);
      --text-primary: #f7fafc;
      --text-secondary: #a0aec0;
      --accent-primary: #1dd1a1;
      --accent-secondary: #48dbfb;
      --border-color: rgba(255, 255, 255, 0.15);
    }}
    * {{ margin: 0; padding: 0; box-sizing: border-box; }}
    html {{ scroll-behavior: smooth; }}
    body {{ font-family: 'Inter', sans-serif; line-height: 1.8; background: var(--bg-primary); color: var(--text-primary); transition: var(--transition); min-height: 100vh; position: relative; overflow-x: hidden; }}
    .particles {{ position: fixed; top: 0; left: 0; width: 100%; height: 100%; pointer-events: none; z-index: -1; }}
    .particle {{ position: absolute; border-radius: 50%; animation: float 25s infinite linear; opacity: 0; background: rgba(255, 255, 255, 0.6);}}
    @keyframes float {{ 0% {{ transform: translateY(100vh) rotate(0deg); opacity: 0; }} 10%, 90% {{ opacity: 0.6; }} 100% {{ transform: translateY(-10vh) rotate(360deg); opacity: 0; }} }}

    .container {{ max-width: 1000px; margin: 0 auto; padding: 2rem; z-index: 1; }}
    .header {{ text-align: center; margin-bottom: 3rem; position: relative; }}
    .main-title {{ font-size: clamp(2.5rem, 5vw, 4.2rem); font-weight: 800; background: var(--accent-gradient); -webkit-background-clip: text; -webkit-text-fill-color: transparent; background-clip: text; margin-bottom: 1rem; }}

    .theme-toggle {{ position: fixed; top: 2rem; right: 2rem; width: 60px; height: 60px; border: 1px solid var(--border-color); border-radius: 50%; background: var(--bg-secondary); backdrop-filter: blur(15px); box-shadow: var(--shadow-card); cursor: pointer; display: flex; align-items: center; justify-content: center; font-size: 1.4rem; color: var(--accent-primary); transition: var(--transition); z-index: 1000; }}
    .theme-toggle:hover {{ transform: scale(1.15) rotate(180deg); }}

    .content-block {{
        background: var(--bg-secondary); backdrop-filter: blur(20px); border-radius: var(--border-radius);
        box-shadow: var(--shadow-card); border: 2px solid var(--border-color);
        padding: 2rem; margin-bottom: 2rem;
    }}
    .content-block h2 {{ color: var(--text-primary); margin-bottom: 1rem; }}
    .content-block p {{ color: var(--text-secondary); margin-bottom: 1rem; }}

    .lesson-container {{ display: flex; flex-direction: column; gap: 1.5rem; }}
    .topic-card {{ background: var(--bg-secondary); backdrop-filter: blur(20px); border-radius: var(--border-radius); box-shadow: var(--shadow-card); border: 2px solid var(--border-color); overflow: hidden; transition: var(--transition); }}
    .topic-header {{ cursor: pointer; padding: 1.5rem 2rem; display: flex; justify-content: space-between; align-items: center; }}
    .topic-title {{ font-size: 1.3rem; font-weight: 600; color: var(--text-primary); }}
    .expand-icon {{ font-size: 1.2rem; color: var(--text-secondary); transition: var(--transition); }}
    .topic-card.open .expand-icon {{ transform: rotate(180deg); }}
    .topic-content {{ max-height: 0; overflow: hidden; transition: max-height 1.2s ease, padding 1.2s ease; background: var(--bg-tertiary); }}
    .topic-card.open .topic-content {{ max-height: 3500px; padding: 1.5rem 2rem; border-top: 1px solid var(--border-color); }}

    .topic-content h4 {{ color: var(--text-primary); margin-top: 1.5rem; margin-bottom: 0.5rem; font-size: 1.1rem; }}
    .topic-content p, .topic-content li {{ color: var(--text-secondary); line-height: 1.7; }}
    .topic-content ul, .topic-content ol {{ padding-left: 25px; margin-bottom: 1rem; }}
    .topic-content strong {{ color: var(--text-primary); font-weight: 700; }}
    .topic-content em {{ color: var(--text-primary); font-style: italic; }}
    .topic-content blockquote {{
        border-left: 4px solid var(--accent-primary); padding-left: 1.5rem; margin: 1.5rem 0;
        font-style: italic; color: var(--text-secondary); background: rgba(127, 140, 141, 0.05);
        border-radius: 0 8px 8px 0; padding: 1rem 1.5rem;
    }}
    .formula-block {{
        padding: 1.5rem; background: rgba(44, 62, 80, 0.1); border-radius: 10px;
        text-align: center; overflow-x: auto; margin: 1.5rem 0; color: var(--text-primary);
        border: 1px solid var(--border-color);
    }}
    .table-container {{ overflow-x: auto; margin: 1.5rem 0; }}
    table {{ width: 100%; border-collapse: collapse; }}
    th, td {{ padding: 12px 15px; border: 1px solid var(--border-color); text-align: left; }}
    thead th {{ background-color: var(--bg-tertiary); color: var(--text-primary); font-weight: 600; }}
    tbody tr:nth-child(even) {{ background: rgba(127, 140, 141, 0.05); }}

    .MathJax {{
        color: var(--text-primary) !important;
    }}
    .MathJax_Display {{
        color: var(--text-primary) !important;
    }}

    footer {{ text-align: center; margin-top: 4rem; padding-top: 2rem; border-top: 1px solid var(--border-color); }}
    footer p {{ color: var(--text-secondary); font-size: 0.9rem; opacity: 0.8; }}

    @media (max-width: 768px) {{
        .container {{ padding: 1rem; }}
        .topic-header {{ padding: 1rem 1.5rem; }}
        .topic-card.open .topic-content {{ padding: 1rem 1.5rem; }}
        .formula-block {{ padding: 1rem; }}
    }}
  </style>
</head>
<body data-theme="dark">
  <div class="particles" id="particles-container"></div>
  <div class="theme-toggle" id="themeToggleButton" title="Cambiar tema"><i class="fas fa-moon" id="theme-icon"></i></div>
  <div class="container">
    <header class="header">
      <h1 class="main-title">{main_title}</h1>
    </header>

    <div class="lesson-container">
        {dynamic_content_html}
    </div>

    {conclusion_html}

    <footer>
      <p>{footer_text}</p>
    </footer>
  </div>

  <script>
    (function() {{
        const themeToggleButton = document.getElementById('themeToggleButton');
        const themeIcon = document.getElementById('theme-icon');
        const htmlEl = document.documentElement;

        function setTheme(theme) {{
            htmlEl.setAttribute('data-theme', theme);
            localStorage.setItem('theme', theme);
            if (themeIcon) {{ themeIcon.className = theme === 'dark' ? 'fas fa-sun' : 'fas fa-moon'; }}
        }}

        themeToggleButton.addEventListener('click', () => {{
            const newTheme = (htmlEl.getAttribute('data-theme') || 'light') === 'dark' ? 'light' : 'dark';
            setTheme(newTheme);
        }});

        setTheme(localStorage.getItem('theme') || 'dark');

        document.querySelectorAll('.topic-header').forEach(header => {{
            header.addEventListener('click', () => {{
                const card = header.parentElement;
                card.classList.toggle('open');
                if (window.MathJax) {{
                    window.MathJax.typesetPromise([card]).catch(function (err) {{
                        console.log('Error rendering MathJax:', err);
                    }});
                }}
            }});
        }});

        const container = document.getElementById('particles-container');
        if (container) {{
            for (let i = 0; i < 30; i++) {{
                const p = document.createElement('div');
                p.className = 'particle';
                p.style.left = Math.random() * 100 + 'vw';
                const size = (Math.random() * 5 + 2);
                p.style.width = size + 'px';
                p.style.height = size + 'px';
                p.style.animationDelay = Math.random() * -20 + 's';
                p.style.animationDuration = (15 + Math.random() * 10) + 's';
                container.appendChild(p);
            }}
        }}

        const firstTopic = document.querySelector('.topic-card');
        if(firstTopic) {{
            firstTopic.classList.add('open');
            if (window.MathJax) {{
                window.MathJax.typesetPromise([firstTopic]).catch(function (err) {{
                    console.log('Error rendering MathJax:', err);
                }});
            }}
        }}
    }})();
  </script>
</body>
</html>
"""

# --- INYECCIÓN DINÁMICA Y RENDERIZADO FINAL ---
final_html = plantilla_profesional.format(
    main_title="Distribución Binomial: Guía Completa",
    dynamic_content_html=contenido_dinamico_html,
    conclusion_html=conclusion_binomial_html,
    footer_text="Material interactivo Elaborado por Sergio Gevatschnaider"
)

display(HTML(final_html))
```

```python
# -*- coding: utf-8 -*-
"""
Distribución Binomial: Control de Calidad en la Industria de Semiconductores
===========================================================================

Aplicación práctica de la distribución binomial en la industria tecnológica.

Contenido del notebook / script
-------------------------------
1. Instalación y configuración
2. Fundamentos teóricos (resumen impreso para el usuario)
3. Implementación de la distribución binomial (clase BinomialQuality)
4. Visualizaciones interactivas
5. Simulación de control de calidad
6. Comparación teoría vs. práctica
7. Controles interactivos (ipywidgets)
8. Ejercicios prácticos
9. Conclusiones y aplicaciones
10. Limpieza y cierre
"""

# ============================================================================
# 1. INSTALACIÓN Y CONFIGURACIÓN
# ============================================================================

# Nota: Descomente la siguiente línea si ejecuta en un entorno limpio.
# !pip install plotly ipywidgets -q

import warnings
from typing import Dict, Any, List

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import binom
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display

warnings.filterwarnings("ignore")

plt.style.use("seaborn-v0_8")
plt.rcParams["figure.figsize"] = (12, 8)
plt.rcParams["font.size"] = 12

print("Librerías instaladas y configuradas correctamente.")

# ============================================================================
# 2. FUNDAMENTOS TEÓRICOS
# ============================================================================

print(
    """
CASO DE ESTUDIO: Control de Calidad en Semiconductores

Cada chip producido puede ser:
• Funcional (éxito, prob. 1 − p)
• Defectuoso (fracaso, prob. p)

Condiciones para la distribución binomial:
1. Número fijo de pruebas (n)
2. Solo dos resultados posibles
3. Probabilidad constante (p)
4. Pruebas independientes

Parámetros:
• n: tamaño del lote
• p: probabilidad de defecto
• X: número de defectuosos

Fórmula:  P(X = k) = C(n, k) p^k (1 − p)^(n − k)
"""
)

# ============================================================================
# 3. IMPLEMENTACIÓN
# ============================================================================

class BinomialQuality:
    """Clase para análisis de control de calidad."""

    def __init__(self, n: int, p: float):
        self.n = n
        self.p = p
        self.mean = n * p
        self.variance = n * p * (1 - p)
        self.std = np.sqrt(self.variance)

    def pmf(self, k: int) -> float:
        return binom.pmf(k, self.n, self.p)

    def cdf(self, k: int) -> float:
        return binom.cdf(k, self.n, self.p)

    def generate_distribution(self, max_k: int | None = None) -> pd.DataFrame:
        limit = min(self.n, 30) if max_k is None else max_k
        k_vals = np.arange(0, limit + 1)
        probs = [self.pmf(k) for k in k_vals]
        return pd.DataFrame(
            {
                "defectos": k_vals,
                "probabilidad": probs,
                "probabilidad_pct": np.array(probs) * 100,
            }
        )

    def simulate_batches(self, num_batches: int = 100) -> pd.DataFrame:
        out: List[Dict[str, Any]] = []
        for b in range(1, num_batches + 1):
            d = np.random.binomial(self.n, self.p)
            out.append({"lote": b, "defectos": d, "calidad": "Aceptable" if d <= 3 else "Rechazado"})
        return pd.DataFrame(out)

    def acceptance_probability(self, max_defects: int = 3) -> float:
        return self.cdf(max_defects)


quality_control = BinomialQuality(50, 0.05)
print("Ejemplo inicial → media:", quality_control.mean, "std:", quality_control.std)
print("Prob. aceptación (≤3 defectos):", quality_control.acceptance_probability())

# ============================================================================
# 4. VISUALIZACIONES INTERACTIVAS
# ============================================================================

def create_interactive_plots(n: int = 50, p: float = 0.05) -> None:

    qc = BinomialQuality(n, p)
    dist = qc.generate_distribution()

    fig = make_subplots(
        rows=2,
        cols=2,
        subplot_titles=(
            "Distribución de Probabilidad",
            "Distribución Acumulada",
            "Estadísticas Clave",
            "Criterio de Aceptación",
        ),
        specs=[[{"secondary_y": False}, {"secondary_y": False}],
               [{"secondary_y": False}, {"type": "domain"}]],  # ← "domain" permite el gráfico de pie
    )

    # Probabilidad
    fig.add_trace(
        go.Bar(x=dist["defectos"], y=dist["probabilidad_pct"], name="Probabilidad (%)", marker_color="steelblue"),
        row=1,
        col=1,
    )

    # Acumulada
    fig.add_trace(
        go.Scatter(x=dist["defectos"], y=dist["probabilidad"].cumsum() * 100, mode="lines+markers", name="Acumulada (%)", line=dict(color="red", width=3)),
        row=1,
        col=2,
    )

    # Estadísticas clave
    stats = ["Media", "Std", "Var"]
    vals = [qc.mean, qc.std, qc.variance]
    fig.add_trace(
        go.Bar(x=stats, y=vals, marker_color=["#FF6B6B", "#4ECDC4", "#45B7D1"], text=[f"{v:.2f}" for v in vals], textposition="auto"),
        row=2,
        col=1,
    )

    # Pie de aceptación
    acc = qc.acceptance_probability() * 100
    fig.add_trace(
        go.Pie(labels=["Aceptable", "Rechazado"], values=[acc, 100 - acc], marker_colors=["#2ECC71", "#E74C3C"], textinfo="label+percent"),
        row=2,
        col=2,
    )

    fig.update_layout(title=f"Control de Calidad (n={n}, p={p:.1%})", template="plotly_white", height=800)
    fig.show()


create_interactive_plots()

# ============================================================================
# 5. SIMULACIÓN DE LOTES
# ============================================================================

def run_quality_simulation(n: int = 50, p: float = 0.05, num_batches: int = 100):

    qc = BinomialQuality(n, p)
    results = qc.simulate_batches(num_batches)
    acc, rej = results[results["calidad"] == "Aceptable"], results[results["calidad"] == "Rechazado"]

    fig = make_subplots(
        rows=2,
        cols=2,
        subplot_titles=(
            "Histograma de Defectos",
            "Evolución Temporal",
            "Comparación Teórica vs Observada",
            "Distribución de Calidad",
        ),
        specs=[[{"secondary_y": False}, {"secondary_y": False}],
               [{"secondary_y": False}, {"type": "domain"}]],
    )

    fig.add_trace(go.Histogram(x=results["defectos"], nbinsx=15, marker_color="lightblue", opacity=0.7), row=1, col=1)

    sample = results.head(50)
    fig.add_trace(go.Scatter(x=sample["lote"], y=sample["defectos"], mode="lines+markers", line=dict(color="red")), row=1, col=2)
    fig.add_hline(y=qc.mean, line_dash="dash", line_color="blue", annotation_text=f"Media teórica {qc.mean:.2f}", row=1, col=2)

    dist = qc.generate_distribution()
    obs_freq = results["defectos"].value_counts().sort_index() / len(results) * 100
    k_vals = dist["defectos"]
    obs = [obs_freq.get(k, 0) for k in k_vals]
    fig.add_trace(go.Bar(x=k_vals, y=dist["probabilidad_pct"], name="Teórica", marker_color="steelblue", opacity=0.7), row=2, col=1)
    fig.add_trace(go.Bar(x=k_vals, y=obs, name="Observada", marker_color="orange", opacity=0.7), row=2, col=1)

    q_counts = results["calidad"].value_counts()
    fig.add_trace(go.Pie(labels=q_counts.index, values=q_counts.values, marker_colors=["#2ECC71", "#E74C3C"], textinfo="label+percent"), row=2, col=2)

    fig.update_layout(title=f"Simulación – {num_batches} lotes", template="plotly_white", height=800)
    fig.show()

    return results


simulation_results = run_quality_simulation()

# ============================================================================
# 6. COMPARACIÓN TEORÍA VS. PRÁCTICA
# ============================================================================

def compare_theory_practice(n: int = 50, p: float = 0.05, sims: int = 5) -> pd.DataFrame:

    qc = BinomialQuality(n, p)
    summary = []
    for i in range(sims):
        sim = qc.simulate_batches(100)
        summary.append({"sim": i + 1, "media": sim["defectos"].mean(), "std": sim["defectos"].std(), "acept": (sim["calidad"] == "Aceptable").mean() * 100})
    df = pd.DataFrame(summary)

    print("Teórico → media:", qc.mean, "std:", qc.std, "acept:", qc.acceptance_probability() * 100)
    print(df)

    return df


compare_theory_practice()

# ============================================================================
# 7. DEMO INTERACTIVA
# ============================================================================

def interactive_demo() -> None:

    n_slider = widgets.IntSlider(value=50, min=10, max=100, step=5, description="n (lote)")
    p_slider = widgets.FloatSlider(value=0.05, min=0.01, max=0.2, step=0.005, description="p defecto")
    m_slider = widgets.IntSlider(value=3, min=1, max=10, step=1, description="máx defectos")

    def update(n: int, p: float, m: int):
        qc = BinomialQuality(n, p)
        acc = qc.cdf(m) * 100
        print(f"Prob. aceptación (≤{m}): {acc:.1f}%")

    display(widgets.interactive_output(update, {"n": n_slider, "p": p_slider, "m": m_slider}))
    display(n_slider, p_slider, m_slider)


interactive_demo()

# ============================================================================
# 8. UTILIDADES PARA EJERCICIOS
# ============================================================================

def solve_exercise(num: int) -> None:

    if num == 1:
        n_vals, p = [20, 40, 60, 80, 100], 0.05
        for n in n_vals:
            qc = BinomialQuality(n, p)
            print(f"n={n} → media={qc.mean:.2f}, std={qc.std:.2f}, aceptación={qc.acceptance_probability() * 100:.1f}%")
    elif num == 3:
        n, p, target = 50, 0.05, 0.9
        qc = BinomialQuality(n, p)
        for m in range(n + 1):
            if qc.cdf(m) >= target:
                print(f"Máx defectos para {target*100}% aceptación: {m}")
                break


solve_exercise(1)
solve_exercise(3)

# ============================================================================
# 9. REPORTE FINAL (resumido)
# ============================================================================

def generate_final_report(n: int = 50, p: float = 0.05):
    qc = BinomialQuality(n, p)
    res = qc.simulate_batches(1000)
    print("Media obs:", res["defectos"].mean(), " vs teórica:", qc.mean)
    return res


generate_final_report()

# ============================================================================
# 10. LIMPIEZA Y CIERRE
# ============================================================================

def cleanup_and_summary() -> None:
    print("Resumen completado. Script listo para producción.")


cleanup_and_summary()

DEMO_CONFIG = {"version": "1.1", "author": "Demostración Estadística", "year": 2025}
print("Configuración:", DEMO_CONFIG)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Función para simular fallos en un sistema de red
def network_failure_simulation(lambda_failures, num_hours):
    # Simula el número de fallos en cada hora usando la distribución de Poisson
    failures_per_hour = np.random.poisson(lambda_failures, num_hours)
    total_failures = np.sum(failures_per_hour)

    # Cálculo de la esperanza y varianza
    esperanza = lambda_failures * num_hours
    varianza = lambda_failures * num_hours

    return failures_per_hour, total_failures, esperanza, varianza

# Función para ejecutar la simulación y mostrar resultados
def run_simulation(lambda_failures, num_hours):
    failures_per_hour, total_failures, esperanza, varianza = network_failure_simulation(lambda_failures, num_hours)

    # Mostrar resultados de la simulación
    print(f"Tasa de fallos por hora: {lambda_failures}")
    print(f"Total de horas simuladas: {num_hours}")
    print(f"Total de fallos observados: {total_failures}")
    print(f"Esperanza matemática de fallos: {esperanza}")
    print(f"Varianza de fallos: {varianza}")

    # Mostrar un gráfico de los fallos por hora
    plt.figure(figsize=(12, 6))
    plt.bar(range(1, num_hours + 1), failures_per_hour, color="blue")
    plt.axhline(esperanza / num_hours, color="red", linestyle="--", label="Esperanza por Hora")
    plt.xlabel("Hora")
    plt.ylabel("Cantidad de Fallos")
    plt.title("Simulación de Fallos en un Sistema de Red")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.show()

# Función para resetear los widgets y limpiar la salida
def reset_simulation(b):
    clear_output(wait=True)
    display(lambda_widget, hours_widget, simulate_button, reset_button)
    print("Valores reiniciados. Listo para una nueva simulación.")

# Widgets para la interfaz interactiva
lambda_widget = widgets.FloatSlider(value=2.0, min=0.5, max=10.0, step=0.5, description="Tasa de Fallos/h:")
hours_widget = widgets.IntSlider(value=24, min=1, max=168, step=1, description="Horas Simuladas:")

# Botón para iniciar la simulación
simulate_button = widgets.Button(description="Simular")
simulate_button.on_click(lambda b: run_simulation(lambda_widget.value, hours_widget.value))

# Botón para resetear los valores
reset_button = widgets.Button(description="Resetear", button_style='warning')
reset_button.on_click(reset_simulation)

# Mostrar interfaz
display(lambda_widget, hours_widget, simulate_button, reset_button)
```

```python
"""
Análisis didáctico de asimetría en distribuciones
=================================================

1. Configuración y utilidades
2. Explicación conceptual de la asimetría
3. Generación de datos ilustrativos
4. Estadísticas y explicación numérica
5. Visualizaciones comparativas
6. Resumen tabular
7. Conclusiones
"""

# -------------------------------------------------------------------
# 1. CONFIGURACIÓN INICIAL
# -------------------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
from scipy import stats
from IPython.display import display, HTML

# Visibilidad y reproducibilidad
np.random.seed(42)
plt.style.use("seaborn-v0_8")
sns.set_palette("colorblind")
plt.rcParams["figure.figsize"] = (15, 9)
plt.rcParams["font.size"] = 11

print("=" * 70)
print("ANÁLISIS DIDÁCTICO DE ASIMETRÍA EN DISTRIBUCIONES")
print("=" * 70, "\n")


# -------------------------------------------------------------------
# 2. FUNCIONES EDUCATIVAS
# -------------------------------------------------------------------
def explicar_concepto_asimetria() -> None:
    """Muestra los conceptos básicos de asimetría con texto estructurado."""
    texto = """
CONCEPTOS BÁSICOS DE ASIMETRÍA
------------------------------------------------------------------
La asimetría describe el grado y la dirección de sesgo en una distribución.

Distribución simétrica   (asimetría ≈ 0)
   • Media ≈ Mediana ≈ Moda
   • Ejemplo típico: distribución normal

Asimetría positiva       (asimetría > 0)
   • Media > Mediana > Moda
   • Cola larga hacia la derecha
   • Ejemplos: ingresos, tiempos de espera

Asimetría negativa       (asimetría < 0)
   • Media < Mediana < Moda
   • Cola larga hacia la izquierda
   • Ejemplos: puntuaciones en exámenes muy fáciles
"""
    print(texto)


def crear_datos_educativos() -> dict[str, dict]:
    """Genera tres conjuntos de datos con perfiles de asimetría distintos."""
    print("Generando datos ilustrativos...\n")

    datos = {
        "simetrica": {
            "data": np.random.normal(loc=50, scale=10, size=1000),
            "descripcion": "Distribución normal, equilibrada",
            "ejemplo": "Altura de personas",
            "color": "steelblue",
        },
        "positiva": {
            "data": np.random.exponential(scale=2, size=1000),
            "descripcion": "Distribución exponencial, cola a la derecha",
            "ejemplo": "Ingresos familiares",
            "color": "indianred",
        },
        "negativa": {
            # Para sesgo a la izquierda reflejamos una exponencial
            "data": 100 - np.random.exponential(scale=2, size=1000),
            "descripcion": "Distribución con cola a la izquierda",
            "ejemplo": "Notas altas en un examen fácil",
            "color": "seagreen",
        },
    }

    return datos


def calcular_estadisticas(data: np.ndarray) -> dict[str, float]:
    """Calcula estadísticas de tendencia, dispersión y asimetría."""
    media = data.mean()
    mediana = np.median(data)
    desv = data.std(ddof=0)
    var = desv**2
    q1, q3 = np.percentile(data, [25, 75])
    iqr = q3 - q1
    asim_pearson = 3 * (media - mediana) / desv if desv else np.nan
    asim_fisher = skew(data)

    return {
        "Media": media,
        "Mediana": mediana,
        "Desv.Std": desv,
        "Varianza": var,
        "Q1": q1,
        "Q3": q3,
        "IQR": iqr,
        "Asimetría (Pearson)": asim_pearson,
        "Asimetría (Fisher)": asim_fisher,
    }


def interpretar_asimetria(asim: float, media: float, mediana: float) -> None:
    """Imprime una interpretación cualitativa del coeficiente de asimetría."""
    print("Interpretación")
    print("--------------")
    if abs(asim) < 0.5:
        print("Distribución aproximadamente simétrica.")
    elif asim > 0:
        print("Distribución con asimetría positiva (sesgo a la derecha).")
    else:
        print("Distribución con asimetría negativa (sesgo a la izquierda).")
    print(f"Diferencia media−mediana: {media - mediana:.2f}\n")


# -------------------------------------------------------------------
# 3. VISUALIZACIONES
# -------------------------------------------------------------------
def visualizacion_comparativa(datos: dict[str, dict]) -> None:
    """Crea histogramas, box‑plots y Q‑Q plots para los tres casos."""
    tipos = ["simetrica", "positiva", "negativa"]
    titulos = [
        "Distribución Simétrica",
        "Asimetría Positiva",
        "Asimetría Negativa",
    ]

    fig = plt.figure(constrained_layout=True, figsize=(18, 12))
    gs = fig.add_gridspec(nrows=3, ncols=3, hspace=0.35, wspace=0.3)

    for col, (tipo, titulo) in enumerate(zip(tipos, titulos)):
        serie = datos[tipo]["data"]
        color = datos[tipo]["color"]
        media, mediana = serie.mean(), np.median(serie)
        asim = skew(serie)

        # Histograma + KDE
        ax_hist = fig.add_subplot(gs[0, col])
        sns.histplot(serie, kde=True, stat="density", color=color, alpha=0.75, ax=ax_hist)
        ax_hist.axvline(media, color="black", linestyle="--", lw=1.5, label=f"Media = {media:.1f}")
        ax_hist.axvline(mediana, color="darkblue", lw=1.5, label=f"Mediana = {mediana:.1f}")
        ax_hist.set_title(f"{titulo}\nCoef. asimetría = {asim:.2f}", fontweight="bold")
        ax_hist.legend()
        ax_hist.grid(alpha=0.3)

        # Box‑plot
        ax_box = fig.add_subplot(gs[1, col])
        sns.boxplot(x=serie, color=color, ax=ax_box)
        ax_box.set_title("Box‑plot", fontsize=10)
        ax_box.grid(alpha=0.3)

        # Q‑Q plot
        ax_qq = fig.add_subplot(gs[2, col])
        stats.probplot(serie, dist="norm", plot=ax_qq)
        ax_qq.set_title("Q‑Q plot frente a normal", fontsize=10)
        ax_qq.grid(alpha=0.3)

    fig.suptitle("Comparación visual de asimetría", fontsize=16, y=0.98)
    plt.show()


# -------------------------------------------------------------------
# 4. RESUMEN TABULAR
# -------------------------------------------------------------------
def resumen_comparativo(datos: dict[str, dict]) -> pd.DataFrame:
    """Devuelve un DataFrame resumen de las tres distribuciones."""
    filas = []
    for clave, info in datos.items():
        fila = calcular_estadisticas(info["data"])
        fila["Tipo"] = clave.capitalize()
        fila["Descripción"] = info["descripcion"]
        filas.append(fila)

    df = pd.DataFrame(filas).set_index("Tipo")
    df = df.round({"Media": 2, "Mediana": 2, "Desv.Std": 2, "Asimetría (Fisher)": 3})
    return df


# -------------------------------------------------------------------
# 5. CONCLUSIONES
# -------------------------------------------------------------------
def mostrar_conclusiones() -> None:
    texto = """
CONCLUSIONES DIDÁCTICAS
------------------------------------------------------------------
• Relación media‑mediana:
  ─ Simétrica      → valores muy próximos.
  ─ Asimetría +    → media mayor que mediana.
  ─ Asimetría –    → media menor que mediana.

• Herramientas gráficas complementarias:
  ─ Histograma y curva KDE     → forma general.
  ─ Box‑plot                   → valores atípicos e IQR.
  ─ Q‑Q plot                   → comparación con normalidad.

• Regla práctica:
  ─ |asimetría| < 0.5  → casi simétrica
  ─ 0.5–1.0           → moderada asimetría
  ─ > 1.0             → alta asimetría

Aplicaciones típicas: análisis financiero, estudios de mercado, control de calidad e investigación científica.
"""
    print(texto)


# -------------------------------------------------------------------
# 6. PROGRAMA PRINCIPAL
# -------------------------------------------------------------------
def main() -> None:
    explicar_concepto_asimetria()

    datos = crear_datos_educativos()

    print("\nESTADÍSTICAS INDIVIDUALES\n" + "-" * 70)
    for nombre, info in datos.items():
        print(f"\n{info['descripcion']}  (ejemplo: {info['ejemplo']})")
        stats_dict = calcular_estadisticas(info["data"])
        for k, v in stats_dict.items():
            if isinstance(v, float):
                print(f"{k:>20}: {v:>8.2f}")
        interpretar_asimetria(stats_dict["Asimetría (Fisher)"],
                              stats_dict["Media"],
                              stats_dict["Mediana"])
        print("-" * 60)

    visualizacion_comparativa(datos)

    df = resumen_comparativo(datos)
    print("\nRESUMEN COMPARATIVO\n" + "-" * 70)
    display(HTML(df.to_html(classes="table table-striped", border=0)))

    mostrar_conclusiones()
    print("\nAnálisis finalizado.")


# -------------------------------------------------------------------
# 7. EJECUCIÓN
# -------------------------------------------------------------------
if __name__ == "__main__":
    main()
```

```python
"""
Comparativa visual de curtosis
==============================

Se ilustran tres perfiles de curtosis:

1. Leptocúrtica  → pico alto y colas gruesas
2. Mesocúrtica   → distribución normal estándar
3. Platicúrtica  → pico bajo y colas finas

Para cada caso se muestran:
• Histograma + KDE superpuesta
• Curva normal ajustada
• Líneas de media y mediana
• Curtosis de Fisher impresa
"""

# ------------------------------------------------------------------
# 1. CONFIGURACIÓN GENERAL
# ------------------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm, kurtosis, skew

np.random.seed(7)
plt.style.use("seaborn-v0_8")
sns.set_palette("colorblind")
plt.rcParams["figure.figsize"] = (18, 6)
plt.rcParams["font.size"] = 11

# ------------------------------------------------------------------
# 2. GENERACIÓN DE DATOS
# ------------------------------------------------------------------
n = 1500

# Leptocúrtica: distribución con colas pesadas
data_lepto = np.random.standard_t(df=3, size=n)           # t de Student con pocas d.f.

# Mesocúrtica: normal estándar
data_meso = np.random.normal(loc=0, scale=1, size=n)

# Platicúrtica: uniforme
data_plati = np.random.uniform(low=-3, high=3, size=n)

datasets = {
    "Leptocúrtica": {"data": data_lepto, "color": "royalblue"},
    "Mesocúrtica": {"data": data_meso, "color": "black"},
    "Platicúrtica": {"data": data_plati, "color": "darkorange"},
}

# ------------------------------------------------------------------
# 3. FUNCIÓN AUXILIAR DE GRÁFICO
# ------------------------------------------------------------------
def plot_dist(ax, serie, label, color):
    """Dibuja histograma, KDE y curva normal ajustada con anotaciones."""
    sns.histplot(
        serie, bins=35, stat="density", alpha=0.65, color=color, kde=True, ax=ax
    )

    # Curva normal ajustada (media y desvío de la muestra)
    mu, sigma = serie.mean(), serie.std(ddof=0)
    x = np.linspace(mu - 4 * sigma, mu + 4 * sigma, 400)
    ax.plot(x, norm.pdf(x, mu, sigma), color=color, linewidth=2.0, label="Normal ajustada")

    # Líneas verticales de media y mediana
    ax.axvline(mu, color="red", linestyle="--", linewidth=1.2, label=f"Media = {mu:.2f}")
    med = np.median(serie)
    ax.axvline(med, color="green", linestyle="-.", linewidth=1.2, label=f"Mediana = {med:.2f}")

    # Cálculo de métricas
    k = kurtosis(serie, fisher=True)
    s = skew(serie)
    ax.set_title(
        f"{label}\nCurtosis = {k:.2f}   |   Asimetría = {s:.2f}",
        fontsize=12,
        fontweight="bold",
    )
    ax.set_xlabel("Valor")
    ax.set_ylabel("Densidad")
    ax.legend(frameon=False)
    ax.grid(alpha=0.3)

# ------------------------------------------------------------------
# 4. VISUALIZACIÓN COMPARATIVA
# ------------------------------------------------------------------
fig, axes = plt.subplots(1, 3, constrained_layout=True)

for ax, (nombre, info) in zip(axes, datasets.items()):
    plot_dist(ax, info["data"], nombre, info["color"])

fig.suptitle("Comparativa de curtosis: leptocúrtica, mesocúrtica y platicúrtica", fontsize=16, y=1.02)
plt.show()

# ------------------------------------------------------------------
# 5. RESUMEN TABULAR
# ------------------------------------------------------------------
filas = []
for nombre, info in datasets.items():
    serie = info["data"]
    filas.append(
        {
            "Distribución": nombre,
            "Media": serie.mean(),
            "Mediana": np.median(serie),
            "Desv.Std": serie.std(ddof=0),
            "Curtosis (Fisher)": kurtosis(serie, fisher=True),
            "Asimetría": skew(serie),
        }
    )

df = pd.DataFrame(filas).set_index("Distribución")
df = df.round({"Media": 2, "Mediana": 2, "Desv.Std": 2, "Curtosis (Fisher)": 2, "Asimetría": 2})

print("\nRESUMEN NUMÉRICO\n" + "-" * 70)
display(df)

"""
Interpretación rápida
---------------------
• Curtosis > 0 (leptocúrtica)  → colas pesadas, mayor probabilidad de valores extremos.
• Curtosis ≈ 0 (mesocúrtica)   → colas similares a la normal.
• Curtosis < 0 (platicúrtica)  → colas ligeras, distribución más aplanada.
"""
```

```python
# Correlación: Tipos, Cálculo e Interpretación ― versión interactiva (Google Colab)

# ======================================================================
# 1. IMPORTS
# ======================================================================
from IPython.display import display, HTML
import html  # para escapar caracteres en los títulos

# ======================================================================
# 2. CONTENIDO FIJO (INTRO & CONCLUSIÓN)
# ======================================================================

introduction_corr_html = r"""
<div class="content-block">
  <h2>¿Qué es la Correlación?</h2>
  <p>La <strong>correlación</strong> mide la intensidad y dirección de la relación estadística entre dos variables cuantitativas. El coeficiente de Pearson (<em>r</em>) cuantifica exclusivamente la <em>relación lineal</em>, mientras que otros coeficientes (Spearman, Kendall) capturan asociaciones <em>monótonas</em>. Todas las métricas se acotan en el intervalo <span class="formula-block-inline">$[-1,\,1]$</span>.</p>
  <ul>
    <li><strong>$r = +1$</strong>: alineación ascendente perfecta.</li>
    <li><strong>$0 &lt; r &lt; 1$</strong>: asociación lineal positiva parcial.</li>
    <li><strong>$r = 0$</strong>: sin relación lineal (puede existir dependencia no lineal).</li>
    <li><strong>$-1 &lt; r &lt; 0$</strong>: asociación lineal negativa.</li>
    <li><strong>$r = -1$</strong>: alineación descendente perfecta.</li>
  </ul>
  <p>Recuerde: <em>correlación no implica causalidad</em>; se requieren diseños experimentales o marcos de inferencia causal para establecerla.</p>
</div>
"""

conclusion_corr_html = r"""
<div class="content-block">
  <h2>Conclusión</h2>
  <p>Evaluar correlaciones va más allá de calcular un número. Exige inspección gráfica, comprobación de supuestos (normalidad, homocedasticidad, independencia), control de variables confusoras y, cuando sea necesario, el uso de coeficientes robustos o no paramétricos. Sólo así se evitan las <em>correlaciones espurias</em> y se fundamentan conclusiones fiables.</p>
</div>
"""

# ======================================================================
# 3. TARJETAS DINÁMICAS
# ======================================================================

corr_definitions_data = [
    # ------------------------------------------------------------------
    # Tarjeta 1 ─ Definición y Propiedades
    # ------------------------------------------------------------------
    {
        "titulo": "1. Definición formal y propiedades del coeficiente $r$",
        "contenido": r"""
          <h4>Formulación matemática</h4>
          <p>Para muestras $(x_i,y_i)$, $i=1,\dots,n$:</p>
          <div class="formula-block">$$
          r \;=\;
          \frac{\displaystyle \sum_{i=1}^n (x_i-\bar{x})(y_i-\bar{y})}
               {\displaystyle (n-1)\,s_x\,s_y},
          \qquad
          s_x^2 = \tfrac{1}{n-1}\!\sum (x_i-\bar{x})^2
          $$</div>
          <h4>Propiedades clave</h4>
          <ul>
            <li><strong>Invarianza</strong> a traslación y escala positiva.</li>
            <li><strong>Simetría</strong>: $\mathrm{Corr}(X,Y)=\mathrm{Corr}(Y,X)$.</li>
            <li>Relación con la pendiente de mínimos cuadrados: $\hat{\beta}_1 = r\,s_y / s_x$.</li>
          </ul>
          <blockquote>Correlación ≠ causalidad.  Se requiere controlar variables confusoras o usar SEM / do‑calculus para inferir causa.</blockquote>
        """
    },

    # ------------------------------------------------------------------
    # Tarjeta 2 ─ Tipos de coeficiente
    # ------------------------------------------------------------------
    {
        "titulo": "2. Coeficientes principales: Pearson, Spearman y Kendall",
        "contenido": r"""
          <div class="table-container">
            <table>
              <thead>
                <tr>
                  <th>Coeficiente</th><th>Datos requeridos</th>
                  <th>Relación captada</th><th>Ventajas</th><th>Limitaciones</th>
                </tr>
              </thead>
              <tbody>
                <tr>
                  <td><strong>Pearson&nbsp;$r$</strong></td>
                  <td>Métricos (intervalo/razón)</td>
                  <td>Lineal</td>
                  <td>Interpretación geométrica; base de regresión</td>
                  <td>Sensible a outliers; asume homocedasticidad</td>
                </tr>
                <tr>
                  <td><strong>Spearman&nbsp;$\rho_s$</strong></td>
                  <td>Rangos (ordinal ↔ rangos)</td>
                  <td>Monótona</td>
                  <td>Robusto a outliers; detecta no lineal monótona</td>
                  <td>Menor potencia si la relación es estrictamente lineal</td>
                </tr>
                <tr>
                  <td><strong>Kendall&nbsp;$\tau$</strong></td>
                  <td>Rangos</td>
                  <td>Monótona (concordancias)</td>
                  <td>Interpretación probabilística clara; alta robustez</td>
                  <td>$\mathcal{O}(n^2)$ en cómputo; valores más bajos que $\rho_s$</td>
                </tr>
              </tbody>
            </table>
          </div>
          <p>Otros coeficientes (biweight, Hoeffding, distancia de energía) detectan dependencias arbitrarias a costa de mayor complejidad.</p>
        """
    },

    # ------------------------------------------------------------------
    # Tarjeta 3 ─ Cálculo práctico de Pearson
    # ------------------------------------------------------------------
    {
        "titulo": "3. Cálculo detallado del coeficiente de Pearson",
        "contenido": r"""
          <h4>Algoritmo estable</h4>
          <ol>
            <li>Centrar: $x'_i=x_i-\bar{x}$, $y'_i=y_i-\bar{y}$.</li>
            <li>Producto cruzado: $S_{xy}=\sum x'_i\,y'_i$.</li>
            <li>Desviaciones estándar: $s_x$, $s_y$.</li>
            <li>Cociente estandarizado: $r = \dfrac{S_{xy}}{(n-1)s_xs_y}\,.$</li>
          </ol>
          <h4>Ejemplo numérico</h4>
          <p>Para $X=(3,4,5)$ y $Y=(2,4,6)$ se obtiene $r=1$ (alineación perfecta).</p>
          <h4>Implementación Python</h4>
          <pre><code class="language-python">import numpy as np
def pearson(x, y):
    x = np.asarray(x, float)
    y = np.asarray(y, float)
    if x.size != y.size:
        raise ValueError("Longitudes distintas")
    n   = x.size
    xc  = x - x.mean()
    yc  = y - y.mean()
    return (xc @ yc) / ((n - 1) * xc.std(ddof=1) * yc.std(ddof=1))</code></pre>
        """
    },

    # ------------------------------------------------------------------
    # Tarjeta 4 ─ Interpretación y significación
    # ------------------------------------------------------------------
    {
        "titulo": "4. Interpretación, prueba de hipótesis e IC de Fisher‑$z$",
        "contenido": r"""
          <h4>Magnitud de $|r|$ (ciencias sociales)</h4>
          <div class="table-container">
            <table>
              <thead><tr><th>|r|</th><th>Calificativo</th></tr></thead>
              <tbody>
                <tr><td>0–0.20</td><td>Despreciable</td></tr>
                <tr><td>0.20–0.40</td><td>Débil</td></tr>
                <tr><td>0.40–0.60</td><td>Moderada</td></tr>
                <tr><td>0.60–0.80</td><td>Fuerte</td></tr>
                <tr><td>0.80–1.00</td><td>Muy fuerte</td></tr>
              </tbody>
            </table>
          </div>
          <h4>Prueba <em>t</em> para $H_0:\rho=0$</h4>
          <div class="formula-block">$$
          t = \frac{r\sqrt{\,n-2\,}}{\sqrt{1-r^{2}}}\;\sim t_{\,n-2}
          $$</div>
          <h4>Intervalo de confianza (transformación Fisher‑$z$)</h4>
          <div class="formula-block">$$
          z = \tfrac{1}{2}\ln\!\bigl(\tfrac{1+r}{1-r}\bigr), \quad
          \text{Var}(z)=\frac{1}{n-3}
          $$</div>
          <h4>Coeficiente de determinación</h4>
          <p>$R^{2}=r^{2}$ representa la fracción de varianza explicada linealmente.</p>
        """
    },

    # ------------------------------------------------------------------
    # Tarjeta 5 ─ Riesgos, correlaciones espurias y buenas prácticas
    # ------------------------------------------------------------------
    {
        "titulo": "5. Precauciones y correlación espuria",
        "contenido": r"""
          <h4>Fuentes de distorsión</h4>
          <ul>
            <li><strong>Outliers</strong>: pueden inflar o cambiar el signo de $r$.</li>
            <li><strong>Heterocedasticidad</strong>: varianza de $Y$ cambia con $X$.</li>
            <li><strong>Series temporales</strong>: autocorrelación genera correlaciones espurias.</li>
            <li><strong>Confusión / causalidad inversa</strong>: variable omitida induce asociación.</li>
          </ul>
          <h4>Mitigaciones recomendadas</h4>
          <ol>
            <li>Análisis robusto (Spearman/Kendall, winsorización).</li>
            <li>Transformaciones o modelos con varianza estructurada.</li>
            <li>Diferenciar series, test de cointegración para variables dependientes del tiempo.</li>
            <li>Diseño experimental, variables instrumentales o SCM para causalidad.</li>
          </ol>
        """
    }
]

# ======================================================================
# 4. FUNCIÓN PARA GENERAR TARJETAS
# ======================================================================

def generar_tarjetas_definiciones(datos):
    tarjetas = []
    for sec in datos:
        titulo = html.escape(sec["titulo"], quote=False)
        titulo = titulo.replace('(', '$(').replace(')', ')$')  # habilita MathJax en títulos
        contenido = sec["contenido"]
        tarjeta = f"""
        <div class="topic-card" aria-expanded="false">
          <button class="topic-header" aria-label="Mostrar/ocultar sección">
            <span class="topic-title">{titulo}</span>
            <i class="fas fa-chevron-down expand-icon" aria-hidden="true"></i>
          </button>
          <div class="topic-content">{contenido}</div>
        </div>
        """
        tarjetas.append(tarjeta)
    return "\n".join(tarjetas)

dynamic_content_html = generar_tarjetas_definiciones(corr_definitions_data)

# ======================================================================
# 5. PLANTILLA PROFESIONAL (mejoras: ARIA, altura dinámica, mobile‑friendly)
# ======================================================================

plantilla_profesional = r"""
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>{main_title}</title>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700;800&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.5.0/css/all.min.css" rel="stylesheet">
  <script>
    window.MathJax = {{
      tex: {{
        inlineMath: [['$', '$'], ['\\(', '\\)']],
        displayMath: [['$$', '$$']],
        processEscapes: true
      }},
      svg: {{ fontCache: 'global' }}
    }};
  </script>
  <script src="https://cdn.jsdelivr.net/npm/mathjax@3/es5/tex-mml-chtml.js" async id="MathJax-script"></script>

  <style>
    :root {{
      --bg-1: linear-gradient(135deg,#2d5be3 0%,#7c4dff 100%);
      --bg-card: rgba(255,255,255,.9);
      --bg-card-dark: rgba(26,32,44,.9);
      --text-1: #2c3e50;
      --text-2: #4a5568;
      --accent: #2d5be3;
      --radius: 20px;
      --shadow: 0 15px 35px rgba(0,0,0,.1);
    }}
    [data-theme="dark"] {{
      --bg-card: rgba(45,55,72,.85);
      --text-1: #f7fafc;
      --text-2: #a0aec0;
      --accent: #7aa0ff;
      --shadow: 0 15px 35px rgba(0,0,0,.3);
    }}
    * {{ box-sizing: border-box; margin: 0; padding: 0; }}
    body {{
      font-family: 'Inter', sans-serif;
      background: var(--bg-1); color: var(--text-1);
      min-height: 100vh; line-height: 1.8; transition: background .4s, color .4s;
    }}
    .container {{ max-width: 950px; margin: 0 auto; padding: 2rem; }}
    h1.main-title {{
      font-size: clamp(2.3rem,5vw,4rem); font-weight: 800;
      background: linear-gradient(135deg,#ffffff 0%,#d4d2ff 100%);
      -webkit-background-clip: text; -webkit-text-fill-color: transparent;
      text-align: center; margin-bottom: 2.5rem;
    }}
    .content-block, .topic-card {{
      background: var(--bg-card); border-radius: var(--radius);
      box-shadow: var(--shadow); border: 1px solid rgba(255,255,255,.3);
      backdrop-filter: blur(12px);
    }}
    .content-block {{ padding: 2rem; margin-bottom: 2rem; }}
    .content-block h2 {{ border-bottom: 2px solid var(--accent); padding-bottom: .5rem; margin-bottom: 1rem; }}
    .content-block p, .content-block li {{ color: var(--text-2); }}
    .formula-block {{ padding: 1rem; background: rgba(0,0,0,.05); border-radius: 10px; margin: 1.2rem 0; }}
    .formula-block-inline {{ white-space: nowrap; }}
    .table-container {{ overflow-x: auto; margin: 1.2rem 0; }}
    table {{ width: 100%; border-collapse: collapse; }}
    th,td {{ padding: 10px 14px; border: 1px solid rgba(0,0,0,.1); }}
    thead th {{ background: rgba(0,0,0,.04); }}

    /* Tarjetas */
    .topic-card + .topic-card {{ margin-top: 1.5rem; }}
    .topic-header {{
      width: 100%; text-align: left; background: none; border: none; cursor: pointer;
      padding: 1.3rem 2rem; display: flex; justify-content: space-between; align-items: center;
      font-size: 1.1rem; font-weight: 600; color: var(--text-1);
    }}
    .topic-content {{ max-height: 0; overflow: hidden; transition: max-height .6s ease, padding .6s ease; background: rgba(255,255,255,.08); }}
    .topic-card[aria-expanded="true"] .topic-content {{ padding: 1.5rem 2rem; }}
    .topic-card .expand-icon {{ transition: transform .6s; color: var(--text-2); }}
    .topic-card[aria-expanded="true"] .expand-icon {{ transform: rotate(180deg); color: var(--accent); }}

    /* Tema toggle */
    .theme-toggle {{
      position: fixed; top: 1.5rem; right: 1.5rem; width: 50px; height: 50px;
      border-radius: 50%; background: var(--bg-card); display: flex; align-items: center; justify-content: center;
      box-shadow: var(--shadow); cursor: pointer; border: 1px solid rgba(255,255,255,.4);
      transition: transform .4s;
    }}
    .theme-toggle:hover {{ transform: scale(1.1) rotate(180deg); }}
    @media (max-width: 600px) {{
      .container {{ padding: 1rem; }}
      .topic-header {{ padding: 1rem 1.3rem; }}
    }}
  </style>
</head>
<body data-theme="dark">
  <button class="theme-toggle" id="themeBtn" aria-label="Cambiar tema">
    <i class="fas fa-moon" id="themeIcon"></i>
  </button>

  <main class="container">
    <h1 class="main-title">{main_title}</h1>
    {introduction_html}
    <section class="lesson-container">{dynamic_content_html}</section>
    {conclusion_html}
    <footer style="text-align:center;margin-top:3rem;color:var(--text-2);font-size:.85rem;">
      {footer_text}
    </footer>
  </main>

  <script>
  // ────────────────── Tema claro/oscuro ───────────────────────────────
  (function() {{
    const body = document.body;
    const btn  = document.getElementById('themeBtn');
    const ico  = document.getElementById('themeIcon');
    function setTheme(t) {{
      body.setAttribute('data-theme', t);
      localStorage.setItem('theme', t);
      ico.className = t === 'dark' ? 'fas fa-sun' : 'fas fa-moon';
    }}
    btn.addEventListener('click', () => {{
      setTheme(body.getAttribute('data-theme') === 'dark' ? 'light' : 'dark');
    }});
    setTheme(localStorage.getItem('theme') || 'dark');
  }})();
  // ────────────────── Tarjetas desplegables ───────────────────────────
  document.querySelectorAll('.topic-card').forEach(card => {{
    const header = card.querySelector('.topic-header');
    const content = card.querySelector('.topic-content');
    header.addEventListener('click', () => {{
      const open = card.getAttribute('aria-expanded') === 'true';
      card.setAttribute('aria-expanded', !open);
      content.style.maxHeight = open ? 0 : content.scrollHeight + 'px';
      if (window.MathJax) {{
        MathJax.typesetPromise([content]).catch(e=>console.warn(e));
      }}
    }});
  }});
  // Tipografía inicial
  window.addEventListener('load', () => {{
    if (window.MathJax) MathJax.typesetPromise();
  }});
  </script>
</body>
</html>
"""

# ======================================================================
# 6. RENDER FINAL
# ======================================================================

final_html = plantilla_profesional.format(
    main_title="Correlación: Fundamentos, Tipos y Buenas Prácticas",
    introduction_html=introduction_corr_html,
    dynamic_content_html=dynamic_content_html,
    conclusion_html=conclusion_corr_html,
    footer_text="Material interactivo Profesor Sergio Gevatschnaider."
)

display(HTML(final_html))
```

```python
"""
Visualización  de la correlación
-----------------------------------------
Se generan tres pares de variables con:

1. Correlación positiva    (pendiente ≈ +2)
2. Correlación negativa    (pendiente ≈ –2)
3. Correlación nula        (ruido aleatorio)

Para cada caso:
• Scatter‑plot con línea de ajuste lineal (OLS)
• Cálculo del coeficiente de Pearson  r  y  R²
• Tabla comparativa de métricas
"""

# -------------------------------------------------------------------
# 1. CONFIGURACIÓN GENERAL
# -------------------------------------------------------------------
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, linregress

np.random.seed(0)
sns.set(style="whitegrid", palette="colorblind")
plt.rcParams["figure.figsize"] = (18, 5)
plt.rcParams["font.size"] = 10

# -------------------------------------------------------------------
# 2. GENERACIÓN DE DATOS
# -------------------------------------------------------------------
n = 150
x_vals = np.linspace(0, 10, n)

data = {
    "Positiva": {
        "x": x_vals,
        "y": 2 * x_vals + np.random.normal(0, 2, n),
        "color": "royalblue",
    },
    "Negativa": {
        "x": x_vals,
        "y": -2 * x_vals + np.random.normal(0, 2, n),
        "color": "indianred",
    },
    "Sin relación": {
        "x": x_vals,
        "y": np.random.normal(0, 5, n),
        "color": "darkorange",
    },
}

# -------------------------------------------------------------------
# 3. FUNCIÓN DE TRAZADO
# -------------------------------------------------------------------
def draw_panel(ax, x, y, label, color):
    """Dibuja scatter‑plot con ajuste lineal y anota métricas."""
    # Ajuste lineal
    slope, intercept, r_value, p_value, stderr = linregress(x, y)
    r2 = r_value**2

    sns.scatterplot(x=x, y=y, color=color, edgecolor="black", alpha=0.65, s=55, ax=ax)
    sns.regplot(
        x=x,
        y=y,
        scatter=False,
        line_kws={"color": "black", "linewidth": 2, "linestyle": "--"},
        ax=ax,
    )

    ax.set_title(label, fontsize=12, fontweight="bold")
    ax.set_xlabel("X")
    ax.set_ylabel("Y")

    # Anotaciones
    txt = (
        f"pendiente = {slope:.2f}\n"
        f"intercepto = {intercept:.2f}\n"
        f"r = {r_value:.2f}\n"
        f"R² = {r2:.2f}"
    )
    ax.text(
        0.05,
        0.95,
        txt,
        transform=ax.transAxes,
        fontsize=9,
        verticalalignment="top",
        bbox=dict(boxstyle="round,pad=0.3", facecolor="white", alpha=0.6),
    )

    ax.grid(alpha=0.4)

    # Devolver métricas para la tabla
    return {
        "Tipo": label,
        "Pendiente": slope,
        "Intercepto": intercept,
        "Coef. Pearson (r)": r_value,
        "p‑value": p_value,
        "R²": r2,
    }


# -------------------------------------------------------------------
# 4. VISUALIZACIÓN COMPARATIVA
# -------------------------------------------------------------------
fig, axes = plt.subplots(1, 3, constrained_layout=True)
metrics = []

for ax, (nombre, info) in zip(axes, data.items()):
    res = draw_panel(ax, info["x"], info["y"], nombre, info["color"])
    metrics.append(res)

fig.suptitle("Comparación de distintos grados de correlación", fontsize=15, y=1.03)
plt.show()

# -------------------------------------------------------------------
# 5. TABLA RESUMEN
# -------------------------------------------------------------------
df_metrics = pd.DataFrame(metrics).set_index("Tipo")
df_metrics = df_metrics.round({"Pendiente": 2, "Intercepto": 2, "Coef. Pearson (r)": 2, "R²": 2, "p‑value": 3})

print("\nRESUMEN DE MÉTRICAS\n" + "-" * 70)
display(df_metrics)
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-

"""
Script para analizar y visualizar la casi-uniformidad estadística
de la salida de SHA-256.

Realiza las siguientes operaciones:
1.  Genera N mensajes aleatorios de 32 bytes.
2.  Calcula el hash SHA-256 para cada mensaje.
3.  Analiza la distribución del primer byte y la visualiza con un histograma.
4.  Analiza la proporción de bits '1' y la muestra en un gráfico circular.
5.  Aplica una prueba de chi-cuadrado para evaluar la uniformidad.
"""

import hashlib
import secrets
import math
from collections import Counter
from typing import Counter as CounterType, Tuple, List

# --- Dependencias Externas ---
# Instalar con: pip install scipy matplotlib seaborn tqdm
try:
    from scipy.stats import chi2
    import matplotlib.pyplot as plt
    import seaborn as sns
    from tqdm import tqdm
except ImportError:
    print("Error: Faltan dependencias. Por favor, instálalas con:")
    print("pip install scipy matplotlib seaborn tqdm")
    exit(1)


# --- Constantes de la Simulación ---
NUM_MUESTRAS = 100_000                # Nº de mensajes aleatorios a generar
BYTES_ENTRADA = 32                    # Longitud de los mensajes de entrada
BITS_POR_DIGEST = 256                 # 32 bytes de salida × 8 bits/byte
GRADOS_LIBERTAD = 255                 # 256 posibles valores de byte - 1

def ejecutar_simulacion(n: int) -> Tuple[CounterType[int], int]:
    """
    Genera n hashes y cuenta las estadísticas básicas.

    Args:
        n: Número de mensajes aleatorios a generar.

    Returns:
        Un tuple conteniendo:
        - Un contador con la frecuencia de cada valor del primer byte.
        - El número total de bits '1' observados en todos los digests.
    """
    contador_bytes = Counter()
    total_de_unos = 0

    print(f"[*] Generando {n:,} hashes SHA-256...")
    # Usamos tqdm para una barra de progreso visualmente atractiva
    for _ in tqdm(range(n), desc="Generando Hashes", unit=" hash"):
        mensaje = secrets.token_bytes(BYTES_ENTRADA)
        digest = hashlib.sha256(mensaje).digest()

        # 1. Contamos la aparición del primer byte
        contador_bytes[digest[0]] += 1

        # 2. Contamos los bits '1' en todo el digest (usando el método optimizado)
        total_de_unos += sum(b.bit_count() for b in digest)

    print("[*] Simulación finalizada.\n")
    return contador_bytes, total_de_unos

def analizar_y_reportar(contador_bytes: CounterType[int], total_de_unos: int, n: int):
    """
    Calcula las métricas estadísticas, presenta los resultados en texto
    y llama a la función de visualización.
    """
    # --- (1) Estadística del primer byte ---
    valor_esperado = n / (GRADOS_LIBERTAD + 1)
    frecuencias_observadas = [contador_bytes.get(i, 0) for i in range(256)]

    chi2_stat = sum((c - valor_esperado)**2 / valor_esperado for c in frecuencias_observadas)
    p_valor = chi2.sf(chi2_stat, GRADOS_LIBERTAD)

    print("=== ANÁLISIS ESTADÍSTICO ===")
    print("\n--- Frecuencia del Primer Byte ---")
    print(f"   Min apariciones = {min(frecuencias_observadas)}")
    print(f"   Max apariciones = {max(frecuencias_observadas)}")
    print(f"   Media esperada  = {valor_esperado:.2f}")
    print(f"   Estadístico χ²  = {chi2_stat:.2f} (gl = {GRADOS_LIBERTAD})")
    print(f"   p-valor         = {p_valor:.4f}")

    if p_valor > 0.05:
        print("   Conclusión: La distribución de bytes es consistente con una distribución uniforme (p > 0.05).")
    else:
        print("   Conclusión: La distribución de bytes se desvía de una distribución uniforme (p <= 0.05).")

    # --- (2) Proporción de bits a 1 en TODA la salida ---
    bits_totales = n * BITS_POR_DIGEST
    p_hat = total_de_unos / bits_totales
    entropia_por_bit = -p_hat * math.log2(p_hat) - (1 - p_hat) * math.log2(1 - p_hat)

    print("\n--- Distribución Global de Bits ---")
    print(f"   Proporción de '1' (p̂) = {p_hat:.5f}  (Ideal: 0.50000)")
    print(f"   Entropía por bit (H)  = {entropia_por_bit:.6f} bits (Ideal: 1.000000)")

    # --- (3) Llamada a la función de visualización ---
    crear_visualizaciones(frecuencias_observadas, total_de_unos, bits_totales, valor_esperado)

def crear_visualizaciones(frecuencias: List[int], total_unos: int, total_bits: int, esperado: float):
    """Genera y muestra los gráficos basados en los datos de la simulación."""

    sns.set_theme(style="whitegrid")

    # --- Gráfico 1: Histograma de Frecuencia de Bytes ---
    plt.figure(figsize=(15, 7))
    ax = sns.barplot(x=list(range(256)), y=frecuencias, color="cornflowerblue")
    ax.axhline(y=esperado, color='red', linestyle='--', label=f'Frecuencia Esperada ({esperado:.2f})')

    plt.title(f'Distribución del Primer Byte de {NUM_MUESTRAS:,} Hashes SHA-256', fontsize=16)
    plt.xlabel('Valor del Byte (0-255)', fontsize=12)
    plt.ylabel('Frecuencia Observada', fontsize=12)
    # Optimiza las etiquetas del eje x para que no se superpongan
    ax.set_xticks(range(0, 256, 16))
    ax.tick_params(axis='x', rotation=45)
    plt.legend()
    plt.tight_layout() # Ajusta el gráfico para que quepa todo

    # --- Gráfico 2: Proporción Global de Bits ---
    total_ceros = total_bits - total_unos
    labels = [f'Bits "1"\n({total_unos:,})', f'Bits "0"\n({total_ceros:,})']
    sizes = [total_unos, total_ceros]
    colors = ['gold', 'lightskyblue']
    explode = (0.01, 0.01) # Ligeramente separados para mejor estética

    plt.figure(figsize=(8, 8))
    plt.pie(sizes, explode=explode, labels=labels, colors=colors,
            autopct='%1.4f%%', shadow=True, startangle=140,
            textprops={'fontsize': 12})
    plt.title(f'Proporción Global de Bits en {total_bits:,} bits totales', fontsize=16)
    plt.axis('equal')  # Asegura que el gráfico sea un círculo.

    print("\n[*] Mostrando visualizaciones. Cierra las ventanas de los gráficos para salir.")
    plt.show()

def main():
    """Función principal del script."""
    # Nota: para Python < 3.9, bit_count() no existe. El código se ha adaptado a Python >= 3.10
    # que es común hoy en día. Si se necesita compatibilidad, se puede usar la versión del script original.
    contador_bytes, total_de_unos = ejecutar_simulacion(NUM_MUESTRAS)
    analizar_y_reportar(contador_bytes, total_de_unos, NUM_MUESTRAS)

if __name__ == "__main__":
    main()
```
