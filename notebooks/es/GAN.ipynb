```python
from IPython.display import display, HTML

# Código HTML
html_gan = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Redes Adversarias Generativas (GAN)</title>
    <style>
        body {
            font-family: 'Georgia', serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            background-image: url('https://www.transparenttextures.com/patterns/light-paper-fibers.png');
            background-size: cover;
            background-attachment: fixed;
        }
        .navbar {
            background-color: #0066cc;
            color: white;
            padding: 10px 15px;
            text-align: center;
            font-size: 1.2em;
        }
        h1 {
            color: #333;
            text-align: center;
            font-size: 2.5em;
            margin-bottom: 20px;
        }
        h2 {
            color: #0056b3;
            margin-top: 15px;
        }
        ul {
            font-size: 1.1em;
            margin-bottom: 20px;
            padding-left: 20px;
        }
        p {
            margin-bottom: 15px;
            font-size: 1.1em;
        }
        pre {
            background-color: #f4f4f4;
            padding: 10px;
            border-radius: 5px;
            overflow-x: auto;
            font-size: 1em;
        }
    </style>
</head>
<body>
    <div class="navbar">
        Redes Adversarias Generativas (GAN)
    </div>

    <h1>Introducción</h1>
    <p>Las <strong>Redes Adversarias Generativas (Generative Adversarial Networks, GAN)</strong> son una arquitectura introducida por <strong>Ian Goodfellow en 2014</strong>. Su objetivo es generar datos sintéticos que sean similares a un conjunto de datos existentes. Están compuestas por dos redes neuronales principales:</p>
    <ul>
        <li><strong>Generador:</strong> Crea datos sintéticos (por ejemplo, imágenes).</li>
        <li><strong>Discriminador:</strong> Evalúa si los datos son reales o generados.</li>
    </ul>

    <h2>1. Generador</h2>
    <p>El Generador toma como entrada ruido aleatorio y produce datos similares al conjunto de datos de entrenamiento.</p>
    <ul>
        <li><strong>Entrada:</strong> Un vector de ruido.</li>
        <li><strong>Salida:</strong> Datos generados (e.g., una imagen).</li>
    </ul>

    <h2>2. Discriminador</h2>
    <p>El Discriminador recibe datos (reales o generados) y predice si son auténticos o falsos.</p>
    <ul>
        <li><strong>Entrada:</strong> Un dato (e.g., imagen).</li>
        <li><strong>Salida:</strong> Probabilidad de que el dato sea real.</li>
    </ul>

    <h2>3. Analogía: Falsificador vs. Policía</h2>
    <p>Podemos imaginar el Generador como un falsificador de billetes y el Discriminador como un policía:</p>
    <ul>
        <li>El falsificador intenta crear billetes indistinguibles de los reales.</li>
        <li>El policía intenta atrapar los billetes falsos.</li>
        <li>Ambos mejoran mutuamente: el falsificador mejora sus billetes y el policía mejora sus métodos de detección.</li>
    </ul>

    <h2>4. Proceso de entrenamiento</h2>
    <p>El entrenamiento de una GAN implica alternar entre:</p>
    <ul>
        <li><strong>Entrenar el Discriminador:</strong> Aprender a diferenciar entre datos reales y generados.</li>
        <li><strong>Entrenar el Generador:</strong> Mejorar en crear datos que el Discriminador clasifique como reales.</li>
    </ul>

    <h3>4.1 Entrenamiento del Discriminador</h3>
    <p>El Discriminador se entrena para clasificar correctamente datos reales y generados:</p>
    <ul>
        <li>Se calcula la pérdida usando una función como la entropía cruzada.</li>
        <li>Se optimizan los pesos del Discriminador para minimizar esta pérdida.</li>
    </ul>

    <h3>4.2 Entrenamiento del Generador</h3>
    <p>El Generador se entrena para engañar al Discriminador:</p>
    <ul>
        <li>Su objetivo es maximizar la probabilidad de que el Discriminador clasifique datos generados como reales.</li>
        <li>Se calcula la pérdida en función de la salida del Discriminador y se optimiza el Generador para minimizarla.</li>
    </ul>

    <h3>4.3 Iteración</h3>
    <p>El proceso completo consiste en alternar entre:</p>
    <ul>
        <li>Actualizar el Discriminador.</li>
        <li>Actualizar el Generador.</li>
    </ul>

    <h2>5. Resumen</h2>
    <ul>
        <li><strong>GAN:</strong> Un Generador y un Discriminador entrenados de manera competitiva.</li>
        <li><strong>Objetivo:</strong> Lograr que el Generador produzca datos indistinguibles de los reales.</li>
        <li><strong>Aplicaciones:</strong> Generación de imágenes, mejora de resolución, síntesis de audio, etc.</li>
    </ul>
</body>
</html>
"""

# Renderizar el HTML en Google Colab
display(HTML(html_gan))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.datasets import mnist
from IPython.display import clear_output  # Para limpiar la salida en Colab

# --- 1. Cargar el conjunto de datos MNIST ---
(x_train, _), (_, _) = mnist.load_data()
x_train = x_train / 127.5 - 1.0  # Normalizar a rango [-1, 1]
x_train = np.expand_dims(x_train, axis=-1)  # Expandir la dimensión para [28, 28, 1]
x_train = x_train[:10000]  # Usar un subconjunto más pequeño para acelerar el entrenamiento

BUFFER_SIZE = 10000
BATCH_SIZE = 128

# Crear un dataset aleatorio de MNIST con caché y prefetch
dataset = tf.data.Dataset.from_tensor_slices(x_train).shuffle(BUFFER_SIZE).batch(BATCH_SIZE).cache().prefetch(buffer_size=tf.data.AUTOTUNE)

# --- 2. Crear el modelo Generador ---
def build_generator():
    model = tf.keras.Sequential([
        layers.Dense(7 * 7 * 256, use_bias=False, input_shape=(100,)),
        layers.BatchNormalization(),
        layers.LeakyReLU(),

        layers.Reshape((7, 7, 256)),  # Forma inicial [7, 7, 256]

        layers.Conv2DTranspose(128, (5, 5), strides=(1, 1), padding='same', use_bias=False),
        layers.BatchNormalization(),
        layers.LeakyReLU(),

        layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False),
        layers.BatchNormalization(),
        layers.LeakyReLU(),

        layers.Conv2DTranspose(1, (5, 5), strides=(2, 2), padding='same', use_bias=False, activation='tanh')
    ])
    return model

# --- 3. Crear el modelo Discriminador ---
def build_discriminator():
    model = tf.keras.Sequential([
        layers.Conv2D(64, (5, 5), strides=(2, 2), padding='same', input_shape=[28, 28, 1]),
        layers.LeakyReLU(),
        layers.Dropout(0.3),

        layers.Conv2D(128, (5, 5), strides=(2, 2), padding='same'),
        layers.LeakyReLU(),
        layers.Dropout(0.3),

        layers.Flatten(),
        layers.Dense(1)
    ])
    return model

# Instanciar modelos
generator = build_generator()
discriminator = build_discriminator()

# --- 4. Función de pérdida y optimizadores ---
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=True)

def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    return real_loss + fake_loss

def generator_loss(fake_output):
    return cross_entropy(tf.ones_like(fake_output), fake_output)

generator_optimizer = tf.keras.optimizers.Adam(1e-4)
discriminator_optimizer = tf.keras.optimizers.Adam(1e-4)

# --- 5. Entrenamiento de la GAN ---
EPOCHS = 10  # Reducido para velocidad
noise_dim = 100
num_examples_to_generate = 16  # Cambiado a 16 para generar una cuadrícula de 4x4
seed = tf.random.normal([num_examples_to_generate, noise_dim])  # Semilla para generar imágenes consistentes

@tf.function
def train_step(images):
    noise = tf.random.normal([BATCH_SIZE, noise_dim])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        generated_images = generator(noise, training=True)
        real_output = discriminator(images, training=True)
        fake_output = discriminator(generated_images, training=True)

        gen_loss = generator_loss(fake_output)
        disc_loss = discriminator_loss(real_output, fake_output)

    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))

def train(dataset, epochs):
    for epoch in range(epochs):
        for image_batch in dataset:
            train_step(image_batch)

        # Limpiar la salida y mostrar imágenes generadas
        clear_output(wait=True)
        generate_and_save_images(generator, epoch + 1, seed)

        print(f"Epoch {epoch + 1} completada")

def generate_and_save_images(model, epoch, test_input):
    predictions = model(test_input, training=False)

    # Calcular filas y columnas para la cuadrícula
    num_images = predictions.shape[0]
    grid_size = int(np.sqrt(num_images))  # Cuadrícula cuadrada (ej. 4x4 para 16 imágenes)

    fig = plt.figure(figsize=(grid_size, grid_size))
    for i in range(num_images):
        plt.subplot(grid_size, grid_size, i + 1)
        plt.imshow(predictions[i, :, :, 0] * 127.5 + 127.5, cmap='gray')
        plt.axis('off')

    plt.show()

# Entrenar el modelo
train(dataset, EPOCHS)


```

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras import layers
from IPython.display import clear_output

# --- 1. Crear el conjunto de datos: puntos en un círculo ---
def generate_real_data(num_points):
    angles = np.linspace(0, 2 * np.pi, num_points)
    x = np.cos(angles) + np.random.normal(0, 0.05, num_points)  # Coordenada x con ruido
    y = np.sin(angles) + np.random.normal(0, 0.05, num_points)  # Coordenada y con ruido
    return np.stack([x, y], axis=1).astype('float32')  # Apilar coordenadas x, y en 2D

num_points = 1000
real_data = generate_real_data(num_points)

# Visualizar los puntos reales
plt.scatter(real_data[:, 0], real_data[:, 1], alpha=0.6)
plt.title("Datos reales: Puntos distribuidos en un círculo")
plt.axis('equal')
plt.show()

# --- 2. Crear el modelo Generador ---
def build_generator():
    model = tf.keras.Sequential([
        layers.Dense(16, activation='relu', input_shape=(2,)),
        layers.Dense(32, activation='relu'),
        layers.Dense(2, activation='linear')  # Dos salidas: x e y
    ])
    return model

# --- 3. Crear el modelo Discriminador ---
def build_discriminator():
    model = tf.keras.Sequential([
        layers.Dense(32, activation='relu', input_shape=(2,)),
        layers.Dense(16, activation='relu'),
        layers.Dense(1, activation='sigmoid')  # Salida: probabilidad de real o falso
    ])
    return model

# Instanciar modelos
generator = build_generator()
discriminator = build_discriminator()

# --- 4. Función de pérdida y optimizadores ---
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=False)

def discriminator_loss(real_output, fake_output):
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    return real_loss + fake_loss

def generator_loss(fake_output):
    return cross_entropy(tf.ones_like(fake_output), fake_output)

generator_optimizer = tf.keras.optimizers.Adam(1e-4)
discriminator_optimizer = tf.keras.optimizers.Adam(1e-4)

# --- 5. Entrenamiento de la GAN ---
EPOCHS = 1000
BATCH_SIZE = 64

@tf.function
def train_step(real_points):
    # Generar puntos falsos
    noise = tf.random.normal([BATCH_SIZE, 2])  # Ruido en 2D (entrada para el Generador)

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        generated_points = generator(noise, training=True)

        # Pasar puntos reales y falsos al Discriminador
        real_output = discriminator(real_points, training=True)
        fake_output = discriminator(generated_points, training=True)

        # Calcular pérdidas
        gen_loss = generator_loss(fake_output)
        disc_loss = discriminator_loss(real_output, fake_output)

    # Calcular y aplicar gradientes
    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))

    return gen_loss, disc_loss

def train(dataset, epochs):
    for epoch in range(epochs):
        gen_loss_avg = 0
        disc_loss_avg = 0

        for batch in dataset:
            gen_loss, disc_loss = train_step(batch)
            gen_loss_avg += gen_loss
            disc_loss_avg += disc_loss

        # Mostrar progreso cada 100 épocas
        if (epoch + 1) % 100 == 0 or epoch == epochs - 1:
            clear_output(wait=True)
            print(f"Época {epoch + 1}/{epochs}")
            print(f"Pérdida del Generador: {gen_loss_avg / len(dataset):.4f}")
            print(f"Pérdida del Discriminador: {disc_loss_avg / len(dataset):.4f}")
            generate_and_plot_points(generator)

def generate_and_plot_points(generator):
    # Generar puntos falsos
    noise = tf.random.normal([1000, 2])
    generated_points = generator(noise, training=False).numpy()

    # Graficar puntos reales y falsos
    plt.figure(figsize=(6, 6))
    plt.scatter(real_data[:, 0], real_data[:, 1], alpha=0.4, label="Datos reales (círculo)")
    plt.scatter(generated_points[:, 0], generated_points[:, 1], alpha=0.6, label="Datos generados (GAN)")
    plt.title("Puntos reales vs. Puntos generados")
    plt.axis('equal')
    plt.legend()
    plt.show()

# Crear dataset de entrenamiento
real_dataset = tf.data.Dataset.from_tensor_slices(real_data).shuffle(num_points).batch(BATCH_SIZE)

# Entrenar la GAN
train(real_dataset, EPOCHS)
```

```python
from IPython.display import display, HTML

# Código HTML
html_noise_gan = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Ruido Aleatorio en GAN</title>
    <style>
        body {
            font-family: 'Georgia', serif;
            line-height: 1.8;
            margin: 20px;
            padding: 20px;
            background-color: #f9f9f9;
            background-image: url('https://www.transparenttextures.com/patterns/light-paper-fibers.png');
            background-size: cover;
            background-attachment: fixed;
        }
        .navbar {
            background-color: #0066cc;
            color: white;
            padding: 10px 15px;
            text-align: center;
            font-size: 1.2em;
        }
        h1 {
            color: #333;
            text-align: center;
            font-size: 2.5em;
            margin-bottom: 20px;
        }
        h2 {
            color: #0056b3;
            margin-top: 15px;
        }
        ul {
            font-size: 1.1em;
            margin-bottom: 20px;
            padding-left: 20px;
        }
        p {
            margin-bottom: 15px;
            font-size: 1.1em;
        }
        pre {
            background-color: #f4f4f4;
            padding: 10px;
            border-radius: 5px;
            overflow-x: auto;
            font-size: 1em;
        }
    </style>
</head>
<body>
    <div class="navbar">
        Ruido Aleatorio en GAN
    </div>

    <h1>¿Qué es el ruido aleatorio?</h1>
    <p>El <strong>ruido aleatorio</strong> es una parte esencial en las Redes Adversarias Generativas (GAN). Sirve como entrada para el Generador, permitiéndole aprender a transformar un vector de números aleatorios en datos sintéticos realistas.</p>

    <h2>Propósito del ruido aleatorio</h2>
    <ul>
        <li><strong>Crear variabilidad:</strong> Permite generar ejemplos variados en lugar de repetir siempre el mismo.</li>
        <li><strong>Introducir incertidumbre:</strong> Ayuda al modelo a explorar un espacio latente significativo.</li>
        <li><strong>Simular distribuciones reales:</strong> El Generador aprende a mapear ruido a una distribución de datos similar a la del conjunto real.</li>
        <li><strong>Facilitar la generalización:</strong> Evita que la GAN memorice los datos de entrenamiento y fomenta la creación de ejemplos novedosos.</li>
    </ul>

    <h2>¿Qué es el ruido en términos técnicos?</h2>
    <p>El ruido aleatorio es un vector generado a partir de una distribución estadística conocida. Las distribuciones más comunes son:</p>
    <ul>
        <li><strong>Distribución uniforme:</strong> Valores distribuidos uniformemente en un rango, como [-1, 1].</li>
        <li><strong>Distribución normal (gaussiana):</strong> Valores distribuidos según una curva de campana, con media 0 y desviación estándar 1.</li>
    </ul>
    <p>Por ejemplo, el siguiente código genera ruido con una distribución normal:</p>
    <pre>
    noise = tf.random.normal([BATCH_SIZE, noise_dim])
    </pre>
    <p>Esto crea un vector de tamaño <code>noise_dim</code> para cada ejemplo del batch.</p>

    <h2>Impacto del ruido en las GAN</h2>
    <ul>
        <li><strong>Variabilidad en los datos generados:</strong> Diferentes vectores de ruido producen diferentes ejemplos sintéticos.</li>
        <li><strong>Exploración del espacio latente:</strong> Permite que el Generador aprenda características generales y significativas de los datos reales.</li>
        <li><strong>Simulación de datos reales:</strong> El Generador transforma la distribución inicial del ruido en una distribución que imita los datos originales.</li>
    </ul>

    <h2>Resumen</h2>
    <ul>
        <li>El ruido aleatorio es la <strong>"materia prima"</strong> que el Generador transforma en datos realistas.</li>
        <li>Introduce variabilidad y ayuda al modelo a generalizar, evitando la simple memorización de los datos reales.</li>
        <li>Es crucial para el éxito del Generador al crear datos novedosos y diversos.</li>
    </ul>
</body>
</html>
"""

# Renderizar el HTML en Google Colab
display(HTML(html_noise_gan))
```
