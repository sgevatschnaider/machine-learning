```python
import re
from datetime import datetime
from typing import Tuple, Optional
import sys

class CalculadoraDescuentos:
    """
    Calculadora de descuentos con validaciones avanzadas y interfaz mejorada.
    """

    # Configuración de descuentos
    DESCUENTOS = {
        'alto_volumen': {
            'espana': 50,
            'europa': 25,
            'umbral': 50
        },
        'medio_volumen': {
            'espana': 20,
            'europa': 10,
            'umbral': 10
        },
        'bajo_volumen': {
            'descuento': 0,
            'umbral': 1
        }
    }

    DESTINOS_VALIDOS = ['españa', 'europa', 'spain', 'europe']

    def __init__(self):
        self.historial = []

    def normalizar_destino(self, destino: str) -> Optional[str]:
        """
        Normaliza y valida el destino de envío.

        Args:
            destino (str): Destino introducido por el usuario

        Returns:
            str: Destino normalizado ('espana' o 'europa') o None si es inválido
        """
        if not destino:
            return None

        destino_limpio = re.sub(r'[^\w\s]', '', destino.lower().strip())

        # Mapeo de variaciones comunes
        mapeo_destinos = {
            'españa': 'espana',
            'espana': 'espana',
            'spain': 'espana',
            'esp': 'espana',
            'es': 'espana',
            'europa': 'europa',
            'europe': 'europa',
            'eu': 'europa',
            'eur': 'europa'
        }

        return mapeo_destinos.get(destino_limpio)

    def validar_unidades(self, unidades: int) -> Tuple[bool, str]:
        """
        Valida el número de unidades.

        Args:
            unidades (int): Número de unidades

        Returns:
            Tuple[bool, str]: (Es válido, mensaje de error si aplica)
        """
        if unidades < 0:
            return False, "El número de unidades no puede ser negativo"
        elif unidades == 0:
            return False, "El número de unidades debe ser mayor a 0"
        elif unidades > 10000:
            return False, "Número de unidades excesivo (máximo 10,000)"

        return True, ""

    def calcular_descuento(self, unidades: int, destino: str) -> dict:
        """
        Calcula el porcentaje de descuento basado en el número de unidades y destino.

        Args:
            unidades (int): Número de unidades solicitadas
            destino (str): Destino de envío

        Returns:
            dict: Información completa del descuento calculado
        """
        # Validar unidades
        es_valido, mensaje_error = self.validar_unidades(unidades)
        if not es_valido:
            return {
                'error': True,
                'mensaje': mensaje_error,
                'descuento': 0,
                'categoria': 'error'
            }

        # Normalizar destino
        destino_norm = self.normalizar_destino(destino)
        if not destino_norm:
            return {
                'error': True,
                'mensaje': "Destino no válido. Use 'España' o 'Europa'",
                'descuento': 0,
                'categoria': 'error'
            }

        # Lógica del árbol de decisión
        if unidades >= self.DESCUENTOS['alto_volumen']['umbral']:
            # Alto volumen (>50 unidades)
            descuento = self.DESCUENTOS['alto_volumen'][destino_norm]
            categoria = 'alto_volumen'
            nivel = 'Premium'
        elif unidades >= self.DESCUENTOS['medio_volumen']['umbral']:
            # Medio volumen (10-50 unidades)
            descuento = self.DESCUENTOS['medio_volumen'][destino_norm]
            categoria = 'medio_volumen'
            nivel = 'Estándar'
        else:
            # Bajo volumen (1-9 unidades)
            descuento = self.DESCUENTOS['bajo_volumen']['descuento']
            categoria = 'bajo_volumen'
            nivel = 'Básico'

        # Guardar en historial
        registro = {
            'fecha': datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
            'unidades': unidades,
            'destino': destino_norm,
            'descuento': descuento,
            'categoria': categoria
        }
        self.historial.append(registro)

        return {
            'error': False,
            'descuento': descuento,
            'categoria': categoria,
            'nivel': nivel,
            'unidades': unidades,
            'destino': destino_norm,
            'ahorro_estimado': self.calcular_ahorro(unidades, descuento),
            'mensaje': self.generar_mensaje(unidades, destino_norm, descuento, nivel)
        }

    def calcular_ahorro(self, unidades: int, descuento: int, precio_base: float = 10.0) -> dict:
        """
        Calcula el ahorro estimado basado en un precio base por unidad.

        Args:
            unidades (int): Número de unidades
            descuento (int): Porcentaje de descuento
            precio_base (float): Precio base por unidad

        Returns:
            dict: Información del ahorro
        """
        precio_total = unidades * precio_base
        ahorro = precio_total * (descuento / 100)
        precio_final = precio_total - ahorro

        return {
            'precio_original': precio_total,
            'ahorro': ahorro,
            'precio_final': precio_final,
            'porcentaje': descuento
        }

    def generar_mensaje(self, unidades: int, destino: str, descuento: int, nivel: str) -> str:
        """
        Genera un mensaje descriptivo del descuento aplicado.

        Args:
            unidades (int): Número de unidades
            destino (str): Destino normalizado
            descuento (int): Porcentaje de descuento
            nivel (str): Nivel de descuento

        Returns:
            str: Mensaje formateado
        """
        if descuento == 0:
            return f"Sin descuento aplicado para {unidades} unidades"

        return (f"¡Descuento aplicado!\n"
                f"   • Unidades: {unidades}\n"
                f"   • Destino: {destino.title()}\n"
                f"   • Descuento: {descuento}%\n"
                f"   • Nivel: {nivel}")

    def mostrar_historial(self):
        """Muestra el historial de cálculos realizados."""
        if not self.historial:
            print("No hay cálculos en el historial")
            return

        print("\nHISTORIAL DE CÁLCULOS")
        print("=" * 50)

        for i, registro in enumerate(self.historial[-5:], 1):  # Últimos 5
            print(f"{i}. {registro['fecha']}")
            print(f"   Unidades: {registro['unidades']} | "
                  f"Destino: {registro['destino'].title()} | "
                  f"Descuento: {registro['descuento']}%")

    def interfaz_usuario(self):
        """Interfaz de usuario mejorada con menú interactivo."""
        print("=" * 60)
        print("CALCULADORA DE DESCUENTOS INTELIGENTE")
        print("=" * 60)

        while True:
            print("\nOPCIONES DISPONIBLES:")
            print("1. Calcular descuento")
            print("2. Ver historial")
            print("3. Mostrar tabla de descuentos")
            print("4. Salir")

            try:
                opcion = input("\nSeleccione una opción (1-4): ").strip()

                if opcion == '1':
                    self.ejecutar_calculo()
                elif opcion == '2':
                    self.mostrar_historial()
                elif opcion == '3':
                    self.mostrar_tabla_descuentos()
                elif opcion == '4':
                    print("\n¡Gracias por usar la calculadora!")
                    break
                else:
                    print("Opción no válida. Intente de nuevo.")

            except KeyboardInterrupt:
                print("\n\nPrograma interrumpido por el usuario.")
                break
            except Exception as e:
                print(f"Error inesperado: {e}")

    def ejecutar_calculo(self):
        """Ejecuta el cálculo de descuento con validaciones."""
        try:
            # Solicitar unidades
            unidades_str = input("\nIngrese el número de unidades: ").strip()

            if not unidades_str:
                print("Debe ingresar un número de unidades.")
                return

            unidades = int(unidades_str)

            # Solicitar destino
            destino = input("Ingrese el destino (España/Europa): ").strip()

            if not destino:
                print("Debe ingresar un destino.")
                return

            # Calcular descuento
            resultado = self.calcular_descuento(unidades, destino)

            # Mostrar resultado
            if resultado['error']:
                print(f"\nError: {resultado['mensaje']}")
            else:
                print(f"\n{resultado['mensaje']}")

                # Mostrar ahorro estimado
                ahorro = resultado['ahorro_estimado']
                if ahorro['ahorro'] > 0:
                    print(f"\nAHORRO ESTIMADO (precio base: €{ahorro['precio_original']:.2f}):")
                    print(f"   • Ahorro: €{ahorro['ahorro']:.2f}")
                    print(f"   • Precio final: €{ahorro['precio_final']:.2f}")

        except ValueError:
            print("Por favor, ingrese un número válido para las unidades.")
        except Exception as e:
            print(f"Error: {e}")

    def mostrar_tabla_descuentos(self):
        """Muestra la tabla de descuentos disponibles."""
        print("\nTABLA DE DESCUENTOS")
        print("=" * 50)
        print("UNIDADES          ESPAÑA       EUROPA")
        print("-" * 50)
        print("1-9 unidades        0%           0%")
        print("10-50 unidades      20%          10%")
        print("51+ unidades        50%          25%")
        print("=" * 50)

# Función principal para ejecutar la calculadora
def main():
    """Función principal del programa."""
    calculadora = CalculadoraDescuentos()

    # Verificar si se ejecuta en modo interactivo
    if len(sys.argv) > 1 and sys.argv[1] == '--test':
        # Modo de prueba
        print("MODO DE PRUEBA")
        casos_prueba = [
            (5, 'España'),
            (25, 'Europa'),
            (75, 'España'),
            (100, 'Europa'),
            (-5, 'España'),
            (15, 'destino_invalido')
        ]

        for unidades, destino in casos_prueba:
            resultado = calculadora.calcular_descuento(unidades, destino)
            print(f"Unidades: {unidades}, Destino: {destino}")
            print(f"Resultado: {resultado.get('mensaje', 'Error')}")
            print("-" * 30)
    else:
        # Modo interactivo
        calculadora.interfaz_usuario()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from typing import Tuple, List, Optional
import seaborn as sns


class KNNVisualizer:
    """
    Visualizador avanzado para el algoritmo K-Nearest Neighbors.

    Permite visualizar datos de clasificación, fronteras de decisión,
    y análisis de vecinos más cercanos con diferentes valores de k.
    """

    def __init__(self, figsize: Tuple[int, int] = (18, 7), style: str = 'seaborn-v0_8-whitegrid'):
        """
        Inicializa el visualizador.

        Args:
            figsize: Tamaño de la figura (ancho, alto)
            style: Estilo de matplotlib
        """
        self.figsize = figsize
        plt.style.use(style)
        # Colores ajustados para mejor contraste
        self.colors = ['#0077B6', '#F77F00', '#D62828', '#FCBF49', '#EAE2B7']
        self.markers = ['s', 'o', '^', 'D', 'v']

    def generate_sample_data(self, n_samples: int = 50, n_classes: int = 2,
                           random_state: int = 42) -> Tuple[np.ndarray, np.ndarray]:
        """
        Genera datos de ejemplo para clasificación.

        Args:
            n_samples: Número de muestras por clase
            n_classes: Número de clases
            random_state: Semilla para reproducibilidad

        Returns:
            Tupla con características (X) y etiquetas (y)
        """
        np.random.seed(random_state)

        X = []
        y = []

        centers = [(2, 2), (5, 5), (2, 5), (5, 2), (3.5, 3.5)][:n_classes]

        for i, (cx, cy) in enumerate(centers):
            noise_scale = 0.8 + 0.4 * np.random.rand()
            class_data = np.random.multivariate_normal(
                [cx, cy],
                [[noise_scale, 0.2], [0.2, noise_scale]],
                n_samples
            )
            X.append(class_data)
            y.extend([i] * n_samples)

        return np.vstack(X), np.array(y)

    def calculate_distances(self, point: np.ndarray, data: np.ndarray) -> np.ndarray:
        """
        Calcula distancias euclidianas desde un punto a todos los puntos del dataset.
        """
        return np.sqrt(np.sum((data - point) ** 2, axis=1))

    def get_k_nearest_neighbors(self, point: np.ndarray, X: np.ndarray,
                               y: np.ndarray, k: int) -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
        """
        Obtiene los k vecinos más cercanos para un punto dado.
        """
        distances = self.calculate_distances(point, X)
        nearest_indices = np.argsort(distances)[:k]
        return nearest_indices, distances[nearest_indices], y[nearest_indices]

    def plot_decision_boundary(self, X: np.ndarray, y: np.ndarray, k: int,
                             ax: plt.Axes, resolution: int = 100) -> None:
        """
        Dibuja la frontera de decisión del clasificador k-NN.
        """
        knn = KNeighborsClassifier(n_neighbors=k)
        knn.fit(X, y)

        x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx, yy = np.meshgrid(np.linspace(x_min, x_max, resolution),
                            np.linspace(y_min, y_max, resolution))

        mesh_points = np.c_[xx.ravel(), yy.ravel()]
        Z = knn.predict(mesh_points)
        Z = Z.reshape(xx.shape)

        unique_classes = np.unique(y)
        ax.contourf(xx, yy, Z, alpha=0.2, levels=len(unique_classes)-1,
                   colors=self.colors[:len(unique_classes)])
        ax.contour(xx, yy, Z, colors='black', linewidths=0.7, alpha=0.5)

    def visualize_knn(self, X: np.ndarray, y: np.ndarray,
                     query_point: Optional[np.ndarray] = None,
                     k_values: List[int] = [3, 5, 9],
                     show_decision_boundary: bool = True,
                     title: str = "Visualización K-Nearest Neighbors") -> plt.Figure:
        """
        Visualiza el algoritmo k-NN con diferentes valores de k.
        """
        n_plots = len(k_values)
        fig, axes = plt.subplots(1, n_plots, figsize=self.figsize, squeeze=False)
        axes = axes.flatten()

        if query_point is None:
            query_point = np.array([np.mean(X[:, 0]), np.mean(X[:, 1])])

        unique_classes = np.unique(y)

        for idx, k in enumerate(k_values):
            ax = axes[idx]

            if show_decision_boundary:
                self.plot_decision_boundary(X, y, k, ax)

            for i, class_label in enumerate(unique_classes):
                class_mask = y == class_label
                ax.scatter(X[class_mask, 0], X[class_mask, 1],
                          color=self.colors[i], marker=self.markers[i],
                          s=60, alpha=0.9, edgecolor='black', linewidth=0.8,
                          label=f'Clase {class_label}')

            nearest_indices, distances, nearest_labels = self.get_k_nearest_neighbors(
                query_point, X, y, k
            )

            # (NUEVO) Dibujar líneas desde el punto de consulta a cada vecino
            for neighbor_idx in nearest_indices:
                ax.plot([query_point[0], X[neighbor_idx, 0]],
                        [query_point[1], X[neighbor_idx, 1]],
                        color='gray', linestyle=':', linewidth=1.5, zorder=3)

            ax.scatter(X[nearest_indices, 0], X[nearest_indices, 1],
                      s=250, facecolors='none', edgecolors='red',
                      linewidth=3, alpha=1.0, label='k-Vecinos', zorder=4)

            ax.scatter(query_point[0], query_point[1],
                      color='#32CD32', marker='*', s=400,
                      edgecolor='black', linewidth=1.5,
                      label='Punto Consulta', zorder=5)

            kth_distance = distances[-1]
            circle = plt.Circle(query_point, kth_distance,
                              color='red', linestyle='--',
                              linewidth=2.5, fill=False, alpha=0.8)
            ax.add_patch(circle)

            predicted_class = np.bincount(nearest_labels).argmax()
            confidence = np.bincount(nearest_labels)[predicted_class] / k

            ax.set_aspect('equal')
            ax.set_title(f'k = {k}\nPredicción: Clase {predicted_class} ({confidence:.1%} confianza)', fontsize=14, pad=15)
            ax.set_xlabel('Característica 1', fontsize=12)
            if idx == 0:
                ax.set_ylabel('Característica 2', fontsize=12)

            # Conteo vecinos
            neighbor_counts = dict(zip(*np.unique(nearest_labels, return_counts=True)))
            counts_str_parts = []
            for c in unique_classes: # Itera para mantener el orden de las clases
                count = neighbor_counts.get(c, 0)
                if count > 0:
                    counts_str_parts.append(f'Clase {c}: {count}')

            counts_str = ", ".join(counts_str_parts) if counts_str_parts else "Ninguno"

            stats_text = (f'Distancia máxima: {kth_distance:.2f}\n'
                          f'Vecinos por clase: {counts_str}')

            ax.text(0.03, 0.97, stats_text, transform=ax.transAxes,
                   fontsize=10, verticalalignment='top',
                   bbox=dict(boxstyle='round,pad=0.4', facecolor='white', alpha=0.85))

            if idx == 0:
              ax.legend(fontsize=9, loc='lower left')


        fig.suptitle(title, fontsize=20, y=1.02)
        plt.tight_layout(rect=[0, 0, 1, 0.96])
        return fig

    def analyze_k_performance(self, X: np.ndarray, y: np.ndarray,
                            k_range: range = range(1, 21),
                            test_size: float = 0.3,
                            random_state: int = 42) -> plt.Figure:
        """
        Analiza el rendimiento del clasificador k-NN para diferentes valores de k.
        """
        from sklearn.model_selection import train_test_split

        X_train, X_test, y_train, y_test = train_test_split(
            X, y, test_size=test_size, random_state=random_state, stratify=y
        )

        train_scores = []
        test_scores = []

        for k in k_range:
            knn = KNeighborsClassifier(n_neighbors=k)
            knn.fit(X_train, y_train)
            train_scores.append(accuracy_score(y_train, knn.predict(X_train)))
            test_scores.append(accuracy_score(y_test, knn.predict(X_test)))

        fig, ax = plt.subplots(figsize=(12, 7))
        ax.plot(k_range, train_scores, 'o-', label='Precisión en Entrenamiento',
                color='#0077B6', linewidth=2, markersize=6)
        ax.plot(k_range, test_scores, 's-', label='Precisión en Validación',
                color='#D62828', linewidth=2, markersize=6)

        best_k = list(k_range)[np.argmax(test_scores)]
        best_score = max(test_scores)

        ax.axvline(x=best_k, color='green', linestyle='--', alpha=0.7,
                  label=f'Mejor k = {best_k} (Precisión = {best_score:.3f})')

        ax.set_xlabel('Número de Vecinos (k)', fontsize=12)
        ax.set_ylabel('Precisión', fontsize=12)
        ax.set_title('Análisis de Rendimiento de k-NN vs. Precisión', fontsize=16, pad=20)
        ax.legend(fontsize=11)
        ax.grid(True, which='both', linestyle='--', linewidth=0.5)
        ax.set_xticks(k_range)
        ax.set_xlim(min(k_range)-0.5, max(k_range)+0.5)
        ax.set_ylim(min(min(test_scores), min(train_scores)) * 0.95, 1.01)

        plt.tight_layout()
        return fig


# Ejemplo de uso
if __name__ == "__main__":
    # Crear visualizador
    visualizer = KNNVisualizer()

    # Generar datos de ejemplo con 3 clases para un caso más interesante
    X, y = visualizer.generate_sample_data(n_samples=40, n_classes=3, random_state=42)

    # Punto de consulta
    query_point = np.array([3.8, 3.8])

    # Visualizar k-NN con diferentes valores de k
    fig1 = visualizer.visualize_knn(
        X, y, query_point=query_point,
        k_values=[3, 5, 9],
        show_decision_boundary=True,
        title="Análisis k-NN: Impacto del Parámetro k"
    )

    # Análisis de rendimiento
    fig2 = visualizer.analyze_k_performance(X, y, k_range=range(1, 26))

    plt.show()
```

```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider
from typing import List, Union

# --- 1. MEJORA ESTRUCTURAL: Encapsulación en una Clase ---
# Se encapsula la lógica del perceptrón en una clase para una mejor
# organización, reutilización y claridad.

class Perceptron:
    """
    Representa un Perceptrón simple con una función de activación escalón.

    Attributes:
        weights (np.ndarray): El vector de pesos (W).
        bias (float): El término de sesgo (b).
    """
    def __init__(self, weights: Union[List[float], np.ndarray], bias: float):
        """
        Inicializa el Perceptrón con pesos y sesgo.

        Args:
            weights (Union[List[float], np.ndarray]): Lista o array de pesos.
            bias (float): Valor del sesgo.
        """
        self.weights = np.asarray(weights)
        self.bias = float(bias)

    def _step_activation(self, z: float) -> int:
        """
        Función de activación escalón (Heaviside).

        Args:
            z (float): La suma ponderada (entrada neta).

        Returns:
            int: 1 si z es mayor o igual a 0, de lo contrario 0.
        """
        return 1 if z >= 0 else 0

    def predict(self, X: Union[List[float], np.ndarray]) -> (float, int):
        """
        Calcula la salida del perceptrón para un vector de entrada dado.

        Args:
            X (Union[List[float], np.ndarray]): Vector de características de entrada.

        Returns:
            tuple[float, int]: Una tupla conteniendo la suma ponderada (z) y
                               la salida final activada (0 o 1).
        """
        X = np.asarray(X)
        # Sumarización ponderada de las entradas
        z = np.dot(X, self.weights) + self.bias
        # Aplicar la función de activación
        output = self._step_activation(z)
        return z, output

# --- 2. Esta función crea una visualización mucho más rica y explicativa.

def visualize_perceptron_step(x1: float, x2: float, x3: float,
                              w1: float, w2: float, w3: float, b: float):
    """
    Función interactiva que crea y visualiza el comportamiento de un perceptrón.
    Se llama cada vez que un slider se mueve.
    """
    inputs = np.array([x1, x2, x3])
    weights = np.array([w1, w2, w3])

    # Crear una instancia del perceptrón con los valores actuales de los sliders
    perceptron = Perceptron(weights=weights, bias=b)

    # Obtener la suma ponderada (z) y la salida final
    z, output = perceptron.predict(inputs)

    # --- Creación de la Visualización ---
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6), gridspec_kw={'width_ratios': [3, 2]})
    fig.suptitle('Análisis Interactivo de un Perceptrón Simple', fontsize=16, weight='bold')

    # --- Gráfico 1: Cálculo de la Suma Ponderada (z) ---
    components = [x1*w1, x2*w2, x3*w3, b]
    labels = [f'x₁*w₁\n({x1:.1f}*{w1:.1f})',
              f'x₂*w₂\n({x2:.1f}*{w2:.1f})',
              f'x₃*w₃\n({x3:.1f}*{w3:.1f})',
              f'Bias (b)\n({b:.1f})']
    colors = ['#4A90E2', '#50E3C2', '#F5A623', '#D0021B']

    bars = ax1.bar(labels, components, color=colors, alpha=0.8)
    ax1.axhline(0, color='black', linewidth=0.8, linestyle='--') # Línea en y=0

    # Ecuación matemática en el título
    title_eq = f'$z = (x_1 \\cdot w_1) + (x_2 \\cdot w_2) + (x_3 \\cdot w_3) + b$'
    ax1.set_title(f'{title_eq}\n$z = {z:.2f}$', fontsize=14)
    ax1.set_ylabel('Valor Contribuido')

    # Añadir etiquetas de valor sobre cada barra
    for bar in bars:
        yval = bar.get_height()
        ax1.text(bar.get_x() + bar.get_width()/2.0, yval + np.sign(yval)*0.05, f'{yval:.2f}',
                 ha='center', va='bottom' if yval >= 0 else 'top', fontsize=9)

    # Ajustar límites del eje y para mejor visualización
    max_val = max(abs(c) for c in components) * 1.2 + 0.1
    ax1.set_ylim(-max_val, max_val)

    # --- Gráfico 2: Función de Activación y Salida ---
    # Dibujar la función de activación escalón
    x_step = np.linspace(-3, 3, 100)
    y_step = np.where(x_step >= 0, 1, 0)
    ax2.plot(x_step, y_step, lw=2, color='#7ED321', label='Función Activación Escalón')

    # Marcar el punto de la suma ponderada 'z' en el eje x
    ax2.plot([z, z], [0, output], 'r--', lw=2, alpha=0.7) # Línea vertical desde z
    ax2.plot([z], [output], 'ro', markersize=12, label=f'Salida: {output}') # Punto de salida

    # Título explicativo
    comparison = '≥' if z >= 0 else '<'
    reason = f'Como $z={z:.2f} {comparison} 0$'
    ax2.set_title(f'Activación y Salida\n{reason}', fontsize=14)
    ax2.set_xlabel('Suma Ponderada (z)')
    ax2.set_ylabel('Salida Activada')
    ax2.set_ylim(-0.2, 1.2)
    ax2.set_xlim(-3, 3)
    ax2.grid(True, linestyle='--', alpha=0.5)
    ax2.legend(loc='upper left')

    plt.tight_layout(rect=[0, 0, 1, 0.96]) # Ajustar para el supertítulo
    plt.show()

# --- 3. MEJORA ANALÍTICA: Widgets Interactivos con Descripciones Claras ---
# Se utilizan descripciones más claras con subíndices para una mejor legibilidad.

interact(
    visualize_perceptron_step,
    x1=FloatSlider(value=1.0, min=-1, max=1, step=0.1, description='Entrada x₁:'),
    x2=FloatSlider(value=0.5, min=-1, max=1, step=0.1, description='Entrada x₂:'),
    x3=FloatSlider(value=-1.0, min=-1, max=1, step=0.1, description='Entrada x₃:'),
    w1=FloatSlider(value=0.5, min=-1, max=1, step=0.1, description='Peso w₁:'),
    w2=FloatSlider(value=-0.8, min=-1, max=1, step=0.1, description='Peso w₂:'),
    w3=FloatSlider(value=0.3, min=-1, max=1, step=0.1, description='Peso w₃:'),
    b=FloatSlider(value=-0.2, min=-1, max=1, step=0.1, description='Bias b:')
);
```

```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider, Dropdown
from typing import List, Union, Tuple, Dict

# --- CLASE PERCEPTRÓN (MODIFICADA LIGERAMENTE PARA 2D) ---
class Perceptron:
    """
    Representa un Perceptrón simple optimizado para visualización 2D.
    """
    def __init__(self, weights: Union[List[float], np.ndarray], bias: float):
        self.weights = np.asarray(weights)
        self.bias = float(bias)

    def _step_activation(self, z: float) -> int:
        return 1 if z >= 0 else 0

    def predict(self, X: Union[List[float], np.ndarray]) -> int:
        X = np.asarray(X)
        z = np.dot(X, self.weights) + self.bias
        return self._step_activation(z)

    def get_decision_boundary(self, x_range: Tuple[float, float]) -> Tuple[np.ndarray, np.ndarray]:
        """
        Calcula las coordenadas (x, y) de la línea de decisión.
        La línea se define por la ecuación: w1*x1 + w2*x2 + b = 0
        Despejando x2 (y): y = (-w1*x1 - b) / w2
        """
        x_coords = np.array(x_range)

        # Evitar división por cero si w2 es 0 (línea vertical)
        if abs(self.weights[1]) < 1e-6:
            # La línea es vertical en x = -b / w1
            if abs(self.weights[0]) > 1e-6:
                y_coords = np.array([-1, 2]) # Rango y arbitrario
                x_coords = np.array([-self.bias / self.weights[0]] * 2)
            else: # Si ambos pesos son 0, no hay línea
                return np.array([]), np.array([])
        else:
            y_coords = (-self.weights[0] * x_coords - self.bias) / self.weights[1]

        return x_coords, y_coords

# --- DATOS DE LAS PUERTAS LÓGICAS ---
# Un diccionario para acceder fácilmente a los datos de cada puerta
GATE_DATA: Dict[str, Dict] = {
    "AND": {
        "X": np.array([[0, 0], [0, 1], [1, 0], [1, 1]]),
        "y": np.array([0, 0, 0, 1])
    },
    "OR": {
        "X": np.array([[0, 0], [0, 1], [1, 0], [1, 1]]),
        "y": np.array([0, 1, 1, 1])
    },
    "XOR": {
        "X": np.array([[0, 0], [0, 1], [1, 0], [1, 1]]),
        "y": np.array([0, 1, 1, 0])
    }
}

# --- FUNCIÓN DE VISUALIZACIÓN PRINCIPAL (REDISEÑADA) ---
def visualize_gate_perceptron(gate: str, w1: float, w2: float, b: float):
    """
    Visualización interactiva que muestra cómo un perceptrón intenta resolver
    una puerta lógica, destacando la frontera de decisión.
    """
    data = GATE_DATA[gate]
    X, y_true = data["X"], data["y"]

    perceptron = Perceptron(weights=[w1, w2], bias=b)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7), gridspec_kw={'width_ratios': [2, 1.5]})
    fig.suptitle(f'Análisis de Perceptrón: Puerta Lógica {gate}', fontsize=18, weight='bold')

    # --- Gráfico 1: Frontera de Decisión en el Espacio de Entradas ---
    ax1.set_title("Frontera de Decisión y Clasificación", fontsize=14)

    # Colores para la visualización
    colors = ['#D62828', '#0077B6'] # Rojo para clase 0, Azul para clase 1

    # Dibujar los puntos de datos (0,0), (0,1), etc.
    for i in range(len(X)):
        ax1.scatter(X[i, 0], X[i, 1], c=colors[y_true[i]], s=200, edgecolor='black', zorder=5)

    # Dibujar la línea de decisión
    x_line, y_line = perceptron.get_decision_boundary((-0.5, 1.5))
    ax1.plot(x_line, y_line, color='#32CD32', lw=3, label=f'Línea: ${w1:.1f}x_1 + {w2:.1f}x_2 + {b:.1f} = 0$')

    # Rellenar las áreas de decisión
    ax1.fill_between(x_line, y_line, 2, color=colors[1], alpha=0.1)
    ax1.fill_between(x_line, y_line, -1, color=colors[0], alpha=0.1)

    ax1.set_xlim(-0.5, 1.5)
    ax1.set_ylim(-0.5, 1.5)
    ax1.set_xlabel('$x_1$', fontsize=12)
    ax1.set_ylabel('$x_2$', fontsize=12)
    ax1.grid(True, linestyle='--', alpha=0.6)
    ax1.legend(loc='upper left')
    ax1.set_aspect('equal', adjustable='box')


    # --- Gráfico 2: Resumen de Resultados ---
    ax2.set_title("Resultados y Precisión", fontsize=14)
    ax2.axis('off') # Ocultar ejes, solo mostraremos texto

    correct_predictions = 0
    for i in range(len(X)):
        point = X[i]
        true_label = y_true[i]
        pred_label = perceptron.predict(point)

        is_correct = (pred_label == true_label)
        if is_correct:
            correct_predictions += 1

        result_text = f"Entrada ({point[0]},{point[1]}): Pred.={pred_label} (Target={true_label})"
        status_text = "✅ CORRECTO" if is_correct else "❌ INCORRECTO"

        # Mostrar resultado con color
        text_color = 'green' if is_correct else 'red'
        ax2.text(0.05, 0.8 - i*0.15, f"{result_text}\n    {status_text}",
                 fontsize=12, color=text_color, verticalalignment='top')

    accuracy = correct_predictions / len(X)
    ax2.text(0.5, 0.1, f'Precisión Total: {accuracy:.0%}',
             fontsize=16, weight='bold', ha='center',
             bbox=dict(facecolor='gold', alpha=0.3, boxstyle='round,pad=0.5'))

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()

# --- WIDGETS INTERACTIVOS ---
interact(
    visualize_gate_perceptron,
    gate=Dropdown(options=['AND', 'OR', 'XOR'], value='XOR', description='Puerta Lógica:'),
    w1=FloatSlider(value=1.0, min=-2, max=2, step=0.1, description='Peso w₁:'),
    w2=FloatSlider(value=1.0, min=-2, max=2, step=0.1, description='Peso w₂:'),
    b=FloatSlider(value=-1.5, min=-2, max=2, step=0.1, description='Bias b:')
);
```

```python
# ==============================================================================
#  BLOQUE DE CÓDIGO ÚNICO Y COMPLETO PARA GOOGLE COLAB
#  Tema: Cómo un Perceptrón Multicapa (MLP) Resuelve el Problema XOR
#  Autor: Experto en Machine Learning (IA)
#  Descripción: Este recurso interactivo demuestra visual y numéricamente
#               cómo un MLP transforma un problema no linealmente separable (XOR)
#               en uno linealmente separable, sentando las bases del Deep Learning.
# ==============================================================================

# --- PASO 0: IMPORTACIONES NECESARIAS ---
from IPython.display import display, HTML
import html
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import io
import base64

# --- FUNCIONES AUXILIARES PARA VISUALIZACIONES Y SIMULACIÓN ---

# 1. Función de activación escalón (step function)
def step_function(x):
    """Función de activación escalón. Devuelve 1 si x > 0, de lo contrario 0."""
    return np.where(x > 0, 1, 0)

# 2. Definición del MLP con pesos y sesgos predefinidos
def mlp_xor(x1, x2):
    """Calcula la salida de un MLP diseñado para XOR, retornando valores intermedios."""
    # Pesos y sesgos para la capa oculta (h1, h2)
    # h1 debe activarse para (1, 0)
    w_h1 = np.array([1, -1])
    b_h1 = -0.5
    # h2 debe activarse para (0, 1)
    w_h2 = np.array([-1, 1])
    b_h2 = -0.5

    # Pesos y sesgos para la capa de salida (y)
    # y debe ser OR de h1 y h2
    w_y = np.array([1, 1])
    b_y = -0.5

    # Propagación hacia adelante (forward pass)
    input_vector = np.array([x1, x2])

    # Capa oculta
    h1_pre_activation = np.dot(w_h1, input_vector) + b_h1
    h1_output = step_function(h1_pre_activation)

    h2_pre_activation = np.dot(w_h2, input_vector) + b_h2
    h2_output = step_function(h2_pre_activation)

    hidden_output_vector = np.array([h1_output, h2_output])

    # Capa de salida
    y_pre_activation = np.dot(w_y, hidden_output_vector) + b_y
    y_output = step_function(y_pre_activation)

    return {
        "input": (x1, x2),
        "h1_pre": h1_pre_activation, "h1_out": h1_output,
        "h2_pre": h2_pre_activation, "h2_out": h2_output,
        "y_pre": y_pre_activation, "y_out": y_output
    }

# 3. Función para generar los gráficos de los espacios de características
def generar_graficos_espacios_base64():
    """Genera dos gráficos: el espacio original y el espacio transformado por la capa oculta."""
    # Datos de entrada
    xor_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
    xor_labels = np.array([0, 1, 1, 0])

    # Transformar los datos usando la capa oculta
    transformed_points = np.array([mlp_xor(x[0], x[1])['h1_out'] for x in xor_inputs]), \
                         np.array([mlp_xor(x[0], x[1])['h2_out'] for x in xor_inputs])
    transformed_points = np.vstack(transformed_points).T

    # Crear DataFrames para graficar fácilmente
    df_original = pd.DataFrame(xor_inputs, columns=['x1', 'x2'])
    df_original['label'] = xor_labels

    df_transformed = pd.DataFrame(transformed_points, columns=['h1', 'h2'])
    df_transformed['label'] = xor_labels

    # Configurar la figura con dos subplots
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5.5))

    # Gráfico 1: Espacio Original
    sns.scatterplot(data=df_original, x='x1', y='x2', hue='label', style='label', s=200, palette={0: '#3498db', 1: '#e74c3c'}, ax=ax1)
    ax1.set_title('Espacio Original (No Linealmente Separable)', fontsize=14, weight='bold')
    ax1.set_xlabel('Entrada x1', fontsize=12)
    ax1.set_ylabel('Entrada x2', fontsize=12)
    ax1.set_xlim(-0.5, 1.5)
    ax1.set_ylim(-0.5, 1.5)
    ax1.legend(title='Clase')

    # Gráfico 2: Espacio Transformado por la Capa Oculta
    sns.scatterplot(data=df_transformed, x='h1', y='h2', hue='label', style='label', s=200, palette={0: '#3498db', 1: '#e74c3c'}, ax=ax2)
    # Dibujar la frontera de decisión en el espacio transformado (h1 + h2 - 0.5 = 0)
    ax2.plot([0.5, -0.5], [-0.5, 0.5], 'g--', lw=3, label='Frontera de Decisión Lineal')
    ax2.set_title('Espacio Oculto (Linealmente Separable)', fontsize=14, weight='bold')
    ax2.set_xlabel('Salida Neurona Oculta h1', fontsize=12)
    ax2.set_ylabel('Salida Neurona Oculta h2', fontsize=12)
    ax2.set_xlim(-0.5, 1.5)
    ax2.set_ylim(-0.5, 1.5)
    ax2.legend(title='Clase')

    plt.tight_layout()

    # Guardar en base64
    buf = io.BytesIO()
    fig.savefig(buf, format='png', bbox_inches='tight')
    plt.close(fig)
    buf.seek(0)
    img_base64 = base64.b64encode(buf.read()).decode('utf-8')
    return f'<div style="text-align:center; margin: 1rem 0;"><img src="data:image/png;base64,{img_base64}" alt="Espacios XOR"/></div>'


# --- PASO 1: ESTRUCTURA DE DATOS CON CONTENIDO PROFUNDO ---
mlp_contenido = [
    {
        "titulo": "1. El Salto Conceptual: Del Perceptrón Simple al Multicapa (MLP)",
        "descripcion": """
            <p>Un <strong>Perceptrón Multicapa (MLP)</strong> es una red neuronal artificial que supera la limitación fundamental del Perceptrón simple. Está compuesto por al menos tres tipos de capas:</p>
            <ol>
                <li><strong>Capa de Entrada:</strong> Recibe las características iniciales del problema (ej. \(x_1, x_2\)). No realiza cómputo.</li>
                <li><strong>Capa(s) Oculta(s):</strong> El corazón del MLP. Son capas intermedias de neuronas que aplican una transformación a los datos. Es aquí donde ocurre la "magia" del aprendizaje no lineal.</li>
                <li><strong>Capa de Salida:</strong> Produce el resultado final de la red (ej. una clasificación).</li>
            </ol>
            <h4>La Idea Clave: Composición de Funciones</h4>
            <p>Cada neurona en un MLP realiza una transformación lineal seguida de una <strong>función de activación no lineal</strong> (como Sigmoide, ReLU o, en nuestro ejemplo simplificado, una función escalón):</p>
            <p style="text-align:center; font-size:1.2em;">\( \text{salida} = \sigma(\mathbf{w} \cdot \mathbf{x} + b) \)</p>
            <p>Al apilar estas neuronas en capas, la red aprende a <strong>componer estas funciones no lineales</strong>. La salida de una capa se convierte en la entrada de la siguiente. Esta composición jerárquica permite al MLP aproximar cualquier función continua compleja, incluyendo aquellas con fronteras de decisión no lineales como XOR.</p>
        """
    },
    {
        "titulo": "2. El Fracaso del Perceptrón Simple con XOR (Recordatorio)",
        "descripcion": """
            <p>Como vimos anteriormente, un Perceptrón simple define una única frontera de decisión lineal. El problema XOR, sin embargo, requiere una frontera no lineal para separar sus clases. Los cuatro puntos de XOR, representados visualmente, no pueden ser divididos en dos grupos por una única línea recta.</p>
            <p>Este fue el desafío que Minsky y Papert destacaron en su libro <em>"Perceptrons"</em> (1969), lo que llevó a una disminución en la investigación de redes neuronales, un periodo conocido como el primer "invierno de la IA". La solución no era un modelo más grande, sino uno conceptualmente diferente: el MLP.</p>
        """
    },
    {
        "titulo": "3. La Solución del MLP: Transformar el Espacio",
        "descripcion": f"""
            <p>La estrategia del MLP para resolver XOR es brillante y elegante: en lugar de tratar de trazar una frontera compleja en el espacio original, <strong>la capa oculta transforma el espacio de entrada en un nuevo espacio de características donde el problema SÍ es linealmente separable.</strong></p>
            <ol>
                <li>La <strong>capa oculta</strong> actúa como un extractor de características. Cada neurona oculta aprende a detectar un patrón específico en los datos. Para XOR, una neurona podría aprender a activarse solo para el caso (1,0) y la otra solo para (0,1).</li>
                <li>La <strong>capa de salida</strong> ya no opera sobre las entradas originales (\(x_1, x_2\)), sino sobre las salidas de la capa oculta (\(h_1, h_2\)). En este nuevo "espacio oculto", la tarea se vuelve trivial: simplemente necesita realizar una operación OR sobre las activaciones de las neuronas ocultas.</li>
            </ol>
            <h4>Visualización de la Transformación</h4>
            <p>El siguiente gráfico lo demuestra. A la izquierda, el espacio original. A la derecha, el espacio después de que los puntos han sido procesados por la capa oculta. Observa cómo los puntos rojos (clase 1) se mueven a una nueva ubicación donde una simple línea verde puede separarlos perfectamente de los puntos azules (clase 0).</p>
            {generar_graficos_espacios_base64()}
        """
    },
    {
        "titulo": "4. Simulación Numérica: Un MLP para XOR Paso a Paso",
        "descripcion": """
            <p>Construyamos un MLP con 2 neuronas en la capa oculta (\(h_1, h_2\)) y 1 neurona de salida (\(y\)). Usaremos la función de activación escalón y pesos/sesgos elegidos manualmente para ilustrar el concepto.</p>
            <h4>Arquitectura y Pesos:</h4>
            <ul>
                <li><strong>Neurona oculta \(h_1\):</strong> Diseñada para activarse solo con la entrada (1, 0). Ecuación: \( h_1 = \text{step}(1 \cdot x_1 - 1 \cdot x_2 - 0.5) \).</li>
                <li><strong>Neurona oculta \(h_2\):</strong> Diseñada para activarse solo con la entrada (0, 1). Ecuación: \( h_2 = \text{step}(-1 \cdot x_1 + 1 \cdot x_2 - 0.5) \).</li>
                <li><strong>Neurona de salida \(y\):</strong> Diseñada para realizar una operación OR sobre \(h_1\) y \(h_2\). Ecuación: \( y = \text{step}(1 \cdot h_1 + 1 \cdot h_2 - 0.5) \).</li>
            </ul>
            <h4>Calculadora Interactiva de la Red:</h4>
            <p>Selecciona una entrada para ver cómo fluyen los valores a través de la red:</p>
        """
    },
    {
        "titulo": "5. Importancia Histórica y Técnica",
        "descripcion": """
            <p>La demostración de que un MLP podía resolver XOR fue un punto de inflexión que revitalizó el campo de las redes neuronales. Este "simple" ejemplo encapsula los principios más profundos del Deep Learning:</p>
            <ul>
                <li><strong>Aprendizaje de Representaciones:</strong> La red no solo aprende a clasificar, sino que aprende a <strong>crear nuevas representaciones</strong> de los datos (el espacio oculto) que hacen que el problema sea más fácil de resolver.</li>
                <li><strong>Composición Jerárquica:</strong> Las redes profundas apilan capas para construir representaciones cada vez más abstractas. La primera capa podría aprender bordes, la siguiente formas, y así sucesivamente. XOR es el caso más simple de esta jerarquía.</li>
                <li><strong>Nacimiento de Backpropagation:</strong> Para entrenar estos MLP automáticamente (en lugar de elegir los pesos a mano), se desarrolló el algoritmo de <strong>retropropagación del error (Backpropagation)</strong>, que sigue siendo el motor de entrenamiento de casi todas las redes neuronales modernas.</li>
            </ul>
            <p>En resumen, resolver XOR no es solo un ejercicio académico; es la puerta de entrada para comprender el poder y la elegancia de las redes neuronales profundas.</p>
        """
    }
]

# --- PASO 2: MOTOR DE TRANSFORMACIÓN (Python -> HTML) ---
def generar_html_secciones_mlp(datos_secciones):
    html_generado = []
    for seccion in datos_secciones:
        titulo_escapado = html.escape(seccion['titulo'])
        contenido_html = seccion.get('descripcion', '')
        plantilla_seccion = f"""
        <div class="topic-card">
            <div class="topic-header">
                <span class="topic-title">{titulo_escapado}</span>
                <i class="fas fa-chevron-down expand-icon"></i>
            </div>
            <div class="topic-content" id="content-mlp-{seccion['titulo'][0]}">
                {contenido_html}
            </div>
        </div>
        """
        html_generado.append(plantilla_seccion)
    return "\n".join(html_generado)

# --- PASO 3: PRE-CÁLCULO DEL CONTENIDO DINÁMICO ---
contenido_dinamico_html = generar_html_secciones_mlp(mlp_contenido)

# --- PASO 4: PLANTILLA MAESTRA HTML ---
plantilla_html_completa = """
<!DOCTYPE html>
<html lang="es">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>MLP Resuelve XOR</title>
  <link href="https://fonts.googleapis.com/css2?family=Inter:wght@300;400;500;600;700;800&family=JetBrains+Mono:wght@400;500&display=swap" rel="stylesheet">
  <link href="https://cdnjs.cloudflare.com/ajax/libs/font-awesome/6.4.0/css/all.min.css" rel="stylesheet">
  <script> MathJax = {{ tex: {{ inlineMath: [['\\\\(', '\\\\)']], displayMath: [['\\\\[', '\\\\]']] }} }}; </script>
  <script id="MathJax-script" async src="https://cdn.jsdelivr.net/npm/mathjax@3/es5/tex-mml-chtml.js"></script>
  <style>
    :root {{ --bg-primary: linear-gradient(135deg, #1d4350 0%, #a43931 100%); --bg-secondary: rgba(255, 255, 255, 0.88); --bg-tertiary: rgba(248, 250, 252, 0.85); --text-primary: #2c3e50; --text-secondary: #34495e; --text-light: #ffffff; --accent-primary: #a43931; --accent-secondary: #1d4350; --accent-gradient: linear-gradient(135deg, var(--accent-primary) 0%, var(--accent-secondary) 100%); --border-color: rgba(226, 232, 240, 0.9); --shadow-card: 0 15px 35px rgba(0, 0, 0, 0.08); --border-radius: 20px; --transition: all 0.4s cubic-bezier(0.25, 0.8, 0.25, 1); }}
    [data-theme="dark"] {{ --bg-primary: linear-gradient(135deg, #232526 0%, #414345 100%); --bg-secondary: rgba(26, 32, 44, 0.88); --bg-tertiary: rgba(45, 55, 72, 0.8); --text-primary: #f7fafc; --text-secondary: #a0aec0; --accent-primary: #ff8c00; --accent-secondary: #f7fafc; --border-color: rgba(255, 255, 255, 0.15); }}
    * {{ margin: 0; padding: 0; box-sizing: border-box; }} html {{ scroll-behavior: smooth; }} body {{ font-family: 'Inter', sans-serif; line-height: 1.8; background: var(--bg-primary); color: var(--text-primary); transition: var(--transition); min-height: 100vh; position: relative; overflow-x: hidden; }}
    .particles {{ position: fixed; top: 0; left: 0; width: 100%; height: 100%; pointer-events: none; z-index: -1; }} .particle {{ position: absolute; border-radius: 50%; animation: float 25s infinite linear; opacity: 0; background: rgba(255, 255, 255, 0.6); }} @keyframes float {{ 0% {{ transform: translateY(100vh) rotate(0deg); opacity: 0; }} 10%, 90% {{ opacity: 0.6; }} 100% {{ transform: translateY(-10vh) rotate(360deg); opacity: 0; }} }}
    .container {{ max-width: 1000px; margin: 0 auto; padding: 2rem; z-index: 1; }} .header {{ text-align: center; margin-bottom: 3rem; position: relative; }}
    .main-title {{ font-size: clamp(2.5rem, 5vw, 4.2rem); font-weight: 800; background: var(--accent-gradient); -webkit-background-clip: text; -webkit-text-fill-color: transparent; background-clip: text; margin-bottom: 1rem; }}
    .subtitle {{ font-size: 1.4rem; color: var(--text-light); font-weight: 400; opacity: 0.95; text-shadow: 0 2px 4px rgba(0, 0, 0, 0.3); max-width: 900px; margin: auto; }}
    .theme-toggle {{ position: fixed; top: 2rem; right: 2rem; width: 60px; height: 60px; border: 1px solid var(--border-color); border-radius: 50%; background: var(--bg-secondary); backdrop-filter: blur(15px); box-shadow: var(--shadow-card); cursor: pointer; display: flex; align-items: center; justify-content: center; font-size: 1.4rem; color: var(--accent-primary); transition: var(--transition); z-index: 1000; }}
    .theme-toggle:hover {{ transform: scale(1.15) rotate(180deg); box-shadow: 0 25px 50px rgba(0, 0, 0, 0.12), 0 0 30px rgba(102, 126, 234, 0.3); }}
    .lesson-container {{ display: flex; flex-direction: column; gap: 1.5rem; }} .topic-card {{ background: var(--bg-secondary); backdrop-filter: blur(20px); border-radius: var(--border-radius); box-shadow: var(--shadow-card); border: 2px solid var(--border-color); overflow: hidden; transition: var(--transition); }}
    .topic-header {{ cursor: pointer; padding: 1.5rem 2rem; display: flex; justify-content: space-between; align-items: center; }} .topic-title {{ font-size: 1.3rem; font-weight: 600; color: var(--text-primary); }}
    .expand-icon {{ font-size: 1.2rem; color: var(--text-secondary); transition: var(--transition); }} .topic-card.open .expand-icon {{ transform: rotate(180deg); }}
    .topic-content {{ max-height: 0; overflow: hidden; transition: max-height 1.8s ease, padding 1.2s ease; background: var(--bg-tertiary); }}
    .topic-card.open .topic-content {{ max-height: 4500px; padding: 1.5rem 2rem; border-top: 1px solid var(--border-color); }}
    .topic-content p, .topic-content li {{ color: var(--text-secondary); margin-bottom: 1rem; }} .topic-content ul, .topic-content ol {{ padding-left: 25px; margin-bottom: 1rem; }}
    .topic-content strong {{ color: var(--text-primary); font-weight: 600; }} .topic-content h4 {{ margin-top: 1.5rem; margin-bottom: .5rem; font-size: 1.1rem; color: var(--text-primary); border-left: 3px solid var(--accent-primary); padding-left: 10px; }}
    .jupyter-widgets {{ margin-top: 1rem; background: var(--bg-secondary); padding: 1rem; border-radius: 8px; }} .jupyter-widgets .widget-label {{ color: var(--text-primary); }}
    footer {{ text-align: center; margin-top: 4rem; padding-top: 2rem; border-top: 1px solid var(--border-color); }} footer p {{ color: var(--text-secondary); font-size: 0.9rem; opacity: 0.8; }}
  </style>
</head>
<body data-theme="dark">
  <div class="particles" id="particles-container"></div>
  <div class="theme-toggle" id="themeToggleButton" title="Cambiar tema"><i class="fas fa-moon" id="theme-icon"></i></div>
  <div class="container">
    <header class="header">
      <h1 class="main-title">MLP Resuelve XOR</h1>
      <p class="subtitle">La Clave para Entender el Aprendizaje No Lineal.</p>
    </header>
    <div class="lesson-container">
        {ml_contenido_html}
    </div>
    <footer> <p>Recurso interactivo generado por Experto en IA</p> </footer>
  </div>
  <script>
    (function() {{
        const themeToggleButton = document.getElementById('themeToggleButton');
        const themeIcon = document.getElementById('theme-icon');
        const bodyEl = document.body;
        function setTheme(theme) {{ bodyEl.setAttribute('data-theme', theme); localStorage.setItem('theme', theme); if (themeIcon) {{ themeIcon.className = theme === 'dark' ? 'fas fa-sun' : 'fas fa-moon'; }} }}
        themeToggleButton.addEventListener('click', () => {{ setTheme(bodyEl.getAttribute('data-theme') === 'dark' ? 'light' : 'dark'); }});
        setTheme(localStorage.getItem('theme') || 'dark');
        document.querySelectorAll('.topic-header').forEach(header => {{
            header.addEventListener('click', () => {{
                const card = header.parentElement;
                card.classList.toggle('open');
                if (card.classList.contains('open') && window.MathJax) {{ setTimeout(() => window.MathJax.typesetPromise([card]), 0); }}
                const contentId = header.nextElementSibling.id;
                if (contentId === 'content-mlp-4' && card.classList.contains('open')) {{
                    const widgetContainer = document.getElementById('widget-placeholder-mlp');
                    const contentDiv = document.getElementById(contentId);
                    if (widgetContainer) {{ contentDiv.appendChild(widgetContainer); }}
                }}
            }});
        }});
        const container = document.getElementById('particles-container');
        if (container) {{ for (let i = 0; i < 30; i++) {{ const p = document.createElement('div'); p.className = 'particle'; p.style.left = Math.random() * 100 + 'vw'; p.style.width = (Math.random() * 5 + 2) + 'px'; p.style.height = p.style.width; p.style.animationDelay = Math.random() * -20 + 's'; p.style.animationDuration = (15 + Math.random() * 10) + 's'; container.appendChild(p); }} }}
        const firstTopic = document.querySelector('.topic-card');
        if (firstTopic) {{ firstTopic.classList.add('open'); if (window.MathJax) {{ setTimeout(() => window.MathJax.typesetPromise([firstTopic]), 100); }} }}
    }})();
  </script>
</body>
</html>
"""

# --- PASO 5: WIDGET INTERACTIVO Y RENDERIZADO FINAL ---
# Definir el widget para la calculadora interactiva
def mlp_calculator(input_pair):
    x1, x2 = input_pair
    results = mlp_xor(x1, x2)

    html_out = f"""
    <div style="font-family: 'JetBrains Mono', monospace; line-height: 1.6;">
    <strong>Entrada: (x1, x2) = ({x1}, {x2})</strong><br><br>
    <u>Capa Oculta:</u><br>
    h1 = step(1*({x1}) - 1*({x2}) - 0.5) = step({results['h1_pre']:.1f}) = <strong>{results['h1_out']}</strong><br>
    h2 = step(-1*({x1}) + 1*({x2}) - 0.5) = step({results['h2_pre']:.1f}) = <strong>{results['h2_out']}</strong><br><br>
    <u>Capa de Salida:</u><br>
    y = step(1*({results['h1_out']}) + 1*({results['h2_out']}) - 0.5) = step({results['y_pre']:.1f}) = <strong style='color: #e74c3c; font-size:1.2em;'>{results['y_out']}</strong>
    </div>
    """
    display(HTML(html_out))

mlp_widget_output = widgets.Output()
with mlp_widget_output:
    interact(mlp_calculator,
             input_pair=widgets.Dropdown(
                 options=[('Entrada (0, 0)', (0,0)), ('Entrada (0, 1)', (0,1)),
                          ('Entrada (1, 0)', (1,0)), ('Entrada (1, 1)', (1,1))],
                 description='Selecciona:',
                 value=(0,0)
             ));

# Renderizar el HTML principal
display(HTML(plantilla_html_completa.format(ml_contenido_html=contenido_dinamico_html)))

# Renderizar el contenedor para el widget y el widget mismo
widget_placeholder_html = """
<div id="widget-placeholder-mlp" style="min-height: 150px;"></div>
<script>
    const widgetOutputEl = document.getElementById('widget-placeholder-mlp');
    if (widgetOutputEl) {
        const widgetView = document.querySelector('.jupyter-widgets-view');
        if (widgetView && !widgetView.parentElement.id) {
            widgetOutputEl.appendChild(widgetView);
        }
    }
</script>
"""
display(HTML(widget_placeholder_html))
display(mlp_widget_output)
```

```python
import itertools

# --- 1. Funciones Auxiliares para Visualización ---

def to_binary(s: str) -> str:
    """Convierte una cadena de texto a su representación binaria (UTF-8)."""
    return ' '.join(format(c, '08b') for c in s.encode('utf-8'))

# --- 2. Clase para el Cifrador XOR ---

class XORCipher:
    """
    Una clase simple para demostrar el cifrado XOR de clave simétrica.

    Este cifrador toma un mensaje y una clave, y aplica la operación XOR
    byte por byte. La misma operación sirve para cifrar y descifrar.
    """

    def _process(self, text: str, key: str) -> bytes:
        """
        El núcleo del proceso de cifrado/descifrado.

        Args:
            text (str): El texto a procesar (ya sea claro o cifrado).
            key (str): La clave simétrica.

        Returns:
            bytes: El resultado de la operación XOR como una cadena de bytes.
        """
        text_bytes = text.encode('utf-8')
        key_bytes = key.encode('utf-8')

        # Si la clave es más corta que el mensaje, se repite cíclicamente.
        # itertools.cycle crea un iterador que repite la clave: 'KEY' -> K, E, Y, K, E, Y, ...
        key_stream = itertools.cycle(key_bytes)

        # Aplicamos la operación XOR byte a byte entre el texto y el flujo de la clave.
        # zip empareja cada byte del texto con el siguiente byte del flujo de la clave.
        # La expresión `p_byte ^ k_byte` es donde ocurre la magia del XOR.
        return bytes([p_byte ^ k_byte for p_byte, k_byte in zip(text_bytes, key_stream)])

    def encrypt(self, plaintext: str, key: str) -> bytes:
        """
        Cifra un mensaje de texto plano.

        Args:
            plaintext (str): El mensaje original que se quiere ocultar.
            key (str): La clave secreta.

        Returns:
            bytes: El mensaje cifrado (ciphertext) como una cadena de bytes.
                   Normalmente no se puede decodificar a texto legible.
        """
        print("--- Proceso de Cifrado ---")
        print(f"Mensaje Original: '{plaintext}'")
        print(f"Clave Secreta:   '{key}'\n")

        return self._process(plaintext, key)

    def decrypt(self, ciphertext: bytes, key: str) -> str:
        """
        Descifra un mensaje cifrado.

        Args:
            ciphertext (bytes): El mensaje cifrado.
            key (str): La misma clave secreta que se usó para cifrar.

        Returns:
            str: El mensaje original recuperado.
        """
        print("\n--- Proceso de Descifrado ---")
        print(f"Mensaje Cifrado: {ciphertext}")
        print(f"Clave Secreta:   '{key}'\n")

        decrypted_bytes = self._process(ciphertext.decode('latin1'), key) # Usamos latin1 para manejar cualquier byte
        return decrypted_bytes.decode('utf-8')


# --- 3. Demostración Práctica ---

if __name__ == "__main__":
    # --- Configuración Inicial ---
    mi_cifrador = XORCipher()
    mensaje_secreto = "El perceptron no puede con XOR!"
    clave_simetrica = "ALGEBRA"

    print("="*60)
    print("DEMOSTRACIÓN DE CRIPTOGRAFÍA SIMÉTRICA CON XOR")
    print("="*60)

    # --- Paso 1: Cifrar el mensaje ---
    mensaje_cifrado = mi_cifrador.encrypt(mensaje_secreto, clave_simetrica)
    print(f"Resultado Cifrado (bytes): {mensaje_cifrado}")
    # Nota: Intentar imprimir esto como texto normal daría caracteres extraños o errores.
    # print(f"Resultado Cifrado (como texto): {mensaje_cifrado.decode('latin1')}") # <-- ¡Texto basura!

    # --- Paso 2: Descifrar el mensaje ---
    mensaje_recuperado = mi_cifrador.decrypt(mensaje_cifrado, clave_simetrica)
    print(f"Resultado Descifrado: '{mensaje_recuperado}'")

    # --- Paso 3: Verificación ---
    print("\n--- Verificación Final ---")
    print(f"¿El mensaje recuperado es igual al original? {'✅ SÍ' if mensaje_secreto == mensaje_recuperado else '❌ NO'}")
    print("="*60)

    # --- Explicación Visual del Proceso (Primeros caracteres) ---
    print("\n¿CÓMO FUNCIONA? VEAMOS EL PRIMER CARÁCTER:")
    char_m = mensaje_secreto[0]
    char_k = clave_simetrica[0]
    byte_m = ord(char_m)
    byte_k = ord(char_k)
    byte_c = byte_m ^ byte_k # Cifrado

    print(f"  Mensaje ('{char_m}'): {byte_m:08b} ({byte_m})")
    print(f"  Clave   ('{char_k}'): {byte_k:08b} ({byte_k})")
    print( "XOR -----------------------------")
    print(f"  Cifrado ('?'): {byte_c:08b} ({byte_c})\n")

    # Ahora, desciframos
    byte_d = byte_c ^ byte_k # Descifrado
    print("Para descifrar, hacemos lo mismo otra vez:")
    print(f"  Cifrado ('?'): {byte_c:08b} ({byte_c})")
    print(f"  Clave   ('{char_k}'): {byte_k:08b} ({byte_k})")
    print( "XOR -----------------------------")
    print(f"  Original('{chr(byte_d)}'): {byte_d:08b} ({byte_d})")
```
