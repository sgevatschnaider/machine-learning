```python
from IPython.display import display, HTML

html_code = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <title>Probabilidad: Lanzamiento de Dos Dados</title>
    <style>
        body { font-family: Arial, sans-serif; }
        .container { text-align: center; padding: 20px; }
        .dice { font-size: 2em; }
        .result { font-size: 1.2em; margin-top: 10px; }
        button { padding: 10px 20px; font-size: 1em; margin-top: 20px; }
    </style>
</head>
<body>
    <div class="container">
        <h1>Simulaci√≥n de Probabilidad: Lanzamiento de Dos Dados</h1>
        <p>Intenta obtener "dos 6" lanzando dos dados. Observa la probabilidad.</p>

        <div id="dice" class="dice">üé≤ üé≤</div>
        <button onclick="rollDice()">Lanzar Dados</button>
        <button onclick="resetGame()">Reiniciar</button>

        <div id="result" class="result"></div>
        <div id="probability" class="result"></div>
    </div>

    <script>
        let totalRolls = 0;
        let successfulRolls = 0;

        function rollDice() {
            const die1 = Math.floor(Math.random() * 6) + 1;
            const die2 = Math.floor(Math.random() * 6) + 1;
            document.getElementById("dice").innerHTML = `üé≤ ${die1} üé≤ ${die2}`;

            totalRolls++;
            if (die1 === 6 && die2 === 6) {
                successfulRolls++;
            }

            const probability = (successfulRolls / totalRolls * 100).toFixed(2);
            document.getElementById("result").innerHTML = `Intentos: ${totalRolls}, "Dos 6": ${successfulRolls}`;
            document.getElementById("probability").innerHTML = `Probabilidad Observada de "dos 6": ${probability}%`;
        }

        function resetGame() {
            totalRolls = 0;
            successfulRolls = 0;
            document.getElementById("dice").innerHTML = "üé≤ üé≤";
            document.getElementById("result").innerHTML = "";
            document.getElementById("probability").innerHTML = "";
        }
    </script>
</body>
</html>
"""

display(HTML(html_code))
```

```python
import random
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Funci√≥n para simular lanzamientos y calcular la probabilidad emp√≠rica
def simulate_rolls(num_rolls):
    successful_rolls = 0
    empirical_probabilities = []

    for i in range(1, num_rolls + 1):
        die1 = random.randint(1, 6)
        die2 = random.randint(1, 6)
        if die1 == 6 and die2 == 6:
            successful_rolls += 1
        # Calculamos y almacenamos la probabilidad emp√≠rica en cada paso
        empirical_probability = (successful_rolls / i) * 100
        empirical_probabilities.append(empirical_probability)

    return empirical_probabilities

# Funci√≥n para actualizar el gr√°fico con el n√∫mero de lanzamientos seleccionado
def update_plot(num_rolls):
    empirical_probabilities = simulate_rolls(num_rolls)
    theoretical_probability = [2.78] * num_rolls  # Probabilidad te√≥rica constante

    plt.figure(figsize=(12, 6))
    plt.plot(empirical_probabilities, label='Probabilidad Emp√≠rica (Observada)', color='#1f77b4', linewidth=2)
    plt.plot(theoretical_probability, color='red', linestyle='--', label='Probabilidad Te√≥rica (2.78%)')
    plt.xlabel('N√∫mero de Lanzamientos', fontsize=12)
    plt.ylabel('Probabilidad (%)', fontsize=12)
    plt.ylim(0, 5)  # Limitar el rango y del eje para claridad
    plt.title(f'Convergencia de la Probabilidad Emp√≠rica hacia la Te√≥rica con {num_rolls} Lanzamientos', fontsize=14)
    plt.legend(loc="upper right", fontsize=10)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# Deslizador interactivo para seleccionar el n√∫mero de lanzamientos
num_rolls_slider = widgets.IntSlider(
    value=100,
    min=10,
    max=10000,
    step=10,
    description='Lanzamientos:',
    style={'description_width': 'initial'}
)

# Bot√≥n de reinicio
reset_button = widgets.Button(
    description="Reiniciar",
    button_style='warning',
    tooltip="Reiniciar gr√°fico"
)

# Funci√≥n de reinicio para establecer el deslizador en su valor inicial
def reset(event):
    num_rolls_slider.value = 100  # Valor inicial del deslizador

reset_button.on_click(reset)

# Conectar el deslizador y el bot√≥n a la funci√≥n de actualizaci√≥n del gr√°fico
ui = widgets.HBox([num_rolls_slider, reset_button])
out = widgets.interactive_output(update_plot, {'num_rolls': num_rolls_slider})

# Mostrar la interfaz y el gr√°fico
display(ui, out)
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

# Configuraci√≥n de la simulaci√≥n
num_lanzamientos_monedas = 1000  # N√∫mero de lanzamientos de la moneda
num_lanzamientos_dado = 100  # N√∫mero de lanzamientos del dado

# Simulaci√≥n Frecuentista: Lanzamiento de una moneda
def frecuentista_monedas(num_lanzamientos):
    resultados = {"cara": 0, "ceca": 0}
    frecuencias = []

    for _ in range(num_lanzamientos):
        resultado = random.choice(["cara", "ceca"])
        resultados[resultado] += 1
        frecuencia_cara = resultados["cara"] / (resultados["cara"] + resultados["ceca"])
        frecuencias.append(frecuencia_cara * 100)  # Convertimos a porcentaje

    return frecuencias

# Simulaci√≥n Bayesiana: C√°lculo de probabilidad de una moneda justa
def bayesiana_monedas(num_caras, num_lanzamientos):
    # Par√°metros previos de Bayes (creemos que la moneda es justa al 50%)
    alpha, beta = 1, 1
    posterior = (num_caras + alpha) / (num_lanzamientos + alpha + beta)
    return posterior * 100  # Probabilidad en porcentaje

# Variable Aleatoria: Posibles resultados al lanzar un dado
def resultados_dado():
    valores = [1, 2, 3, 4, 5, 6]
    conteo = {i: 0 for i in valores}
    resultados = []

    for _ in range(num_lanzamientos_dado):
        resultado = random.choice(valores)
        conteo[resultado] += 1
        resultados.append(resultado)

    return resultados, conteo

# Ejecutar simulaci√≥n Frecuentista
frecuencias = frecuentista_monedas(num_lanzamientos_monedas)

# Calcular probabilidad bayesiana despu√©s de num_lanzamientos_monedas lanzamientos
num_caras = frecuencias[-1] * num_lanzamientos_monedas / 100  # Usamos el √∫ltimo valor de frecuencia
prob_bayesiana = bayesiana_monedas(num_caras, num_lanzamientos_monedas)

# Ejecutar simulaci√≥n de lanzamientos de dado
resultados_dado, conteo_dado = resultados_dado()

# Visualizaci√≥n
plt.figure(figsize=(14, 6))

# Gr√°fico de probabilidad frecuentista (moneda)
plt.subplot(1, 2, 1)
plt.plot(frecuencias, label="Frecuencia de 'Cara'")
plt.axhline(50, color='red', linestyle='--', label="Probabilidad Te√≥rica (50%)")
plt.xlabel("N√∫mero de Lanzamientos")
plt.ylabel("Frecuencia de 'Cara' (%)")
plt.title("Probabilidad Frecuentista: Lanzamientos de una Moneda")
plt.legend()
plt.grid()

# Mostrar resultado de probabilidad bayesiana
print(f"Probabilidad Bayesiana de que la moneda sea justa: {prob_bayesiana:.2f}%")

# Visualizaci√≥n de la variable aleatoria: resultados de un dado
plt.subplot(1, 2, 2)
plt.bar(conteo_dado.keys(), conteo_dado.values(), color="skyblue")
plt.xlabel("Resultados posibles (1 a 6)")
plt.ylabel("Frecuencia")
plt.title("Distribuci√≥n de resultados al lanzar un dado")
plt.grid(axis='y')

plt.tight_layout()
plt.show()
```

```python
import random
import matplotlib.pyplot as plt

def calcular_apuesta_justa():
    """Calcula la apuesta justa basada en el promedio esperado de un dado de 6 caras."""
    apuesta_justa = sum(range(1, 7)) / 6  # Promedio de 1, 2, 3, 4, 5, 6
    return apuesta_justa

def lanzar_dado():
    """Simula el lanzamiento de un dado."""
    return random.randint(1, 6)

def jugar(apuesta, costo_participacion, num_juegos):
    """Simula el juego m√∫ltiples veces y retorna la ganancia neta total."""
    ganancia_total = 0
    for _ in range(num_juegos):
        resultado = lanzar_dado()
        ganancia = resultado - costo_participacion
        ganancia_total += ganancia
    return ganancia_total

def simulacion_apuesta():
    # Paso 1: Calcular y mostrar la apuesta justa
    apuesta_justa = calcular_apuesta_justa()
    print(f"La apuesta justa para el juego es: {apuesta_justa}")

    # Paso 2: Solicitar al usuario la cantidad a apostar
    try:
        apuesta_usuario = float(input("¬øCu√°nto deseas apostar por juego?: "))
    except ValueError:
        print("Por favor, ingresa un n√∫mero v√°lido.")
        return

    # Validar la apuesta del usuario
    if apuesta_usuario < apuesta_justa:
        print(f"La apuesta ingresada ({apuesta_usuario}) es menor a la apuesta justa ({apuesta_justa}). Apuesta rechazada.")
        return
    else:
        print(f"Apuesta aceptada. Se proceder√° con una simulaci√≥n con diferentes valores de intentos.")

    # Valores de intentos para la simulaci√≥n
    intentos = [100, 1000, 10000, 100000]
    ganancias = []

    # Simulaci√≥n de la apuesta para diferentes cantidades de juegos
    for num_juegos in intentos:
        ganancia = jugar(apuesta=apuesta_usuario, costo_participacion=apuesta_usuario, num_juegos=num_juegos)
        ganancias.append(ganancia)
        print(f"\n--- Resultados para {num_juegos} juegos ---")
        print(f"Ganancia total del jugador: {ganancia}")
        if ganancia > 0:
            print("El jugador ha tenido una ganancia positiva.")
        elif ganancia < 0:
            print("El jugador ha tenido una p√©rdida.")
        else:
            print("El juego result√≥ en empate total.")

    # Visualizaci√≥n de los resultados
    plt.figure(figsize=(10, 6))
    plt.plot(intentos, ganancias, marker="o", linestyle="-")
    plt.axhline(0, color="black", linestyle="--", linewidth=0.8, label="Punto de Equilibrio")
    plt.title("Ganancias Totales del Jugador para Distintos N√∫meros de Juegos")
    plt.xlabel("N√∫mero de Juegos")
    plt.ylabel("Ganancia Total ($)")
    plt.xscale("log")
    plt.legend()
    plt.grid(True)
    plt.show()

# Ejecutar la simulaci√≥n
simulacion_apuesta()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Funci√≥n para simular el juego de dados con diferentes apuestas
def dice_game_simulation(money, bet, games, bet_type="Pass Line", strategy="Fixed Bet"):
    profit = [money]
    initial_money = money
    wins, losses = 0, 0

    # Funciones de reglas seg√∫n el tipo de apuesta
    def pass_line_bet():
        nonlocal money, wins, losses
        die1, die2 = np.random.randint(1, 7), np.random.randint(1, 7)
        dice_sum = die1 + die2
        if dice_sum in [7, 11]:  # Gana en 7 u 11
            money += bet
            wins += 1
        elif dice_sum in [2, 3, 12]:  # Pierde en 2, 3 o 12
            money -= bet
            losses += 1
        else:  # Punto establecido
            point = dice_sum
            while True:
                die1, die2 = np.random.randint(1, 7), np.random.randint(1, 7)
                dice_sum = die1 + die2
                if dice_sum == point:
                    money += bet
                    wins += 1
                    break
                elif dice_sum == 7:
                    money -= bet
                    losses += 1
                    break

    def dont_pass_bet():
        nonlocal money, wins, losses
        die1, die2 = np.random.randint(1, 7), np.random.randint(1, 7)
        dice_sum = die1 + die2
        if dice_sum in [2, 3]:  # Gana en 2 o 3
            money += bet
            wins += 1
        elif dice_sum in [7, 11]:  # Pierde en 7 o 11
            money -= bet
            losses += 1
        elif dice_sum == 12:  # Push (empate)
            pass
        else:  # Punto establecido
            point = dice_sum
            while True:
                die1, die2 = np.random.randint(1, 7), np.random.randint(1, 7)
                dice_sum = die1 + die2
                if dice_sum == 7:
                    money += bet
                    wins += 1
                    break
                elif dice_sum == point:
                    money -= bet
                    losses += 1
                    break

    # Estrategias de apuestas
    fibonacci_sequence = [1, 1]
    def update_fibonacci_sequence():
        fibonacci_sequence.append(fibonacci_sequence[-1] + fibonacci_sequence[-2])

    current_bet = bet
    for game in range(games):
        if strategy == "Martingala" and losses > 0:
            current_bet = min(current_bet * 2, money)
        elif strategy == "Fibonacci" and losses > 0:
            update_fibonacci_sequence()
            current_bet = min(bet * fibonacci_sequence[min(len(fibonacci_sequence) - 1, game)], money)
        else:
            current_bet = bet

        # Selecci√≥n del tipo de apuesta
        if bet_type == "Pass Line":
            pass_line_bet()
        elif bet_type == "Don't Pass":
            dont_pass_bet()

        profit.append(money)
        if money <= 0:
            break

    # C√°lculo de estad√≠sticas
    final_money = profit[-1]
    gain_loss_ratio = (final_money - initial_money) / initial_money * 100
    print(f"Dinero final: {final_money}")
    print(f"Ganancia/P√©rdida (%): {gain_loss_ratio:.2f}%")
    print(f"Juegos ganados: {wins}, Juegos perdidos: {losses}")
    print(f"M√°ximo dinero alcanzado: {max(profit)}, M√≠nimo dinero alcanzado: {min(profit)}")

    # Graficar resultados
    plt.figure(figsize=(12, 6))
    plt.plot(profit, color='green' if gain_loss_ratio >= 0 else 'red', linewidth=2)
    plt.axhline(initial_money, color='blue', linestyle='--', label='Dinero Inicial')
    plt.xlabel("N√∫mero de Juegos")
    plt.ylabel("Dinero")
    plt.title("Simulaci√≥n de Juego de Dados")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

    # Histograma de distribuci√≥n
    plt.figure(figsize=(10, 5))
    plt.hist(profit, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
    plt.xlabel("Dinero")
    plt.ylabel("Frecuencia")
    plt.title("Distribuci√≥n del Dinero a lo Largo de la Simulaci√≥n")
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

# Funci√≥n para manejar la interacci√≥n de widgets
def on_simulate_button_clicked(b):
    clear_output(wait=True)
    display(ui)  # Mostrar los widgets nuevamente despu√©s de limpiar la salida

    # Obtener los valores de los widgets
    money = money_widget.value
    bet = bet_widget.value
    games = games_widget.value
    bet_type = bet_type_dropdown.value
    strategy = strategy_dropdown.value

    # Ejecutar la simulaci√≥n con los valores seleccionados
    dice_game_simulation(money, bet, games, bet_type, strategy)

# Funci√≥n para resetear los widgets
def on_reset_button_clicked(b):
    clear_output(wait=True)
    display(ui)
    print("Valores reiniciados. Listo para una nueva simulaci√≥n.")

# Widgets de entrada
money_widget = widgets.IntText(value=1000, description="Dinero inicial:")
bet_widget = widgets.IntText(value=50, description="Apuesta por juego:")
games_widget = widgets.IntSlider(value=100, min=10, max=1000, step=10, description="N√∫mero de juegos:")
bet_type_dropdown = widgets.Dropdown(
    options=["Pass Line", "Don't Pass"],
    value="Pass Line",
    description="Tipo de Apuesta:"
)
strategy_dropdown = widgets.Dropdown(
    options=["Fixed Bet", "Martingala", "Fibonacci"],
    value="Fixed Bet",
    description="Estrategia:"
)

# Botones
simulate_button = widgets.Button(description="Simular")
simulate_button.on_click(on_simulate_button_clicked)
reset_button = widgets.Button(description="Reiniciar", button_style='warning')
reset_button.on_click(on_reset_button_clicked)

# Contenedor de interfaz
ui = widgets.VBox([money_widget, bet_widget, games_widget, bet_type_dropdown, strategy_dropdown, simulate_button, reset_button])

# Mostrar interfaz de usuario
display(ui)
```

```python
import random
import matplotlib.pyplot as plt

def calcular_apuesta_justa():
    """Calcula la apuesta justa basada en el promedio esperado de un dado de 6 caras."""
    apuesta_justa = sum(range(1, 7)) / 6  # Promedio de 1, 2, 3, 4, 5, 6
    return apuesta_justa

def lanzar_dado():
    """Simula el lanzamiento de un dado."""
    return random.randint(1, 6)

def jugar(apuesta, costo_participacion, num_juegos):
    """Simula el juego m√∫ltiples veces y retorna la ganancia neta total y la lista de ganancias."""
    ganancias = []
    for _ in range(num_juegos):
        resultado = lanzar_dado()
        ganancia = resultado - costo_participacion
        ganancias.append(ganancia)
    ganancia_total = sum(ganancias)
    return ganancia_total, ganancias

def calcular_varianza(ganancias):
    """Calcula la varianza de las ganancias."""
    media = sum(ganancias) / len(ganancias)
    varianza = sum((x - media) ** 2 for x in ganancias) / len(ganancias)
    return varianza

def simulacion_apuesta():
    # Paso 1: Calcular y mostrar la apuesta justa
    apuesta_justa = calcular_apuesta_justa()
    print(f"La apuesta justa para el juego es: {apuesta_justa}")

    # Paso 2: Solicitar al usuario la cantidad a apostar
    try:
        apuesta_usuario = float(input("¬øCu√°nto deseas apostar por juego?: "))
    except ValueError:
        print("Por favor, ingresa un n√∫mero v√°lido.")
        return

    # Validar la apuesta del usuario
    if apuesta_usuario < apuesta_justa:
        print(f"La apuesta ingresada ({apuesta_usuario}) es menor a la apuesta justa ({apuesta_justa}). Apuesta rechazada.")
        return
    else:
        print(f"Apuesta aceptada. Se proceder√° con una simulaci√≥n con diferentes valores de intentos.")

    # Valores de intentos para la simulaci√≥n
    intentos = [100, 1000, 10000, 100000]
    ganancias_totales = []
    varianzas = []

    # Simulaci√≥n de la apuesta para diferentes cantidades de juegos
    for num_juegos in intentos:
        ganancia_total, ganancias = jugar(apuesta=apuesta_usuario, costo_participacion=apuesta_usuario, num_juegos=num_juegos)
        varianza = calcular_varianza(ganancias)

        ganancias_totales.append(ganancia_total)
        varianzas.append(varianza)

        print(f"\n--- Resultados para {num_juegos} juegos ---")
        print(f"Ganancia total del jugador: {ganancia_total}")
        print(f"Varianza de las ganancias: {varianza}")

        if ganancia_total > 0:
            print("El jugador ha tenido una ganancia positiva.")
        elif ganancia_total < 0:
            print("El jugador ha tenido una p√©rdida.")
        else:
            print("El juego result√≥ en empate total.")

    # Visualizaci√≥n de los resultados
    plt.figure(figsize=(10, 6))
    plt.plot(intentos, ganancias_totales, marker="o", linestyle="-", label="Ganancias Totales")
    plt.plot(intentos, varianzas, marker="x", linestyle="--", label="Varianza de las Ganancias", color="orange")
    plt.axhline(0, color="black", linestyle="--", linewidth=0.8, label="Punto de Equilibrio")
    plt.title("Ganancias Totales y Varianza para Distintos N√∫meros de Juegos")
    plt.xlabel("N√∫mero de Juegos")
    plt.ylabel("Valor")
    plt.xscale("log")
    plt.legend()
    plt.grid(True)
    plt.show()

# Ejecutar la simulaci√≥n
simulacion_apuesta()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Funci√≥n para simular lanzamientos de dados y contar n√∫meros primos
def prime_number_simulation(num_rolls):
    prime_count = 0  # Contador de n√∫meros primos (2, 3, 5)
    prime_numbers = {2, 3, 5}  # Conjunto de n√∫meros primos en un dado de 6 caras

    for _ in range(num_rolls):
        roll = np.random.randint(1, 7)  # Generar un n√∫mero aleatorio entre 1 y 6
        if roll in prime_numbers:
            prime_count += 1  # Contar como √©xito si es primo

    return prime_count

# Funci√≥n para ejecutar la simulaci√≥n y mostrar resultados
def run_simulation(num_rolls, bet_primes):
    prime_count = prime_number_simulation(num_rolls)
    esperanza_primos = num_rolls * 0.5  # Esperanza matem√°tica
    varianza_primos = num_rolls * 0.5 * (1 - 0.5)  # Varianza

    # Mostrar resultados en texto
    print(f"Total de lanzamientos: {num_rolls}")
    print(f"N√∫meros primos obtenidos: {prime_count}")
    print(f"Esperanza matem√°tica de n√∫meros primos: {esperanza_primos}")
    print(f"Varianza de n√∫meros primos: {varianza_primos}")

    # Resultados de la apuesta
    if prime_count >= bet_primes:
        print(f"¬°Ganaste! Obtuvimos al menos {bet_primes} n√∫meros primos.")
    else:
        print(f"Perdiste. No se lograron los {bet_primes} n√∫meros primos apostados.")

    # Gr√°fico de resultados
    plt.bar(["N√∫meros Primos", "Otros"], [prime_count, num_rolls - prime_count], color=["green", "red"])
    plt.xlabel("Resultado")
    plt.ylabel("Frecuencia")
    plt.title("Distribuci√≥n de N√∫meros Primos y Otros Resultados")
    plt.show()

# Funci√≥n para reiniciar los widgets y limpiar la salida
def reset_simulation(b):
    clear_output(wait=True)
    display(num_rolls_widget, bet_primes_widget, simulate_button, reset_button)
    print("Valores reiniciados. Listo para una nueva simulaci√≥n.")

# Widgets interactivos
num_rolls_widget = widgets.IntSlider(value=20, min=10, max=100, step=5, description="N¬∫ de Lanzamientos:")
bet_primes_widget = widgets.IntSlider(value=5, min=1, max=20, description="Apuesta (Primos):")

# Bot√≥n para iniciar la simulaci√≥n
simulate_button = widgets.Button(description="Simular")
simulate_button.on_click(lambda b: run_simulation(num_rolls_widget.value, bet_primes_widget.value))

# Bot√≥n para resetear los valores
reset_button = widgets.Button(description="Resetear", button_style='warning')
reset_button.on_click(reset_simulation)

# Mostrar interfaz
display(num_rolls_widget, bet_primes_widget, simulate_button, reset_button)
```

```python
from IPython.display import display, HTML

# Contenido HTML para explicar la Distribuci√≥n de Poisson
html_content = """
<!DOCTYPE html>
<html lang="es">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Distribuci√≥n de Poisson</title>
    <style>
        body { font-family: Arial, sans-serif; line-height: 1.6; color: #333; margin: 20px; background-color: #f4f4f9; }
        h1 { color: #0056b3; text-align: center; font-size: 2.5em; margin-bottom: 20px; }
        h2 { color: #333; font-size: 1.6em; background-color: #e0f7fa; padding: 10px; border-radius: 8px; margin-top: 30px; cursor: pointer; }
        p { font-size: 1.1em; margin-bottom: 15px; }
        ul { margin-left: 20px; }
        .formula { font-size: 1.3em; background-color: #e8f0fe; padding: 10px; border-radius: 5px; text-align: center; margin: 15px 0; }
        .content { display: none; margin-top: 10px; padding-left: 10px; }
        .example { background-color: #f0fff0; padding: 10px; border-left: 5px solid #4caf50; margin: 15px 0; }
    </style>
</head>
<body>

    <h1>Distribuci√≥n de Poisson</h1>

    <h2 onclick="toggleContent('definicion')">¬øQu√© es la Distribuci√≥n de Poisson?</h2>
    <div id="definicion" class="content">
        <p>La <strong>Distribuci√≥n de Poisson</strong> es una distribuci√≥n de probabilidad discreta que describe la probabilidad de un n√∫mero determinado de eventos que ocurren en un intervalo de tiempo o espacio dado, bajo las siguientes condiciones:</p>
        <ul>
            <li>Los eventos ocurren de forma independiente entre s√≠.</li>
            <li>La tasa de ocurrencia de eventos es constante en el tiempo o espacio considerado.</li>
            <li>Dos eventos no pueden ocurrir simult√°neamente.</li>
        </ul>
        <p>La Distribuci√≥n de Poisson es √∫til para modelar eventos raros o inusuales, como llamadas telef√≥nicas en un centro de atenci√≥n, fallos en un sistema de red, o llegada de clientes a una tienda.</p>
    </div>

    <h2 onclick="toggleContent('formula')">F√≥rmula de la Distribuci√≥n de Poisson</h2>
    <div id="formula" class="content">
        <p>La probabilidad de observar exactamente <em>k</em> eventos en un intervalo de tiempo o espacio dado se calcula con la f√≥rmula de Poisson:</p>
        <div class="formula">
            P(X = k) = (Œª^k * e^(-Œª)) / k!
        </div>
        <p>donde:</p>
        <ul>
            <li><em>P(X = k)</em>: Probabilidad de observar exactamente <em>k</em> eventos.</li>
            <li><em>Œª</em>: Tasa promedio de ocurrencia de eventos en el intervalo de tiempo o espacio.</li>
            <li><em>e</em>: La constante de Euler, aproximadamente igual a 2.718.</li>
            <li><em>k!</em>: Factorial de <em>k</em>, que es el n√∫mero de eventos observados.</li>
        </ul>
    </div>

    <h2 onclick="toggleContent('esperanza_varianza')">Esperanza y Varianza en la Distribuci√≥n de Poisson</h2>
    <div id="esperanza_varianza" class="content">
        <p>En la Distribuci√≥n de Poisson, la <strong>esperanza matem√°tica</strong> y la <strong>varianza</strong> son iguales y est√°n dadas por el par√°metro <em>Œª</em>:</p>
        <div class="formula">
            E(X) = Œª <br>
            Var(X) = Œª
        </div>
        <p>Esto significa que tanto el valor promedio esperado como la variabilidad de los eventos en el tiempo o espacio son determinados por <em>Œª</em>. Si <em>Œª</em> es alto, se esperan m√°s eventos y una mayor variabilidad; si <em>Œª</em> es bajo, los eventos ser√°n menos frecuentes y con menor dispersi√≥n.</p>
    </div>

    <h2 onclick="toggleContent('ejemplo')">Ejemplo: Fallos en un Sistema de Red</h2>
    <div id="ejemplo" class="content example">
        <p>Imaginemos que queremos modelar la cantidad de fallos en un sistema de red. Supongamos que, en promedio, ocurren 2 fallos por hora. La tasa de fallos por hora es <em>Œª = 2</em>.</p>
        <p>Usando la distribuci√≥n de Poisson, podemos calcular la probabilidad de que ocurran exactamente <em>k</em> fallos en una hora espec√≠fica. Por ejemplo:</p>
        <ul>
            <li><strong>Probabilidad de 0 fallos:</strong> P(X = 0) = (2^0 * e^(-2)) / 0! ‚âà 0.1353</li>
            <li><strong>Probabilidad de 1 fallo:</strong> P(X = 1) = (2^1 * e^(-2)) / 1! ‚âà 0.2707</li>
            <li><strong>Probabilidad de 2 fallos:</strong> P(X = 2) = (2^2 * e^(-2)) / 2! ‚âà 0.2707</li>
            <li><strong>Probabilidad de 3 fallos:</strong> P(X = 3) = (2^3 * e^(-2)) / 3! ‚âà 0.1804</li>
        </ul>
        <p>Este modelo es √∫til para evaluar la frecuencia de eventos como fallos en un sistema, ya que proporciona una probabilidad de ocurrencia para cada posible cantidad de eventos en un intervalo determinado.</p>
    </div>

    <h2 onclick="toggleContent('aplicaciones')">Aplicaciones de la Distribuci√≥n de Poisson</h2>
    <div id="aplicaciones" class="content">
        <p>La Distribuci√≥n de Poisson es ampliamente utilizada en diferentes √°reas debido a su capacidad para modelar eventos raros o poco frecuentes. Algunos ejemplos de aplicaciones incluyen:</p>
        <ul>
            <li><strong>Salud P√∫blica:</strong> N√∫mero de nuevos casos de una enfermedad en una poblaci√≥n en un per√≠odo determinado.</li>
            <li><strong>Infraestructura:</strong> Cantidad de fallos en una red de servicios p√∫blicos (agua, electricidad) en un mes.</li>
            <li><strong>Log√≠stica:</strong> Llegada de paquetes o clientes a una instalaci√≥n en un tiempo espec√≠fico.</li>
            <li><strong>Producci√≥n:</strong> N√∫mero de defectos en productos fabricados en una l√≠nea de producci√≥n en un d√≠a.</li>
        </ul>
        <p>Gracias a sus propiedades, la Distribuci√≥n de Poisson es ideal para modelar la ocurrencia de eventos en intervalos de tiempo o espacio en los que la probabilidad de ocurrencia es baja pero constante.</p>
    </div>

    <script>
        function toggleContent(id) {
            var content = document.getElementById(id);
            if (content.style.display === "none" || content.style.display === "") {
                content.style.display = "block";
            } else {
                content.style.display = "none";
            }
        }
        document.querySelectorAll('.content').forEach(div => div.style.display = 'none');
    </script>
</body>
</html>
"""

# Mostrar el contenido HTML en Google Colab
display(HTML(html_content))
```

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

# Funci√≥n para simular fallos en un sistema de red
def network_failure_simulation(lambda_failures, num_hours):
    # Simula el n√∫mero de fallos en cada hora usando la distribuci√≥n de Poisson
    failures_per_hour = np.random.poisson(lambda_failures, num_hours)
    total_failures = np.sum(failures_per_hour)

    # C√°lculo de la esperanza y varianza
    esperanza = lambda_failures * num_hours
    varianza = lambda_failures * num_hours

    return failures_per_hour, total_failures, esperanza, varianza

# Funci√≥n para ejecutar la simulaci√≥n y mostrar resultados
def run_simulation(lambda_failures, num_hours):
    failures_per_hour, total_failures, esperanza, varianza = network_failure_simulation(lambda_failures, num_hours)

    # Mostrar resultados de la simulaci√≥n
    print(f"Tasa de fallos por hora: {lambda_failures}")
    print(f"Total de horas simuladas: {num_hours}")
    print(f"Total de fallos observados: {total_failures}")
    print(f"Esperanza matem√°tica de fallos: {esperanza}")
    print(f"Varianza de fallos: {varianza}")

    # Mostrar un gr√°fico de los fallos por hora
    plt.figure(figsize=(12, 6))
    plt.bar(range(1, num_hours + 1), failures_per_hour, color="blue")
    plt.axhline(esperanza / num_hours, color="red", linestyle="--", label="Esperanza por Hora")
    plt.xlabel("Hora")
    plt.ylabel("Cantidad de Fallos")
    plt.title("Simulaci√≥n de Fallos en un Sistema de Red")
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.7)
    plt.show()

# Funci√≥n para resetear los widgets y limpiar la salida
def reset_simulation(b):
    clear_output(wait=True)
    display(lambda_widget, hours_widget, simulate_button, reset_button)
    print("Valores reiniciados. Listo para una nueva simulaci√≥n.")

# Widgets para la interfaz interactiva
lambda_widget = widgets.FloatSlider(value=2.0, min=0.5, max=10.0, step=0.5, description="Tasa de Fallos/h:")
hours_widget = widgets.IntSlider(value=24, min=1, max=168, step=1, description="Horas Simuladas:")

# Bot√≥n para iniciar la simulaci√≥n
simulate_button = widgets.Button(description="Simular")
simulate_button.on_click(lambda b: run_simulation(lambda_widget.value, hours_widget.value))

# Bot√≥n para resetear los valores
reset_button = widgets.Button(description="Resetear", button_style='warning')
reset_button.on_click(reset_simulation)

# Mostrar interfaz
display(lambda_widget, hours_widget, simulate_button, reset_button)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew
import pandas as pd

# Configuraci√≥n de Seaborn
sns.set(style="whitegrid")

# Paso 1: Generar distribuciones
np.random.seed(42)  # Para reproducibilidad

# Distribuci√≥n Sim√©trica (Normal)
data_symmetric = np.random.normal(loc=0, scale=1, size=1000)

# Distribuci√≥n con Asimetr√≠a Positiva (Exponencial)
data_positive_skew = np.random.exponential(scale=1, size=1000)

# Distribuci√≥n con Asimetr√≠a Negativa (Inversa de Exponencial)
data_negative_skew = -np.random.exponential(scale=1, size=1000)

# Paso 2: Crear una funci√≥n para graficar y calcular medidas
def plot_distribution(data, title):
    # Calcular medidas de tendencia central y dispersi√≥n
    mean = np.mean(data)
    median = np.median(data)
    std_dev = np.std(data)

    # Calcular coeficiente de asimetr√≠a de Pearson y de Fisher
    skewness_pearson = 3 * (mean - median) / std_dev
    skewness_fisher = skew(data)

    # Crear figura y subplots
    fig, axs = plt.subplots(1, 2, figsize=(15, 6))
    fig.suptitle(title, fontsize=16)

    # Gr√°fico de Histograma
    sns.histplot(data, kde=True, ax=axs[0], color="skyblue")
    axs[0].axvline(mean, color='red', linestyle='--', label=f'Media = {mean:.2f}')
    axs[0].axvline(median, color='green', linestyle='-', label=f'Mediana = {median:.2f}')
    axs[0].legend()
    axs[0].set_title("Histograma con Media y Mediana")
    axs[0].set_xlabel("Valor")
    axs[0].set_ylabel("Frecuencia")

    # Gr√°fico de Box Plot
    sns.boxplot(data=data, ax=axs[1], color="lightcoral")
    axs[1].axvline(mean, color='red', linestyle='--', label='Media')
    axs[1].axvline(median, color='green', linestyle='-', label='Mediana')
    axs[1].legend()
    axs[1].set_title("Box Plot con Media y Mediana")
    axs[1].set_xlabel("Valor")

    # Mostrar resultados
    plt.figtext(0.15, -0.1, f"Coeficiente de Asimetr√≠a de Pearson: {skewness_pearson:.2f}", fontsize=12)
    plt.figtext(0.15, -0.15, f"Coeficiente de Asimetr√≠a de Fisher: {skewness_fisher:.2f}", fontsize=12)
    plt.show()

# Paso 3: Graficar y analizar cada distribuci√≥n
plot_distribution(data_symmetric, "Distribuci√≥n Sim√©trica (Normal)")
plot_distribution(data_positive_skew, "Distribuci√≥n con Asimetr√≠a Positiva (Exponencial)")
plot_distribution(data_negative_skew, "Distribuci√≥n con Asimetr√≠a Negativa (Inversa de Exponencial)")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, kurtosis

# Generaci√≥n de datos para las distintas curtosis

# Leptoc√∫rtica (pico alto, colas gruesas)
data_lepto = np.random.normal(0, 0.5, 1000) ** 2  # Datos con distribuci√≥n m√°s concentrada en el centro

# Mesoc√∫rtica (distribuci√≥n normal est√°ndar)
data_meso = np.random.normal(0, 1, 1000)

# Platic√∫rtica (pico bajo, colas finas)
data_plati = np.random.uniform(-2, 2, 1000)  # Distribuci√≥n uniforme para menor concentraci√≥n en el centro

# C√°lculo de la curtosis
kurtosis_lepto = kurtosis(data_lepto)
kurtosis_meso = kurtosis(data_meso)
kurtosis_plati = kurtosis(data_plati)

# Crear las distribuciones para graficar
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Funci√≥n para graficar cada distribuci√≥n y mostrar la curtosis
def plot_distribution(ax, data, color, title, kurtosis_value):
    ax.hist(data, bins=30, density=True, color=color, alpha=0.6)
    x = np.linspace(-3, 3, 1000)
    ax.plot(x, norm.pdf(x, np.mean(data), np.std(data)), color=color)
    ax.axvline(0, color="red", linestyle="--", linewidth=1)
    ax.set_title(f"{title}\nCurtosis: {kurtosis_value:.2f}")
    ax.set_xlabel("Valor")
    ax.set_ylabel("Densidad")

# Leptoc√∫rtica
plot_distribution(axs[0], data_lepto, "blue", "Leptoc√∫rtica", kurtosis_lepto)

# Mesoc√∫rtica
plot_distribution(axs[1], data_meso, "black", "Mesoc√∫rtica", kurtosis_meso)

# Platic√∫rtica
plot_distribution(axs[2], data_plati, "orange", "Platic√∫rtica", kurtosis_plati)

# Configuraci√≥n de gr√°ficos
for ax in axs:
    ax.grid(True, linestyle="--", alpha=0.7)

plt.suptitle("Diferentes tipos de Curtosis: Leptoc√∫rtica, Mesoc√∫rtica y Platic√∫rtica", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Configuraci√≥n de Seaborn para un estilo visual agradable
sns.set(style="whitegrid")

# Paso 1: Generar datos para cada tipo de correlaci√≥n

# Correlaci√≥n Positiva
np.random.seed(0)
x_positive = np.linspace(0, 10, 100)
y_positive = 2 * x_positive + np.random.normal(0, 2, 100)

# Correlaci√≥n Negativa
x_negative = np.linspace(0, 10, 100)
y_negative = -2 * x_negative + np.random.normal(0, 2, 100)

# Correlaci√≥n Cero (Sin relaci√≥n)
x_zero = np.linspace(0, 10, 100)
y_zero = np.random.normal(0, 5, 100)

# Paso 2: Crear funci√≥n para graficar y mostrar correlaci√≥n
def plot_correlation(x, y, title):
    # Calcular el coeficiente de correlaci√≥n de Pearson
    corr_coef, _ = pearsonr(x, y)

    # Crear el scatter plot
    plt.figure(figsize=(8, 6))
    sns.scatterplot(x=x, y=y, color="dodgerblue", s=60, alpha=0.7, edgecolor="black")

    # Agregar l√≠nea de tendencia
    sns.regplot(x=x, y=y, scatter=False, color="red", line_kws={"linewidth": 2, "linestyle": "--"})

    # T√≠tulo y etiquetas
    plt.title(f"{title}\nCoeficiente de Correlaci√≥n de Pearson: {corr_coef:.2f}", fontsize=14)
    plt.xlabel("Variable X")
    plt.ylabel("Variable Y")
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Paso 3: Graficar y analizar cada tipo de correlaci√≥n
plot_correlation(x_positive, y_positive, "Correlaci√≥n Positiva")
plot_correlation(x_negative, y_negative, "Correlaci√≥n Negativa")
plot_correlation(x_zero, y_zero, "Correlaci√≥n Cero (Sin Relaci√≥n)")
```
