```python
from IPython.display import display, HTML

html_content = """
<!DOCTYPE html>
<html>
<head>
    <style>
        body {
            font-family: Arial, sans-serif;
            line-height: 1.6;
            color: #333;
        }
        h1, h2, h3, h4 {
            color: #2C3E50;
        }
        p {
            margin-bottom: 15px;
        }
        ul {
            list-style-type: none;
            padding-left: 0;
        }
        li {
            margin-bottom: 10px;
        }
        .sublist {
            list-style-type: disc;
            margin-left: 20px;
        }
        .highlight {
            font-weight: bold;
            color: #2980B9;
        }
    </style>
</head>
<body>

<h1>Paradigma EDA (Análisis Exploratorio de Datos)</h1>
<p>El <span class="highlight">paradigma EDA</span> (Análisis Exploratorio de Datos, por sus siglas en inglés: <em>Exploratory Data Analysis</em>) es un enfoque en estadística y ciencia de datos que se centra en examinar y comprender conjuntos de datos de manera visual e intuitiva antes de realizar análisis estadísticos formales o modelos predictivos. Este paradigma fue introducido por el estadístico John Tukey en la década de 1970 y es fundamental en la exploración inicial de los datos en cualquier proyecto de análisis.</p>

<h2>Objetivos del Paradigma EDA</h2>
<p>El EDA tiene varios objetivos clave:</p>
<ul>
    <li><span class="highlight">Identificar patrones y relaciones</span>: Busca descubrir patrones, tendencias y relaciones iniciales entre variables que puedan orientar el análisis posterior.</li>
    <li><span class="highlight">Detección de anomalías y valores atípicos</span>: Permite identificar datos fuera de lo común (outliers) o errores de registro que podrían afectar el análisis o los modelos predictivos.</li>
    <li><span class="highlight">Comprensión de la distribución de datos</span>: Ayuda a evaluar cómo se distribuyen los datos, identificar posibles asimetrías, sesgos, y si existen concentraciones de datos en ciertos rangos.</li>
    <li><span class="highlight">Verificación de supuestos estadísticos</span>: Ayuda a comprobar si los datos cumplen con los supuestos necesarios para aplicar métodos estadísticos o modelos específicos (por ejemplo, normalidad, homocedasticidad).</li>
    <li><span class="highlight">Selección de técnicas de análisis</span>: El EDA permite elegir las técnicas estadísticas o de modelado adecuadas, dependiendo de la estructura y características de los datos observados.</li>
</ul>

<h2>Herramientas y Técnicas del EDA</h2>
<p>El EDA emplea una variedad de herramientas, principalmente visuales y estadísticas descriptivas, para explorar los datos. Entre las herramientas comunes se encuentran:</p>
<ul>
    <li><span class="highlight">Gráficos de dispersión (scatter plots)</span>: Para visualizar relaciones entre dos variables numéricas.</li>
    <li><span class="highlight">Histogramas y diagramas de densidad</span>: Para ver la distribución de datos en una variable continua.</li>
    <li><span class="highlight">Diagramas de caja (box plots)</span>: Para observar la distribución de datos, los rangos intercuartílicos y los valores atípicos.</li>
    <li><span class="highlight">Tablas de frecuencias y gráficos de barras</span>: Útiles para analizar la distribución de variables categóricas.</li>
    <li><span class="highlight">Matrices de correlación y diagramas de correlación</span>: Para evaluar relaciones entre múltiples variables.</li>
</ul>

<h2>Importancia del Paradigma EDA</h2>
<p>El EDA es fundamental porque proporciona una comprensión profunda y exploratoria de los datos que guía todas las etapas posteriores de análisis. Sin un EDA adecuado, es fácil pasar por alto patrones importantes o malinterpretar el contexto y las limitaciones de los datos, lo cual puede llevar a modelos inexactos o conclusiones incorrectas. En el contexto de la ciencia de datos y el aprendizaje automático, el EDA es también esencial para preparar y limpiar los datos, ya que ayuda a transformar y ajustar los datos para que sean adecuados para el modelado.</p>

<p>En resumen, el paradigma EDA es una fase crítica en cualquier análisis de datos, que permite conocer y explorar los datos de manera intuitiva y detallada, preparándolos adecuadamente para un análisis más riguroso y fundamentado.</p>

</body>
</html>
"""

display(HTML(html_content))
```

```python
import seaborn as sns
import pandas as pd

# Cargar el dataset de propinas
tips = sns.load_dataset("tips")

# Clasificación de variables
qualitative = []  # Variables cualitativas (categóricas)
quantitative = []  # Variables cuantitativas (numéricas)

for column in tips.columns:
    if pd.api.types.is_numeric_dtype(tips[column]):
        # Variables cuantitativas (numéricas)
        quantitative.append(column)
    else:
        # Variables cualitativas (categóricas)
        qualitative.append(column)

# Mostrar clasificación de las variables
print("Variables Cualitativas (Categóricas):", qualitative)
print("Variables Cuantitativas (Numéricas):", quantitative)

```

```python
from IPython.display import display, HTML

html_content = """
<!DOCTYPE html>
<html>
<head>
    <style>
        body {
            font-family: Arial, sans-serif;
            line-height: 1.6;
            color: #333;
        }
        h1, h2, h3 {
            color: #2C3E50;
            margin-bottom: 10px;
        }
        p {
            margin-bottom: 15px;
        }
        ul {
            margin-top: 5px;
            margin-bottom: 15px;
        }
        li {
            margin-bottom: 5px;
        }
        .highlight {
            font-weight: bold;
            color: #2980B9;
        }
        .section-title {
            color: #2C3E50;
            font-size: 1.2em;
            margin-top: 20px;
            font-weight: bold;
        }
    </style>
</head>
<body>

<h1>Definiciones y Propósitos de Gráficos en Análisis Univariado</h1>

<h2>1. Histograma</h2>
<p>Un <span class="highlight">histograma</span> es una representación gráfica que muestra la distribución de una variable cuantitativa dividiendo el rango de datos en intervalos llamados "bins" o "clases". Cada "bin" representa un rango específico de valores, y la altura de cada barra indica la frecuencia o el número de observaciones que caen dentro de ese intervalo.</p>
<ul>
    <li><span class="highlight">Eje X (horizontal)</span>: Muestra los intervalos de la variable.</li>
    <li><span class="highlight">Eje Y (vertical)</span>: Indica la frecuencia (número de observaciones) para cada intervalo.</li>
    <li><span class="highlight">Objetivo</span>: Permitir una visualización rápida de la distribución de la variable, identificar patrones, concentraciones de datos, simetrías, sesgos y valores extremos.</li>
    <li><span class="highlight">Aplicación</span>: Utilizado para datos cuantitativos continuos y discretos.</li>
</ul>
<p>El histograma es útil para evaluar la forma de la distribución de los datos y permite identificar si la variable sigue una distribución normal (simétrica) o si está sesgada (asimétrica).</p>

<h2>2. Diagrama de Densidad (KDE - Kernel Density Estimation)</h2>
<p>Un <span class="highlight">diagrama de densidad</span> es una curva que estima la distribución de probabilidad de una variable cuantitativa, suavizando las frecuencias observadas en los datos. Este gráfico es una representación continua y suave de la distribución, a diferencia del histograma que es discreto.</p>
<ul>
    <li><span class="highlight">Eje X (horizontal)</span>: Representa los valores de la variable.</li>
    <li><span class="highlight">Eje Y (vertical)</span>: Muestra la densidad estimada, que es una aproximación a la frecuencia relativa de los valores.</li>
    <li><span class="highlight">Kernel Density Estimation (KDE)</span>: La curva KDE utiliza funciones de "kernel" para suavizar los datos y dar una estimación continua de la densidad.</li>
    <li><span class="highlight">Objetivo</span>: Visualizar la densidad de la distribución y facilitar la comparación de la forma de la distribución con otros conjuntos de datos.</li>
    <li><span class="highlight">Aplicación</span>: Adecuado para variables cuantitativas continuas.</li>
</ul>
<p>El diagrama de densidad es ideal para observar la forma de la distribución sin la dependencia de intervalos específicos, proporcionando una visión más suave de los datos.</p>

<h2>3. Diagrama de Caja (Box Plot)</h2>
<p>Un <span class="highlight">diagrama de caja</span>, también conocido como <em>box plot</em>, es una representación gráfica que resume la distribución de una variable cuantitativa a través de sus cuartiles y permite identificar valores atípicos.</p>
<ul>
    <li><span class="highlight">Caja</span>: Representa el rango intercuartílico (IQR), que abarca del percentil 25 (Q1) al percentil 75 (Q3).</li>
    <li><span class="highlight">Línea en la caja (mediana)</span>: La línea dentro de la caja indica la mediana (percentil 50%) de los datos.</li>
    <li><span class="highlight">Bigotes</span>: Suelen extenderse hasta el valor mínimo y máximo dentro de 1.5 veces el rango intercuartílico (IQR).</li>
    <li><span class="highlight">Puntos Atípicos</span>: Los valores que caen fuera del rango de los bigotes se muestran como puntos separados.</li>
    <li><span class="highlight">Objetivo</span>: Mostrar la distribución de los datos, resaltando la mediana, la dispersión y los valores atípicos.</li>
    <li><span class="highlight">Aplicación</span>: Utilizado para variables cuantitativas continuas o discretas.</li>
</ul>
<p>El diagrama de caja es útil para comparar distribuciones entre diferentes categorías y para identificar rápidamente valores extremos.</p>

<h2>4. Gráfico de Barras</h2>
<p>Un <span class="highlight">gráfico de barras</span> es una representación gráfica que muestra la frecuencia o el conteo de categorías de una variable cualitativa. Cada barra representa una categoría distinta y la altura de la barra indica el número de observaciones o la proporción de esa categoría en los datos.</p>
<ul>
    <li><span class="highlight">Eje X (horizontal)</span>: Muestra las categorías de la variable cualitativa.</li>
    <li><span class="highlight">Eje Y (vertical)</span>: Indica la frecuencia o el conteo de cada categoría.</li>
    <li><span class="highlight">Barra</span>: Cada barra representa una categoría específica, y su altura es proporcional al conteo de observaciones en esa categoría.</li>
    <li><span class="highlight">Objetivo</span>: Proporcionar una comparación visual de la frecuencia de distintas categorías y mostrar la moda (categoría más frecuente).</li>
    <li><span class="highlight">Aplicación</span>: Adecuado para variables cualitativas (nominales y ordinales).</li>
</ul>
<p>El gráfico de barras es ideal para observar la popularidad o distribución de categorías en una variable, permitiendo una comparación rápida entre diferentes grupos o subcategorías.</p>

<p>Estos gráficos juntos permiten explorar y comprender la estructura y patrones de los datos en profundidad, facilitando la interpretación y análisis de cada tipo de variable.</p>

</body>
</html>
"""

# Muestra el contenido HTML en Google Colab
display(HTML(html_content))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Cargar el dataset de propinas
tips = sns.load_dataset("tips")

# Análisis univariado de variables cuantitativas, incluyendo asimetría y curtosis
def univariate_analysis_numeric(data, column):
    print(f"\nAnálisis univariado de la variable numérica: {column}")
    print(data[column].describe())  # Medidas estadísticas descriptivas

    # Calcular asimetría y curtosis
    skewness = skew(data[column].dropna())
    kurt = kurtosis(data[column].dropna())
    print(f"Asimetría (Skewness) de {column}: {skewness:.2f}")
    print(f"Curtosis (Kurtosis) de {column}: {kurt:.2f}")

    # Histograma y diagrama de densidad
    plt.figure(figsize=(10, 5))
    sns.histplot(data[column], kde=True)
    plt.title(f"Distribución de {column} (Asimetría: {skewness:.2f}, Curtosis: {kurt:.2f})")
    plt.xlabel(column)
    plt.ylabel("Frecuencia")
    plt.show()

    # Diagrama de caja (box plot)
    plt.figure(figsize=(10, 2))
    sns.boxplot(data[column])
    plt.title(f"Diagrama de caja de {column}")
    plt.show()

# Análisis univariado de variables cualitativas
def univariate_analysis_categorical(data, column):
    print(f"\nAnálisis univariado de la variable cualitativa: {column}")
    print(data[column].value_counts())  # Frecuencias de cada categoría

    # Gráfico de barras
    plt.figure(figsize=(10, 5))
    sns.countplot(x=column, data=data)
    plt.title(f"Distribución de {column}")
    plt.xlabel(column)
    plt.ylabel("Frecuencia")
    plt.show()

# Análisis de variables cuantitativas (total_bill, tip, size)
quantitative_columns = ["total_bill", "tip", "size"]
for column in quantitative_columns:
    univariate_analysis_numeric(tips, column)

# Análisis de variables cualitativas (sex, smoker, day, time)
qualitative_columns = ["sex", "smoker", "day", "time"]
for column in qualitative_columns:
    univariate_analysis_categorical(tips, column)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest

# Cargar el dataset de propinas
tips = sns.load_dataset("tips")

# Configurar el estilo de los gráficos para mayor claridad
sns.set(style="whitegrid")

# 1. Detección de Anomalías en Variables Numéricas Usando Diagramas de Caja (Box Plots)
numerical_columns = ["total_bill", "tip", "size"]

plt.figure(figsize=(15, 5))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(1, 3, i)
    sns.boxplot(y=tips[column])
    plt.title(f"Diagrama de Caja de {column}")
    plt.ylabel(column)
plt.tight_layout()
plt.show()

# 2. Análisis Z-Score para Identificar Outliers Numéricos
def detect_anomalies_z_score(data, threshold=3):
    anomalies = {}
    for column in numerical_columns:
        # Calcular el Z-score
        z_scores = np.abs((data[column] - data[column].mean()) / data[column].std())
        # Detectar valores que superen el umbral de 3
        anomalies[column] = data[z_scores > threshold][column]
    return anomalies

# Aplicar la función para detectar anomalías
anomalies = detect_anomalies_z_score(tips)
print("Anomalías detectadas mediante Z-score:")
for column, anomaly_data in anomalies.items():
    print(f"\n{column}:\n{anomaly_data}")

# 3. Gráficos de Dispersión para Identificar Anomalías en Relaciones Bivariadas
plt.figure(figsize=(8, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.title("Relación entre Total de la Cuenta y Propina")
plt.xlabel("Total de la Cuenta (total_bill)")
plt.ylabel("Propina (tip)")
plt.show()

# 4. DBSCAN para Detectar Anomalías en el Espacio Multidimensional
# Seleccionar variables numéricas y estandarizarlas
numeric_data = tips[["total_bill", "tip", "size"]]
scaler = StandardScaler()
numeric_data_scaled = scaler.fit_transform(numeric_data)

# Aplicar DBSCAN para detectar anomalías
dbscan = DBSCAN(eps=1.5, min_samples=5)  # Ajusta `eps` y `min_samples` según los datos
labels = dbscan.fit_predict(numeric_data_scaled)

# Agregar los resultados al DataFrame
tips["cluster"] = labels

# Visualización de los resultados de DBSCAN
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='cluster', palette="deep", style="cluster", s=100)
plt.title("Detección de Anomalías usando DBSCAN")
plt.xlabel("Total de la Cuenta (total_bill)")
plt.ylabel("Propina (tip)")
plt.legend(title="Cluster")
plt.show()

# 5. Isolation Forest para Detectar Anomalías en Datos Multivariados
# Aplicar Isolation Forest para detectar anomalías
iso_forest = IsolationForest(contamination=0.05, random_state=42)
outliers = iso_forest.fit_predict(numeric_data_scaled)

# Agregar el resultado de Isolation Forest al DataFrame
tips["outlier"] = outliers

# Visualización de los outliers detectados
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='outlier', palette={1: 'blue', -1: 'red'}, style="outlier", s=100)
plt.title("Detección de Anomalías usando Isolation Forest")
plt.xlabel("Total de la Cuenta (total_bill)")
plt.ylabel("Propina (tip)")
plt.legend(title="Anomalía (1: Normal, -1: Anomalía)")
plt.show()

# Resumen de interpretación de las técnicas de detección de anomalías
print("Resumen de Técnicas de Detección de Anomalías:")
print("1. Box Plots: Visualiza valores atípicos individuales en cada variable numérica.")
print("2. Z-score: Identifica valores fuera de 3 desviaciones estándar en total_bill, tip y size.")
print("3. Gráfico de Dispersión: Permite identificar visualmente puntos fuera de la tendencia entre total_bill y tip.")
print("4. DBSCAN: Detecta anomalías multivariadas como ruido en clustering de densidad.")
print("5. Isolation Forest: Marca anomalías en datos multivariados utilizando un modelo basado en árboles.")
```

```python
import seaborn as sns

# Cargar el dataset de propinas
tips = sns.load_dataset("tips")

# Agregar una columna para calcular el porcentaje de propina
tips['tip_percentage'] = (tips['tip'] / tips['total_bill']) * 100

# Definir una propina alta como superior al 20%
tips['high_tip'] = tips['tip_percentage'] > 20

# Calcular las probabilidades necesarias para aplicar el Teorema de Bayes

# P(fumador): Probabilidad de que un cliente sea fumador
p_fumador = (tips['smoker'] == 'Yes').mean()

# P(propina alta): Probabilidad de que un cliente deje una propina alta
p_high_tip = tips['high_tip'].mean()

# P(propina alta | fumador): Probabilidad de dejar una propina alta dado que el cliente es fumador
p_high_tip_given_fumador = tips[tips['smoker'] == 'Yes']['high_tip'].mean()

# Aplicar el Teorema de Bayes
# P(fumador | propina alta) = P(propina alta | fumador) * P(fumador) / P(propina alta)
p_fumador_given_high_tip = (p_high_tip_given_fumador * p_fumador) / p_high_tip

print(f"Probabilidad de que un cliente sea fumador dado que dejó una propina alta: {p_fumador_given_high_tip:.2f}")
```
