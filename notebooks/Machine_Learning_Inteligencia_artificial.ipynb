```python
#!/usr/bin/env python3
"""
Sistema Combinacional - Implementación de Puertas Lógicas y Circuitos
Un sistema combinacional es aquel donde la salida depende únicamente de las entradas actuales,
sin memoria ni estados previos.
"""

from typing import List, Dict, Any
import itertools


class PuertaLogica:
    """Clase base para todas las puertas lógicas."""

    def __init__(self, nombre: str):
        self.nombre = nombre

    def evaluar(self, *entradas) -> bool:
        """Método abstracto que debe ser implementado por cada puerta."""
        raise NotImplementedError("Debe implementar el método evaluar")

    def tabla_verdad(self, num_entradas: int) -> List[Dict[str, Any]]:
        """Genera la tabla de verdad para la puerta lógica."""
        tabla = []
        for combinacion in itertools.product([False, True], repeat=num_entradas):
            salida = self.evaluar(*combinacion)
            entrada_dict = {f'A{i}': entrada for i, entrada in enumerate(combinacion)}
            entrada_dict['Salida'] = salida
            tabla.append(entrada_dict)
        return tabla


class AND(PuertaLogica):
    """Puerta AND: Salida verdadera solo si todas las entradas son verdaderas."""

    def __init__(self):
        super().__init__("AND")

    def evaluar(self, *entradas) -> bool:
        return all(entradas)


class OR(PuertaLogica):
    """Puerta OR: Salida verdadera si al menos una entrada es verdadera."""

    def __init__(self):
        super().__init__("OR")

    def evaluar(self, *entradas) -> bool:
        return any(entradas)


class NOT(PuertaLogica):
    """Puerta NOT: Invierte la entrada."""

    def __init__(self):
        super().__init__("NOT")

    def evaluar(self, entrada) -> bool:
        return not entrada


class NAND(PuertaLogica):
    """Puerta NAND: NOT AND - Salida falsa solo si todas las entradas son verdaderas."""

    def __init__(self):
        super().__init__("NAND")

    def evaluar(self, *entradas) -> bool:
        return not all(entradas)


class NOR(PuertaLogica):
    """Puerta NOR: NOT OR - Salida verdadera solo si todas las entradas son falsas."""

    def __init__(self):
        super().__init__("NOR")

    def evaluar(self, *entradas) -> bool:
        return not any(entradas)


class XOR(PuertaLogica):
    """Puerta XOR: Salida verdadera si el número de entradas verdaderas es impar."""

    def __init__(self):
        super().__init__("XOR")

    def evaluar(self, *entradas) -> bool:
        return sum(entradas) % 2 == 1


class XNOR(PuertaLogica):
    """Puerta XNOR: NOT XOR - Salida verdadera si el número de entradas verdaderas es par."""

    def __init__(self):
        super().__init__("XNOR")

    def evaluar(self, *entradas) -> bool:
        return sum(entradas) % 2 == 0


class CircuitoCombinacional:
    """Clase para crear circuitos combinacionales complejos."""

    def __init__(self, nombre: str):
        self.nombre = nombre
        self.puertas = {}
        self.conexiones = {}

    def agregar_puerta(self, id_puerta: str, puerta: PuertaLogica):
        """Agrega una puerta lógica al circuito."""
        self.puertas[id_puerta] = puerta

    def conectar(self, salida_de: str, entrada_a: str, pin_entrada: int = 0):
        """Conecta la salida de una puerta a la entrada de otra."""
        if entrada_a not in self.conexiones:
            self.conexiones[entrada_a] = {}
        self.conexiones[entrada_a][pin_entrada] = salida_de

    def evaluar(self, entradas_primarias: Dict[str, bool]) -> Dict[str, bool]:
        """Evalúa el circuito con las entradas primarias dadas."""
        resultados = entradas_primarias.copy()

        # Evaluación en orden topológico (simplificado)
        cambios = True
        while cambios:
            cambios = False
            for id_puerta, puerta in self.puertas.items():
                if id_puerta not in resultados:
                    # Verificar si todas las entradas están disponibles
                    entradas_puerta = []
                    if id_puerta in self.conexiones:
                        entradas_disponibles = True
                        for pin in sorted(self.conexiones[id_puerta].keys()):
                            fuente = self.conexiones[id_puerta][pin]
                            if fuente in resultados:
                                entradas_puerta.append(resultados[fuente])
                            else:
                                entradas_disponibles = False
                                break

                        if entradas_disponibles:
                            resultados[id_puerta] = puerta.evaluar(*entradas_puerta)
                            cambios = True

        return resultados


# Ejemplos de circuitos combinacionales específicos

class Sumador1Bit:
    """Sumador de 1 bit (Half Adder)."""

    def __init__(self):
        self.xor = XOR()
        self.and_gate = AND()

    def sumar(self, a: bool, b: bool) -> tuple:
        """Suma dos bits y retorna (suma, acarreo)."""
        suma = self.xor.evaluar(a, b)
        acarreo = self.and_gate.evaluar(a, b)
        return suma, acarreo

    def tabla_verdad(self):
        """Genera la tabla de verdad del sumador."""
        tabla = []
        for a in [False, True]:
            for b in [False, True]:
                suma, acarreo = self.sumar(a, b)
                tabla.append({
                    'A': a, 'B': b,
                    'Suma': suma, 'Acarreo': acarreo
                })
        return tabla


class Multiplicador2x2:
    """Multiplicador de 2x2 bits."""

    def __init__(self):
        self.and_gates = [AND() for _ in range(4)]
        self.sumadores = [Sumador1Bit() for _ in range(2)]

    def multiplicar(self, a1: bool, a0: bool, b1: bool, b0: bool) -> tuple:
        """Multiplica dos números de 2 bits."""
        # Productos parciales
        p00 = self.and_gates[0].evaluar(a0, b0)
        p01 = self.and_gates[1].evaluar(a0, b1)
        p10 = self.and_gates[2].evaluar(a1, b0)
        p11 = self.and_gates[3].evaluar(a1, b1)

        # Bit menos significativo
        resultado_0 = p00

        # Segundo bit
        suma1, acarreo1 = self.sumadores[0].sumar(p01, p10)
        resultado_1 = suma1

        # Tercer bit
        suma2, acarreo2 = self.sumadores[1].sumar(p11, acarreo1)
        resultado_2 = suma2

        # Bit más significativo
        resultado_3 = acarreo2

        return resultado_3, resultado_2, resultado_1, resultado_0


class Decodificador2a4:
    """Decodificador de 2 a 4 líneas."""

    def __init__(self):
        self.not_gates = [NOT() for _ in range(2)]
        self.and_gates = [AND() for _ in range(4)]

    def decodificar(self, a1: bool, a0: bool, enable: bool = True) -> tuple:
        """Decodifica la entrada de 2 bits a 4 salidas."""
        if not enable:
            return False, False, False, False

        not_a1 = self.not_gates[0].evaluar(a1)
        not_a0 = self.not_gates[1].evaluar(a0)

        y0 = self.and_gates[0].evaluar(not_a1, not_a0)
        y1 = self.and_gates[1].evaluar(not_a1, a0)
        y2 = self.and_gates[2].evaluar(a1, not_a0)
        y3 = self.and_gates[3].evaluar(a1, a0)

        return y0, y1, y2, y3


def mostrar_tabla_verdad(tabla: List[Dict[str, Any]], titulo: str):
    """Función auxiliar para mostrar tablas de verdad."""
    print(f"\n{titulo}")
    print("-" * 40)

    if not tabla:
        return

    # Encabezados
    encabezados = list(tabla[0].keys())
    print(" | ".join(f"{h:^8}" for h in encabezados))
    print("-" * (len(encabezados) * 11 - 1))

    # Filas
    for fila in tabla:
        valores = []
        for valor in fila.values():
            if isinstance(valor, bool):
                valores.append("1" if valor else "0")
            else:
                valores.append(str(valor))
        print(" | ".join(f"{v:^8}" for v in valores))


def demo_puertas_logicas():
    """Demostración de puertas lógicas básicas."""
    print("=== DEMOSTRACIÓN DE PUERTAS LÓGICAS ===")

    puertas = [
        (AND(), 2), (OR(), 2), (NOT(), 1),
        (NAND(), 2), (NOR(), 2), (XOR(), 2), (XNOR(), 2)
    ]

    for puerta, num_entradas in puertas:
        tabla = puerta.tabla_verdad(num_entradas)
        mostrar_tabla_verdad(tabla, f"Puerta {puerta.nombre}")


def demo_circuitos_combinacionales():
    """Demostración de circuitos combinacionales complejos."""
    print("\n=== DEMOSTRACIÓN DE CIRCUITOS COMBINACIONALES ===")

    # Sumador de 1 bit
    sumador = Sumador1Bit()
    tabla_sumador = sumador.tabla_verdad()
    mostrar_tabla_verdad(tabla_sumador, "Sumador de 1 Bit (Half Adder)")

    # Multiplicador 2x2
    multiplicador = Multiplicador2x2()
    print("\nMultiplicador 2x2 bits:")
    print("-" * 40)
    print("A1 A0 | B1 B0 | P3 P2 P1 P0 | Decimal")
    print("-" * 40)

    for a1 in [False, True]:
        for a0 in [False, True]:
            for b1 in [False, True]:
                for b0 in [False, True]:
                    resultado = multiplicador.multiplicar(a1, a0, b1, b0)
                    # Convertir a decimal
                    a_decimal = (2 * a1) + a0
                    b_decimal = (2 * b1) + b0
                    resultado_decimal = (8 * resultado[0]) + (4 * resultado[1]) + (2 * resultado[2]) + resultado[3]

                    print(f" {int(a1)} {int(a0)}  |  {int(b1)} {int(b0)}  | " +
                          f" {int(resultado[0])} {int(resultado[1])} {int(resultado[2])} {int(resultado[3])}  |   {a_decimal} × {b_decimal} = {resultado_decimal}")

    # Decodificador 2 a 4
    decodificador = Decodificador2a4()
    print("\nDecodificador 2 a 4 líneas:")
    print("-" * 40)
    print("A1 A0 | Y3 Y2 Y1 Y0")
    print("-" * 40)

    for a1 in [False, True]:
        for a0 in [False, True]:
            salidas = decodificador.decodificar(a1, a0)
            print(f" {int(a1)} {int(a0)}  | {int(salidas[3])} {int(salidas[2])} {int(salidas[1])} {int(salidas[0])}")


def demo_circuito_personalizado():
    """Demostración de un circuito combinacional personalizado."""
    print("\n=== CIRCUITO COMBINACIONAL PERSONALIZADO ===")

    # Crear un circuito que implementa la función: F = (A AND B) OR (NOT C)
    circuito = CircuitoCombinacional("Función F = (A AND B) OR (NOT C)")

    # Agregar puertas
    circuito.agregar_puerta("and1", AND())
    circuito.agregar_puerta("not1", NOT())
    circuito.agregar_puerta("or1", OR())

    # Conectar el circuito
    circuito.conectar("A", "and1", 0)
    circuito.conectar("B", "and1", 1)
    circuito.conectar("C", "not1", 0)
    circuito.conectar("and1", "or1", 0)
    circuito.conectar("not1", "or1", 1)

    # Evaluar todas las combinaciones
    print("Función: F = (A AND B) OR (NOT C)")
    print("-" * 30)
    print("A B C | F")
    print("-" * 30)

    for a in [False, True]:
        for b in [False, True]:
            for c in [False, True]:
                entradas = {"A": a, "B": b, "C": c}
                resultados = circuito.evaluar(entradas)
                f = resultados.get("or1", False)
                print(f"{int(a)} {int(b)} {int(c)} | {int(f)}")


if __name__ == "__main__":
    # Ejecutar todas las demostraciones
    demo_puertas_logicas()
    demo_circuitos_combinacionales()
    demo_circuito_personalizado()

    print("\n=== CARACTERÍSTICAS DE LOS SISTEMAS COMBINACIONALES ===")
    print("1. La salida depende únicamente de las entradas actuales")
    print("2. No tienen memoria ni estados internos")
    print("3. No pueden realizar operaciones secuenciales")
    print("4. Son deterministas: misma entrada → misma salida")
    print("5. Respuesta instantánea (sin considerar retardos de propagación)")
```

```python
def finite_automaton(input_string):
    state = 0  # Estado inicial
    for char in input_string:
        if char == 'a':
            state = 1 - state  # Alterna entre estados 0 y 1
    return state == 0  # Acepta si termina en estado 0

# Ejemplo de uso
print("\nAutómata Finito - Número Par de 'a's")
print("Cadena 'aa' aceptada:", finite_automaton("aa"))
print("Cadena 'aaa' aceptada:", finite_automaton("aaa"))
print("Cadena 'aaaa' aceptada:", finite_automaton("aaaa"))
```

```python
def pushdown_automaton(expression):
    stack = []
    for char in expression:
        if char == '(':
            stack.append(char)
        elif char == ')':
            if not stack:
                return False  # Si la pila está vacía, no está balanceado
            stack.pop()
    return len(stack) == 0  # Acepta si la pila está vacía al final

# Ejemplo de uso
print("\nAutómata con Pila - Paréntesis Balanceados")
print("Expresión '(())' balanceada:", pushdown_automaton("(())"))
print("Expresión '(()' balanceada:", pushdown_automaton("(()"))
print("Expresión '())(' balanceada:", pushdown_automaton("())("))
```

```python
#!/usr/bin/env python3
"""
Autómata Finito - Sistema de Estados y Transiciones
Los autómatas finitos pueden manejar cálculos secuenciales simples con un conjunto
finito de estados y transiciones basadas en entradas.
"""

from typing import Dict, List, Set, Optional, Tuple, Any
from enum import Enum
import re
from dataclasses import dataclass


class TipoAutomata(Enum):
    """Tipos de autómatas finitos."""
    DETERMINISTA = "AFD"  # Autómata Finito Determinista
    NO_DETERMINISTA = "AFN"  # Autómata Finito No Determinista


@dataclass
class Transicion:
    """Representa una transición en el autómata."""
    estado_origen: str
    simbolo: str
    estado_destino: str

    def __str__(self):
        return f"δ({self.estado_origen}, {self.simbolo}) = {self.estado_destino}"


class AutomataFinito:
    """Clase base para autómatas finitos."""

    def __init__(self, nombre: str, tipo: TipoAutomata):
        self.nombre = nombre
        self.tipo = tipo
        self.estados: Set[str] = set()
        self.alfabeto: Set[str] = set()
        self.transiciones: Dict[Tuple[str, str], Set[str]] = {}
        self.estado_inicial: Optional[str] = None
        self.estados_finales: Set[str] = set()
        self.estado_actual: Optional[str] = None
        self.historial: List[Tuple[str, str, str]] = []  # (estado, símbolo, nuevo_estado)

    def agregar_estado(self, estado: str, es_inicial: bool = False, es_final: bool = False):
        """Agrega un estado al autómata."""
        self.estados.add(estado)
        if es_inicial:
            self.estado_inicial = estado
        if es_final:
            self.estados_finales.add(estado)

    def agregar_simbolo(self, simbolo: str):
        """Agrega un símbolo al alfabeto."""
        self.alfabeto.add(simbolo)

    def agregar_transicion(self, estado_origen: str, simbolo: str, estado_destino: str):
        """Agrega una transición al autómata."""
        if estado_origen not in self.estados:
            raise ValueError(f"Estado origen '{estado_origen}' no existe")
        if estado_destino not in self.estados:
            raise ValueError(f"Estado destino '{estado_destino}' no existe")

        clave = (estado_origen, simbolo)
        if clave not in self.transiciones:
            self.transiciones[clave] = set()
        self.transiciones[clave].add(estado_destino)
        self.alfabeto.add(simbolo)

    def reiniciar(self):
        """Reinicia el autómata al estado inicial."""
        self.estado_actual = self.estado_inicial
        self.historial = []

    def obtener_transiciones_desde(self, estado: str) -> List[Transicion]:
        """Obtiene todas las transiciones desde un estado dado."""
        transiciones = []
        for (estado_orig, simbolo), estados_dest in self.transiciones.items():
            if estado_orig == estado:
                for estado_dest in estados_dest:
                    transiciones.append(Transicion(estado_orig, simbolo, estado_dest))
        return transiciones

    def es_estado_final(self, estado: str) -> bool:
        """Verifica si un estado es final."""
        return estado in self.estados_finales

    def mostrar_estructura(self):
        """Muestra la estructura del autómata."""
        print(f"\n=== AUTÓMATA: {self.nombre} ({self.tipo.value}) ===")
        print(f"Estados: {sorted(self.estados)}")
        print(f"Alfabeto: {sorted(self.alfabeto)}")
        print(f"Estado inicial: {self.estado_inicial}")
        print(f"Estados finales: {sorted(self.estados_finales)}")
        print("\nTransiciones:")
        for (estado, simbolo), destinos in sorted(self.transiciones.items()):
            for destino in sorted(destinos):
                print(f"  δ({estado}, '{simbolo}') = {destino}")


class AFD(AutomataFinito):
    """Autómata Finito Determinista."""

    def __init__(self, nombre: str):
        super().__init__(nombre, TipoAutomata.DETERMINISTA)

    def agregar_transicion(self, estado_origen: str, simbolo: str, estado_destino: str):
        """En AFD, cada transición debe ser única."""
        clave = (estado_origen, simbolo)
        if clave in self.transiciones and self.transiciones[clave]:
            raise ValueError(f"Transición δ({estado_origen}, '{simbolo}') ya existe en AFD")
        super().agregar_transicion(estado_origen, simbolo, estado_destino)

    def procesar_simbolo(self, simbolo: str) -> bool:
        """Procesa un símbolo y cambia de estado."""
        if self.estado_actual is None:
            self.estado_actual = self.estado_inicial

        clave = (self.estado_actual, simbolo)
        if clave in self.transiciones:
            nuevo_estado = list(self.transiciones[clave])[0]  # AFD: solo un estado destino
            self.historial.append((self.estado_actual, simbolo, nuevo_estado))
            self.estado_actual = nuevo_estado
            return True
        return False

    def reconocer_cadena(self, cadena: str) -> bool:
        """Reconoce si una cadena es aceptada por el autómata."""
        self.reiniciar()

        for simbolo in cadena:
            if not self.procesar_simbolo(simbolo):
                return False

        return self.estado_actual in self.estados_finales

    def obtener_camino(self, cadena: str) -> List[str]:
        """Obtiene el camino de estados seguido para una cadena."""
        self.reiniciar()
        camino = [self.estado_actual]

        for simbolo in cadena:
            if self.procesar_simbolo(simbolo):
                camino.append(self.estado_actual)
            else:
                break

        return camino


class AFN(AutomataFinito):
    """Autómata Finito No Determinista."""

    def __init__(self, nombre: str):
        super().__init__(nombre, TipoAutomata.NO_DETERMINISTA)

    def procesar_simbolo_desde_estados(self, estados_actuales: Set[str], simbolo: str) -> Set[str]:
        """Procesa un símbolo desde un conjunto de estados."""
        nuevos_estados = set()
        for estado in estados_actuales:
            clave = (estado, simbolo)
            if clave in self.transiciones:
                nuevos_estados.update(self.transiciones[clave])
        return nuevos_estados

    def reconocer_cadena(self, cadena: str) -> bool:
        """Reconoce si una cadena es aceptada por el autómata."""
        estados_actuales = {self.estado_inicial}

        for simbolo in cadena:
            estados_actuales = self.procesar_simbolo_desde_estados(estados_actuales, simbolo)
            if not estados_actuales:
                return False

        return bool(estados_actuales.intersection(self.estados_finales))


# Ejemplos específicos de autómatas finitos

class ValidadorEmail(AFD):
    """Autómata para validar estructura básica de email."""

    def __init__(self):
        super().__init__("Validador de Email")
        self._construir_automata()

    def _construir_automata(self):
        # Estados
        self.agregar_estado("inicio", es_inicial=True)
        self.agregar_estado("usuario")
        self.agregar_estado("arroba")
        self.agregar_estado("dominio")
        self.agregar_estado("punto")
        self.agregar_estado("extension", es_final=True)

        # Transiciones
        # Caracteres de usuario (letras y números)
        for char in "abcdefghijklmnopqrstuvwxyz0123456789":
            self.agregar_transicion("inicio", char, "usuario")
            self.agregar_transicion("usuario", char, "usuario")
            self.agregar_transicion("arroba", char, "dominio")
            self.agregar_transicion("dominio", char, "dominio")
            self.agregar_transicion("punto", char, "extension")
            self.agregar_transicion("extension", char, "extension")

        # Símbolo @
        self.agregar_transicion("usuario", "@", "arroba")

        # Símbolo .
        self.agregar_transicion("dominio", ".", "punto")


class ContadorBinario(AFD):
    """Autómata que cuenta números binarios módulo 3."""

    def __init__(self):
        super().__init__("Contador Binario Módulo 3")
        self._construir_automata()

    def _construir_automata(self):
        # Estados representan el resto de la división por 3
        self.agregar_estado("q0", es_inicial=True, es_final=True)  # resto 0
        self.agregar_estado("q1")  # resto 1
        self.agregar_estado("q2")  # resto 2

        # Transiciones para dígitos binarios
        # Cuando leemos un '0', multiplicamos por 2
        self.agregar_transicion("q0", "0", "q0")  # (0 * 2) mod 3 = 0
        self.agregar_transicion("q1", "0", "q2")  # (1 * 2) mod 3 = 2
        self.agregar_transicion("q2", "0", "q1")  # (2 * 2) mod 3 = 1

        # Cuando leemos un '1', multiplicamos por 2 y sumamos 1
        self.agregar_transicion("q0", "1", "q1")  # (0 * 2 + 1) mod 3 = 1
        self.agregar_transicion("q1", "1", "q0")  # (1 * 2 + 1) mod 3 = 0
        self.agregar_transicion("q2", "1", "q2")  # (2 * 2 + 1) mod 3 = 2


class DetectorPatron(AFD):
    """Autómata que detecta el patrón 'aba' en una cadena."""

    def __init__(self):
        super().__init__("Detector de Patrón 'aba'")
        self._construir_automata()

    def _construir_automata(self):
        self.agregar_estado("q0", es_inicial=True)
        self.agregar_estado("q1")
        self.agregar_estado("q2")
        self.agregar_estado("q3", es_final=True)

        # Transiciones para detectar 'aba'
        self.agregar_transicion("q0", "a", "q1")
        self.agregar_transicion("q0", "b", "q0")

        self.agregar_transicion("q1", "a", "q1")
        self.agregar_transicion("q1", "b", "q2")

        self.agregar_transicion("q2", "a", "q3")
        self.agregar_transicion("q2", "b", "q0")

        self.agregar_transicion("q3", "a", "q1")
        self.agregar_transicion("q3", "b", "q2")


class MaquinaExpendedora(AFD):
    """Autómata que simula una máquina expendedora simple."""

    def __init__(self, precio_producto: int):
        super().__init__(f"Máquina Expendedora (${precio_producto})")
        self.precio_producto = precio_producto
        self._construir_automata()

    def _construir_automata(self):
        # Estados representan el dinero acumulado
        for i in range(self.precio_producto + 1):
            es_final = (i == self.precio_producto)
            self.agregar_estado(f"${i}", es_inicial=(i == 0), es_final=es_final)

        # Transiciones para monedas
        for dinero_actual in range(self.precio_producto):
            # Moneda de $1
            if dinero_actual + 1 <= self.precio_producto:
                self.agregar_transicion(f"${dinero_actual}", "1", f"${dinero_actual + 1}")

            # Moneda de $5
            if dinero_actual + 5 <= self.precio_producto:
                self.agregar_transicion(f"${dinero_actual}", "5", f"${dinero_actual + 5}")

            # Moneda de $10
            if dinero_actual + 10 <= self.precio_producto:
                self.agregar_transicion(f"${dinero_actual}", "10", f"${dinero_actual + 10}")


class AnalizadorLexicografico(AFD):
    """Autómata que reconoce identificadores válidos en programación."""

    def __init__(self):
        super().__init__("Analizador Lexicográfico")
        self._construir_automata()

    def _construir_automata(self):
        self.agregar_estado("inicio", es_inicial=True)
        self.agregar_estado("identificador", es_final=True)

        # Debe comenzar con letra o guión bajo
        for char in "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ_":
            self.agregar_transicion("inicio", char, "identificador")
            self.agregar_transicion("identificador", char, "identificador")

        # Puede contener números después del primer carácter
        for char in "0123456789":
            self.agregar_transicion("identificador", char, "identificador")


def probar_automata(automata: AutomataFinito, cadenas_prueba: List[str]):
    """Prueba un autómata con varias cadenas."""
    print(f"\n=== PRUEBAS PARA {automata.nombre} ===")

    for cadena in cadenas_prueba:
        if isinstance(automata, AFD):
            resultado = automata.reconocer_cadena(cadena)
            camino = automata.obtener_camino(cadena)
            print(f"Cadena: '{cadena}' -> {'ACEPTADA' if resultado else 'RECHAZADA'}")
            print(f"  Camino: {' -> '.join(camino)}")
        else:
            resultado = automata.reconocer_cadena(cadena)
            print(f"Cadena: '{cadena}' -> {'ACEPTADA' if resultado else 'RECHAZADA'}")


def demo_automatas_basicos():
    """Demostración de autómatas básicos."""
    print("=== DEMOSTRACIÓN DE AUTÓMATAS FINITOS ===")

    # Autómata simple que acepta cadenas con número par de 'a's
    automata_par = AFD("Número Par de 'a's")
    automata_par.agregar_estado("par", es_inicial=True, es_final=True)
    automata_par.agregar_estado("impar")

    automata_par.agregar_transicion("par", "a", "impar")
    automata_par.agregar_transicion("impar", "a", "par")
    automata_par.agregar_transicion("par", "b", "par")
    automata_par.agregar_transicion("impar", "b", "impar")

    automata_par.mostrar_estructura()
    probar_automata(automata_par, ["", "a", "aa", "aaa", "aba", "abab", "baba"])


def demo_aplicaciones_practicas():
    """Demostración de aplicaciones prácticas."""
    print("\n=== APLICACIONES PRÁCTICAS ===")

    # Validador de email
    validador = ValidadorEmail()
    validador.mostrar_estructura()
    probar_automata(validador, [
        "user@domain.com",
        "test@example.org",
        "@domain.com",
        "user@",
        "user.domain.com",
        "a@b.c"
    ])

    # Contador binario
    contador = ContadorBinario()
    contador.mostrar_estructura()
    print("\n--- Números binarios divisibles por 3 ---")
    probar_automata(contador, [
        "0", "1", "10", "11", "100", "101", "110", "111", "1000", "1001"
    ])

    # Detector de patrón
    detector = DetectorPatron()
    detector.mostrar_estructura()
    probar_automata(detector, [
        "aba", "ababa", "baba", "aaba", "abab", "abababa", "bababa"
    ])

    # Máquina expendedora
    maquina = MaquinaExpendedora(15)
    maquina.mostrar_estructura()
    print("\n--- Secuencias de monedas para producto de $15 ---")
    probar_automata(maquina, [
        "111111111111111",  # 15 monedas de $1
        "51111111111",      # $5 + 10 monedas de $1
        "1051",             # $10 + $5
        "10111",            # $10 + 3 monedas de $1
        "555",              # 3 monedas de $5
        "10"                # Solo $10 (insuficiente)
    ])

    # Analizador lexicográfico
    analizador = AnalizadorLexicografico()
    analizador.mostrar_estructura()
    probar_automata(analizador, [
        "variable", "var123", "_private", "MyClass", "2invalid", "valid_name", "a1b2c3"
    ])


def demo_conversion_afn_afd():
    """Demostración de AFN y conversión conceptual."""
    print("\n=== AUTÓMATA FINITO NO DETERMINISTA ===")

    # AFN que acepta cadenas que terminan en 'ab'
    afn = AFN("Termina en 'ab'")
    afn.agregar_estado("q0", es_inicial=True)
    afn.agregar_estado("q1")
    afn.agregar_estado("q2", es_final=True)

    # Transiciones no deterministas
    afn.agregar_transicion("q0", "a", "q0")
    afn.agregar_transicion("q0", "b", "q0")
    afn.agregar_transicion("q0", "a", "q1")  # No determinista: desde q0 con 'a' puede ir a q0 o q1
    afn.agregar_transicion("q1", "b", "q2")

    afn.mostrar_estructura()
    probar_automata(afn, [
        "ab", "aab", "bab", "abab", "aba", "bb", "a", "b", "abcab"
    ])


def mostrar_propiedades_automatas():
    """Muestra las propiedades de los autómatas finitos."""
    print("\n=== PROPIEDADES DE LOS AUTÓMATAS FINITOS ===")
    print("1. Estados finitos: Conjunto limitado de estados posibles")
    print("2. Transiciones: Cambios de estado basados en entrada")
    print("3. Memoria limitada: Solo recuerdan el estado actual")
    print("4. Reconocimiento: Pueden reconocer lenguajes regulares")
    print("5. Determinismo: AFD vs AFN")
    print("\n=== APLICACIONES COMUNES ===")
    print("• Validación de patrones (emails, teléfonos)")
    print("• Análisis léxico en compiladores")
    print("• Protocolos de comunicación")
    print("• Sistemas de control simple")
    print("• Máquinas expendedoras")
    print("• Reconocimiento de texto")


if __name__ == "__main__":
    # Ejecutar todas las demostraciones
    demo_automatas_basicos()
    demo_aplicaciones_practicas()
    demo_conversion_afn_afd()
    mostrar_propiedades_automatas()
```
