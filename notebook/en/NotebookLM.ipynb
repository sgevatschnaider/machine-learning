```python
from IPython.display import display, HTML
from ipywidgets import Button, VBox, Output

# Define the texts
english_text = """
<h2>English Version</h2>
<p>Imagine transforming a complex document into a podcast-like experience in just minutes. With Google’s NotebookLM, this vision is now a reality. By harnessing cutting-edge AI tools such as SoundStorm and AudioLM, NotebookLM revolutionizes content creation, turning text into dynamic, lifelike audio discussions.</p>
<p>Whether you're summarizing lengthy reports, improving accessibility, or developing engaging educational materials, NotebookLM offers unmatched precision and efficiency. This article delves into the technologies behind this tool and the innovative processes that make it possible.</p>
<a href="https://economiayetica.blogspot.com/2024/11/transforming-text-into-podcasts-with.html" target="_blank" style="text-decoration:none;">
    <button style="background-color: blue; color: white; padding: 10px 20px; border: none; cursor: pointer;">Read more on my blog</button>
</a>
"""

spanish_text = """
<h2>Versión en Español</h2>
<p>Imagina transformar un documento complejo en una experiencia similar a un podcast en cuestión de minutos. Con NotebookLM de Google, esta visión se ha convertido en una realidad. Al aprovechar herramientas de IA de vanguardia, como SoundStorm y AudioLM, NotebookLM revoluciona la creación de contenido, convirtiendo texto en discusiones de audio dinámicas y realistas.</p>
<p>Ya sea que estés resumiendo informes extensos, mejorando la accesibilidad o desarrollando materiales educativos atractivos, NotebookLM ofrece una precisión y eficiencia sin igual. Este artículo profundiza en las tecnologías detrás de esta herramienta y los procesos innovadores que la hacen posible.</p>
<a href="https://economiayetica.blogspot.com/2024/11/transforming-text-into-podcasts-with.html" target="_blank" style="text-decoration:none;">
    <button style="background-color: blue; color: white; padding: 10px 20px; border: none; cursor: pointer;">Lee más en mi blog</button>
</a>
"""

# Function to display content based on language
def display_content(language):
    if language == 'en':
        display(HTML(english_text))
    elif language == 'es':
        display(HTML(spanish_text))

# Create buttons
button_en = Button(description='English', style={'button_color': 'blue', 'font_color': 'white'}, layout={'width': '200px'})
button_es = Button(description='Spanish', style={'button_color': 'blue', 'font_color': 'white'}, layout={'width': '200px'})

# Output widget
output = Output()

# Event handlers
def on_button_en_clicked(b):
    with output:
        output.clear_output()
        display_content('en')

def on_button_es_clicked(b):
    with output:
        output.clear_output()
        display_content('es')

# Attach event handlers
button_en.on_click(on_button_en_clicked)
button_es.on_click(on_button_es_clicked)

# Display buttons and output
display(VBox([button_en, button_es, output]))

```
